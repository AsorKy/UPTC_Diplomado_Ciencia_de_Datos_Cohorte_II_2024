## Mquinas de Soporte Vectorial (Support Vector Machines, SVM): Explicacin y Formalismo

Las **mquinas de soporte vectorial (SVM)** son un modelo supervisado utilizado tanto para clasificacin como para regresin, especialmente eficaz en problemas de clasificacin binaria. Su objetivo es encontrar un **hiperplano ptimo** que separe las clases de manera que la distancia entre los puntos ms cercanos de cada clase al hiperplano (margen) sea mxima, mejorando as la generalizacin del modelo.

### 1. Concepto Bsico

La SVM busca un **hiperplano** que separe los puntos de diferentes clases en el espacio de caractersticas. En un problema de clasificacin lineal, el hiperplano se define como una lnea en 2D o un plano en 3D que maximiza la separacin entre las dos clases. Esta separacin o **margen mximo** permite que el modelo sea robusto ante nuevas muestras y minimice el riesgo de error.

### 2. Formalismo Matemtico

Dado un conjunto de datos con \( n \) puntos \( (x_i, y_i) \), donde \( x_i \) es un punto de caractersticas y \( y_i \) su clase (normalmente \( y_i = \pm 1 \)), el objetivo de una SVM es encontrar un hiperplano definido por el vector normal \( w \) y el trmino de sesgo \( b \) que maximice el margen. El hiperplano cumple la siguiente ecuacin:

$$
w \cdot x + b = 0
$$

Para maximizar el margen, las SVM requieren resolver el siguiente problema de optimizacin, minimizando \( \| w \|^2 / 2 \) sujeto a la restriccin de que todos los puntos queden correctamente clasificados:

$$
y_i (w \cdot x_i + b) \geq 1, \; \forall i
$$

Este problema se convierte en una optimizacin cuadrtica que puede resolverse mediante el **mtodo del multiplicador de Lagrange**, con el objetivo de encontrar el mejor \( w \) y \( b \) que maximicen el margen y cumplan las restricciones de clasificacin.

### 3. Formalismo Computacional

Desde una perspectiva computacional, la SVM resuelve la optimizacin cuadrtica utilizando algoritmos de programacin convexa, los cuales son eficientes para encontrar la solucin ptima del hiperplano. En la prctica:

1. **Vector de Soporte**: Durante el entrenamiento, solo algunos puntos son crticos para definir el margen mximo; estos puntos se llaman **vectores de soporte** y determinan la posicin del hiperplano.
  
2. **Funcin de Kernel**: En problemas donde los datos no son separables linealmente, las SVM emplean funciones de kernel, como el **kernel radial (RBF)** o el **kernel polinomial**, que transforman los datos a un espacio de mayor dimensin donde s sean linealmente separables.

3. **Solucin Computacional**: Los algoritmos SVM utilizan mtodos como el **algoritmo SMO (Sequential Minimal Optimization)** para resolver problemas de gran tamao al dividir la optimizacin en subproblemas ms pequeos, permitiendo su ejecucin en paralelo para mejorar la eficiencia.

### 4. Funcin de Kernel: Expansin a Problemas No Lineales

Para problemas no lineales, las SVM utilizan la **tcnica del kernel**, que proyecta los datos a un espacio de mayor dimensin donde s pueden separarse linealmente. La funcin de kernel $$ K(x_i, x_j) $$ calcula el producto interno en el espacio proyectado sin necesidad de realizar la transformacin explcita, lo que reduce el costo computacional. Los kernels ms comunes incluyen:

- **Kernel Lineal**: $$ K(x_i, x_j) = x_i \cdot x_j $$
- **Kernel Polinomial**: $$ K(x_i, x_j) = (x_i \cdot x_j + 1)^d $$
- **Kernel RBF (Radial Basis Function)**: $$ K(x_i, x_j) = \exp(-\gamma \| x_i - x_j \|^2) $$

### 5. Ventajas y Desventajas de las SVM

**Ventajas**:
- **Eficiencia**: Son efectivos en espacios de alta dimensionalidad y se adaptan bien a problemas linealmente separables.
- **Robustez**: Gracias al uso de vectores de soporte y el margen mximo, las SVM son robustas ante el sobreajuste, especialmente cuando el nmero de caractersticas supera el de muestras.
- **Flexibilidad con Kernels**: La tcnica de kernel permite que las SVM se apliquen a problemas no lineales de forma eficaz.

**Desventajas**:
- **Costo Computacional**: Entrenar una SVM puede ser lento en conjuntos de datos grandes, debido a la necesidad de resolver un problema de optimizacin cuadrtica.
- **Sensibilidad a Parmetros**: El desempeo de las SVM depende en gran medida de la eleccin del kernel y de parmetros como \( C \) y \( \gamma \), que deben ajustarse cuidadosamente.

### 6. Implementacin en Python

A continuacin, mostramos cmo implementar una SVM con `scikit-learn` en Python para un problema de clasificacin binaria.


```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Cargar el conjunto de datos
data = datasets.load_iris()
X = data.data[data.target != 2]
y = data.target[data.target != 2]

# Dividir en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrenar la SVM con un kernel lineal
model = SVC(kernel='linear', C=1)
model.fit(X_train, y_train)

# Realizar predicciones y evaluar
y_pred = model.predict(X_test)
print(f"Precisin del modelo SVM: {accuracy_score(y_test, y_pred):.4f}")
```

# Clasificacin con Maquinas de soporte Vectorial

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import pandas as pd
from sklearn import preprocessing
```

```python
#Apertura de archivo csv a un dataframe de pandas

#revisar la ruta para tu caso particular
ruta = "/content/Social_Network_Ads.csv"
dataset = pd.read_csv(ruta)
dataset.head(5)
```

```python
#Sacar las variables independientes
X = dataset.iloc[:,[2,3]].values

# Sacar la variable dependientes
y = dataset.iloc[:,4].values
```

```python
#Division de Datos - entrenamiento y validacion

#herramienta para dividir los datos
from sklearn.model_selection import train_test_split

#divide los datos en 20% para la validacion y se colocar una semilla para hacer la division
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# Escalado de Variables - Estandarizacion y Normalizacion

#Es preferible utilizalo en algortimos que usan el calculo de distancia - pitagoras

#Estandarizacion
sc_X = preprocessing.StandardScaler()

X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

## Kernel Lineal

```python
from sklearn.svm import SVC
#Ajuste del clasificdor con Maquinas de soporte vectorial

clasificador = SVC(kernel="linear", random_state=0)
clasificador.fit(X_train, y_train)
```

```python
#Prediccion de los resultados con el conjusto de testing

y_pre = clasificador.predict(X_test)
#respuesta del modelo
print(y_pre)
#Valores dados por los datos de testing
print(y_test)
```

```python
#Comprobacion del resultado - Matriz de confucion
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pre)
cm
```

```python
# Representacin grfica de los resultados del algoritmo en el Conjunto de Entrenamiento

X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, clasificador.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Clasificador (Conjunto de Entrenamiento)')
plt.xlabel('Edad')
plt.ylabel('Sueldo Estimado')
plt.legend()
plt.show()
```

```python
# Representacin grfica de los resultados del algoritmo en el Conjunto de Testing
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, clasificador.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Clasificador (Conjunto de Test)')
plt.xlabel('Edad')
plt.ylabel('Sueldo Estimado')
plt.legend()
plt.show()
```

# SVM

Las mquinas de vectores soporte (SVM) son un conjunto de mtodos de aprendizaje supervisado utilizados para __clasificacin__, __regresin__ y deteccin de valores atpicos.

Empecemos con un ejemplo de clasificacin y veamos los argumentos del clasificador [SVM Classifier Doc](https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html)

- **C:** Parmetro de regularizacin (fijado por defecto en 1.0). La fuerza de la regularizacin es inversamente proporcional a C. Debe ser estrictamente positivo.

- **kernel:** Especifica el tipo de kernel que se utilizar en el algoritmo. `linear` usar un kernel lineal simple, `rbf` (por defecto) aplicar el truco de kernel visto en las diapositivas de teora. Otros kernels no cubiertos en la teora tambin estn disponibles y puedes crear tu propio kernel y usarlo aqu tambin.

- gamma:** Valor del hiperparmetro usado slo con los ncleos `rbf`, `poly` y `sigmoid`. Se establece por defecto a `scale` que utiliza la siguiente frmula 1 / (n_caractersticas * X.var()) como valor de gamma. si `gamma=auto`, utiliza 1 / n_caractersticas. Tambin puede pasar un valor gamma personalizado como float, `gamma=0.1` tambin se considera un buen valor por defecto.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.axes._axes import _log as matplotlib_axes_logger
matplotlib_axes_logger.setLevel('ERROR')
```

```python
!gdown "1r7ZqEH9kqsoNGW3hFumWetLUoTMMbLiq"
dataset = pd.read_csv('data_SVM.csv')
dataset.head()
```

```python
dataset["Purchased"].value_counts()
```

Utilizaremos como caractersticas `Age` y `EstimatedSalary`. Nuestro objetivo de prediccin ser Comprado.


```python
X = dataset.iloc[:, [2, 3]].values
Y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=0)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=Y)
```

```python
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
from sklearn.svm import SVC

classifier = SVC(C=1, kernel='linear', random_state=1)
classifier.fit(X_train, y_train)
```

```python
# Ahora es el momento de utilizar nuestro modelo entrenado en muestras no vistas
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
# Usando esta frmula podemos medir el % de muestras correctamente clasificadas
# Veremos ms adelante mejores formas de medir el rendimiento de nuestro clasificador
accuracy = sum(y_pred == y_test) / len(y_pred)
print(f"Accuracy: {100 * accuracy:.2f}%")
```

```python
# Comprobemos la precisin en el conjunto de entrenamiento
y_pred_train = classifier.predict(X_train)
accuracy = sum(y_pred_train == y_train) / len(y_pred_train)
print(f"Accuracy: {100 * accuracy:.2f}%")
```

**Consejo avanzado:** Puede utilizar el siguiente cdigo para analizar ms a fondo cmo nuestro modelo SVM con Kernel Lineal est clasificando sus datos. Este fragmento mostrar los lmites de decisin del modelo.


```python
from matplotlib.colors import ListedColormap

X_Set, Y_Set = X_train, y_train
X1, X2 = np.meshgrid(
    np.arange(start=X_Set[:, 0].min()-1, stop=X_Set[:, 0].max()+1, step=0.01),
    np.arange(start=X_Set[:, 1].min()-1, stop=X_Set[:, 1].max()+1, step=0.01)
)

plt.figure(figsize=(10,5))
plt.contourf(
    X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
    alpha=0.75, cmap=ListedColormap(('red', 'green'))
)
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(Y_Set)):
    plt.scatter(
        X_Set[Y_Set == j, 0], X_Set[Y_Set == j, 1],
        c=ListedColormap(('red', 'green'))(i), label=j
    )
plt.title('Support Vector Machine (Train)', fontsize=20)
plt.xlabel('Age', fontsize=15)
plt.ylabel('Estimated Salary', fontsize=15)
plt.legend()
plt.show()
```
