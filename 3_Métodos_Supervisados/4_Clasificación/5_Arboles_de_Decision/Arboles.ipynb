## rboles de Decisin

Un rbol de decisin es un modelo predictivo que se utiliza tanto para clasificacin como para regresin. A continuacin, explicamos su funcionamiento de manera terica y computacional:

### 1. Concepto Clave

El modelo de rboles de decisin divide el espacio de caractersticas en regiones ms pequeas y homogneas mediante "divisiones" o "splits". Cada divisin se realiza seleccionando la caracterstica que mejor separa los datos de acuerdo a una medida como **gini** o **entropa** (para clasificacin) o **error cuadrtico** (para regresin).

#### Frmulas de Clasificacin:

- ndice de Gini: $$  Gini(t) = 1 - \sum_{i=1}^{n} p_i^2  $$
  
- Entropa: $$ Entropa(t) = - \sum_{i=1}^{n} p_i \log(p_i) $$
Donde \( p_i \) es la proporcin de instancias de la clase \( i \) en el nodo \( t \).

### 2. Algoritmo del rbol de Decisin

1. **Inicio del rbol**: Todos los datos estn en la raz.
2. **Bsqueda de la Mejor Divisin**: Para cada nodo, el algoritmo evala todas las caractersticas y selecciona la divisin que mejor separa los datos de acuerdo a la medida elegida (gini, entropa, etc.).
3. **Divisin**: Divide el nodo en dos hijos, repitiendo el proceso hasta que no se puedan realizar ms divisiones (criterio de parada: nmero mnimo de instancias o profundidad mxima).
4. **Prediccin**: Para hacer una prediccin, los datos nuevos siguen el camino del rbol desde la raz hasta una hoja.

## Crea tu modelo de rbol

Si echas un vistazo a los parmetros que puede tomar el [DecisionTreeClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html), puede que te sorprendas, as que, veamos algunos de ellos.

- criterio: Este parmetro determina cmo se medir la impureza de una divisin. El valor por defecto es gini pero tambin se puede utilizar entropa como mtrica para la impureza.

- Separador: Esta es la forma en que el rbol de decisin busca las caractersticas para una separacin. El valor por defecto es mejor. Es decir, para cada nodo, el algoritmo considera todas las caractersticas y elige la mejor divisin. Si decide establecer el parmetro de divisin en aleatorio, entonces se considerar un subconjunto aleatorio de caractersticas. La divisin se realizar por la mejor caracterstica dentro del subconjunto aleatorio. El tamao del subconjunto aleatorio viene determinado por el parmetro max_features. De aqu proviene en parte el nombre de Random Forest.

- Profundidad_mx: Determina la profundidad mxima del rbol. El valor por defecto es ninguno. Esto suele dar lugar a rboles de decisin sobreajustados. El parmetro de profundidad es una de las formas en que podemos regularizar el rbol, o limitar la forma en que crece para evitar el sobreajuste.

- min_samples_split: El nmero mnimo de muestras que debe contener un nodo para considerar la divisin. El valor por defecto es dos. Puede utilizar este parmetro para regularizar su rbol.

- min_samples_leaf: El nmero mnimo de muestras necesarias para ser considerado un nodo hoja. El valor por defecto es uno. Utilice este parmetro para limitar el crecimiento del rbol.

- caractersticas_max: El nmero de caractersticas a considerar cuando se busca la mejor divisin. Si no se establece este valor, el rbol de decisin considerar todas las caractersticas disponibles para realizar la mejor divisin. Dependiendo de su aplicacin, a menudo es una buena idea ajustar este parmetro. [Aqu hay un artculo](https://stats.stackexchange.com/questions/324370/references-on-number-of-features-to-use-in-random-forest-regression) que recomienda cmo establecer max_features.



```python
# Creamos el modelo de rbol de decisin
tree_model = DecisionTreeClassifier(criterion='gini', max_depth=3, min_samples_split=5, min_samples_leaf=6, random_state=10)
# Entrenamos el modelo con los datos
tree_model.fit(X_train, y_train);
```

### Interpretacin de la Matriz de Confusin

Esta es una matriz de confusin para un problema de clasificacin binaria. La matriz representa los resultados de un modelo de clasificacin y se interpreta de la siguiente manera:

|            | Prediccin Positiva | Prediccin Negativa |
|------------|---------------------|---------------------|
| **Clase Positiva** | 62 (Verdaderos Positivos - VP) | 3 (Falsos Negativos - FN) |
| **Clase Negativa** | 6 (Falsos Positivos - FP) | 29 (Verdaderos Negativos - VN) |

#### Descripcin:

- **62**: Verdaderos positivos (VP) - El modelo predijo correctamente 62 casos de la clase positiva.
- **6**: Falsos positivos (FP) - El modelo predijo que 6 casos eran de la clase positiva, pero en realidad eran de la clase negativa.
- **3**: Falsos negativos (FN) - El modelo predijo que 3 casos eran de la clase negativa, pero en realidad pertenecan a la clase positiva.
- **29**: Verdaderos negativos (VN) - El modelo predijo correctamente 29 casos de la clase negativa.

Con esta matriz, se pueden calcular mtricas importantes para evaluar el rendimiento del modelo, como la **precisin**, **sensibilidad** (recall) y **especificidad**.

#### Ejemplo de clculos:

- **Precisin**: Mide qu tan precisas son las predicciones positivas:
  \[
  \text{Precisin} = \frac{VP}{VP + FP} = \frac{62}{62 + 6} = 0.91
  \]

- **Sensibilidad** (o recall): Mide cuntos de los verdaderos positivos fueron correctamente clasificados:
  \[
  \text{Sensibilidad} = \frac{VP}{VP + FN} = \frac{62}{62 + 3} = 0.95
  \]

- **Especificidad**: Mide cuntos de los verdaderos negativos fueron correctamente clasificados:
  \[
  \text{Especificidad} = \frac{VN}{VN + FP} = \frac{29}{29 + 6} = 0.83
  \]

## Entropa y Ganancia de Informacin

### Entropa

Definamos entropa para una distribucin probabilista:

$$H(Y) = - \sum_{i=1}^k P(Y = y_i) log_2 P(Y = y_i)$$

```python
def entropy(probs):
    return - np.sum(probs * np.log2(probs))
```

Veamos posibles entropas para el problema de tirar una moneda adulterada:

```python
entropy(np.array([0.5, 0.5]))
```

```python
entropy(np.array([0.01, 0.99]))
```

```python
X = np.linspace(0, 1)[1:-1]
plt.plot(X, [entropy([x, 1-x]) for x in X])
plt.xlabel('P(Y=y_1)')
plt.ylabel('entropy')
plt.show()
```

### Entropa de un Dataset

Un dataset define una distribucin emprica. La entropa del dataset es entones la entropa de la distribucin asociada. Definamos el clculo de la distribucin, y luego redefinamos entropa:

```python
def probs(y):
    _, counts = np.unique(y, return_counts=True)
    probs = counts / counts.sum()
    return probs
```

Definimos un dataset de 6 elementos con dos atributos $X_1$ y $X_2$ con dos tipos de etiquetas:

$False: -1$ y $True: 1$.

Siendo las etiquetas resultantes del mismo:


```python
y = np.array([1, 1, 1, 1, 1, -1])
```

Mediante la funcin "probs" podemos calcular la probilidad de cada clase en este dataset:

```python
probs(y)
```

Esos resultados nos informan que hay un desbalance en el mismo (hay ms probabilidad de una de las clases).

Ahora obtengamos la entropa del dataset:

```python
def entropy(y):
    p = probs(y)
    return - np.sum(p * np.log2(p))
```

```python
entropy(y)
```

### Entropa Condicional

Definamos entropa condicional:

$$H(Y|X) = - \sum_{j=1}^v P(X = x_j) \sum_{i=1}^k P(Y = y_i | X = x_j) log_2 P(Y = y_i | X = x_j)$$

Equivalentemente,

$$H(Y|X) = \sum_{j=1}^v P(X = x_j) H(Y|X = x_j)$$

Tomaremos $X$ binaria ($v=2$), por lo que la entropa condicional tendr slo dos trminos.


```python
def cond_entropy(y1, y2):
    size = y1.shape[0] + y2.shape[0]
    return y1.shape[0] / size * entropy(y1) + y2.shape[0] / size * entropy(y2)
```

Esta funcin toma como argumento los dos subconjuntos de datos que se forman al considerar una de las variables como condicin.

Si analizmos la varibale $Y$ donde la $X_1$ es $True$, tenemos:

```python
y_X1true = np.array([1,1,1,1]) #X1 = True
```

de igual manera, la varibale $Y$  donde la $X_1$ es $False$, tenemos:

```python
y_X1false =  np.array([1,-1]) #X1 = False
```

Haciendo uso de la funcin "cond_entropy" podemos obtener la entropa condicional de la variable $Y$ dado $X_1$

```python
cond_entropy(y_X1true,y_X1false)  # x1
```

De igual manera podemos calcular la entropa condicional dado la variable $X_2$

```python
y_X2true = np.array([1,1,1]) #X2 = True
y_X2false = np.array([1,1,-1]) #X2 = False
cond_entropy(y_X2true, y_X2false)  # x2
```

### Ganancia de Informacin

La ganancia de informacin ser simplemente la diferencia entre entropa y entropa condicional:

```python
def information_gain(y1, y2):
    y = np.concatenate((y1,y2))
    return entropy(y) - cond_entropy(y1,y2)
```

Podemos calcular la ganancia de informacin que implicara dividir el dataset tomando como $X_1$ como nodo referencia. Para ello hacemos uso de la funcin "information_gain" que toma los dos subconjuntos de etiquetas que se formaran al tomar una u otra variable como referencia.

Para el caso de $X_1$:

```python
information_gain(y_X1true, y_X1false)  # x1
```

Para el caso de $X_2$:


```python
information_gain(y_X2true, y_X2false)  # x2
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_decision_boundary(pred_func, X, y):
    # Establecer los lmites del grfico
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    h = 0.01  # Paso en la malla
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    # Predecir las clases para cada punto en la malla
    Z = pred_func(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    # Trazar la superficie de decisin
    plt.contourf(xx, yy, Z, cmap=plt.cm.Spectral, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())

# Usar la funcin para visualizar la frontera de decisin
plt.xlabel("X[0]")
plt.ylabel("X[1]")
plot_decision_boundary(lambda x: model.predict(x), X, y_true)
```

Para hacer una divisin sobre una variable real usaremos un valor "threshold" (umbral):

```python
def split(X, y, i, threshold):
    y1 = y[X[:, i] > threshold]
    y2 = y[X[:, i] <= threshold]
    return y1, y2
```

definimos la "split" que toma como argumento el dataset con sus etiquetas, el indice de la variable que usaremos para dividir el dataset y el umbral para realizarlo y nos devuelve dos subconjuntos de este dataset.

Si utilizamos la variable $X_1$ con un umbral de $0.00$ obtenemos:

```python
y1, y2 = split(X_train, y_train, 0, 0.00)
```

```python
y1,y2
```

### Buscando la Mejor Divisin

Ilustraremos un paso en la construccin del rbol de decisin.

Probemos muchos threshold para ambas variables y seleccionemos la mejor divisin.

En este caso buscaremos en una grilla uniforme de valores, pero existen tcnicas mejores.

```python
np.linspace(-2.5, 2.5, 11)
```

```python
best_ig = 0

for i in [0, 1]:
    for threshold in np.linspace(-2.5, 2.5, 11):
        y1, y2 = split(X_train, y_train, i, threshold)
        ig = information_gain(y1, y2)
        print(f'i={i}\tthreshold={threshold:+00.2f}\tig={ig:.2f}')

        if ig >= best_ig:
            best_ig = ig
            best_feature = i
            best_threshold = threshold

print('Mejor divisin:')
print(f'feature={best_feature}, threshold={best_threshold}, ig={best_ig:00.2f}')
```

Dividamos los datos de acuerdo a esta frontera:

```python
best_feature, best_threshold
```

Podemos ver los datos en el grfico

```python
plt.scatter(X_train[y_train==1, 0], X_train[y_train==1, 1], color="royalblue", label="1")
plt.scatter(X_train[y_train==-1, 0], X_train[y_train==-1, 1], color="red", label="-1")

plt.axhline(y=0, xmin=-3, xmax=3, color='green', linestyle='-.', linewidth=2)

plt.xlabel("X[0]")
plt.ylabel("X[1]")
plt.grid()
plt.legend()
plt.show()
```

```python
y1, y2 = split(X_train, y_train, best_feature, best_threshold)
```

Con esta divisin, la entropa baja considerablemente:

```python
entropy(y_train)
```

```python
cond_entropy(y1, y2)
```
