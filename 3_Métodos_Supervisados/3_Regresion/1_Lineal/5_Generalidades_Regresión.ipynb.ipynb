**<h2 style="color: #007ACC;">Introduccin</h2>**

La **regresin** es una tarea fundamental en el aprendizaje automtico que busca modelar y predecir valores continuos mediante la relacin entre variables. Es ampliamente utilizada en reas como prediccin de precios, anlisis de riesgos y estimaciones cientficas. Este curso explora una seleccin de modelos de regresin esenciales, conocidos por su eficacia y versatilidad:

1. **Regresin Lineal Mltiple**: Base de la regresin, ideal para relaciones lineales entre mltiples variables.
2. **Regresin Polinomial**: Extiende la linealidad al capturar relaciones no lineales.
3. **Regresin LASSO**: Modelo regularizado que favorece la simplicidad al eliminar variables menos relevantes.
4. **Regresin Ridge**: Maneja la multicolinealidad al reducir el impacto de variables altamente correlacionadas.
5. **Regresin con rboles de Decisin**: Introduccin a modelos no lineales basados en reglas jerrquicas.
6. **Regresin con Bosques Aleatorios**: Combina mltiples rboles para mejorar precisin y robustez.
7. **Regresin con SVM**: Utiliza hiperplanos para capturar relaciones complejas en espacios transformados.

A travs de ejemplos prcticos, entenderemos por qu estos modelos son herramientas indispensables en problemas de regresin en el machine learning.

**<h2 id="section1" style="color: #007ACC;">1. Regresin Lineal Mltiple</h2>**

Si bien el caso de regresin lineal simple es interesante y permite explorar la relacin entre dos variables, es un escenario muy idealista y no permite abordar los casos ms reales. De esta manera, extendemos la definicin a mltiples variables y definimos el modelo de regresin lineal mltiple como:

$$y_{i} = \beta_{0} + \beta_{1} x_{1} + \beta_{2} x_{2} + \cdots + \beta_{k} x_{k} + \epsilon_{i}$$

Considerando una muestra de $N$ individuos, el modelo se puede expresar matricialmente como

$$ \mathbf{Y} = \mathbf{X}\mathbf{\beta} + \mathbf{e} =  \begin{pmatrix}
1 & x_{11} &  \cdots& x_{1k} \\
1 & x_{21} & \cdots & x_{2k} \\
 \vdots & \vdots & \ddots & \vdots \\
1 & x_{n1} & \cdots & x_{nk} \\
\end{pmatrix} \begin{pmatrix}
 \beta_0 \\
 \beta_1 \\
 \vdots \\
 \beta_k
\end{pmatrix} + \begin{pmatrix}
 \epsilon_1 \\
 \epsilon_2 \\
 \vdots \\
 \epsilon_N
\end{pmatrix} $$

Y la funcin de prdida de **Mnimos Cuadrados Ordinarios** es:

$$ \mathbf{e} ^{T} \mathbf{e}= (\mathbf{Y} - \mathbf{X}\mathbf{\beta})^{T} (\mathbf{Y} - \mathbf{X}\mathbf{\beta})$$

Y la solucin del problema de minimizacin es:

$$\mathbf{\beta}^{*} = \left(\mathbf{X}^T \mathbf{X} \right)^{-1} \mathbf{X}^{T} \mathbf{Y}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression

# Generating random data
np.random.seed(0)
n_samples = 100
X1 = 2 * np.random.rand(n_samples, 1)
X2 = 3 * np.random.rand(n_samples, 1)
y = 4 + 3 * X1 + 2 * X2 + np.random.randn(n_samples, 1)

# Fit linear regression model
reg = LinearRegression()
X = np.hstack((X1, X2))
reg.fit(X, y)
coef = reg.coef_
intercept = reg.intercept_

# Create meshgrid for plotting
x1_range = np.linspace(0, 2, 10)
x2_range = np.linspace(0, 3, 10)
x1_mesh, x2_mesh = np.meshgrid(x1_range, x2_range)
y_pred_mesh = intercept + coef[0][0] * x1_mesh + coef[0][1] * x2_mesh

# Plotting
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X1, X2, y, color='red', label='Puntos de datos')
ax.plot_surface(x1_mesh, x2_mesh, y_pred_mesh, alpha=0.5, rstride=100, cstride=100, color='blue')

ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('Y')

plt.title('Regresin Lineal con dos regresores')
plt.legend()
plt.show()
```

**<h2 id="section2" style="color: #007ACC;">2. Regresin polinmica y logartmica</h2>**




Una vez comprendido el modelo de regresin lineal, vale la pena pensar en si la misma teora es til para explicar relaciones no lineales entre variables, y la respuesta es "s". En particular, considere los siguientes modelos de regresin:

$$y_{i} = \beta_{0} + \beta_{1} x_{i1} + \beta_{2} x_{i2}^{2} + \beta_{3} x_{i3}^{3} + \epsilon_{i}$$
$$y_{i} = \beta_{0} + \beta_{1} ln(x_{i}) + z_{i} + \epsilon$$
$$ln(y_{i}) = \beta_{0} + ln(x_{i}) + \epsilon$$

Aunque, pueda ser sorprendente (y, contraintuitivo al comienzo), **TODOS ESTOS EJEMPLOS SON CASOS PARTICULARES DEL MODELO DE REGRESIN LINEAL**. Para entender porqu, utilicemos el primer ejemplo, y definamos un modelo de regresin lineal con tres regresores:

$$y_{i} = \beta_{0} + \beta_{1} x_{i} + \beta_2 z_{i} + \beta_{3} w_{i} + \epsilon_i$$

Claramente, esta ecuacin es lineal en $x$, $z$ y $w$, as que basta con hacer $z = x^2$ y $w = x^3$ para obtener el modelo de regresin de nuestro ejemplo. Y, el mismo ejercicio puede hacerse aplicando cualquier otro tipo de funcin (P.ej. Logartmica, exponencial, radical).

**Interpretacin de los parmetros del modelo logartmico**

| Nombre del modelo | Ecuacin | Escala de y | Escala de x | Interpretacin de $\beta$|
| --- | --- | --- | --- | --- |
| $$\text{lin - lin}$$ | $$y = \alpha + \beta x$$ | $$y$$ | $$x$$ | Si $x$ cambia en una unidad, $y$ cambia en $\beta$ unidades|
| $$\text{log - lin}$$ | $$ln(y) = \alpha + \beta x$$ | $$ln(y)$$ | $$x$$ |Si $x$ cambia en una unidad, $y$ cambia en $\beta * 100$ % |
| $$\text{lin - log}$$ | $$y = \alpha + \beta ln(x)$$ | $$y$$ | $$ln(x)$$ | Si $x$ cambia en un $1 \%$, $y$ cambia en $\beta /100$ unidades |
| $$\text{log - log}$$ | $$ln(y) = \alpha + \beta ln(x)$$ | $$ln(y)$$ | $$ln(x)$$ | Si $x$ cambia en un $1 \%$, $y$ cambia en un $\beta \%$|

**<h2 id="section3" style="color: #007ACC;">3. Regresin LASSO</h2>**

La regresin LASSO es un caso de regularizacin del modelo de regresin lineal, tpicamente conocido tambin como regularizacin L1. En particular, corresponde al mismo modelo que ya vimos previamente, pero definimos la funcin de prdida como:

$$h(\mathbf{\beta}) = \mathbf{e}^T \mathbf{e} + \alpha \sum_{i=1}^{N} \vert \beta_{i} \vert$$

Esta funcin de prdida corresponde a encontrar:

$$\mathbf{\beta}^{*} = \arg \min \mathbf{e}^{T} \mathbf{e} \hspace{0.1in} \text{ s.a.} \sum_{i = 1}^N \vert \beta_{i} \vert < \alpha$$

Una caracterstica importante de este modelo es que hemos introducido el parmetro $\alpha$, que indica qu tanto queremos restringir el problema de optimizacin. Al hacer esto, forzamos al algoritmo de Mnimos Cuadrados Ordinarios a encontrar valores pequeos en varios de los parmetros a estimar, y en algunos casos los hace ser $0$. En muchos casos, esta tcnica mejora los resultados de prediccin del modelo de regresin lineal y evita el sobreentrenamiento.

En aplicaciones explicativas (no predictivas), la regresin LASSO ayuda a determinar qu regresores son ms relevantes para explicar la variable dependiente, $y$.

```python
np.random.seed(42)
m = 100
x = 6 * np.random.rand(m, 1) - 3
y = 0.5 * x ** 2 + x + 2 + 3*np.random.randn(m, 1)
```

```python
from sklearn.metrics import mean_squared_error


np.random.seed(42)
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X ** 2 + X + 2 + np.random.randn(m, 1)

# Polynomial features
poly_features = PolynomialFeatures(degree=10, include_bias=False)
X_poly = poly_features.fit_transform(X)

# Linear regression
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)

# Plot original data
plt.plot(X, y, 'o', label='Datos originales', markersize=5)

mse_dict = {}

# Lasso regression with different alpha values
alpha_values = [0.1, 1, 10]
for alpha, style in zip(alpha_values, ('x', 'b.', '+', '>')):
    lasso_reg = Lasso(alpha=alpha)
    lasso_reg.fit(X_poly, y)
    y_pred = lasso_reg.predict(X_poly)
    mse = mean_squared_error(y, y_pred)

    model_name = f'Regresin LASSO (alpha: {alpha})'
    mse_dict[model_name] = mse

    # Plot Lasso regression lines
    plt.plot(X, y_pred, style, label=f'Lasso Regression (alpha={alpha})', markersize=3)

plt.plot(X, lin_reg.predict(X_poly), 'o',label='Linear Regression', color='r', markersize=3)
plt.legend()
plt.show()

mse = mean_squared_error(y, lin_reg.predict(X_poly))

mse_dict['Regresin clsica'] = mse
```

```python
import pandas as pd
mse_df = pd.DataFrame.from_dict(mse_dict, orient='index')
mse_df.columns = ['MSE']
mse_df
```

**<h2 id="section4" style="color: #007ACC;">4. Regresin Ridge</h2>**


La regresin Ridge es otro caso de regularizacin del modelo de regresin lineal, tpicamente conocido tambin como regularizacin L2. En este caso, la funcin de prdida es

$$h(\beta) = \mathbf{e}^{T} \mathbf{e} + \alpha \sum_{i = 1}^{N} \beta_{i} ^{2}$$

Esta funcin de prdida corresponde a encontrar:

$$\arg \min \mathbf{e}^{T} \mathbf{e} \hspace{0.1in} \text{ s.a.} \sum_{i = 1}^N  \beta_{i}^{2} < \alpha$$

Para este caso, el mtodo de Mnimos Cuadrados Ordinarios genera el estimador

$$\beta^{*} = \left( \mathbf{X}^{T} \mathbf{X} + \alpha I \right)^{-1} \mathbf{X}^{T} \mathbf{Y}$$

dnde $I$ corresponde a la matriz identidad, por lo que

$$\alpha I =  \begin{pmatrix}
\alpha & 0 & \cdots & 0 \\
 0 &  \alpha & \cdots  & 0 \\
 \vdots & \vdots & \ddots & 0 \\
  0& 0 & \cdots & \alpha \\
\end{pmatrix}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error


np.random.seed(42)
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X ** 2 + X + 2 + np.random.randn(m, 1)

# Polynomial features
poly_features = PolynomialFeatures(degree=10, include_bias=False)
X_poly = poly_features.fit_transform(X)

# Linear regression
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)

# Plot original data
plt.plot(X, y, 'o', label='Original Data', markersize=5)

mse_dict = {}

# Lasso regression with different alpha values
alpha_values = [0.1, 1, 10]
for alpha, style in zip(alpha_values, ('x', 'b.', '+', '>')):
    ridge_reg = Ridge(alpha=alpha)
    ridge_reg.fit(X_poly, y)
    y_pred = ridge_reg.predict(X_poly)
    mse = mean_squared_error(y, y_pred)

    model_name = f'Regresin Ridge (alpha: {alpha})'
    mse_dict[model_name] = mse

    # Plot Lasso regression lines
    plt.plot(X, y_pred, style, label=f'Regresin Ridge (alpha={alpha})', markersize=3)

plt.plot(X, lin_reg.predict(X_poly), 'o',label='Regresin Lineal', color='r', markersize=3)
plt.legend()
plt.show()

mse = mean_squared_error(y, lin_reg.predict(X_poly))

mse_dict['Regresin clsica'] = mse
```

```python
import pandas as pd
mse_df = pd.DataFrame.from_dict(mse_dict, orient='index')
mse_df.columns = ['MSE']
mse_df
```

**<h2 id="section7" style="color: #007ACC;">7. Regresin con SVM - Introduccin</h2>**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn import preprocessing
```

```python
# apertura de archivo
ruta = "Position_Salaries.csv"

dataset = pd.read_csv(ruta)
dataset.head(5)
```

```python
#se asignan las variable dependiente e independiente
X = dataset.iloc[:, 1:2].values.reshape(-1,1) # Matriz de caracteristicas, debe ser una matriz
y = dataset.iloc[:,2].values.reshape(-1,1)
```

```python
# Escalado de Variables - Estandarizacion y Normalizacion

sc_X = preprocessing.StandardScaler()
X = sc_X.fit_transform(X)
sc_y = preprocessing.StandardScaler()
y = sc_y.fit_transform(y)

print("X",X,"\n", "y",y )
```

```python
#Ajustar la regresion con una maquina de soporte vectorial-kernel Gaussiano

#Para este modelo de deben estandarizar los datos a utilizar
regresion = SVR(kernel="rbf")
regresion.fit(X,y)
```

```python
#visualizacion de los datos del modelo SVM kernel gaussiano
plt.scatter(X,y, color="red")
plt.plot(X, regresion.predict(X), color="blue")

plt.title("Modelo de Regresion SVM-Gaussiano")
plt.ylabel("Sueldo en $")
plt.xlabel("Posicion del Empleado")
plt.show()

#No es una aproximacion buena, pero mejor que el kernel polinomial
```

```python
#Prediccion del modelo, para escala salarial de 6.5
arreglo = np.array([[6.5]])
y_pre = regresion.predict(sc_X.transform(arreglo))
y_pre = sc_y.inverse_transform(y_pre.reshape(-1,1))
y_pre
```

```python
#Ajustar la regresion con una maquina de soporte vectorial-kernel Gaussiano

#Para este modelo de deben estandarizar los datos a utilizar
regresion = SVR(kernel="poly")
regresion.fit(X,y)
```

```python
#visualizacion de los datos del modelo SVM kernel gaussiano
plt.scatter(X,y, color="red")
plt.plot(X, regresion.predict(X), color="blue")

plt.title("Modelo de Regresion SVM-Gaussiano")
plt.xlabel("Sueldo en $")
plt.ylabel("Posicion del Empleado")
plt.show()

#Es mejor la aproximacion con el kernel Gaussiano
```
