Supongamos que tenemos el siguiente sistema de ecuaciones lineales el cul queremos resolver.

\begin{align}
2x + 3y &= 1 \\\
5x - y &= 3
\end{align}

````python
A = np.array([[2, 3], [5, -1]])
print(A)
````

````python
b = np.array([1, 3]).reshape(2, 1)
print(b)
````

````python
x = la.solve(A, b)
print(x)
````

````python
A @ Ainv
````

Y multiplicar $A^{-1} \mathbf{b}$ para resolver $\mathbf{x}$:

````python
x = Ainv @ b
print(x)
````

Vamos a trazar algunos vectores!

````python
def plot_vectors(vectors, colors):
    plt.figure()
    plt.axvline(x=0, color='lightgray')
    plt.axhline(y=0, color='lightgray')

    for i in range(len(vectors)):
        x = np.concatenate([[0,0],vectors[i]])
        plt.quiver([x[0]], [x[1]], [x[2]], [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=colors[i],)
````

````python
# Definamos un vector
x = np.array([3, 1])

# Funciones
plot_vectors([x], ['lightblue'])
plt.xlim(-1, 5)
_ = plt.ylim(-1, 3)
````

**<h2 id="section2" style="color: #007ACC;">2. Derivadas</h2>**


Definimos una funcin llamada `numerical_derivative` que calcula la derivada numrica de una funcin en un punto especfico utilizando tres posibles mtodos: **centrado**, **hacia adelante** o **hacia atrs**. La derivada numrica aproxima el valor de la derivada real mediante la diferencia de los valores de la funcin en puntos cercanos.

### **Parmetros:**
- `function`: es la funcin para la cual se quiere calcular la derivada.
- `point`: es el punto en el cual se quiere evaluar la derivada.
- `method`: es el mtodo numrico que se utilizar para calcular la derivada. Los valores posibles son:
  - `'central'`: usa la frmula de diferencias centradas.
  - `'forward'`: usa la frmula de diferencias hacia adelante.
  - `'backward'`: usa la frmula de diferencias hacia atrs.
- `h`: es el tamao del paso, que representa una pequea variacin en el valor de `point` (por defecto es 0.01).

### **Mtodos de clculo:**
1. **Mtodo centrado** (`'central'`): calcula la derivada aproximada utilizando valores a la izquierda y derecha del punto. La frmula es:

   $$
   f'(x) \approx \frac{f(x + h) - f(x - h)}{2h}
   $$

2. **Mtodo hacia adelante** (`'forward'`): calcula la derivada aproximada usando el valor del punto y un punto hacia adelante (ms `h`). La frmula es:

   $$
   f'(x) \approx \frac{f(x + h) - f(x)}{h}
   $$

3. **Mtodo hacia atrs** (`'backward'`): calcula la derivada aproximada usando el valor del punto y un punto hacia atrs (menos `h`). La frmula es:

   $$
   f'(x) \approx \frac{f(x) - f(x - h)}{h}
   $$

Si se proporciona un valor de mtodo que no sea `'central'`, `'forward'` o `'backward'`, el cdigo lanza un error (`ValueError`).

````python
def numerical_derivative(function, point, method='central', h=0.01):
    if method == 'central':
        return (function(point + h) - function(point - h)) / (2 * h)
    elif method == 'forward':
        
        return (function(point + h) - function(point)) / h
    elif method == 'backward':
        return (function(point) - function(point - h)) / h
    else:
        raise ValueError("Method must be 'central', 'forward' or 'backward'.")
````

1. **`x = np.linspace(0.1, 5*np.pi, 100)`**:

2. **`numerical = numerical_derivative(np.log, x)`**:

**<h2 id="section3" style="color: #007ACC;">3. Clculo de Gradientes</h2>**


````python
x = [1, 2, 3, 4, 5, 6]
y = np.array([1, 2, 4, 7, 11, 16])
````

````python
j = np.gradient(y)
j
````

El gradiente se define como (cambio en y)/(cambio en x).

**x**, aqu, es el ndice de la lista, por lo que la diferencia entre valores adyacentes es 1.

En los lmites, se calcula la primera diferencia. Esto significa que en cada extremo de la matriz, el gradiente dado es simplemente, la diferencia entre los dos valores finales (dividido por 1).

Lejos de los lmites, el gradiente para un ndice particular se obtiene tomando la diferencia entre los valores de cada lado y dividindola por 2.

As, el gradiente de y, arriba, se calcula as:
```python
j[0] = (y[1]-y[0])/1 = (2-1)/1 = 1
j[1] = (y[2]-y[0])/2 = (4-1)/2 = 1.5
j[2] = (y[3]-y[1])/2 =.(7-2)/2 = 2,5
j[3] = (y[4]-y[2])/2 =.(11-4)/2 = 3,5
j[4] = (y[5]-y[3])/2 . (16-7)/2 = 4,5
j[5] = (y[5]-y[4])/1 = (16-11)/1 = 5
```
