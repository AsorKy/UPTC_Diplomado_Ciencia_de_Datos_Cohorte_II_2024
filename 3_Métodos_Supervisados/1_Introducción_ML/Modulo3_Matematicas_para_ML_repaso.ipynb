### <!-- Subsection 1.6 -->
**<h3 id="subsection16" style="color: #003366;">1.6. Solución de Ecuaciones Lineales </h3>**

---
Solución de ecuaciones lineales

---

Supongamos que tenemos el siguiente sistema de ecuaciones lineales el cuál queremos resolver.

\begin{align}
2x + 3y &= 1 \\\
5x - y &= 3
\end{align}

````python
A = np.array([[2, 3], [5, -1]])
print(A)
````

````python
b = np.array([1, 3]).reshape(2, 1)
print(b)
````

````python
x = la.solve(A, b)
print(x)
````

````python
#Podemos verificar la solución calculando la inversa de A:
Ainv = la.inv(A)
print(Ainv)
````

````python
A @ Ainv
````

Y multiplicar $A^{-1} \mathbf{b}$ para resolver $\mathbf{x}$:

````python
x = Ainv @ b
print(x)
````

Vectores y Espacios

¡Vamos a trazar algunos vectores!

````python
def plot_vectors(vectors, colors):
    plt.figure()
    plt.axvline(x=0, color='lightgray')
    plt.axhline(y=0, color='lightgray')

    for i in range(len(vectors)):
        x = np.concatenate([[0,0],vectors[i]])
        plt.quiver([x[0]], [x[1]], [x[2]], [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=colors[i],)
````

````python
# Definamos un vector
x = np.array([3, 1])

# Funciones
plot_vectors([x], ['lightblue'])
plt.xlim(-1, 5)
_ = plt.ylim(-1, 3)
````

````python
# Segundo vector
y = np.array([6, 4])

# Funciones
plot_vectors([x, y], ['lightblue', 'lightgreen'])
plt.xlim(-1, 8)
_ = plt.ylim(-1, 6)
````

````python
# Suma
x = np.array([3, 6])
y = np.array([6, 4])

vec_sum = x + y
vec_sum
````

````python
# Resultado
plot_vectors([x, y, vec_sum], ['lightblue', 'lightgreen', 'red'])
plt.xlim(-1, 10)
_ = plt.ylim(-1, 11)
````

````python
# Para la multiplicacion por escalar
x = np.array([3, 1])

x_2 = x * 2
x_2
````

````python
# Resultado
plot_vectors([x_2, x], ['lightgray', 'blue'])
plt.xlim(-1, 8)
_ = plt.ylim(-1, 4)
````

**<h2 id="section2" style="color: #007ACC;">2. Derivadas</h2>**


Definimos una función llamada `numerical_derivative` que calcula la derivada numérica de una función en un punto específico utilizando tres posibles métodos: **centrado**, **hacia adelante** o **hacia atrás**. La derivada numérica aproxima el valor de la derivada real mediante la diferencia de los valores de la función en puntos cercanos.

### **Parámetros:**
- `function`: es la función para la cual se quiere calcular la derivada.
- `point`: es el punto en el cual se quiere evaluar la derivada.
- `method`: es el método numérico que se utilizará para calcular la derivada. Los valores posibles son:
  - `'central'`: usa la fórmula de diferencias centradas.
  - `'forward'`: usa la fórmula de diferencias hacia adelante.
  - `'backward'`: usa la fórmula de diferencias hacia atrás.
- `h`: es el tamaño del paso, que representa una pequeña variación en el valor de `point` (por defecto es 0.01).

### **Métodos de cálculo:**
1. **Método centrado** (`'central'`): calcula la derivada aproximada utilizando valores a la izquierda y derecha del punto. La fórmula es:

   $$
   f'(x) \approx \frac{f(x + h) - f(x - h)}{2h}
   $$

2. **Método hacia adelante** (`'forward'`): calcula la derivada aproximada usando el valor del punto y un punto hacia adelante (más `h`). La fórmula es:

   $$
   f'(x) \approx \frac{f(x + h) - f(x)}{h}
   $$

3. **Método hacia atrás** (`'backward'`): calcula la derivada aproximada usando el valor del punto y un punto hacia atrás (menos `h`). La fórmula es:

   $$
   f'(x) \approx \frac{f(x) - f(x - h)}{h}
   $$

Si se proporciona un valor de método que no sea `'central'`, `'forward'` o `'backward'`, el código lanza un error (`ValueError`).

````python
def numerical_derivative(function, point, method='central', h=0.01):
    if method == 'central':
        return (function(point + h) - function(point - h)) / (2 * h)
    elif method == 'forward':
        
        return (function(point + h) - function(point)) / h
    elif method == 'backward':
        return (function(point) - function(point - h)) / h
    else:
        raise ValueError("Method must be 'central', 'forward' or 'backward'.")
````

3. **`plt.figure(figsize=(15, 7))`**:

Ahora comparamos diferentes formas de calcular la derivada de la función logarítmica \( \log(x) \). Específicamente, se grafican las derivadas aproximadas y la derivada real en el mismo gráfico.

1. **`x = np.linspace(0.1, 5*np.pi, 100)`**:
   - Genera un array de 100 puntos igualmente espaciados en el intervalo entre 0.1 y \( 5\pi \), que se utiliza como los valores de \( x \) sobre los cuales se calcularán las derivadas.

2. **`numerical = numerical_derivative(np.log, x)`**:
   - Calcula la derivada numérica de \( \log(x) \) utilizando la función `numerical_derivative` que se explicó anteriormente. Esta función utiliza el método de diferencias centradas por defecto.

3. **`by_rule = 1 / x`**:
   - Calcula el valor exacto de la derivada de \( \log(x) \) según la regla matemática, que es \( \frac{d}{dx} \log(x) = \frac{1}{x} \). Esto sirve como referencia para comparar con los métodos numéricos.

4. **`numpy_gradient = np.gradient(np.log(x), 0.15)`**:
   - Utiliza la función `np.gradient()` de NumPy para calcular la derivada numérica de \( \log(x) \). El segundo argumento (0.15) es el tamaño del paso utilizado para el cálculo de la derivada. Este método proporciona otra aproximación numérica de la derivada.

5. **`plt.figure(figsize=(15, 7))`**:
   - Establece el tamaño de la gráfica a 15 de ancho y 7 de alto, permitiendo una mejor visualización de los resultados.

6. **`plt.plot(x, numerical, 'r.', label='Central Difference', markersize=12)`**:
   - Dibuja los puntos que representan la derivada numérica calculada por el método de diferencias centradas, con marcadores rojos (`'r.'`) de tamaño 12.

7. **`plt.plot(x, by_rule, 'b', label='True Value')`**:
   - Dibuja la derivada real de \( \log(x) \) (que es \( \frac{1}{x} \)) con una línea azul (`'b'`). Este es el valor de referencia para comparar los otros métodos.

8. **`plt.plot(x, np.log(x), 'g', label='Log(x)')`**:
   - Dibuja la función \( \log(x) \) con una línea verde (`'g'`), para mostrar la curva original que estamos diferenciando.

9. **`plt.plot(x, numpy_gradient, 'y.', label='np.gradient()')`**:
   - Dibuja los puntos que representan la derivada calculada con `np.gradient()` con marcadores amarillos (`'y.'`).

10. **`plt.legend(loc='best')`**:
    - Añade una leyenda que explica qué representa cada curva en la gráfica. La leyenda se coloca automáticamente en la mejor posición.

11. **`plt.show()`**:
    - Muestra, así como visualizar la función original.

````python
x = np.linspace(0.1, 5*np.pi, 100)

numerical = numerical_derivative(np.log, x)
by_rule = 1 / x

numpy_gradient = np.gradient(np.log(x), 0.15)

plt.figure(figsize=(15, 7))
plt.plot(x, numerical, 'r.', label='Central Difference', markersize=12)
plt.plot(x, by_rule, 'b', label='True Value')
plt.plot(x, np.log(x), 'g', label='Log(x)')
plt.plot(x, numpy_gradient, 'y.', label='np.gradient()')
plt.legend(loc='best')
plt.show()
````

**<h2 id="section3" style="color: #007ACC;">3. Cálculo de Gradientes</h2>**


````python
x = [1, 2, 3, 4, 5, 6]
y = np.array([1, 2, 4, 7, 11, 16])
````

````python
j = np.gradient(y)
j
````

El gradiente se define como (cambio en y)/(cambio en x).

**x**, aquí, es el índice de la lista, por lo que la diferencia entre valores adyacentes es 1.

En los límites, se calcula la primera diferencia. Esto significa que en cada extremo de la matriz, el gradiente dado es simplemente, la diferencia entre los dos valores finales (dividido por 1).

Lejos de los límites, el gradiente para un índice particular se obtiene tomando la diferencia entre los valores de cada lado y dividiéndola por 2.

Así, el gradiente de y, arriba, se calcula así:
```python
j[0] = (y[1]-y[0])/1 = (2-1)/1 = 1
j[1] = (y[2]-y[0])/2 = (4-1)/2 = 1.5
j[2] = (y[3]-y[1])/2 =.(7-2)/2 = 2,5
j[3] = (y[4]-y[2])/2 =.(11-4)/2 = 3,5
j[4] = (y[5]-y[3])/2 . (16-7)/2 = 4,5
j[5] = (y[5]-y[4])/1 = (16-11)/1 = 5
```

````python
A = np.array([[1, 2, 6], [3, 4, 5], [3, 4, 5]])
````

````python
np.gradient(A)
````

Visualización de Gradientes

---

````python
x = np.arange(-1, 1, 0.1)
y = np.arange(-1, 1, 0.1)
X, Y = np.meshgrid(x, y, indexing='ij')
Z = X**2 + Y**2

gradx, grady = np.gradient(Z)

n = 20
l = np.array([0.0, 0.5, 1.0, 1.5, 2.0])

plt.figure(figsize=(10, 8))
plt.contourf(X, Y, Z, n)
plt.contour(X, Y, Z, levels=l, colors='k', linewidths=1, linestyles='solid')
plt.quiver(X, Y, gradx, grady);
````
