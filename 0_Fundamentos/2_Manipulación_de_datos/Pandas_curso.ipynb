Un DataFrame es una estructura de datos bidimensional de Pandas que se asemeja a una tabla u hoja de clculo. ---
Manejo de datos nulos

---

Los datos nulos, tambin conocidos como valores faltantes o `NaN` (Not a Number), son valores que faltan o no estn presentes en un conjunto de datos. En Pandas, los datos nulos se representan como `NaN`. Es importante tener en cuenta los datos nulos por las siguientes razones:

* De no ser corregidos, pueden comprometer la intergridad de los datos e ingresar sesgos en el anlisis de los msmos.

* De no ser corregidos, comprometen la presicin del anlisis y del modelo de prediccin.

* Los datos nulos pueden tener significados diferentes dependiendo del contexto del conjunto de datos. Al comprender y manejar los datos nulos adecuadamente, se puede interpretar correctamente el conjunto de datos y tomar decisiones informadas basadas en los resultados del anlisis.

```python
# Creacin de un dataframe con datos nulos
import numpy as np
import pandas as pd

datos = {'var_1':[1,2,3,np.nan],
         'var_2':[4, np.nan,6,7],
         'var_3':['a','b','c', None],
         'var_4':[1,4,3,2]}
df_test = pd.DataFrame(datos)
df_test
```

Para identificar que columnas y el nmero de valores nulos que posee cada variable, podemos emplear el mtodo `.info()`.

```python
# Deteccin de valores faltantes mediante .info()
df_test.info()
```

Por otra parte, es posible identificar concretamente la posicin de los valores faltanes mediante el mtodo `.isnull()`.

```python
# Identificar datos nulos en el dataframe
df_test.isnull()
```

```python
# Identificar datos nulos con un valor numrico
df_test.isnull()*1
```

```python
# Identificar datos no nulos en un dataset
df_test.notnull()
```

```python
# Determinar la cantidad de valores nulos o faltantes por columna
df_test.isnull().sum()
```

El manejo de valores nulos o faltantes es un tema bastante amplio, que implica procesos de interpretacin de dichos valores para determinar su relevancia y procesos relacionados con la imputacin de estos valores segn diferentes tcnicas. De forma general, cuando nos enfrentamos a un dataset con valores nulos o faltantes, pueden tomarse las siguientes acciones segn el contexto de los datos y la finalidad de la accin:

* Eliminacin de datos faltantes: Posible mediante mtodos como `.dropna()`.
* Reemplazo de valores faltantes: Posible mediante mtodos como `.fillna().
* Imputacin de datos faltantes: Posible mediante mtodos como `.fillna()` e `.interpolate()`.

```python
# Eliminacin de valores faltantes (eliminacin por filas)
df_test_1 = df_test.copy()
df_test_1.dropna(axis = 0, inplace=True)
df_test_1
```

```python
# Eliminacin de valores faltantes (eliminacin por columnas)
df_test_2 = df_test.copy()
df_test_2.dropna(axis=1, inplace=True)
df_test_2
```

```python
# Remplazo de valores faltantes por un string
df_test_1 = df_test.copy()
df_test_1.fillna('Missing', inplace=True)
df_test_1
```

```python
# Remplazo de valores faltantes por un string
df_test_1 = df_test.copy()
df_test_1['var_1'].fillna(0.0, inplace=True)
df_test_1['var_2'].fillna(0.0, inplace=True)
df_test_1['var_3'].fillna('Missing', inplace=True)
df_test_1
```

```python
df_test_1.info()
```

```python
# Imputacin de valores por promedio
df_test_1 = df_test.copy()
df_test_1.fillna(df_test_1.mean(), inplace=True)
df_test_1
```

El mtodo `.interpolate()` en Pandas se utiliza para interpolar valores faltantes en un DataFrame. La interpolacin es el proceso de estimar valores desconocidos en funcin de valores conocidos mediante mtodos como regresin lineal o polinomial.

```python
# Imputacin de valores por interpolate
df_test_1 = df_test.copy()
df_test_1.interpolate(inplace=True)
df_test_1
```

```python
# Crear un DataFrame con valores faltantes
data = {'A': [1, 2, None, 4, None, 6],
        'B': [None, 5, None, None, 8, 9]}
df_incomplete = pd.DataFrame(data)

print(df_incomplete)
# Interpolar valores faltantes utilizando interpolacin lineal
df_interpolado = df_incomplete.interpolate(method='linear', axis=0)
df_interpolado
```

**<h3 id="subsection14" style="color: #003366;">1.4. Manejo de Duplicados</h3>**

---
Manejo de duplicados

---

Los valores duplicados son elementos que aparecen ms de una vez en un conjunto de datos. Estos duplicados pueden estar presentes en una o ms columnas del conjunto de datos y pueden representar una variedad de situaciones, como errores de entrada, registros duplicados o datos repetidos. Detectar y manejar valores duplicados es fundamental en el anlisis de datos por varias razones:

* Asegurar la integridad y calidad de los datos: Los valores duplicados pueden distorsionar la integridad de los datos y conducir a anlisis incorrectos. Identificar y eliminar estos duplicados ayuda a garantizar que los resultados del anlisis sean precisos y confiables.

* Asegurar la consistencia de los resultados: Al eliminar los valores duplicados, se asegura que los resultados del anlisis sean coherentes y replicables.

* Asegurar la eficiencia en el procesamiento de datos: Los valores duplicados pueden aumentar la complejidad y el tiempo de procesamiento durante el anlisis de datos. Al eliminar los duplicados, se simplifica el conjunto de datos y se mejora la eficiencia en el procesamiento y la manipulacin de los datos.

```python
# Creacin de un dataset con valores duplicados
data = {
    'A': [1, 2, 3, 4, 5, 2, 7, 8, 9, 2],
    'B': ['a', 'b', 'c', 'd', 'e', 'b', 'g', 'h', 'i', 'b'],
    'C': [10.5, 20.3, 15.8, 7.2, 3.4, 20.3, 18.9, 22.1, 30.6, 20.3]
}

df_test = pd.DataFrame(data)
df_test
```

La deteccin de valores duplicados se da con base a la existencia de regstros cuyos valores de variables son exactamente los msmos.

```python
# Detectar filas duplicadas en el DataFrame
duplicados = df_test.duplicated()
duplicados
```

No obstante, antes de proceder, es importante saber que es posible detectar los valores nicos dentro de cada columna.

```python
# Valores nicos dentro de una columna
df_test['B'].unique()
```

```python
# Cuntos valores posibles toma una variable
df_test['B'].nunique()
```

En general, el manejo de duplicados suele consistir en la eliminacin de los msmos. Para ello, empleamos el mtodo `.drop_duplicates()`.

```python
# Eliminacin simple de duplicados
df_test_1 = df_test.copy()
df_test_1.drop_duplicates(inplace=True)
df_test_1
```

Para reindexar el dataframe resultante despus de eliminar los duplicados, empleamos el parmetro `ignore_index`.

```python
# Eliminacin simple de duplicados
df_test_1 = df_test.copy()
df_test_1.drop_duplicates(inplace=True, ignore_index=True)
df_test_1
```

Dentro de Pandas, es posible eliminar duplicados de acuerdo con un conjunto de variables o columnas especificas mediante el parmetro `subset`.

```python
# Eliminacin de duplicados de acuerdo con un conjunto de columnas
data = {
    'A': [1, 2, 3, 2, 4, 5, 2],
    'B': ['a','b','c','b','e','f','b'],
    'C': np.random.randn(7)
}

df_test = pd.DataFrame(data)
print(df_test)

df_subset = df_test.drop_duplicates(subset = ['A','B'], ignore_index=True)
df_subset
```

El parmetro `keep` en `drop_duplicates()` permite especificar qu duplicados mantener. Las opciones son `'first'` para mantener la primera aparicin de cada duplicado, `'last'` para mantener la ltima aparicin y `False` para eliminar todas los duplicados.

```python
# Uso del comando Keep
df_subset = df_test.drop_duplicates(subset=['A','B'], keep='last')
df_subset
```

---
Combinacin de datos

---

La combinacin de DataFrames y la agrupacin de datos son dos tcnicas esenciales en el anlisis y manipulacin de datos con Pandas. Estas operaciones permiten realizar tareas complejas de manejo y anlisis de datos de manera eficiente y efectiva.

La combinacin de DataFrames implica la unin de dos o ms conjuntos de datos basados en una o ms columnas comunes. Esto permite combinar la informacin de diferentes fuentes de datos en un solo DataFrame, lo que facilita el anlisis y la interpretacin de los datos. Las operaciones comunes de combinacin de DataFrames incluyen la fusin (`merge`), la concatenacin (`concat`) y la unin (`join`).

* Fusin (`merge`): Combina filas de dos DataFrames basndose en una columna comn.
* Concatenacin (`concat`): Une dos o ms DataFrames a lo largo de un eje especfico (filas o columnas).
* Unin (`join`): Combina columnas de dos DataFrames basndose en sus ndices.

```python
# Crear DataFrames de ejemplo
df1 = pd.DataFrame({'A': [1, 2, 3],
                    'B': ['a', 'b', 'c']})

df2 = pd.DataFrame({'C': [4, 5, 6],
                    'D': ['d', 'e', 'f']})
```

```python
# Concatenacin Vertical (a lo largo de las filas)
result_vertical = pd.concat([df1, df2], axis=0, ignore_index=True)
print("Concatenacin Vertical:")
result_vertical
```

```python
# Concatenacin Horizontal (a lo largo de las columnas)
result_horizontal = pd.concat([df1, df2], axis=1)
print("Concatenacin Horizontal:")
result_horizontal
```

Las fusiones de dataframes se conssten en combinar dataframes de acuerdo a caractersticas comnes a los dataframes implicados a diferencia de la concatenacin la cul simplemente consste en unir dataframes bien sea vertical u horizontalmente.

```python
# DataFrame df1
df1 = pd.DataFrame({'clave': ['A', 'B', 'C', 'D'],
                    'valor': [1, 2, 3, 4]})

# DataFrame df2
df2 = pd.DataFrame({'clave': ['B', 'D', 'E', 'F'],
                    'otro_valor': [5, 6, 7, 8]})
print(df1)
print(df2)
```

En la fusin por columna comn, los DataFrames se fusionan utilizando una columna que es comn a ambos DataFrames. Esta columna se especifica utilizando el argumento `on`. La fusin combina las filas de los DataFrames donde los valores en la columna comn coinciden.

```python
# Fusin por columna comn
merged_df = pd.merge(df1, df2, on='clave')
print("Fusin por Columna Comn:")
merged_df
```

En la fusin interna, tambin conocida como fusin por defecto, se conservan solo las filas que tienen claves coincidentes en ambos DataFrames. Es decir, solo se incluyen las filas donde los valores en la columna comn coinciden entre los DataFrames. Las filas que no tienen coincidencias en ambos DataFrames se descartan.

```python
# Fusin interna
merged_inner = pd.merge(df1, df2, on='clave', how='inner')
print("Fusin Interna:")
merged_inner
```

En la fusin externa, se conservan todas las filas de ambos DataFrames, incluso aquellas que no tienen coincidencias en la columna comn. Se crean valores nulos (`NaN`) en las columnas correspondientes para las filas que no tienen coincidencias en el otro DataFrame.

```python
# Fusin externa
merged_outer = pd.merge(df1, df2, on='clave', how='outer')
print("Fusin Externa:")
merged_outer
```

En la fusin izquierda, tambin conocida como fusin izquierda externa, se conservan todas las filas del DataFrame izquierdo y solo las filas del DataFrame derecho que tienen claves coincidentes. Las filas que no tienen coincidencias en el DataFrame derecho tendrn valores nulos en las columnas correspondientes.

```python
# Fusin izquierda
merged_left = pd.merge(df1, df2, on='clave', how='left')
print("Fusin Izquierda:")
merged_left
```

En la fusin derecha, tambin conocida como fusin derecha externa, se conservan todas las filas del DataFrame derecho y solo las filas del DataFrame izquierdo que tienen claves coincidentes. Las filas que no tienen coincidencias en el DataFrame izquierdo tendrn valores nulos en las columnas correspondientes.

```python
# Fusin derecha
merged_right = pd.merge(df1, df2, on='clave', how='right')
print("Fusin Derecha:")
merged_right
```

El mtodo `join()` en Pandas proporciona una forma conveniente de combinar DataFrames utilizando sus ndices como clave de unin. Aunque es menos flexible que `merge()`, es til cuando necesitas fusionar DataFrames basados en sus ndices sin tener que especificar columnas especficas. La diferencia principal entre `merge()` y `join()` radica en cmo se selecciona la clave de unin: columnas especficas para `merge()` e ndices para `join()`.

```python
# DataFrame df1
df1 = pd.DataFrame({'clave': ['A', 'B', 'C', 'D'],
                    'valor': [1, 2, 3, 4]}
                   , index = ['a','b','c','d']
                   )

# DataFrame df2
df2 = pd.DataFrame({'clave': ['B', 'D', 'E', 'F'],
                    'otro_valor': [5, 6, 7, 8]}
                   , index = ['c','d','e','f']
                   )
print(df1)
print(df2)
```

```python
# Inner Join
try:
  inner_join = df1.join(df2, how='inner')
except:
  print('Agrega los sufijos izquierdo y derecho para diferenciar columnas:')
  inner_join = df1.join(df2, how='inner', lsuffix='_left', rsuffix='_right')

print("Inner Join:")
inner_join
```

```python
# Outer Join
try:
  outer_join = df1.join(df2, how='outer')
except:
  print('Agrega los sufijos izquierdo y derecho para diferenciar columnas:')
  outer_join = df1.join(df2, how='outer', lsuffix='_left', rsuffix='_right')

print("Outer Join:")
outer_join
```

```python
# Left Join
try:
  left_join = df1.join(df2, how='left')
except:
  print('Agrega los sufijos izquierdo y derecho para diferenciar columnas:')
  left_join = df1.join(df2, how='left', lsuffix='_left', rsuffix='_right')

print("left Join:")
left_join
```

```python
# right Join
try:
  right_join = df1.join(df2, how='right')
except:
  print('Agrega los sufijos izquierdo y derecho para diferenciar columnas:')
  right_join = df1.join(df2, how='right', lsuffix='_left', rsuffix='_right')

print("right Join:")
right_join
```

Agrupacin de datos

---
La agrupacin de datos implica dividir los datos en grupos basados en ciertas caractersticas y luego aplicar una funcin a cada grupo de datos. Esta tcnica es til para resumir y analizar datos de manera eficiente, especialmente cuando se trabaja con conjuntos de datos grandes. La operacin principal para la agrupacin de datos es el mtodo `groupby()`, que permite dividir un DataFrame en grupos basados en los valores de una o ms columnas y luego aplicar funciones de agregacin a estos grupos.


```python
# Datos de ejemplo
tips_data = sns.load_dataset('tips')
tips_data.info()
```

```python
# Agrupacin por variable categrica
grupos = tips_data.groupby('day')
tips_data.day.cat.categories
```

Para obtener un grupo en especfico, empleamos el mtodo `get_group()`.

```python
# Obtenemos los grupos
grupos_lista = []
for nombre_grupo, grupo in grupos:
  grupos_lista.append(grupo)

grupos_lista[0].head()
```

Una vez los grupos con base a una variable son identificados, es posible aplicar mtodos a cada grupo.

```python
# Promedios para cada agrupacin
grupos.mean()
```

```python
# Estadsticos bsicos para cada agrupacin
grupos.describe()['total_bill']
```


El comando `.agg()` en Pandas se utiliza para realizar agregaciones de datos en un DataFrame o en grupos de datos de tal manera que permite calcular una o ms estadsticas agregadas para una o ms columnas, lo que facilita el resumen y anlisis de grandes conjuntos de datos.

```python
tips_data.columns
```

```python
# Agregacin y estadsticos
grupos = tips_data.groupby('day')
variables_de_interes = {'total_bill': 'mean', 'tip': 'mean'}
sub_grupos = grupos.agg(variables_de_interes)

sub_grupos
```

El comando `.agg` tambien permite generar agrupaciones con mltiples estadsticas.

```python
# Agregacin con mltiples estadsticas
grupos = tips_data.groupby('day')
consulta = {'total_bill' : ['min' , 'max'], 'tip': ['mean','median']}
consulta_anidada = grupos.agg(consulta)

consulta_anidada
```

Finalmente, el comando `.groupby` permite realizar agrupaciones anidadas, es decir, agrupaciones con base a mltiples variables.

```python
# Agrupaciones con base a mltiples variables
variables_a_agrupar = ['day', 'time', 'sex']
grupos = tips_data.groupby(variables_a_agrupar)['tip'].mean()

grupos
```

```python
# Agrupaciones con base a mltiples variables
variables_a_agrupar = ['day', 'time', 'sex']
grupos = tips_data.groupby(variables_a_agrupar)['tip'].describe()

grupos
```

Mtodos apply

---


El mtodo `apply()` en Pandas se utiliza para aplicar una funcin arbitraria a los elementos de un DataFrame o una Serie. Esta funcin puede ser una funcin integrada de Python, una funcin definida por el usuario o una funcin lambda. El mtodo `apply()` es til para realizar operaciones personalizadas en los datos de un DataFrame y para realizar transformaciones ms complejas que no estn directamente disponibles mediante los mtodos integrados de Pandas. Su sintaxis bsica es `DataFrame.apply(func, axis=0, raw=False, result_type=None, args=(), **kwargs)`.

* `func`: Funcin a aplicar a cada elemento del DataFrame o la Serie.

* `axis`: Eje a lo largo del cual se aplica la funcin. `0` para aplicar la funcin a cada columna y `1` para aplicarla a cada fila (solo en DataFrame).

* `raw`: Si es `True`, pasa cada fila o columna como una serie a la funcin. Por defecto es `False`.

* `result_type`: Tipo de datos de retorno esperado despus de aplicar la funcin. Puede ser `'expand'`, `'reduce'` o `'broadcast'`.

* `args, kwargs`: Argumentos adicionales que se pasan a la funcin.
