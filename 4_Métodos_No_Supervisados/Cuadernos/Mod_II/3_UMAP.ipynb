<h2 style="color: #007ACC;"> Introducción </h2>

UMAP (Uniform Manifold Approximation and Projection) es una técnica de reducción de dimensionalidad utilizada para visualizar y analizar datos de alta dimensión. Desarrollada por [Leland McInnes, John Healy y James Melville](https://arxiv.org/pdf/1802.03426), UMAP es particularmente efectiva para descubrir estructuras subyacentes en conjuntos de datos complejos, lo que la hace popular en campos como la biología computacional, el análisis de textos y la visualización de grandes volúmenes de datos.

UMAP trabaja bajo el supuesto de que los datos de alta dimensión se encuentran en una variedad de dimensiones más bajas, conocida como un colector. Utiliza conceptos matemáticos avanzados, como la teoría de la homología persistente y el análisis topológico de datos, para construir un gráfico que represente las relaciones locales entre los puntos de datos. A partir de este gráfico, UMAP proyecta los datos en un espacio de menor dimensión, preservando tanto las estructuras globales como las locales.

Las principales ventajas de UMAP incluyen su velocidad y escalabilidad, su capacidad para capturar tanto la estructura global como los detalles locales de los datos, y su flexibilidad para adaptarse a diferentes tipos de datos y métricas de distancia. Estas características han hecho de UMAP una herramienta esencial en el análisis exploratorio de datos y en la creación de visualizaciones intuitivas y reveladoras.

En resumen, UMAP es una herramienta poderosa y versátil para reducir la dimensionalidad de datos complejos, proporcionando una ventana clara hacia las estructuras intrincadas que pueden ocultarse en conjuntos de datos de alta dimensión.

<h2 style="color: #007ACC;"> Funcionamiento del UMAP </h2>

UMAP es una técnica de reducción de dimensionalidad que se ha vuelto muy popular en el campo del análisis de datos y la visualización. A continuación, se explica en detalle cómo funciona UMAP.

<h3 id="subsection1-1" style="color: #003366;">Fundamentos Teóricos</h3>

UMAP se basa en conceptos de topología algebraica y teoría de grafos. Su objetivo principal es preservar la estructura local de los datos en un espacio de alta dimensionalidad al proyectarlos en un espacio de menor dimensionalidad. Esto se logra manteniendo tanto la proximidad de los puntos cercanos como la estructura global en la medida de lo posible.

<h3 id="subsection1-1" style="color: #003366;">Construcción del Grafo de Vecindad</h3>

El primer paso en UMAP es construir un grafo de vecindad basado en la distancia entre los puntos en el espacio de alta dimensionalidad. Este grafo se construye mediante los siguientes pasos:

1. **Definir el Número de Vecinos**: Se selecciona un parámetro \( k \) que define el número de vecinos más cercanos para cada punto. Este parámetro es crucial, ya que controla la granularidad de la estructura local que UMAP intenta preservar.

2. **Cálculo de la Distancia**: Se calculan las distancias entre todos los puntos en el espacio de alta dimensionalidad utilizando una métrica de distancia (por ejemplo, la distancia euclidiana).

3. **Construcción del Grafo**: Para cada punto, se identifican los \( k \) vecinos más cercanos y se conectan mediante aristas en el grafo.

<h3 id="subsection1-1" style="color: #003366;">Cálculo de la Simplicidad Local</h3>

UMAP utiliza un enfoque basado en simplicidad local para modelar la estructura local de los datos. Para ello, realiza los siguientes pasos:

1. **Escalado Local**: UMAP calcula una escala local para cada punto que determina cómo de amplio o estrecho es el entorno local. Esto se hace utilizando una distribución adaptativa (por ejemplo, la distribución uniforme).

2. **Ponderación de las Aristas**: Las aristas del grafo se ponderan en función de las escalas locales calculadas. Esto permite que las distancias locales se ajusten de manera que se preserve mejor la estructura local en el espacio reducido.


<h3 id="subsection1-1" style="color: #003366;">Embedding en el Espacio de Baja Dimensionalidad</h3>

Una vez construido el grafo de vecindad y calculada la simplicidad local, UMAP procede a proyectar los datos en un espacio de baja dimensionalidad (generalmente 2D o 3D) mediante los siguientes pasos:

1. **Optimización**: UMAP utiliza un algoritmo de optimización basado en gradiente para encontrar una representación en el espacio de baja dimensionalidad que minimice la discrepancia entre las estructuras locales del espacio de alta y baja dimensionalidad.

2. **Función de Costo**: La función de costo que UMAP minimiza está diseñada para preservar la proximidad de los puntos cercanos en el espacio original y mantener una estructura global coherente.

3. **Iteraciones**: El algoritmo de optimización itera hasta que la representación en el espacio de baja dimensionalidad converge a una configuración que satisface los criterios de proximidad y simplicidad local.

<h3 id="subsection1-1" style="color: #003366;">Interpretación y Visualización</h3>

El resultado final de UMAP es una representación de los datos en un espacio de baja dimensionalidad que puede ser visualizada y analizada fácilmente. Esta representación preserva la estructura local de los datos, lo que facilita la identificación de patrones, clústeres y relaciones en los datos originales.

<h3 id="subsection1-1" style="color: #003366;"> Ventajas y Limitaciones</h3>

**Ventajas:**

- **Eficiencia**: UMAP es muy eficiente en términos de tiempo de computación, lo que lo hace adecuado para grandes conjuntos de datos.
- **Preservación de la Estructura Local**: UMAP es efectivo para mantener la proximidad de puntos cercanos, lo que es útil para tareas de clustering y visualización.
- **Flexibilidad**: Permite ajustar el parámetro \( k \) para controlar el nivel de detalle en la estructura local.

**Limitaciones:**

- **Sensibilidad a Parámetros**: La elección del número de vecinos \( k \) y otros parámetros puede afectar significativamente los resultados.
- **Estructura Global**: Aunque UMAP preserva bien la estructura local, puede no ser tan efectivo en mantener la estructura global de los datos en algunos casos.

En resumen, UMAP es una técnica poderosa y flexible para la reducción de dimensionalidad que se basa en principios de topología y teoría de grafos para preservar la estructura local de los datos. Su eficiencia y capacidad para manejar grandes volúmenes de datos lo hacen una herramienta valiosa en el análisis y visualización de datos.

<h3 id="subsection1-1" style="color: #003366;"> Ejemplo 1: datos iris</h3>

````python
import umap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
````

````python
# Cargar el conjunto de datos Iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
````

````python
# Normalizar los datos
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df.iloc[:, :-1])
````

````python
# Configurar y aplicar UMAP
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=2, random_state=42)
embedding = reducer.fit_transform(df_scaled)

# Convertir el resultado a un DataFrame
embedding_df = pd.DataFrame(embedding, columns=['UMAP1', 'UMAP2'])
embedding_df['species'] = df['species']
````

````python
# Visualizar el resultado usando Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='UMAP1', y='UMAP2', hue='species', palette='viridis', data=embedding_df)
plt.title('UMAP projection of the Iris dataset')
plt.show()
````

<h3 id="subsection1-1" style="color: #003366;"> Ejemplo 3: datos de dígitos</h3>

````python
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
````

````python
# Cargar el conjunto de datos MNIST
mnist = fetch_openml('mnist_784', version=1)
X, y = mnist.data, mnist.target

# Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reducir la dimensionalidad inicial con PCA para acelerar UMAP
pca = PCA(n_components=50)
X_pca = pca.fit_transform(X_scaled)
````

````python
# Configurar y aplicar UMAP
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=2, random_state=42)
embedding = reducer.fit_transform(X_pca)

# Convertir el resultado a un DataFrame
embedding_df = pd.DataFrame(embedding, columns=['UMAP1', 'UMAP2'])
embedding_df['digit'] = y
````

````python
# Visualizar el resultado usando Seaborn
plt.figure(figsize=(12, 10))
sns.scatterplot(x='UMAP1', y='UMAP2', hue='digit', palette='tab10', data=embedding_df, legend='full', alpha=0.3)
plt.title('UMAP projection of the MNIST dataset')
plt.show()
````

<h2 style="color: #007ACC;"> Parámetros básicos de UMAP </h2>

1.	n_neighbors: Este parámetro determina el número de vecinos más cercanos que se considerarán para cada punto al construir el grafo de vecindad. Afecta la forma en que UMAP equilibra la preservación de la estructura local versus la estructura global de los datos. Un valor bajo de n_neighbors hace que UMAP capture más detalles locales, mientras que un valor alto captura mejor la estructura global.
  
2.	min_dist: Este parámetro controla la distancia mínima entre los puntos en el espacio reducido. Afecta la dispersión de los datos en el espacio de baja dimensionalidad. Un valor bajo de min_dist mantiene los puntos más cercanos entre sí, lo que resulta en una mayor compactación de los clústeres. Un valor alto separa más los puntos, mostrando una estructura más extendida.

3.	n_components: Define el número de dimensiones en las que se proyectarán los datos. Comúnmente, se elige 2 o 3 para facilitar la visualización, pero se pueden utilizar valores más altos para otras aplicaciones.

4.	metric: La métrica de distancia utilizada para calcular las distancias entre los puntos en el espacio de alta dimensionalidad. Las métricas comunes incluyen ‘euclidean’, ‘manhattan’, ‘cosine’, entre otras. La elección de la métrica puede afectar significativamente los resultados.

5.	random_state: Una semilla para el generador de números aleatorios que se utiliza para la inicialización y para garantizar la reproducibilidad de los resultados.

6.	transform_seed: Controla la semilla para el método de transformación de nuevos puntos. Asegura que las transformaciones de datos nuevos sean consistentes con la transformación original.

````python
# Cargar el conjunto de datos MNIST
mnist = fetch_openml('mnist_784', version=1)
X, y = mnist.data, mnist.target

# Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reducir la dimensionalidad inicial con PCA para acelerar UMAP
pca = PCA(n_components=50)
X_pca = pca.fit_transform(X_scaled)
````

````python
# Configurar y aplicar UMAP con diferentes valores de n_neighbors y min_dist
reducer_1 = umap.UMAP(n_neighbors=5, min_dist=0.1, n_components=2, random_state=42)
embedding_1 = reducer_1.fit_transform(X_pca)

reducer_2 = umap.UMAP(n_neighbors=30, min_dist=0.3, n_components=2, random_state=42)
embedding_2 = reducer_2.fit_transform(X_pca)
````

````python
# Convertir los resultados a DataFrames
embedding_df_1 = pd.DataFrame(embedding_1, columns=['UMAP1', 'UMAP2'])
embedding_df_1['digit'] = y

embedding_df_2 = pd.DataFrame(embedding_2, columns=['UMAP1', 'UMAP2'])
embedding_df_2['digit'] = y

# Visualizar los resultados usando Seaborn
plt.figure(figsize=(12, 10))

plt.subplot(1, 2, 1)
sns.scatterplot(x='UMAP1', y='UMAP2', hue='digit', palette='tab10', data=embedding_df_1, legend='full', alpha=0.3)
plt.title('UMAP with n_neighbors=5, min_dist=0.1')

plt.subplot(1, 2, 2)
sns.scatterplot(x='UMAP1', y='UMAP2', hue='digit', palette='tab10', data=embedding_df_2, legend='full', alpha=0.3)
plt.title('UMAP with n_neighbors=30, min_dist=0.3')

plt.show()
````
