<h2 style="color: #007ACC;"> Introduccin </h2>

UMAP (Uniform Manifold Approximation and Projection) es una tcnica de reduccin de dimensionalidad utilizada para visualizar y analizar datos de alta dimensin. Desarrollada por [Leland McInnes, John Healy y James Melville](https://arxiv.org/pdf/1802.03426), UMAP es particularmente efectiva para descubrir estructuras subyacentes en conjuntos de datos complejos, lo que la hace popular en campos como la biologa computacional, el anlisis de textos y la visualizacin de grandes volmenes de datos.

UMAP trabaja bajo el supuesto de que los datos de alta dimensin se encuentran en una variedad de dimensiones ms bajas, conocida como un colector. Utiliza conceptos matemticos avanzados, como la teora de la homologa persistente y el anlisis topolgico de datos, para construir un grfico que represente las relaciones locales entre los puntos de datos. A partir de este grfico, UMAP proyecta los datos en un espacio de menor dimensin, preservando tanto las estructuras globales como las locales.

Las principales ventajas de UMAP incluyen su velocidad y escalabilidad, su capacidad para capturar tanto la estructura global como los detalles locales de los datos, y su flexibilidad para adaptarse a diferentes tipos de datos y mtricas de distancia. Estas caractersticas han hecho de UMAP una herramienta esencial en el anlisis exploratorio de datos y en la creacin de visualizaciones intuitivas y reveladoras.

En resumen, UMAP es una herramienta poderosa y verstil para reducir la dimensionalidad de datos complejos, proporcionando una ventana clara hacia las estructuras intrincadas que pueden ocultarse en conjuntos de datos de alta dimensin.

<h2 style="color: #007ACC;"> Funcionamiento del UMAP </h2>

UMAP es una tcnica de reduccin de dimensionalidad que se ha vuelto muy popular en el campo del anlisis de datos y la visualizacin. A continuacin, se explica en detalle cmo funciona UMAP.

<h3 id="subsection1-1" style="color: #003366;">Fundamentos Tericos</h3>

UMAP se basa en conceptos de topologa algebraica y teora de grafos. Su objetivo principal es preservar la estructura local de los datos en un espacio de alta dimensionalidad al proyectarlos en un espacio de menor dimensionalidad. Esto se logra manteniendo tanto la proximidad de los puntos cercanos como la estructura global en la medida de lo posible.

<h3 id="subsection1-1" style="color: #003366;">Construccin del Grafo de Vecindad</h3>

El primer paso en UMAP es construir un grafo de vecindad basado en la distancia entre los puntos en el espacio de alta dimensionalidad. Este grafo se construye mediante los siguientes pasos:

1. **Definir el Nmero de Vecinos**: Se selecciona un parmetro \( k \) que define el nmero de vecinos ms cercanos para cada punto. Este parmetro es crucial, ya que controla la granularidad de la estructura local que UMAP intenta preservar.

2. **Clculo de la Distancia**: Se calculan las distancias entre todos los puntos en el espacio de alta dimensionalidad utilizando una mtrica de distancia (por ejemplo, la distancia euclidiana).

3. **Construccin del Grafo**: Para cada punto, se identifican los \( k \) vecinos ms cercanos y se conectan mediante aristas en el grafo.

<h3 id="subsection1-1" style="color: #003366;">Clculo de la Simplicidad Local</h3>

UMAP utiliza un enfoque basado en simplicidad local para modelar la estructura local de los datos. Para ello, realiza los siguientes pasos:

1. **Escalado Local**: UMAP calcula una escala local para cada punto que determina cmo de amplio o estrecho es el entorno local. Esto se hace utilizando una distribucin adaptativa (por ejemplo, la distribucin uniforme).

2. **Ponderacin de las Aristas**: Las aristas del grafo se ponderan en funcin de las escalas locales calculadas. Esto permite que las distancias locales se ajusten de manera que se preserve mejor la estructura local en el espacio reducido.


<h3 id="subsection1-1" style="color: #003366;">Embedding en el Espacio de Baja Dimensionalidad</h3>

Una vez construido el grafo de vecindad y calculada la simplicidad local, UMAP procede a proyectar los datos en un espacio de baja dimensionalidad (generalmente 2D o 3D) mediante los siguientes pasos:

1. **Optimizacin**: UMAP utiliza un algoritmo de optimizacin basado en gradiente para encontrar una representacin en el espacio de baja dimensionalidad que minimice la discrepancia entre las estructuras locales del espacio de alta y baja dimensionalidad.

2. **Funcin de Costo**: La funcin de costo que UMAP minimiza est diseada para preservar la proximidad de los puntos cercanos en el espacio original y mantener una estructura global coherente.

3. **Iteraciones**: El algoritmo de optimizacin itera hasta que la representacin en el espacio de baja dimensionalidad converge a una configuracin que satisface los criterios de proximidad y simplicidad local.

<h3 id="subsection1-1" style="color: #003366;">Interpretacin y Visualizacin</h3>

El resultado final de UMAP es una representacin de los datos en un espacio de baja dimensionalidad que puede ser visualizada y analizada fcilmente. Esta representacin preserva la estructura local de los datos, lo que facilita la identificacin de patrones, clsteres y relaciones en los datos originales.

<h3 id="subsection1-1" style="color: #003366;"> Ventajas y Limitaciones</h3>

**Ventajas:**

- **Eficiencia**: UMAP es muy eficiente en trminos de tiempo de computacin, lo que lo hace adecuado para grandes conjuntos de datos.
- **Preservacin de la Estructura Local**: UMAP es efectivo para mantener la proximidad de puntos cercanos, lo que es til para tareas de clustering y visualizacin.
- **Flexibilidad**: Permite ajustar el parmetro \( k \) para controlar el nivel de detalle en la estructura local.

**Limitaciones:**

- **Sensibilidad a Parmetros**: La eleccin del nmero de vecinos \( k \) y otros parmetros puede afectar significativamente los resultados.
- **Estructura Global**: Aunque UMAP preserva bien la estructura local, puede no ser tan efectivo en mantener la estructura global de los datos en algunos casos.

En resumen, UMAP es una tcnica poderosa y flexible para la reduccin de dimensionalidad que se basa en principios de topologa y teora de grafos para preservar la estructura local de los datos. Su eficiencia y capacidad para manejar grandes volmenes de datos lo hacen una herramienta valiosa en el anlisis y visualizacin de datos.

<h3 id="subsection1-1" style="color: #003366;"> Ejemplo 1: datos iris</h3>

````python
import umap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
````

````python
# Cargar el conjunto de datos Iris
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
````

````python
# Normalizar los datos
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df.iloc[:, :-1])
````

````python
# Configurar y aplicar UMAP
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=2, random_state=42)
embedding = reducer.fit_transform(df_scaled)

# Convertir el resultado a un DataFrame
embedding_df = pd.DataFrame(embedding, columns=['UMAP1', 'UMAP2'])
embedding_df['species'] = df['species']
````

````python
# Visualizar el resultado usando Seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(x='UMAP1', y='UMAP2', hue='species', palette='viridis', data=embedding_df)
plt.title('UMAP projection of the Iris dataset')
plt.show()
````

<h3 id="subsection1-1" style="color: #003366;"> Ejemplo 3: datos de dgitos</h3>

````python
from sklearn.datasets import fetch_openml
from sklearn.decomposition import PCA
````

````python
# Cargar el conjunto de datos MNIST
mnist = fetch_openml('mnist_784', version=1)
X, y = mnist.data, mnist.target

# Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reducir la dimensionalidad inicial con PCA para acelerar UMAP
pca = PCA(n_components=50)
X_pca = pca.fit_transform(X_scaled)
````

````python
# Configurar y aplicar UMAP
reducer = umap.UMAP(n_neighbors=15, min_dist=0.1, n_components=2, random_state=42)
embedding = reducer.fit_transform(X_pca)

# Convertir el resultado a un DataFrame
embedding_df = pd.DataFrame(embedding, columns=['UMAP1', 'UMAP2'])
embedding_df['digit'] = y
````

````python
# Visualizar el resultado usando Seaborn
plt.figure(figsize=(12, 10))
sns.scatterplot(x='UMAP1', y='UMAP2', hue='digit', palette='tab10', data=embedding_df, legend='full', alpha=0.3)
plt.title('UMAP projection of the MNIST dataset')
plt.show()
````

<h2 style="color: #007ACC;"> Parmetros bsicos de UMAP </h2>

1.	n_neighbors: Este parmetro determina el nmero de vecinos ms cercanos que se considerarn para cada punto al construir el grafo de vecindad. Afecta la forma en que UMAP equilibra la preservacin de la estructura local versus la estructura global de los datos. Un valor bajo de n_neighbors hace que UMAP capture ms detalles locales, mientras que un valor alto captura mejor la estructura global.
  
2.	min_dist: Este parmetro controla la distancia mnima entre los puntos en el espacio reducido. Afecta la dispersin de los datos en el espacio de baja dimensionalidad. Un valor bajo de min_dist mantiene los puntos ms cercanos entre s, lo que resulta en una mayor compactacin de los clsteres. Un valor alto separa ms los puntos, mostrando una estructura ms extendida.

3.	n_components: Define el nmero de dimensiones en las que se proyectarn los datos. Comnmente, se elige 2 o 3 para facilitar la visualizacin, pero se pueden utilizar valores ms altos para otras aplicaciones.

4.	metric: La mtrica de distancia utilizada para calcular las distancias entre los puntos en el espacio de alta dimensionalidad. Las mtricas comunes incluyen euclidean, manhattan, cosine, entre otras. La eleccin de la mtrica puede afectar significativamente los resultados.

5.	random_state: Una semilla para el generador de nmeros aleatorios que se utiliza para la inicializacin y para garantizar la reproducibilidad de los resultados.

6.	transform_seed: Controla la semilla para el mtodo de transformacin de nuevos puntos. Asegura que las transformaciones de datos nuevos sean consistentes con la transformacin original.

````python
# Cargar el conjunto de datos MNIST
mnist = fetch_openml('mnist_784', version=1)
X, y = mnist.data, mnist.target

# Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reducir la dimensionalidad inicial con PCA para acelerar UMAP
pca = PCA(n_components=50)
X_pca = pca.fit_transform(X_scaled)
````

````python
# Configurar y aplicar UMAP con diferentes valores de n_neighbors y min_dist
reducer_1 = umap.UMAP(n_neighbors=5, min_dist=0.1, n_components=2, random_state=42)
embedding_1 = reducer_1.fit_transform(X_pca)

reducer_2 = umap.UMAP(n_neighbors=30, min_dist=0.3, n_components=2, random_state=42)
embedding_2 = reducer_2.fit_transform(X_pca)
````

````python
# Convertir los resultados a DataFrames
embedding_df_1 = pd.DataFrame(embedding_1, columns=['UMAP1', 'UMAP2'])
embedding_df_1['digit'] = y

embedding_df_2 = pd.DataFrame(embedding_2, columns=['UMAP1', 'UMAP2'])
embedding_df_2['digit'] = y

# Visualizar los resultados usando Seaborn
plt.figure(figsize=(12, 10))

plt.subplot(1, 2, 1)
sns.scatterplot(x='UMAP1', y='UMAP2', hue='digit', palette='tab10', data=embedding_df_1, legend='full', alpha=0.3)
plt.title('UMAP with n_neighbors=5, min_dist=0.1')

plt.subplot(1, 2, 2)
sns.scatterplot(x='UMAP1', y='UMAP2', hue='digit', palette='tab10', data=embedding_df_2, legend='full', alpha=0.3)
plt.title('UMAP with n_neighbors=30, min_dist=0.3')

plt.show()
````
