Conceptos Básicos de Estadística Inferencial

---

La **estadística inferencial** es una rama de la estadística que se enfoca en hacer inferencias, predicciones o generalizaciones sobre una población a partir de una muestra de datos. A diferencia de la estadística descriptiva, que se limita a describir y resumir los datos disponibles, la estadística inferencial va un paso más allá, utilizando los datos de la muestra para hacer afirmaciones o tomar decisiones sobre la población de la que provienen esos datos.

La estadística inferencial es esencial en la investigación científica, ya que permite a los investigadores y profesionales tomar decisiones informadas basadas en datos. En lugar de limitarse a describir lo que ha ocurrido, la estadística inferencial permite predecir lo que podría ocurrir en el futuro y evaluar la validez de las afirmaciones sobre una población más amplia. Esto tiene aplicaciones en diversas áreas, incluyendo la medicina, la psicología, la economía, las ciencias sociales, la ingeniería, y muchas más.

Por ejemplo, en la investigación médica, la estadística inferencial es vital para determinar si un nuevo tratamiento es efectivo y si los resultados obtenidos en un grupo de pacientes pueden generalizarse a toda la población. En la industria, se utiliza para controlar la calidad, optimizar procesos y predecir la demanda de productos.

Dentro de los objetivos o tareas de la estadística inferencial encontramos:

1. **Estimación de Parámetros:** Uno de los objetivos principales de la estadística inferencial es estimar los parámetros desconocidos de una población, como la media, la varianza o la proporción. Estas estimaciones se realizan a partir de los estadísticos calculados en una muestra. Por ejemplo, si se desea conocer la media de una población, pero solo se tiene acceso a una muestra, se utiliza la media muestral como una estimación de la media poblacional.

2. **Pruebas de Hipótesis:** Otro objetivo clave es la realización de pruebas de hipótesis, que permiten a los investigadores evaluar si existe suficiente evidencia en los datos muestrales para aceptar o rechazar una hipótesis sobre la población. Por ejemplo, si un científico quiere saber si un nuevo medicamento es más efectivo que un tratamiento estándar, puede usar una prueba de hipótesis para determinar si los resultados observados en un grupo de pacientes son lo suficientemente convincentes como para generalizar esa conclusión a todos los pacientes.

3. **Predicción:** La estadística inferencial también se utiliza para hacer predicciones sobre futuros eventos o comportamientos basados en los datos actuales. Por ejemplo, un modelo de regresión puede predecir el rendimiento académico de un estudiante en función de su rendimiento pasado y otras variables relacionadas.

4. **Generalización:** A partir de una muestra representativa, se pueden hacer generalizaciones sobre una población entera. Esto es crucial en muchas áreas de la ciencia, la economía, la medicina, y otros campos donde no es práctico o posible observar a toda la población.

### **Fundamentos de la Estadística Inferencial:**

A lo largo del módulo, nos oscuparemos de diferenetes aspectos de la estadística inferencial que nos capacitarán con las herraientas básicas necesarias para llevar a cabo estimaciones de parámetros, pruebas de hipótesis, predicciones y generalizaciones de una población a partir de muestras. Así pues, para poder comenzar es crucial conocer los siguientes conceptos:

- **Muestras y Poblaciones:** La estadística inferencial se basa en la relación entre una muestra y su población correspondiente. Una población es el conjunto completo de elementos o individuos que son objeto de estudio, mientras que una muestra es un subconjunto representativo de esa población. La clave para la inferencia es que *la muestra debe ser representativa de la población para que las conclusiones sean válidas*.

- **Sesgo y Precisión:** El sesgo se refiere a *cualquier error sistemático que puede distorsionar los resultados*. Por ejemplo, si una muestra no es representativa de la población (sesgo de selección), las inferencias hechas a partir de esa muestra pueden ser incorrectas. La precisión se refiere a *la consistencia o reproducibilidad de los resultados de la inferencia*. Se mejora aumentando el tamaño de la muestra y minimizando las fuentes de error.

- **Error Muestral:** Cuando se estima un parámetro poblacional usando un estadístico muestral, siempre habrá una diferencia entre el valor estimado y el valor verdadero del parámetro, conocida como error muestral. Este error es inevitable, pero se puede medir y manejar dentro de un rango aceptable usando el concepto de intervalos de confianza.

### **Tipos de Estadística Inferencial:**

Los tipos de estadística inferencial son:

1. **Inferencia Paramétrica:** Supone que los datos provienen de una población que sigue una distribución conocida (como la distribución normal) y utiliza esta suposición para hacer inferencias. Ejemplos incluyen la estimación de la media poblacional y la realización de pruebas de hipótesis para medias y proporciones.

2. **Inferencia No Paramétrica:** No requiere suposiciones estrictas sobre la forma de la distribución de la población y es más flexible, pero generalmente menos poderosa que la inferencia paramétrica. Ejemplos incluyen la prueba de rangos con signo de Wilcoxon y la prueba de chi-cuadrado.

3. **Métodos Bayesianos:** Esta es una forma de inferencia que combina datos muestrales con información previa (a priori) usando el teorema de Bayes. Es particularmente útil cuando se dispone de información previa sólida o cuando se realizan inferencias en situaciones complejas o con datos limitados.

### **Relación entre la Estadística Descriptiva y la Teoría de la Probabilidad**

La estadística inferencial y la teoría de la probabilidad están íntimamente relacionadas, ya que la estadística inferencial se basa en los principios de la probabilidad para realizar inferencias y tomar decisiones sobre una población a partir de una muestra. A continuación, se describe la relación entre estas dos disciplinas:

- **1. Fundamento Teórico:** La teoría de la probabilidad proporciona el marco matemático que permite modelar la incertidumbre y el azar, conceptos esenciales en la estadística inferencial. En esencia, la probabilidad cuantifica la incertidumbre asociada a los resultados de eventos aleatorios. La estadística inferencial, por su parte, *utiliza esta teoría para hacer inferencias sobre una población a partir de una muestra de datos y los expresa como probabilidades que dan cuenca de la certidumbre de la estimación o hipótesis*.

- **2. Distribuciones Muestrales:** La estadística inferencial utiliza distribuciones muestrales, que son distribuciones de probabilidad de estadísticos (como la media o la proporción) obtenidos a partir de múltiples muestras de una población. Estas distribuciones muestrales son fundamentales para realizar estimaciones y pruebas de hipótesis ya que modelan el comportamiento del fenómeno sobre el cual se pretende inferir.

- **3. Estimación e Intervalos de Confianza:** Los intervalos de confianza, una herramienta inferencial importante los cuáles están construidos utilizando distribuciones de probabilidad. Un intervalo de confianza para una media, por ejemplo, se basa en la distribución de probabilidad de la media muestral, que es derivada de la teoría de la probabilidad. La probabilidad juega un papel crucial en la interpretación de estos intervalos, que indican el *rango de valores en el que, con un cierto nivel de confianza (por ejemplo, 95%), se espera que se encuentre el parámetro poblacional*.

- **4. Pruebas de Hipótesis:** Al formular una hipótesis nula y una hipótesis alternativa, *se utiliza la probabilidad para determinar la probabilidad de observar un resultado tan extremo como el observado, o más extremo, bajo la hipótesis nula*. El p-valor, una medida crucial en las pruebas de hipótesis, es una probabilidad calculada utilizando la distribución de probabilidad del estadístico de prueba.

- **5. Regresión y Modelos Probabilísticos:** La regresión y otros modelos estadísticos son construidos bajo supuestos probabilísticos. Por ejemplo, en la regresión lineal simple, se asume que el término de error tiene una distribución normal con media cero y varianza constante. Este supuesto probabilístico es fundamental para hacer inferencias sobre los coeficientes de regresión y para construir intervalos de confianza y realizar pruebas de hipótesis sobre estos coeficientes.


### **Conceptos Fundamentales de la Estadística Inferencial**

- **Población y Muestra**: 
  - **Población**: Conjunto completo de todos los elementos que se desea estudiar.
  - **Muestra**: Subconjunto de la población seleccionado para el estudio.

- **Parámetros y Estadísticos**:
  - **Parámetro**: Medida descriptiva calculada a partir de toda la población (por ejemplo, la media poblacional).
  - **Estadístico**: Medida descriptiva calculada a partir de una muestra (por ejemplo, la media muestral).

- **Distribución de Frecuencia y de Probabilidad**: Organización de datos que muestra la frecuencia (número de veces) de ocurrencia de cada valor o rango de valores en un conjunto de datos. Por otra parte, la distribución de probabilidad siendo en el caso discreto, la versión normalizada de la distribución de frecuencia y en el caso contínuo, la función de densidad normalizada de probabilidad, describe cómo se distribuyen los valores de una variable aleatoria de manera que asigna probabilidades de ocurrencia a cada valore de la variable dependiendo de su presencia en el espacio muestral.

- **Distribución Muestral:** Es la distribución de un estadístico como la media o la proporción, obteniendo múltiples muestras de una población. Es una forma de distribución que recopila la información extraída a partir de las muestras de una población empleada para efectuar inferencias y tomar desiciones sobre la población de orígen.

- **Medidas de Resumen**: Métricas numéricas que compactan propiedades de la distribución de frecuencia y evalúan diferentes aspectos de la mísma.
  - **Medidas de Tendencia Central**: Valores que representan un punto central o típico en un conjunto de datos. Estas medidas ayudan a identificar la ubicación central de los datos y proporcionan una indicación del "centro" de la distribución. Entre ellas encontramos la Media, mediana y moda.
  - **Medidas de Dispersión**: Grado de variabilidad o dispersión de los datos en relación con una medida de tendencia central. Estas medidas ayudan a entender la dispersión de los datos y a identificar la consistencia de los mismos. Entre ellas encontramos el Rango, varianza y desviación estándar.
  - **Medidas de Posición Relativa**: Las medidas de posición relativa indican la posición de un valor en relación con el resto del conjunto de datos. Estas medidas incluyen percentiles, cuartiles y deciles.
  - **Medidas de Forma**: Las medidas de forma describen la forma de la distribución de los datos mediante aspectos tales como la Asimetría y curtosis.

- **Estimación:** Corresponde al proceso de inferir el valor de un parámetro poblacional a partir de un estadístico muestral. Exísten dos tipos de estimación:
  - **Estimación puntual:** Proporciona el único valor como estimación del parámetro (por ejemplo, la media meustral como estimación de la media poblacional).

  - **Estimación por Intervalo:** Proporciona un rango de valores dentro del cual se espera que se encuentre el parámetro poblacinal, con un cierto nivel de confianza.

- **Pruebas de Hipótesis:** Las pruebas de hipótesis son procedimientos utilizados para tomar desiciones sobre los parámetros poblacionales basados en datos muestrales (o desiciones sobre la población mísma). Implica la formulación de dos hipótesis:
  - **Hipótesis nula ($H_0$):** Es la hipótesis que se asume como **verdadera** a menos que la evidencia en contra de ella sea fuerte.
  - **Hipótesis alternativa ($H_1$):** Es la **hipótesis que se desea probar** y que se acepta si la hipótesis es rechazada.
  - **p-valor:** Es la probabilidad de obtener un resultado **tan extremo o más extremo** si la hipótesis nula es verdadera. Un p-valor bajo indica que la hipótesis nula es improbable.
  - **Errores Tipo I y Tipo II:** Un error Tipo I ocurre cuando se rechaza incorrectamente la hipótesis nula, mientras que un error Tipo II ocurre cuando no se rechaza la hipótesis nula cuando esta es falsa.

- **Pruebas**:
  - **Pruebas Paramétricas**: Son pruebas de hipótesis que suponen que los datos provienen de una población que sigue una distribución de probabilidad conocida (como la distribución normal).
  - **Pruebas No Paramétricas:** No requieren suposiciones estrictas sobre la distribución de la población y se utilizan cuando las condiciones para las pruebas paramétricas no se cumplen.

- **Inferencia Bayesiana:** La inferencia bayesiana es un enfoque de la estadística inferencial que incorpora la probabilidad previa sobre un parámetro junto con los datos muestrales para actualizar las creencias sobre ese parámetro. Utiliza el Teorema de Bayes para combinar información previa y la probabilidad de la muestra.

---
Mecanísmos de Muestreo

---

Los mecanísmos de muestreo son aquellas maneras en las que se toman muestras de cierto tamaño a partir de una población mediante un enfoque o estrategia de extracción de datos. Así pues, encontramos tres tipos de mecanismos:

1. **Muestreo Aleatorio:** Corresponde a la extracción de $m$ muéstras a partir de una población de tamaño $n$ de manera completamente aleatoria tal que $m<<n$.
2. **Muestreo Sistemático:** Método de muestreo probabilístico en el que se seleccionan elementos de una población a intervalos regulares. Consiste en seleccionar cada $k$ elementos de la población, siendo $k$ una constante predeterminada.
3. **Muestreo Estratificado:** Método de muestreo probabilístico que consiste en dividir la población en distintos subgrupos o estratos en función de determinadas características. El objetivo es garantizar que cada subgrupo esté bien representado en la muestra para obtener estimaciones más exactas y precisas de los parámetros de la población. El muestreo estratificado es ventajoso porque garantiza que cada subgrupo o estrato esté adecuadamente representado en la muestra, lo que permite obtener estimaciones más precisas de los parámetros de población de cada subgrupo. Este método es especialmente útil cuando existen diferencias o variaciones sustanciales entre los subgrupos de la población. Ayuda a reducir la variabilidad del muestreo y puede proporcionar resultados más fiables y precisos en comparación con otras técnicas de muestreo.


Para ilustrar los conceptos de técnicas de muestreo junto con otros conceptos de estadística inferencial desarrollados a lo largo del presente curso, cargaremos un conjunto de datos reales a continuación donde las celdas siguientes ejecutan respectivamente los siguientes procedimientos:

```python
# Localización del archivo de datos
current_dir = os.getcwd()
data_dir = os.path.join(current_dir, 'data', 'tourisnm_engagement.csv')
# Cargamos el archivo de datos
data_raw = pd.read_csv(data_dir)
df_turismo = data_raw.copy()
```

```python
# Definición de mapeos para 'preferred_device'

device_mapping= {'iOS and Android': 'Mobile',
                'iOS' : 'Mobile',
                'ANDROID' : 'Mobile',
                'Android': 'Mobile',
                'Android OS' : 'Mobile',
                'Other': 'Mobile',
                'Others' : 'Mobile',
                'Tab' : 'Mobile'}

df_turismo['preferred_device'] = df_turismo['preferred_device'].replace(device_mapping)

# Definición de mapeos para 'yearly_avg_Outstation_checkins'
df_turismo['yearly_avg_Outstation_checkins']= df_turismo['yearly_avg_Outstation_checkins'].replace('*',np.nan)
df_turismo['yearly_avg_Outstation_checkins'] = pd.to_numeric(df_turismo['yearly_avg_Outstation_checkins'], errors='coerce', downcast='integer')
df_turismo['yearly_avg_Outstation_checkins'].unique()

# Definición de mapeos para 'following_company_page'
page_mapping= {'1': 'Yes',
               '0' : 'No',
               'Yeso' : 'Yes',
               1: 'Yes',
               0: 'No'}
df_turismo['following_company_page'] = df_turismo['following_company_page'].replace(page_mapping)

# Definición de mapeos para 'member_in_family'
df_turismo['member_in_family']= df_turismo['member_in_family'].replace('Three',3)
df_turismo['member_in_family'] = pd.to_numeric(df_turismo['member_in_family'], errors='coerce', downcast='integer')

# Definición de mapeos para 'working_flag'
df_turismo['working_flag']= df_turismo['working_flag'].replace('0','No')
```

```python
# Remplazamos los posibles strings 'nan' y 'NaN' por el valor ausente que pandas reconoce por defecto el cual es np.nan
df_turismo = df_turismo.applymap(lambda x: np.nan if x == 'nan' else x)
df_turismo = df_turismo.applymap(lambda x: np.nan if x == 'NaN' else x)

# Imputación para variables categóricas
cat_columns = df_turismo.select_dtypes(exclude=['int64', 'float64']).columns
for i in cat_columns:
    df_turismo[i].fillna(df_turismo[i].mode()[0], inplace = True)
    
# Imputación para variables numéricas
num_columns = df_turismo.select_dtypes(exclude=['object']).columns
for column in num_columns:
    median_value = df_turismo[column].median()
    df_turismo[column].fillna(median_value, inplace=True)
```

En un panorama diferente y más relacicionado y dependiente de la pregunta de investigación efectuada sobre la población, el **muestreo estratificado** es una técnica en la que la población se divide en subgrupos o estratos que son mutuamente excluyentes y colectivamente exhaustivos. Una vez dividia la población en grupos de interés, se toma una muestra aleatoria en cada estrato. Para realizar un muestreo estratificado se siguen los siguientes pasos:

1.   Identifique la población objetivo: Defina la población de interés de la que desea extraer una muestra.
2.   Determine los factores de estratificación pertinentes: Identifique las características o variables clave que son importantes para dividir la población en subgrupos o estratos homogéneos. Estas variables deben estar relacionadas con la pregunta de investigación o el propósito del estudio.
3.   Divida la población en estratos: Divida la población en estratos mutuamente excluyentes y colectivamente exhaustivos en función de los factores de estratificación identificados. Cada elemento de la población debe pertenecer a un solo estrato.
4.   Determine el tamaño de la muestra: Decida el tamaño de muestra deseado para cada estrato. El tamaño de la muestra puede ser proporcional al tamaño de cada estrato o puede asignarse en función de la variabilidad dentro de cada estrato.
5.   Seleccione aleatoriamente muestras de cada estrato: Seleccione de forma independiente y aleatoria muestras de cada estrato utilizando cualquier técnica de muestreo apropiada (como el muestreo aleatorio simple o el muestreo sistemático).
6.   Combine las muestras de los estratos: Una vez obtenidas las muestras de cada estrato, combínelas para formar la muestra estratificada final.

Por ejemplo, dentro de nuestro conjunto de datos planeamos estratificar dependiendo de la doble agrupación categórica hecha con base a la variable "taken_product" y "prefered_device". A continuación, se define una función de muestreo estratificado donde se especifican las columnas bajo las cuáles se establece la estratificación, los valores de estratificación y las proporciones requeridas para cada valor de estratificación.

```python
def stratified_data(data, columnas_estratificacion, valores_estrat, prop_estrat, random_state=None):
    
    df_estrat = pd.DataFrame(columns = data.columns) # Creamos un data frame vacío con los nombres de las columnas del dataframe
    # Creamos la columna estratificada
    data['estratificado'] = data[columnas_estratificacion].astype(str).agg('_'.join, axis=1)
    
    pos = -1
    for i in range(len(valores_estrat)): # iteración sobre los valores estratificados
        pos += 1
        if pos == len(valores_estrat) - 1: 
            ratio_len = len(data) - len(df_estrat) # si es la iteración final calcula el número de valores de salida tenga el mismo número de filas que de entrada
        else:
            ratio_len = int(len(data) * prop_estrat[i]) # calcula el número de filas según la proporción deseada

        df_filtrado = data[data['estratificado'] ==valores_estrat[i]] # filtra los datos de origen según los valores seleccionados en la estratificación de datos
        df_temp = df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state) # haz un sample de los datos filtrados usando la ratio que hemos calculado
        
        df_estrat = pd.concat([df_estrat, df_temp]) # junta las tablas de sample con la estratificada para producir el resultado final
        
    return df_estrat # Return the stratified, re-sampled data   
```

Supongamos ahora, que deseamos generar una muestra de nuestra población que de acuerdo al criterio de estratificación "take_product-prefered_device", necesitamos que nuestra muestra tenga las siguientes proporciones para cada combinación categórica:

- No_Mobile = 0.8
- Yes_Mobile = 0.14
- No_Laptop = 0.05
- Yes_Laptop = 0.01

```python
# Generamos un muestreo estratificado con proporciones dadas
df_estrat = df_turismo.copy()
cat_cols = df_estrat.select_dtypes(exclude=[int,float]).columns
columnas_estratificacion = [cat_cols[0], cat_cols[1]]
proporciones = {'No_Mobile': 0.8,
                'Yes_Mobile':0.14,
                'No_Laptop': 0.05,
                'Yes_Laptop': 0.01}

# Generamos el muestreo estratificado
df_estrat = stratified_data(df_estrat, columnas_estratificacion, list(proporciones.keys()), list(proporciones.values()), random_state=42)

# Observamos el las proporciones de los datos resultantes del sampleo estartificado
df_estrat['estratificado'].value_counts()/len(df_estrat)
```

Ley de Grandes Números

---

La Ley de los Grandes Números (LGN) es un principio fundamental en la teoría de la probabilidad y la estadística que describe el comportamiento del promedio de una gran cantidad de observaciones independientes e idénticamente distribuidas. Según esta ley, a medida que el tamaño de la muestra aumenta, la media aritmética de las observaciones se aproxima cada vez más al valor esperado o promedio de la población de la que se toma la muestra.

En otras palabras, cuanto mayor sea el número de repeticiones de un experimento aleatorio, más cerca estará la media muestral de la media poblacional verdadera.

### **Tipos de la Ley de los Grandes Números**

Hay dos versiones principales de la Ley de los Grandes Números:

1. **Ley Débil de los Grandes Números (WDLL):**
   - Establece que, para un tamaño de muestra suficientemente grande, la media muestral converge en probabilidad al valor esperado de la variable aleatoria. Esto significa que la probabilidad de que la media muestral difiera del valor esperado por más de un margen pequeño tiende a cero a medida que el tamaño de la muestra aumenta.
   
   - Formalmente, si $X_1, X_2, \dots, X_n$ son variables aleatorias independientes e idénticamente distribuidas con valor esperado $E(X_i) = \mu$, entonces:
     $$
     \lim_{n \to \infty} P\left(\left|\frac{1}{n} \sum_{i=1}^{n} X_i - \mu\right| > \epsilon\right) = 0 \text{ para cualquier } \epsilon > 0.
     $$

2. **Ley Fuerte de los Grandes Números (SLLN):**
   - Establece que la media muestral converge casi seguramente (o con probabilidad 1) al valor esperado de la variable aleatoria a medida que el tamaño de la muestra tiende a infinito. Esto es un resultado más fuerte que la ley débil porque implica que la convergencia no solo es probable, sino que ocurrirá con certeza en casi todos los casos.
   
   - Formalmente, si $X_1, X_2, \dots, X_n$ son variables aleatorias independientes e idénticamente distribuidas con valor esperado $E(X_i) = \mu$, entonces:
     $$
     P\left(\lim_{n \to \infty} \frac{1}{n} \sum_{i=1}^{n} X_i = \mu\right) = 1.
     $$

### **Propiedades de la Ley de los Grandes Números**

Para que la ley de los grandes números sea válida, debe satisfacerse:

1. **Independencia muestral:** Las muestras generadas a partir de una población deben ser tomadas aleatoriamente.
2. **Significancia de tamaño:** El tamaño de la muestra debe ser estadísticamente significativa para dar razón del parámetro a estimar.
3. **Independencia observacional:** Cada observación de la población debe ser independiente de las otras.

Algúnas de las propiedades de la ley de grandes números son:

1. **Convergencia al Valor Esperado:** La LGN asegura que la media muestral converge al valor esperado de la distribución subyacente cuando el tamaño de la muestra es suficientemente grande.

2. **Independencia y Distribución Idéntica:** La LGN requiere que las *variables aleatorias sean independientes y sigan la misma distribución*. Si estas condiciones no se cumplen, la ley puede no ser aplicable.

3. **Tamaño de Muestra:** Aunque la LGN garantiza la convergencia, no especifica la rapidez con que esto ocurre. En la práctica, el tamaño de muestra necesario para que la media muestral sea un buen estimador de la media poblacional puede variar según la distribución de los datos.

4. **Aplicabilidad General:** La LGN es aplicable a una amplia gama de distribuciones, incluyendo aquellas con una varianza infinita, siempre que el valor esperado exista y sea finito.

Teorema del Limite Central

---

El **Teorema del Límite Central (TLC)** es uno de los resultados más fundamentales en estadística y teoría de probabilidades. Este teorema establece que, bajo ciertas condiciones, la distribución de la suma (o media) de un gran número de variables aleatorias independientes e identicamente distribuidas (i.i.d.) tiende a ser aproximadamente normal, independientemente de la distribución original de las variables.

Si $X_1, X_2, ..., X_n$ son variables aleatorias independientes e identicamente distribuidas con media $\mu$ y varianza $\sigma^2$, entonces la distribución de la suma $S_n = X_1 + X_2 + ... + X_n$ tiende a una distribución normal con media $N\mu$ y varianza $N\sigma^2$ cuando $N$ es grande.

Para la media muestral $\bar{X} = \frac{S_n}{n} $, la distribución se aproxima a una distribución normal con media $\mu$ y varianza $\frac{\sigma^2}{n}$. con esto en mente, se tienen las siguientes propiedades:

1. **Convergencia a la Normalidad**: Independientemente de la forma de la distribución original, la distribución de la media muestral se aproximará a una normal si el tamaño de la muestra es suficientemente grande.
2. **Tamaño de la Muestra**: Aunque el TLC se aplica a muestras grandes, en la práctica, una muestra de tamaño 30, 50 o más suele ser suficiente para que la distribución de la media muestral se aproxime a una normal.
3. **Aplicación Práctica**: El TLC permite hacer inferencias sobre la media de poblaciones grandes, incluso cuando la distribución de la población es desconocida.


El ejemplo más simple del teorema del límite central, consiste en generar un numero $N$ determinado de muestras de tamañp $n$ creadas a partir del muestro uniforme y aleatorio para las cuales calculamos la media. Cuando graficamos la distribución de éstas medias obtenemos una distribución normal.

```python
# Parámetros
n_samples = 1000  # Número de muestras
sample_size = 30  # Tamaño de cada muestra

# Simulamos n_samples de medias muestrales de tamaño sample_size
means = []
for _ in range(n_samples):
    sample = np.random.uniform(low=0, high=1, size=sample_size)
    means.append(np.mean(sample))

# Graficamos el histograma de las medias muestrales
graficar_media_muestal(means)
```

Otro ejemplo del TLC, corresponde a la distribución Binomial de $N$ experimentos de Bernoulli con probabilidad de exito $p$. En este escenario el tamaño de la muestra $n$ al aumentar, a pesar de que la distribución Binomial modela las probabilidades para la variable aleatoria $X$ = "Número de caras detectadas en 10 lanzamientos consecutivos", cuando el tamaño de las muestras $n$ aumenta, la distribución Binomial empieza a adquirir una forma normal.

```python
# Definir los posibles resultados de un lanzamiento de moneda
outcomes = ['H', 'T']  # H para "Cara" (Heads), T para "Cruz" (Tails)

# Generar el espacio muestral para tres lanzamientos
sample_space = list(product(outcomes, repeat=10))

# Definir la variable aleatoria X como el número de "caras" (H) en tres lanzamientos
def count_heads(trial):
    return trial.count('H')

# Calcular la función de probabilidad (PMF)
results = [count_heads(trial) for trial in sample_space]
mu, std = np.round(np.mean(results),3), np.round(np.std(results),3)
pmf = Counter(results)

# Normalizar las frecuencias para obtener probabilidades
total_outcomes = len(sample_space)
for x in pmf:
    pmf[x] /= total_outcomes

# Generamos el gráfico de la distribución PMF
print('Valor esperado np = ', {10 * 0.5})
print('Varianza np(1-p) = ', {10 * 0.5*(1-0.5)})
graficacion_TLC(pmf, mu, std)
```

Intervalos de Confianza

---

Un **intervalo de confianza** es un rango de valores que se utiliza para *estimar un parámetro poblacional desconocido con un nivel de certeza especificado*. En otras palabras, proporciona un rango dentro del cual es probable que se encuentre el valor verdadero del parámetro basado en una muestra de datos. Nos permiten:

1. **Estimación de Parámetros**: Los intervalos de confianza permiten establecer el rango de posibles valores de parámetros poblacionales, como la media o la proporción, con un grado de certeza. En lugar de proporcionar un único valor estimado, ofrecen un rango que refleja la incertidumbre de la estimación.

2. **Toma de Decisiones**: Son útiles en la toma de decisiones al proporcionar una medida de la precisión de las estimaciones. Por ejemplo, en un estudio clínico, un intervalo de confianza para la diferencia en eficacia entre dos tratamientos ayuda a determinar si la diferencia observada es estadísticamente significativa y relevante.

3. **Evaluación de Resultados**: Permiten evaluar la variabilidad y la precisión de los resultados obtenidos a partir de una muestra. 

4. **Comunicación de Resultados**: Facilitan la comunicación de la incertidumbre y la fiabilidad de las estimaciones a otros profesionales y al público en general.

Exísten diferentes enfoques para hacer el cálculo de intervalos de confianza, dentro de los cuáles encontramos:

1. **Métodos Basados en la Distribución Normal**: Cuando la muestra es grande o la distribución subyacente es normal, se puede usar la distribución normal estándar para calcular el intervalo de confianza. Se utilizan valores críticos de la distribución normal para determinar el rango.

2. **Métodos Basados en la Distribución $t$ de Student**: Para muestras pequeñas y cuando la desviación estándar poblacional no es conocida, se emplea la distribución $t$ de Student. La distribución $t$ es similar a la normal pero tiene colas más anchas, lo que permite captar la mayor incertidumbre en las estimaciones de pequeñas muestras.

3. **Métodos de Bootstrap**: Técnicas de remuestreo como el bootstrap pueden ser usadas para estimar intervalos de confianza sin asumir una distribución específica para los datos. El método bootstrap involucra la creación de múltiples muestras a partir de la muestra original y el cálculo del intervalo basado en estos datos simulados.

4. **Métodos No Paramétricos**: En casos donde no se puede asumir una distribución específica para los datos, se pueden utilizar métodos no paramétricos, como los intervalos de confianza de percentiles, para estimar la variabilidad del parámetro poblacional.

5. **Métodos Bayesianos**: En la inferencia bayesiana, los intervalos de confianza se basan en distribuciones a posteriori y se denominan intervalos de credibilidad. Estos intervalos reflejan la probabilidad de que el parámetro poblacional se encuentre dentro del intervalo dado los datos observados.

---
Distribución $Z$

---

La **distribución Z**, también conocida como **distribución normal estándar**, es una distribución normal con una media de 0 y una desviación estándar de 1. Es una forma de la distribución normal que se utiliza para *simplificar el análisis de datos y realizar comparaciones entre diferentes conjuntos de datos*. La distribución $Z$ se calcula estandarizando una variable aleatoria normal. Esto se hace transformando los datos de una distribución normal con media $\mu$ y desviación estándar $\sigma$ a una distribución normal estándar con media 0 y desviación estándar 1. La fórmula para convertir una puntuación $X$ en una puntuación $Z$ es:

$$
Z = \frac{X - \mu}{\sigma}
$$


### **Aplicaciones de la Distribución $Z$**

1. **Comparación de Datos**: Permite comparar datos provenientes de diferentes distribuciones normales al convertirlas a una escala común.

2. **Pruebas de Hipótesis**: Se utiliza para calcular valores críticos en pruebas de hipótesis, como la prueba $Z$ para medias y proporciones.

3. **Intervalos de Confianza**: Se emplea para calcular intervalos de confianza para la media cuando la desviación estándar de la población es conocida.

4. **Puntuaciones Estándar**: Permite determinar cuántas desviaciones estándar un dato está alejado de la media, lo que es útil en análisis estadísticos y en la detección de valores atípicos.


Para evidenciar la utilidad e importancia de la distribución $Z$, imaginemos ahora que nuestra variable  $X = \mu_{\text{sample}}$ corresponde al valor del parámetro poblacional que deseamos estimar que en éste caso,es la media de la población. Así, la distribución $Z$ se define como la distribución de probabilidad observacional dada para la estandarización de nuestras medias muestrales.

$$
Z = \frac{\mu_{\text{sample}} - \mu}{\sigma}
$$

Por el teorema del límite central, sabemos que $\mu_{\text{sample}}$ seguirá una distribución normal con media $\mu_{muestral}$ y desviación estándar $\sigma_{muestral}$ tal que esto es cierto cuando el número de muéstras $N$ tiende a infinito. En éste sentido, si quisieramos dar respuesta a la pregunta **¿cuál es el rango de valores de $\mu_{sample}$ tal que el $C$% de las estimaciones del parámetro poblacional $\mu$ se encuentren dentro de dicho rango?**.

Antes de responder ésta pregunta, recordemos que los *percentiles* son aquellos valores de nuestra variable aleatoria que nos permíten determinar un porcentaje de concentración de los valores de dicha variable. En este sentido, la pregunta ha de responderse determinando en efecto como *el percentil que representa nuestro  $C$ porcentaje de confidencia (acumulación de datos) teniendo en cuenta una distribución muestral*. De esta forma, si sabemos o al menos podemos asumir que la *distribución muestral del parámetro poblacional sigue una distribución normal a grandes números de muestras* esposible calcular el valor del percentíl a traves de la distribución $Z$ gracias a que los percentiles de una distribución normal como los de una distribución $Z$ son los mísmos.


#### **Efectos del Tamaño Muestral Sobre los IC**

Ya que la distribución muestral $Z$ puede ayudarnos a calcular los intervalos de confianca (IC), definimos ahora la  **desviación estandar de la distribución muestral** del parámetro poblacional como:

$$
\sigma_{\hat{x}} = \frac{\sigma}{\sqrt{n}}
$$

siendo $\sigma$ la varianza poblacional, $\hat{X}$ la media muestral (o parámetro poblacional a estiamar) y $n$ el tamaño de la muestra. Así pues, si $\sigma_{\hat{x}}$ representa la variabilidad en la estimación de nuestro parámetro, el efecto que el tamaño muestral es simplemente **estrechar la distribución muestral del parámetro poblacional**, lo que se traduce una mayor cantidad de estimaciones estarán centradas al rededor del valor real y la estimación será más precisa.

**<h3 id="subsection222" style="color: #003366;">2.2.2. Error Estándar y Cálculo de Intervalos de Confianza</h3>**

---
Error Estándar y sus Aplicaciones

---

El **Error Estándar** es un concepto fundamental en la estadística inferencial, y su comprensión es clave para interpretar correctamente los resultados de las estimaciones basadas en muestras. El Error Estándar es una medida de la **precisión de una estimación estadística**. Específicamente, cuando hablamos del **Error Estándar de la Media (EEM)**, nos referimos a la dispersión o variabilidad de las medias muestrales alrededor de la media verdadera de la población. En otras palabras, el Error Estándar cuantifica *cuánto varía la media de una muestra particular en relación con la media de la población de la que proviene*. El Error Estándar de la Media se relaciona con la Desviación Estándar de la siguiente manera:

$$
\text{EEM} = \frac{\sigma}{\sqrt{n}} 
$$

Donde $\sigma$ es la desviación estándar de la población y $n$ es el número de observaciones de la muestra poblacional. No obstante ya que la desviación estándar de la población rara vez se conoce, el error estándar de la media suele estimarse como la desviación estándar de la muestra dividida por la raiz cuadrada del tamaño de la muestra.

$$
\text{EEM} = \frac{\sigma_{\text{muestra}}}{\sqrt{n}} 
$$

El Error Estándar indica cómo varían las medias muestrales alrededor de la media poblacional (o un estimado de ella). Un EEM pequeño sugiere que la media muestral está muy cerca de la media poblacional, lo que implica que nuestra muestra es una buena representación de la población. Por el contrario, un EEM alto sugiere mayor variabilidad y menos precisión en nuestra estimación.

El **tamaño de la muestra  determina la precisión de la estimación** ya que a medida que aumenta el tamaño de la muestra, el Error Estándar disminuye, lo que significa que las medias muestrales se vuelven más cercanas a la media verdadera de la población. Esto se debe a que más datos proporcionan una estimación más precisa. El error estándar puede ser empleado para:

- **Construcción de Intervalos de Confianza:** Se utiliza para determinar cuán confiable es una estimación, ayudando a construir intervalos dentro de los cuales es probable que se encuentre el parámetro poblacional.

- **Pruebas de Hipótesis:** Se utiliza para evaluar la significancia estadística de los resultados de una muestra, comparando la media muestral con la media poblacional.


1.**Construcción Intervalos de Confianza**: Se utiliza para determinar cuán confiable es una estimación, ayudando a construir intervalos dentro de los cuales es probable que se encuentre el parámetro poblacional. El intervalo de confianza para la media poblacional, cuando se conoce el Error Estándar, se calcula de la siguiente manera:

$$
IC = \bar{x} \pm Z_{\alpha/2} \cdot \text{EEM} 
$$

Donde:
- $ \bar{x} $ es la media muestral.
- $ Z_{\alpha/2} $ es el valor crítico de la distribución normal estándar correspondiente al nivel de confianza deseado **(por ejemplo, 1.96 para un 95% de confianza)**.
- $ \text{EEM} $ es el Error Estándar de la Media.


2. **Pruebas de Hipótesis**: Las pruebas de hipótesis son un procedimiento estadístico para evaluar si una afirmación sobre un parámetro poblacional es compatible con los datos muestrales. El Error Estándar se utiliza para determinar si la diferencia observada es estadísticamente significativa.

   - **Formulación de Hipótesis**:
      - **Hipótesis nula $H_0$**: No hay efecto o diferencia $ \mu = \mu_0 $.
      - **Hipótesis alternativa $H_a$**: Hay un efecto o diferencia $ \mu \neq \mu_0$.

   - **Cálculo del Estadístico de Prueba**:
      El estadístico de prueba se basa en la diferencia entre la media muestral y el valor hipotético bajo $H_0$:
      $$
      Z = \frac{\bar{x} - \mu_0}{\text{EEM}}
      $$

   - **Determinación del P-valor**:
      - Se calcula el p-valor correspondiente al estadístico $Z$ obtenido.
      - Si el p-valor es menor que el nivel de significancia $\alpha$, se rechaza $H_0$.


Para empezar a ilustrar estos conceptos, primero generemos un experimento aleatorio donde supondrémos que nuestros datos siguen una distribución normal,luego, realizarémos un estimativo de la media muestral a partir de los datos del experimento y generarémos los intervalos de confianza primero a partir del EEM. El error estándar de la media (es decir, el error debido a la estimación de la media poblacional a partir de las medias muestrales) es la desviación estándar de todas las posibles muestras de un tamaño dado escogidos de esa población. Específicamente, *el error estándar de la media se refiere a la desviación estándar de la distribución de la media muestral*. En términos sencillos, indica cuánto se espera que la media muestral varíe de una muestra a otra si se tomaran múltiples muestras de la misma población.

De ésta manera, ya que exíste una equivalencia entre los *percentiles* de la distribución normal y la distribución estandar $Z$, sabémos que $Z_{\alpha/2}$ = 1.96 es el valor crítico de $Z$ tal que *contiene el 95% de ala acumulación de los datos **al rededor de la media de la distribución***. Así, los **márgenes de error** son los valores de la variable aleatoria (estimación del parámetro poblacional) que delimítan el intervalo de confianza tanto por izquierda como por derecha de la media de la estimación poblacional.

$$
ME = Z_{\alpha / 2} \frac{\sigma}{\sqrt{n}}
$$


$$
IC = \bar{x} \pm ME = \bar{x} \pm Z_{\alpha/2} \cdot \text{EEM} 
$$


```python
# Parámetros de la población
population_mean = 100   # Media poblacional
population_std = 15     # Desviación estándar poblacional
n_samples = 10000        # Número de muestras
sample_size = 50        # Tamaño de cada muestra

# Generación de múltiples muestras de la población
sample_means = []
for _ in range(n_samples):
    sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
    sample_means.append(np.mean(sample))

# Cálculo del error estándar de la media
eem = population_std / np.sqrt(sample_size)
eem_sample = np.std(sample_means, ddof=1) / np.sqrt(sample_size)
print(f"Error Estándar de la Media (EEM polacional): {eem:.2f}")
print(f"Error Estándar de la Media (EEM muestral): {eem_sample:.2f}")
print(f"Desviación Estándar Poblacional:", population_std)
print(f"Desviación Estándar Muestral: {np.std(sample_means, ddof=1):.2f}")
print(f"Media Muestral: {np.mean(sample_means):.2f}")
print(f"Media Poblacional:", population_mean)

# Visualización de la distribución de las medias muestrales

graficar_eem(sample_means,population_mean, eem_sample )
```

Para calcular intervalos de confianza para la estimación de un parámetro seguimos los siguientes pasos:

1. Estimamos el parámetro poblacional (como por ejemplo, la media muestral).
2. Calculamos el error estandar por medio de la población o de la muestra.
3. Especificamos un valor de confianza, el cuál es la probabilidad de que en el intervalo de confianza construido, exísta el valor verdadero poblacional. Comúnmente se utilizan los niveles de 90%, 95% o 99%.
4. Especificamos un valor  $Z_{\alpha}$ que refleje el nivel de confianza esperado, el cuál se obtiene a partir de la distribución normal estándar. Por ejemplo, para un 95% de confianza, $Z_{\alpha}$ es aproximadamente 1.96.
5. Calculamos los intervalos de confianza usando la relación: 

$$
IC = \bar{x} \pm Z_{\alpha/2} \cdot \text{EEM} 
$$

```python
# Parámetros de la población
population_mean = 100   # Media poblacional
population_std = 15     # Desviación estándar poblacional
n_samples = 10000        # Número de muestras
sample_size = 50        # Tamaño de cada muestra

# Generación de múltiples muestras de la población
sample_means = []
for _ in range(n_samples):
    sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
    sample_means.append(np.mean(sample))

# Estimación del parámetro poblacional
mean_sample = np.mean(sample_means)

# Cálculo del error estándar de la media (usamos eem poblacional para el IC)
eem = population_std / np.sqrt(sample_size)
eem_sample = np.std(sample_means)/np.sqrt(sample_size)

# Intervalo de confianza del 95%
confianza = 0.95
Z_critical = stats.norm.ppf(1 - (1 - confianza) / 2)  # Cálculo del valor crítico Z
margen_error = Z_critical * eem  # Usamos eem basado en la desviación estándar poblacional
intervalo_confianza = (mean_sample - margen_error, mean_sample + margen_error)

# Cálculo de ocnfianza al 95% mediante scipy.stats
intervalos_stats = stats.norm.interval(confidence=0.95, loc=mean_sample, scale=np.std(sample_means))

# Imprimimos los resultados
print(f"Media muestral: {mean_sample:.2f}")
print(f"Z critico: {Z_critical:.2f}")
print(f"Intervalo de confianza del 95%: {intervalo_confianza}")
print(f"Intervalo de confianza del 95% (stats): {intervalos_stats}")
print(f"Margen de error: {margen_error:.2f}")

# Visualización de la distribución de las medias muestrales
graficar_ID_normal(sample_means, population_mean, intervalo_confianza)
```

El **nivel de confianza** refleja el **grado de certeza asociado con un intervalo de confianza**. El nivel de confianza, a menudo expresado como $1 - \alpha$, representa la proporción de intervalos de confianza que, en un número infinito de muestras, contendrían el verdadero valor del parámetro poblacional. En otras palabras, es **la probabilidad de que el intervalo de confianza calculado a partir de una muestra incluya el valor verdadero del parámetro en cuestión**.

1. **Nivel de Confianza (1 - $α$)**: Si se construyen múltiples intervalos de confianza con un nivel de confianza del 95% ($α$ = 0.05), se espera que el 95% de estos intervalos contengan el valor verdadero del parámetro poblacional. Por lo tanto, un nivel de confianza del 95% implica que existe una probabilidad del 95% de que el intervalo de confianza calculado a partir de una muestra contenga el verdadero valor del parámetro.

2. **Error de Confianza o Significancia ($α$)**: El nivel de confianza es complementario al nivel de **significancia** ($α$), que es la probabilidad de que el intervalo de confianza **no** contenga el valor verdadero del parámetro. Por ejemplo, un nivel de confianza del 95% corresponde a un nivel de significancia del 5% ($α$ = 0.05), lo que significa que existe una probabilidad del 5% de que el intervalo no contenga el valor verdadero.



Con los anteriores conceptos en mente, para responder la pregunta *¿que tamaño muestral ha de tenerse para obtener un determinado margen de error?* empleamos la siguiente inigualdad.

$$
n \geq \left(\frac{Z_{\text{critico}}\sigma}{ME}\right)^2
$$

```python
# Calculo de tamaño muestral
margen_error_deseado = 3
population_std = 15
n = int((1.96 * population_std/margen_error_deseado)**2 )


# Parámetros de la población
population_mean = 100   # Media poblacional
population_std = 15     # Desviación estándar poblacional
n_samples = 10000       # Número de muestras
sample_size = n         # Tamaño de cada muestra

# Generación de múltiples muestras de la población
sample_means = []
for _ in range(n_samples):
    sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
    sample_means.append(np.mean(sample))

# Estimación del parámetro poblacional
mean_sample = np.mean(sample_means)

# Cálculo del error estándar de la media (usamos eem poblacional para el IC)
eem = population_std / np.sqrt(sample_size)
eem_sample = np.std(sample_means)/np.sqrt(sample_size)

# Intervalo de confianza del 95%
confianza = 0.95
Z_critical = stats.norm.ppf(1 - (1 - confianza) / 2)  # Cálculo del valor crítico Z
margen_error = Z_critical * eem  # Usamos eem basado en la desviación estándar poblacional
intervalo_confianza = (mean_sample - margen_error, mean_sample + margen_error)

# Cálculo de ocnfianza al 95% mediante scipy.stats
intervalos_stats = stats.norm.interval(confidence=0.95, loc=mean_sample, scale=np.std(sample_means))

# Imprimimos los resultados
print(f"Media muestral: {mean_sample:.2f}")
print(f"Z critico: {Z_critical:.2f}")
print(f"Intervalo de confianza del 95%: {intervalo_confianza}")
print(f"Intervalo de confianza del 95% (stats): {intervalos_stats}")
print(f"Margen de error: {margen_error:.2f}")

# Visualización de la distribución de las medias muestrales
graficar_ID_normal(sample_means, population_mean, intervalo_confianza)
```

Ahora, entendamos un poco más la diferencia entre **confianza** y **probabilidad**. Por ejemplo si estipulamos la confianza al 95% tenemos:

- (**correcto**) El intervalo de confianza contiene el verdadero parámetro poblacional aproximadamente el 95% de las veces.
- (**incorrecto**) Exíste un 95% de probabilidad que el parámetro poblacional caiga dentro de los intervalos de confianza.

La diferencia es algo sutil, no obstante considere lo siguiente: el parámetro poblacional a estimar **es fijo y no posee una distribución de probabilidad, solo es desconocido** y su relación con la distribución muestral es simplemente que *este valor puede o no estar dentro de los intervalos de confianza*, por otra parte **el estadístico muestral SI tiene una distribución de probabilidad**. Esto quiere decir que **el nivel de confianza se relaciona con la taza de éxito que el intervalo de confianza tiene para capturar/contener el parámetro poblacional**, mas no representa la *probabilidad de que un intervalo de confianza contenga el parámetro poblacional*.

**<h3 id="subsection223" style="color: #003366;">2.2.3. Distribución t-student</h3>**

---
Distribución t-student

---

Hasta este punto, hemos realizado múltiples estimaciones puntuales asumiendo el conocimiento de la desviación estándar de nuestra población, lo cual no es muchas veces acertado ya que en casos de la vida real dícho parámetro no es de nuestro conocimiento. Para afrontar éste problema usamos en lugar de la distribución $Z$ (la cuál hace alución a la población), la distribución **t de Student**, comúnmente conocida como distribución t. Es una distribución simétrica y en forma de campana, similar a la distribución normal pero con colas más gruesas lo cúla significa que hay una *mayor probabilidad de obtener valores extremos en comparación con la distribución normal*.

La distribución $t$ depende de un parámetro llamado **grados de libertad (df)**, que generalmente se define como el *tamaño de la muestra menos uno* ($df = n - 1$). A medida que los grados de libertad aumentan, la distribución $t$ se asemeja cada vez más a la distribución normal estándar. Concrétamente, la distribución $t$ al igual que la distribución $Z$, es una distribución muestral que se genera a partir de una "estandarización" de la variable aleatoria, solo que a diferencia de la distribución $Z$, se emplea diréctamente la varianza muestral $\text{s}^2$ de modo que ésta distribución describe la densidad de probabilidad de la variable aleatoria siguiente:

$$
t = \frac{\hat{x} - \mu} {\text{s}/\sqrt{n}}
$$

siendo $t$ el t-valor, $\hat{x}$ la media muestral, $\mu$ la media poblacional, $s$ la desviación estándar muestral y $n$el tamaño muestral. La distribución $t$ es útil tanto en cásos donde el tamaño de la muestra es pequeño ($n<30$) o cuando la desviación estándar poblacional es desconocída. De manera análoga ala construcción de intervalos de confianza con base a la variable $Z$, cuando se desconoce la desviación estándar de la población, un intervalo de confianza se genera a partir de la distribución t-student de la forma:

$$
IC = \bar{x} \pm t_{\alpha / 2} \frac{s}{\sqrt{n}}
$$


No obstante, la función de densidad de probabilidad de una distribución t-student  es muy diferente en su escritura a la distribución $Z$. Recordemos que si bien la distribución $Z$ es una *distribución normal estándar que modela la variable aleatoria $Z$*, así mísmo, a pesar de que nosotros experimentalmente definamos la variable $t$ con la relación anterior, la función de densidad de probabilidad para una variable $t$ viene dada por la siguiente expresión, donde $\nu$ es el número de grados de libertad:

$$
f(t|\nu) = \frac{\Gamma\left(\frac{\nu + 1}{2}\right)}{\sqrt{\nu \pi} \ \Gamma\left(\frac{\nu}{2}\right)} \left(1 + \frac{t^2}{\nu}\right)^{-\frac{\nu + 1}{2}}
$$

Donde:
- $\Gamma$ es la función Gamma, que generaliza el concepto de factorial a números reales y complejos.
- $\nu$ representa los grados de libertad.
- $t$ es el valor de la variable aleatoria $t$.


La función gamma $\Gamma(x)$ es una generalización del concepto de factorial que se aplica no solo a enteros positivos, sino también a números reales y complejos (excepto los enteros negativos y el cero). La función Gamma es definida mediante la siguiente integral impropia:

$$
\Gamma(x) = \int_0^\infty t^{x-1} e^{-t} \, dt
$$

Donde:
- $x$ es un número real o complejo.
- $t$ es la variable de integración.

Algúnas de sus propiedades son:

1. **Relación con el Factorial**: Para un número entero positivo $n$, la función Gamma está relacionada con el factorial de la siguiente manera:
   $$
   \Gamma(n) = (n-1)!
   $$
   Por ejemplo:
   $$
   \Gamma(5) = 4! = 24
   $$

2. **Desplazamiento**: Existe una propiedad importante llamada la "propiedad de desplazamiento":
   $$
   \Gamma(x+1) = x \cdot \Gamma(x)
   $$
   Esta propiedad es útil para derivar valores de la función Gamma para números no enteros.

3. **Valores Especiales**: Algunos valores especiales de la función Gamma incluyen:
   $$
   \Gamma\left(\frac{1}{2}\right) = \sqrt{\pi}
   $$
   $$
   \Gamma(1) = 1
   $$

A continuación, se muéstra manualmente como generar una PDF para una variable $t$, no obstante el módulo `scipy` nos permite realizar éstos cálculos directamente.

```python
# Función para hallar la PDF de la distribución t
from scipy.special import gamma

def t_distribution_pdf(t, nu):
    numerator = gamma((nu + 1) / 2)
    denominator = np.sqrt(nu * np.pi) * gamma(nu / 2)
    return (numerator / denominator) * (1 + (t ** 2) / nu) ** (- (nu + 1) / 2)

# Parámetros
nu_values = list(range(1, 11))  # Grados de libertad
t_values = np.linspace(-5, 5, 1000)  # Valores de t

# Graficar la distribución t
plt.figure(figsize=(8, 4))
for degree in nu_values:
    # Cálculo de la PDF para cada valor de t
    pdf_values = t_distribution_pdf(t_values, degree)
    plt.plot(t_values, pdf_values, label=f't-distribution (df={degree})')

plt.title("Distribución t de Student")
plt.xlabel("t")
plt.ylabel("Densidad de Probabilidad")
plt.legend()
plt.grid(True)
plt.show()
```

Como se observa en el plot anterior, una peculiaridad de la distribución $t$ es que si bien posee una forma general especificada por su PDF, ésta al depender de los grados de libertad y o de la desviación estándar, posee una forma única para cada tamaño muestral y por ende, es necesario calcular la distribución cada vez que se tenga un tamaño de muestra dado ya que ésto nos permitirá calcular intervalos de confianza acertados para nuestro caso particular.

Para ilustrar el uso de la distribución $t$ para generar intervalos de confianza, vamos a generar dos experimentos aleatorios: el primero, tomará solo una muestra de tamaño $n$ de la distribución poblacional, calculará la desviación estándar muestral a partir de una única muestra y calculará los intervalos de confianza. El segúndo experimento, generará $N$ muestras de tamaño $n$, hallará la desviación estándar muestral de cada muestra y estimará la varianza poblacional como un promedio de las varianzas muestrales para posteriormente calcular el intervalo de confianza. En generál se siguen los siguientes procedimientos:

1. Cálculo de la media y desviación estándar muestral.
2. Definición de los grados de libertad como $df = n -1$.
3. Especificación del nivel de confianza y significancia.
4. Cálculo del error estándar.
5. Cálculo del valor crítico de $t$ teniendo en cuenta los grados de libertad.
6. Cálculo del márgen de error.
7. Cálculo del intervalo de confianza.

```python
# Parámetros de la población
population_mean = 100   # Media poblacional
population_std = 15     # Desviación estándar poblacional
sample_size = 50        # Tamaño de cada muestra

# Generación de una única muestra de la población
sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
mean_sample = np.mean(sample)
sample_std = np.std(sample, ddof=1)

# Grados de libertad
df = sample_size - 1

# Nivel de confianza
confidence_level = 0.95
alpha = 1 - confidence_level

# Error estándar de la media (calculado para la muestra específica)
sem = sample_std / np.sqrt(sample_size)

# Valor crítico t para el nivel de confianza
t_critical = stats.t.ppf(1 - alpha / 2, df)

# Margen de error
margen_error = t_critical * sem

# Intervalo de confianza
intervalo_confianza = (mean_sample - margen_error , mean_sample + margen_error )

# Imprimimos los resultados
print(f"Media muestral: {mean_sample:.2f}")
print(f"t critico: {t_critical:.2f}")
print(f"Intervalo de confianza del 95%: {intervalo_confianza}")
print(f"Margen de error: {margen_error:.2f}")
```

**<h3 id="subsection224" style="color: #003366;">2.2.4. Intervalos de Confianza para Proporciones</h3>**

---
Intervalos de COnfianza para Proporciones

---

Los intervalos de confianza para proporciones se utilizan para estimar el rango dentro del cual se espera que esté la verdadera proporción de una población, basándose en una muestra. Este intervalo proporciona una estimación de la precisión de la proporción muestral y se calcula usando la distribución normal, especialmente cuando el tamaño de la muestra es grande. 

Una proporción poblacional $\hat(p)$ se calcula como el *número de éxitos (eventos de interés)* dividido por el tamaño de la muestra. 
$$
\hat{p} = \frac{x}{n}
$$
donde $x$ es el número de éxitos y $n$ es el tamaño de la muestra. En este sentido, el **error estándar de la proporción** usado para medir la variabilidad de la proporción muestral, se calcula como:

$$
SE = \sqrt{\frac{\hat{p} (1 - \hat{p})}{n}}
$$

De ésta manera, los intervalos de confianza se calculan mediante la **distribución normal estándar** según el nivel de confianza dado, de modo que

$$
ICP = \hat{p} \pm Z_{\alpha / 2} SE
$$


```python
# Parámetros de la muestra
n = 200   # Tamaño de la muestra
x = 120   # Número de éxitos (personas que prefieren el producto)

# Proporción muestral
p_hat = x / n

# Error estándar de la proporción
SE = np.sqrt(p_hat * (1 - p_hat) / n)

# Nivel de confianza
confidence_level = 0.95
alpha = 1 - confidence_level
Z = stats.norm.ppf(1 - alpha / 2)  # Valor crítico Z para el intervalo de confianza

# Margen de error
margin_of_error = Z * SE

# Intervalo de confianza
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)

# Resultados
print(f"Proporción muestral: {p_hat:.2f}")
print(f"Error estándar de la proporción: {SE:.2f}")
print(f"Valor crítico Z: {Z:.2f}")
print(f"Margen de error: {margin_of_error:.2f}")
print(f"Intervalo de confianza del 95%: {confidence_interval}")
```

**<h2 id="section3" style="color: #007ACC;">3. Prueba de Hipótesis</h2>**

---
Prueba de Hipótesis

---

En el mundo de la estadística, tomar decisiones informadas basadas en datos es fundamental para la investigación científica, el análisis de negocios y la toma de decisiones en general. Las pruebas de hipótesis son herramientas poderosas que nos permiten **evaluar si los datos que hemos recolectado apoyan o refutan una afirmación específica sobre una población o un proceso**. Estas pruebas proporcionan un marco sistemático para tomar decisiones basadas en evidencia, en lugar de en suposiciones o intuiciones.

Por ejemplo, un investigador puede querer saber si un nuevo medicamento es más efectivo que el tratamiento estándar. Utilizando una prueba de hipótesis, puede determinar si las diferencias observadas en los resultados entre el nuevo medicamento y el tratamiento estándar son lo suficientemente grandes como para no ser atribuibles al azar. De manera similar, en un entorno empresarial, las pruebas de hipótesis pueden ayudar a evaluar si una nueva estrategia de marketing ha llevado a un aumento significativo en las ventas.

#### **Componentes Principales de una Prueba de Hipótesis**

Una prueba de hipótesis se basa en varios componentes clave táles como:

1. **Hipótesis Nula ($H_0$)**: Es la afirmación que se pone a prueba. Generalmente, esta hipótesis representa una situación de no efecto o no diferencia, y se asume como verdadera hasta que los datos sugieren lo contrario. Por ejemplo, podría ser que no hay diferencia en la eficacia entre dos tratamientos o entre la medición de una media muestral y una media poblacional.

2. **Hipótesis Alternativa ($H_1$ o $H_A$)**: Es la afirmación que se considera si la hipótesis nula es rechazada. Representa una diferencia o efecto que el investigador está tratando de encontrar. En el ejemplo del medicamento, la hipótesis alternativa podría ser que el nuevo medicamento es más efectivo que el tratamiento estándar, o que la medición de un parámetro poblacional como la media, si pose diferencias sustanciales entre el valor estimado a partir de la muestra y el valor poblacional real. 

3. **Nivel de Significancia ($\alpha$)**: Es el umbral para tomar una decisión sobre la hipótesis nula. Representa la probabilidad de cometer un error tipo I, es decir, rechazar la hipótesis nula cuando en realidad es cierta. Comúnmente, se usa un nivel de significancia del 5% ($\alpha = 0.05$).

4. **Estadístico de Prueba**: Es una medida calculada a partir de los datos que se utiliza para tomar decisiones sobre la hipótesis nula. Dependiendo del tipo de prueba y los datos, el estadístico de prueba puede tomar diversas formas táles como el valor $t$ de la distribución $t$-student.

5. **Valor p**: Es la **probabilidad de observar un estadístico de prueba al menos tan extremo como el observado**, bajo la suposición de que la hipótesis nula es cierta. El valor p se compara con el nivel de significancia para tomar una decisión.

6. **Decisión**: Basado en el valor $p$ y el nivel de significancia, se decide si se rechaza o no la hipótesis nula. Si el valor $p$ es menor que $\alpha$, se rechaza la hipótesis nula en favor de la hipótesis alternativa. 


En el proceso de decisión basados en pruebas de hipótesis exíste una sutileza conceptual. En el momento en que se rechaza la hipótesis nula $H_0$ **se acepta la validez de la hipótesis alternativa $H_A$**, por el contrario, al fallar en rechazar la hipótesis alternativa $H_A$ **no significa que $H_0$ sea cierta**, sino que se está concluyendo que **no se tiene suficiente evidencia** para asumir la veracidad de $H_A$.

#### **Cómo Calcular las Pruebas de Hipótesis**

El cálculo de una prueba de hipótesis involucra los siguientes pasos:

1. **Formular las Hipótesis**: Definir claramente las hipótesis nula y alternativa.
2. **Seleccionar el Nivel de Significancia**: Establecer el umbral para decidir si rechazar la hipótesis nula.
3. **Recoger y Analizar los Datos**: Realizar el experimento o recolectar los datos necesarios.
4. **Calcular el Estadístico de Prueba**: Utilizar los datos para calcular un estadístico que refleje la diferencia entre las hipótesis.
5. **Obtener el Valor p**: Calcular la probabilidad asociada con el estadístico de prueba.
6. **Tomar una Decisión**: Comparar el valor p con el nivel de significancia y decidir si se rechaza la hipótesis nula.

---
Tipos de errores

---

Un tipo de error dentro del contexto de toma de decisiones en pruebas de hipótesis, corresponden a rechazos incorrectos de la hipótesis nula o la hipótesis alternativa. Según sea el caso tenemos los siguiente tipos de errores:

- **Error tipo I (falso positivo - error $\alpha$)**: Corresponde al error cometido durante el proceso de evaluación de la evidencia y toma de decisión, donde la hipótesis nula **$H_0$** es rechazada y la hipótesis alternativa **H_A** es tomada como cierta siendo la hipótesis nula **$H_0$** la opción correcta a considerar como verdadera.

- **Error tipo II (falso negativo - error $\beta$)**: Corresponde al error cometido durante el proceso de evaluación de la evidencia y toma de decisión, en donde la hipótesis alternativa **$H_A$** es rechazada y en su lugar, la hipótesis nula **$H_0$** es considerada como verdadera siendo la hipótesis alternativa **$H_A$** la opción correcta a considerar como verdadera.



image.png

Note entonces que los errores tipo I y tipo II no tienen el mísmo impacto sobre el problema en cuestión. Si consideramos que en general, $H_A$ es la hipótesis que deseamos poner a prueba mientras que  $H_0$ representa la hipótesis base, o status-quo, considerar la hipótesis alternativa suele tener peores repercuciones ya que es aquella hipótesis que contiene una consecuencia debido a que "aprueba" una desición a tomar. Por ejemplo, en el caso médico, si $H_A$ corresponde a "validar que un medicamento es mejor que otro y no posee efectos secundarios", un error tipo I generaría percances de salud a la población objetivo de dicho medicamento. En el caso científico, $H_A$ hace alución a validar que cierta hipótesis sobre un fenómeno o sistemasea cierta, así un error tipo I implicaría falsamente dar validez a una teoría.

Con esto en mente, si consideramos que una toma de decisión puede ejecutarse múltiples veces según una situación lo requiera, pero dicha decisión siempre estará basada en la evidencia que respalde a la hipótesis detrás de la mísma, cabría preguntarse **¿cuál es la probabilidad más alta  de errores tipo I que permitiríamos en esta toma de decisiones?** o, en el caso opuesto, **¿cuál es la probabilidad más alta  de errores tipo II que permitiríamos en esta toma de decisiones?**.

En este sentido, formulamos las siguientes definiciones:

1. **Significancia $\alpha$**: Probabilidad máxima de cometer un **error tipo I** tolerada en nuestro experimento. Corresponde al **humbral para rechazar a $H_0$ (o no)**, dada una evidencia que soporte a $H_A$. Tambíen puede interpretarse como la probabilidad de rechazar $H_0$ cuando en realidad es verdadera.

$$
\alpha = \text{max} P(\text{Error tipo I})
$$

2. **Potencia de Prueba**: Probabilidad máxima de cometer un **error tipo II** tolerada en nuestro experimento. Corresponde al **humbral para rechazar $H_A$(o no)**. También puede interpretarse como la probabilidad de rechazar $H_0$ cuando esta es falsa, es decir, es la capacidad de la prueba de detectar un efecto real.

$$
1- \beta
$$

**<h3 id="subsection31" style="color: #003366;">3.1. Tipos de Prueba de Hipótesis</h3>**

---
Tipos de Prueba de Hipótesis

---

En las pruebas de hipótesis, el objetivo es decidir si los datos observados proporcionan suficiente evidencia para rechazar una hipótesis nula (\(H_0\)) a favor de una hipótesis alternativa (\(H_1\)). Dependiendo de la naturaleza de la hipótesis alternativa, las pruebas estadísticas se pueden clasificar en tres tipos:

1. **Right-Tailed Test (Prueba de Cola Derecha)**
2. **Left-Tailed Test (Prueba de Cola Izquierda)**
3. **Two-Sided Test (Prueba de Dos Colas)**

Antes de definir los tipos de test estadísticos de significancia de hipótesis, definmos los siguientes conceptos:

  - **p-valor**: Es la **probabilidad de que, asumiendo $H_0$ es verdadera, el estadístico de prueba tome un valor tan extremo o más extremo que el valor observado**, donde un estadístico de prueba puede ser por ejemplo la media muestral. En este sentido, el $\text{p-avlor}$ es el **área de la distribución del estadístico de prueba más alla de un valor crítico**, es decir, es el **cuantil $1 - \alpha$**.
  - **Valor Crítico:** Corresponden a aquellos valores del estadístico (variable aleatoria) a partir de los cuáles se rechaza la hipótesis nula, es decir, corresponden a aqueelos valores donde $\text{p-value}=\alpha$.

#### 1. **Right-Tailed Test (Prueba de Cola Derecha)**

En una prueba de cola derecha, la hipótesis alternativa ($H_A$) propone que el parámetro de interés es **mayor que el valor especificado en la hipótesis nula** ($H_0$). En otras palabras, estamos interesados en *saber si el valor observado es significativamente mayor que el valor bajo $H_0$*. Tal que en el caso de que nuestro parámetro poblacional de interés es la media muestral, tenemos:

- **Hipótesis:**
  - $H_0$: $\mu = \mu_0$
  - $H_A$: $\mu > \mu_0$

En éste escenario, buscamos entonces verificar la hipótesis alternativa $H_A$ mediante la siguiente probabilidad, donde si $\text{p-value}$ < $\alpha$, rechazamos la hipótesis nula $H_0$  en favor de la hipótesis $H_A$ alternativa. 

$$
\text{p-value} = P(T(X) > t | H_0)
$$

Donde:
  - $T(X)$: estadística de prueba de una variable aleatoria.
  - $t$: valor observado para la estadística.
  - $H_0$: hipótesis nula.

Note entonces siendo $\hat{X}$ el estadístico muestral, $\mu = a$ es el valor del parámetro que respalda a $H_0$ mientras que $\mu = b$ es el valor del parámetro poblacional que respalda a $H_A$. 
$$
\text{p-value} = P(\hat{X} > b | \mu = a)
$$

#### 2. **Left-Tailed Test (Prueba de Cola Izquierda)**

En una prueba de cola izquierda, la hipótesis alternativa ($H_A$) propone que el parámetro de interés es **menor que el valor especificado en la hipótesis nula** ($H_0$). Aquí, estamos interesados en *saber si el valor observado es significativamente menor que el valor bajo $H_0$*.

- **Hipótesis:**
  - $H_0$: $\mu = \mu_0$
  - $H_A$: $\mu < \mu_0$

Para pruebas de cola izquierda, ya que se busca conocer si un valor del estadístico de prueba es estadísticamente menor como para rechazar la hipótesis nula, el valor de desición de significancia $\text{p-value}$ se expresa como:

$$
\text{p-value} = P(T(X) < t | H_0)
$$

#### 3. **Two-Sided Test (Prueba de Dos Colas)**

En una prueba de dos colas, la hipótesis alternativa ($H_A$) propone que el parámetro de interés es **diferente (ya sea mayor o menor) al valor especificado en la hipótesis nula** ($H_0$). Aquí, estamos interesados en *saber si el valor observado es significativamente diferente (en cualquier dirección) del valor bajo* $H_0$.

- **Hipótesis:**
  - $H_0$: $\mu = \mu_0$
  - $H_A$: $\mu \neq \mu_0$

Debido a que en un test de dos colas, se busca una variación estadísticamente significativa bien sea por incremento o decremento, el valor de decisión $\text{p-value}$ se expresa como:

$$
\text{p-value} = P(|T(X) t_0| > |t - t_0| | H_0)
$$

donde $t_0$ es el valor del estadístico que apoya a  $H_0$ mientras que $t$ es el valor observado, el cuál apoya a $H_A$.


Para empezar a ilustrar estos conceptos, supongamos la siguiente situación: en un estudio hecho hace unos años, se determinó que la media poblacional poseía un valor de $\mu = 100$. Años después, otro investigador llegó a nosotros reportando que realizó una medición sobre la población recientemente y observó que la media aumentó, de hecho, ésta adquirió un valor de $\mu = 106$. Si en este ejemplo se asume que la varianza poblacional **es conocida y no ha cambiado**, ¿es la diferencia que el investigador encontró en las medias lo suficientemente significativa estadísticamente para asegurar que la media poblacional aumento?.

Para resolver ésta incógnica, ejecutamos los siguientes procedimientos:

1. Determinamos el valor del parámetro y del estadístico muestral.
2. Ya que la varianza es conocida, calculamos con ella el error estándar.
3. Determinamos el $\text{p-valor}$ como el cuantil $1-\alpha$.
4. Determinamos la significancia estadística según la proporción entre $\text{p-valor}$  y  $\alpha$.

Note a demás, que acontinuación para nuestro ejemplo, asumimos la normalidad de la distribución muestral gracias al teorema del límite central.

```python
# Parámetros conocidos
population_mean = 100  # Media poblacional bajo H0
population_std = 15    # Desviación estándar poblacional (conocida)
sample_size = 30       # Tamaño de la muestra

# Generamos una muestra a partir de la distribución poblacional
np.random.seed(0)
data = np.random.normal(population_mean, population_std, sample_size)

# Calculamos el estadístico de prueba y el error estandar
sample_mean = np.mean(data)
sem = population_std / np.sqrt(sample_size)

# Calculamos el valor crítico para right-tailed test
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha)  # Z crítico ( cuantil 1- alpha )
critical_value_right = population_mean + z_critical * sem  

# Calculamos el valor crítico para left-tailed test
alpha = 0.05 
z_critical_left = stats.norm.ppf(alpha)  # Z crítico ( cuantil alpha )
critical_value_left = population_mean + z_critical_left * sem

# Calculamos el valor crítico para two-sided test
z_critical_two_sided = stats.norm.ppf(1 - alpha / 2)
critical_value_lower = population_mean - z_critical_two_sided * sem
critical_value_upper = population_mean + z_critical_two_sided * sem

# Gráfico right-tail test
graficar_right_normal_test(90, 110, population_mean, sample_mean, sem, critical_value_right)
# Gráfico left-tail test
graficar_left_normal_test(90, 110, population_mean, sample_mean, sem, critical_value_left)
# Gráfico two-tail test    
graficar_two_normal_test(90, 110, population_mean, sample_mean, sem, critical_value_lower, critical_value_upper)
```

En el ejemplo anterior, nos es fácil entender que está sucediendo: cuando se realiza una distribución muestral de un estadístico para estimar un parámetro, por el teorema del límite central tenemos que la distribución muestral sigue una distribución normal. Ésto nos permíte formular una hipótesis nula $H_0$ que se entiende como que **no exíste una diferencia significativa entre el estadístico muestral y el parámetro poblacional** y a su vez, una hipótesis alternativa $H_A$ que se entiende como **exíste una diferencia significativa entre el estadístico y el parámetro en algúna dirección**. Así, mediante el error estándar y los percentiles de la distribución normal, nos es posible encontrar la región de valores de la variable aleatoria (media muestral) los cuáles representan casos extremos que permíten evidenciar diferencias estadísticamente significativas entre el parámetro y la estimación desde la muéstra.

Por otra parte, si empleamos diréctamente la distribución $Z$, el procedimiento para determinar las regiones de rechazo de la hipótesis nula $H_0$ será más fácil ya que los valores críticos corresponden directamente a los percentiles relacionados con la significancia establecida $\alpha = 0.05$. A continuación, modelaremos el mísmo ejemplo anterior pero ésta vez haciendo uso directo de la distribución $Z$.

```python
# Parámetros conocidos
population_mean = 100  # Media poblacional bajo H0
population_std = 15    # Desviación estándar poblacional (conocida)
sample_size = 40

# Datos simulados de la población
np.random.seed(0)
data = np.random.normal(population_mean, population_std, sample_size)

# Estadístico de prueba
sample_mean = np.mean(data)

# Calculamos los valores críticos
sem = population_std / np.sqrt(sample_size)
z_statistic = (sample_mean - population_mean) / sem

# Calculamos el valor crítico para right-tailed test (distribución Z)
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha)

# Calculamos el valor crítico para left-tailed test (distribución Z)
z_critical_left = stats.norm.ppf(alpha)

# Calculamos el valor crítico para two-sided test
z_critical_two_sided = stats.norm.ppf(1 - alpha / 2)
critical_value_lower = - z_critical_two_sided
critical_value_upper = z_critical_two_sided

# Gráfico right-tail test 
graficar_right_z_test(z_critical, z_statistic)
# Gráfico left-tail test 
graficar_left_z_test(z_critical_left, z_statistic)
# Gráfico two-tail test 
graficar_two_z_test(critical_value_lower, critical_value_upper, z_statistic)

```

De manera análoga en razonamiento y procedimiento, en el caso tal que la varianza poblacional sea desconocída, basta con realizar la prueba de hipótesis haciendo uso de la distribución t-student.

```python
# Parámetros conocidos
population_mean = 100  # Media poblacional bajo H0
population_std = 15    # Desviación estándar poblacional (conocida)
sample_size = 40

# Datos simulados de la población
np.random.seed(0)
data = np.random.normal(population_mean, population_std, sample_size)

# Estadístico de prueba
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
sem = sample_std / np.sqrt(sample_size)
t_statistic = (sample_mean - population_mean) / sem


# Calculamos el valor crítico para right-tailed test (t-student)
alpha = 0.05
df = sample_size - 1
t_critical_right = stats.t.ppf(1 - alpha, df)

# Calculamos el valor crítico para left-tailed test (t-student)
t_critical_left = stats.t.ppf(alpha, df)

# Calculamos el valor crítico para two-sided test
t_critical_two_sided = stats.t.ppf(1 - alpha / 2, df)
critical_value_lower = - t_critical_two_sided
critical_value_upper = t_critical_two_sided

# Gráfico right-tail test 
graficar_right_t_test(t_critical_right, t_statistic, df)
# Gráfico left-tail test 
graficar_left_t_test(t_critical_left, t_statistic, df)
# Gráfico two-tail test 
graficar_two_t_test(critical_value_lower, critical_value_upper, t_statistic, df)
```

### <!-- Subsection 3.2 -->
**<h3 id="subsection32" style="color: #003366;">3.2. Poder de una Prueba y Error Tipo II</h3>**

---
Poder de una Prueba, Error Tipo II y la Distribución para $H_A$

---

Hasta éste punto, hemos realizado diferentes ilustraciones a cerca de cómo poner a prueba una hipótesis con vísta a una evidencia disponible. No obstante, en secciones anteriores hablabamos a cerca de cómo muchas veces el error tipo I puede ser más peligroso debido a que la hipótesis alternativa $H_A$ es aquella que nos incita a tomar una acción debido a que es la hipótesis que deseamos poner a prueba. Así pues, nuestras discuciones han estado basadas o referenciadas con respecto al error tipo I y por ende, nos hemos concentrado en realizar la prueba de hipótesis considerando la evidencia **en contra** de $H_0$ gracias a que a través de nuestras pruebas de hipótesis deseamos saber si exíste suficiente evidencia **a favor** de  $H_A$.

Ésto quiere decir, que las distribuciones muestrales que hemos empleado para demostrar el uso del testeo de hipótesis hasta éste punto, corresponden a la evidencia que **sustenta a $H_0$** y por ende, la probabilidad que está contenida dentro de los valores críticos corresponde a **la evidencia disponible a favor de $H_0$ y en contra de $H_A$**. Sin embargo, si bien esta metodología nos permite **establecer el rango de valores da partir de los cuáles rechazamos a H_0 y consideraremos H_A como verdadera**, en la presente subsección verémos que exíste una perspectiva complementaria a ésta de modo tal que ya no nos preguntarémos a cerca de la significancia estadística de la hipótesis nula sino que dirigirémos nuestra atención a **establecer el rango de valores a partir de los cuáles rechazarémos H_A debido a que no contamos con la suficiente evidencia para soportarla**.

$$
\alpha = P(\text{Rechazamos} \text{ } H_0 | H_0 \text{ } \text{es Verdadera}) = \text{probabilidad tolerada de faltos positivos}
$$

$$
\beta = P(\text{ NO Rechazamos} \text{ } H_0 | H_0 \text{ } \text{es Falsa}) = \text{probabilidad tolerada de faltos negativos}
$$

Así, $\beta$ se calcula como la probabilidad de que la estimación del parámetro caiga dentro de la región de **no rechazo de la hipótesis nula H_0** definida por:

$$
\text{Zona de aceptación } H_0 = \left[Z_{\text{H}_{(0,-)}}, Z_{\text{H}_{(0,+)}}\right]
$$

donde $Z_{\text{H}_{0,-/+}}$ es el límite inferior (-) y superior (+) de la región de aceptación de $H_0$ definida por la distribución de hipótesis bajo $H_0$. Es entonces $\beta$ la probabilidad acumulada de la distribución bajo la hipótesis $H_A$ evaluada en ésta región. Por ende, la probabilidad $\beta$ **NO depende de la muéstra sino del nivel de significancia**.

$$
\beta = P(T(X) < t | H_A) 
$$

El **poder de un test** es la  **probabilidad de rechazar correctamente H_0 cuando en realidad es falsa** (probabilidad de verdadero negativo). Es decir, es la capacidad de un test para *detextar un efecto cuando éste realmente exíste* y viene dado por:

$$
P(\text{Rechazamos} \text{ } H_0 | H_0 \text{ } \text{es Falsa}) =  1 - \beta
$$

En éste sentido, el poder del test es **inversamente proporcional al error tipo II** donde un alto $1-\beta$ significa que hay una *baja probabilidad de cometer un error tipo II*. El nivel de significancia $\alpha$ siendo la probabilidad de cometer un error tipo I, se relaciona con el poder de la prueba en la medida de que si disminuimos $\alpha$ (lo que significa que queremos ser más cautelosos para evitar un error tipo I), podríamos estar aumentando $\beta$ lo que disminuye el poder del test. Por otro lado, si aumentamos el poder del test, incrementamos la probabilidad de cometer un error tipo I.

- **Distribución H_0 (en azul)**:
    - Representa la distribución de la media muestral (o estadístico) bajo la hipótesis nula.
    - Las regiones sombreadas en azul corresponden a la **región de rezhado de H_0** y corresponden a la probabilidad del error tipo I $\alpha$.
- **Distribución H_A (en naranja)**:
    - Representa la distribución de la media muestral bajo la hipótesis alternativa.
    - La región sombreada en naranja entre los valores críticos es la **probabilidad de error tipo II (\beta)**
- **Poder del Test**:
    - La porción de la distribución $H_A$ que caue fuera de los valores críticos (es decir, donde rechazaríamos $H_0$) es el poder del test $1-\beta$.
    - Un test con un poder alto tendrá una menor área sombreada en naranja (menor $\beta$).

```python
# Parámetros de la distribución bajo la hipótesis nula (H0)
mu_0 = 100  # Media bajo H0
sigma = 15  # Desviación estándar (conocida)

# Parámetros de la distribución bajo la hipótesis alternativa (H1)
mu_1 = 110  # Media bajo H1 (efecto real)
sample_size = 30

# Nivel de significancia y valor crítico para un test de dos colas
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha / 2)

# Calculamos el valor crítico en la distribución bajo H0
critical_value_left = mu_0 - z_critical * (sigma / np.sqrt(sample_size))
critical_value_right = mu_0 + z_critical * (sigma / np.sqrt(sample_size))

# Generamos datos para las distribuciones bajo H0 y H1
x = np.linspace(90, 120, 1000)
y_H0 = stats.norm.pdf(x, mu_0, sigma / np.sqrt(sample_size))
y_H1 = stats.norm.pdf(x, mu_1, sigma / np.sqrt(sample_size))

# Gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y_H0, label='Distribución bajo H0', color='blue')
plt.plot(x, y_H1, label='Distribución bajo H1', color='orange')

# Sombrear la región de error tipo I (α)
plt.fill_between(x, 0, y_H0, where=(x <= critical_value_left) | (x >= critical_value_right), color='blue', alpha=0.3, label='Error Tipo I (α)')

# Sombrear la región de error tipo II (β)
plt.fill_between(x, 0, y_H1, where=(x > critical_value_left) & (x < critical_value_right), color='orange', alpha=0.3, label='Error Tipo II (β)')

# Sombrear la región de poder del test (1 - β)
plt.fill_between(x, 0, y_H1, where=(x <= critical_value_left) | (x >= critical_value_right), color='green', alpha=0.3, label='Poder del Test (1 - β)')

# Líneas verticales para valores críticos y medias
plt.axvline(critical_value_left, color='red', linestyle='dashed', label='Valor Crítico Izquierdo (Límite inferior de NO rechazo H_0)')
plt.axvline(critical_value_right, color='red', linestyle='dashed', label='Valor Crítico Derecho (Límite superior rechazo H_0)')
plt.axvline(mu_1, color='green', linestyle='dashed', label=f'Media bajo H1 = {mu_1}')

# Título y etiquetas
plt.title('Visualización del Poder del Test con Regiones de Error Tipo I y II')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

Interpretación de Resultados

---

En algúnas circunstancias, la correcta interpretación de las pruebas de hipótesis puede ser algo confusa, así pues resumamos a continuación las correctas interpretaciones de los elementos y resultados obtenidos en las discuciones mencionadas hasta ahora.

Los pasos necesarios para generar una prueba de hipótesis son:

1. **Definir las hipótesis**:
    - Definir la hipótesis (baseline) $H_0$
    - Definir la hipótesis alternativa a probar $H_A$
2. **Definir el tipo de prueba**:
    - Definir un estadístico de prueba.
    - Decidir el nivel de significancia $\alpha$
3. **Computar la estadística de interés**
    - Definir el estadístico en la muestra.
    - Definir el valor $\text{p-value}$.
4. **Definir una conclusión**
    - Si el $\text{p-value}$ es menor a $\alpha$, rechazamos $H_0$

Las correctas interpretaciones de los elementos involucrados en una prueba de hipótesis son:

- **Error tipo I**:
    Se rechaza la hipótesis nula $H_0$ como cierta y se escoge $H_A$ siendo ésta falsa (falso positivo).
- **Error tipo II**:
    Se **falla** en rechazar $H_0$ siendo ésta falsa y se descarta $H_A$ como verdadera (falso negativo).
- **Significancia $\alpha$**:
    Máxima probabilidad admitida de cometer un error tipo I.
- **\text{p-value}**:
    El valor p  **NO** representa la probabilidad de que $H_0$ sea cierta, en su lugar, representa la probabilidad de *observar la data registrada al azar*.
- **Conclusiones en las pruebas de hipótesis**:
    - Al rechazar $H_0$ **se asume que  $H_A$ es verdadera**.
    - Al NO rechazar $H_0$ NO se asume que $H_0$ sea verdadera, solo se **asume que no hay suficiente evidencia para validar $H_A$**



### <!-- Subsection 3.3 -->
**<h3 id="subsection33" style="color: #003366;">3.3. Intervalos de Confianza y Pruebas de Hipótesis Aplicados a un Caso Real</h3>**

---
Cálculo de Intervalos de Confianza a Datos Reales

---

A continuación generaremos los intervalos de confianza para la estimación de la media de una variable de nuestro conjunto de datos real donde emplearemos el error estandar calculado a partir de la desviación estándar poblacional estimada por medio del proceso de muestreo. El primer caso, corresponde a una prueba de hipótesis generada sobre un conjunto de 50 muestras.

```python
# Parámetros de la población
variable = "total_likes_on_outofstation_checkin_received"
data = list(df_turismo[variable].values)
population_mean = np.round(np.mean(data),3)  # media poblacional
population_std  = np.round(np.std(data),3)   # desviación estándar poblacional   
n_samples = 10000                # número de muestras
sample_size = 50                 # tamaño de cada muestra

# Generación de múltiples muestras de la población
sample_means = []
sample_variance = []
for _ in range(n_samples):
    sample = random.sample(data, sample_size)
    sample_means.append(np.mean(sample))
    sample_variance.append(np.std(sample))

# Estimación del parámetro poblacional
mean_sample = np.mean(sample_means)
variance_sample = np.mean(sample_variance)

# Cálculo del error estándar de la media (usamos eem poblacional para el IC)
eem = population_std / np.sqrt(sample_size)
eem_sample = variance_sample/np.sqrt(sample_size)

# Intervalo de confianza del 95%
confianza = 0.95
Z_critical = stats.norm.ppf(1 - (1 - confianza) / 2)  # Cálculo del valor crítico Z
margen_error = Z_critical * eem_sample  # Usamos eem basado en la desviación estándar poblacional
intervalo_confianza = (mean_sample - margen_error, mean_sample + margen_error)

# Cálculo de ocnfianza al 95% mediante scipy.stats
intervalos_stats = stats.norm.interval(confidence=0.95, loc=mean_sample, scale=np.std(sample_means))

# Imprimimos los resultados
print(f"Media poblacional: {population_mean:.2f}")
print(f"Media muestral: {mean_sample:.2f}")
print(f"Error Estándar: {eem_sample:.2f}")
print(f"Z critico: {Z_critical:.2f}")
print(f"Intervalo de confianza del 95%: {intervalo_confianza}")
print(f"Intervalo de confianza del 95% (stats): {intervalos_stats}")

# Visualización de la distribución de las medias muestrales
graficar_ID_normal(sample_means, population_mean, intervalo_confianza)


# Validamos nuestra hipótesis de que nuestra estimación es acertada

# Hipótesis nula H0: mu = 50
mu_0 = population_mean

# Estadístico de prueba Z
Z = (mean_sample - mu_0) / eem_sample
p_value = 2 * (1 - stats.norm.cdf(np.abs(Z)))  # Prueba de dos colas

print(f"Estadístico de prueba Z: {Z:.2f}")
print(f"P-valor: {p_value:.4f}")
print(f"Discrepancia de la estiamción:", np.abs(population_mean - mean_sample))

# Decisión
alpha = 0.05
if p_value < alpha:
    print("Rechazamos la hipótesis nula.")
else:
    print("No podemos rechazar la hipótesis nula.")
```

Ahora, notemos que en el ejemplo anterior usamos un promedio de las medias muestrales para calcular nuestros intervalos de confianza y testear nuestra hipótesis. No obstante es posible generar éstos mísmos cálculos usando solo una muestra.

```python
# Parámetros de la población
variable = "total_likes_on_outofstation_checkin_received"
data = list(df_turismo[variable].values)
sample_size = 50 
sample = random.sample(data, sample_size)

# Estadísticos muestrales
sample_mean = np.mean(sample)
sample_std  = np.std(sample, ddof=1)

# Nivel de confianza
confianza = 0.95
alfa = 1 - confianza

# Grados de libertad
df_grados = sample_size - 1

# Estadístico t crítico
t_critical = stats.t.ppf(1 - alfa/2, df_grados)

# Error estándar de la media
eem = sample_std / np.sqrt(sample_size)

# Margen de error
margen_error = t_critical * eem

# Intervalo de confianza
intervalo_confianza = (sample_mean - margen_error, sample_mean + margen_error)

print(f"Media poblacional: {np.mean(data):.2f}")
print(f"Media muestral: {sample_mean:.2f}")
print(f"Margen de Error: {margen_error:.2f}")
print(f"Intervalo de confianza del 95%: {intervalo_confianza}")
```

En resumen podemos observar que *asumimos la normalidad en la distribución de los estimativos miestrales* fundamentada en el teorema del límite central. Luego, empleamos la distribución $Z$ para mediante el usi de sus **percentiles**, determinar los niveles de confianza (acumulación percentil de datos) dentro de nuestro espacio de muestras. De hecho, vemos las siguientes conclusiones:

1. **Intervalos de Confianza:** Para generar los intervalos de confianza, primero  *calculamos el error esperado del parámetro poblacional para cada muestra (EEM)*, luego, determinamos el *nivel de confianza en el que esperamos que nuestros datos caigan* mediante fijar un porcentaje de acumulacion de datos o *percentil* para luego, determinar el *valor $Z$* correspondiente. Una vez determinado el nivel de confianza, hallamos las fronteras de confianza como  el  *producto entre el parámetro estimado mas/menos el producto entre el valor de confianza $Z$ y el error esperado*.

2. **Prueba de Hipótesis:** En el contexto de la estimación paramétrica, nuestra hipotesis a corroborar es *si nuestra estimación es válida*. Para ello, calculamos el *valor Z generado a partir de la discrepancia entre la estimación y el valor real dividido por el EEM*, luego, usamos ese valor $Z$ para calcular *el  p-valor, el cual es la probabilidad de que nuestra hipotesis sea correcta*.

### <!-- Subsection 3.4 -->
**<h3 id="subsection34" style="color: #003366;">3.4. t-tests</h3>**

---
t-tests y Comparaciones Entre Muestras/Poblaciones

---

Los **t-tests** son un conjunto de pruebas estadísticas que se utilizan para *comparar las medias de uno o dos grupos de datos*. El objetivo principal de un t-test es determinar si *existe una diferencia significativa entre las medias de los grupos comparados*, lo cual puede ayudar a *inferir si esas diferencias se deben a variabilidad aleatoria o si son lo suficientemente grandes como para ser consideradas estadísticamente significativas*.

Un t-test es un tipo de prueba de hipótesis que utiliza la distribución t de Student para determinar si hay diferencias significativas entre las medias de una o dos poblaciones. Las pruebas de hipótesis t-test asumen que los dos conjuntos de datos son:

- Todos los elementos dentro de las muestras de los dos grupos son **distintos**, es decir, no exíste un elemento que pertenezca simultáneamente a dos grupos.
- Cada elemento en ambos samples son **independientes**.
- Las poblaciones están **normalmente** distribuidas.

Dentro de los tipos de t-tests encontramos:

1. **t-Test para Una Muestra (One-sample t-test)**: Determina si la media de una sola muestra difiere significativamente de un valor conocido o hipotético.

2. **t-Test para Muestras Independientes (Independent two-sample t-test)**: Compara las medias de dos grupos independientes para ver si hay diferencias significativas.

3. **t-Test para Muestras Pareadas (Paired-sample t-test)**: Compara las medias de dos medidas relacionadas (como antes y después de un tratamiento) dentro del mismo grupo.

#### **Fundamentos Matemáticos del t-Test**

Supongamos que deseamos comparar dos poblaciones a través de una muestra extraída de cada grupo y determinar si las poblaciones son distíntas a través del análisis de las muestras generadas a partir  de cada una de ella. En éste sentido, tomarémos como el estadístico de prueba la media muestral de cada grupo, donde al desconocer la varianza de ambas poblaciones, la única información a cerca de la variabilidad de los datos con la que contámos es la varianza muestral. Con base a éste propósito podémos generar los siguientes tres escenarios para comprobar la diferencia entre las poblaciones a través de su media:

- **Hipótesis (caso mayor que):**
  - $H_0$: $\mu_A = \mu_B$
  - $H_A$: $\mu_A > \mu_B$

- **Hipótesis (caso menor que):**
  - $H_0$: $\mu_A = \mu_B$
  - $H_A$: $\mu_A < \mu_B$

- **Hipótesis (caso diferente de):**
  - $H_0$: $\mu_A = \mu_B$
  - $H_A$: $\mu_A \neq \mu_B$

Si cada grupo poblacional $A$ y $B$ siguen cada uno una distribución normal tal que $A \sim \mathcal{N}(\mu_A, \sigma_A^2)$ y $B \sim \mathcal{N}(\mu_B, \sigma_B^2)$ respectivamente, observaremos las diferencias de las poblaciones a través de las diferencias de sus medias muestrales. Debido a que ésto significa generar una combinación lineal de dos distribuciones gausianas, se tiene que:

$$
\hat{A} - \hat{B} \sim \mathcal{N}\left(\mu_A - \mu_B, \frac{\sigma_A^2}{n_A} + \frac{\sigma_B^2}{n_B}\right)
$$

Ahora, si quisieramos ejecutar una prueba de hipótesis sería necesario generar una estadística $Z$, no obstante, debido a que no conocemos la varianza poblacional tenémos que usar la varianza muestral y por ende, usar la distribución t-student.

$$
\hat{A} - \hat{B} \sim \mathcal{N}\left(\mu_A - \mu_B, \frac{S_A^2}{n_A} + \frac{S_B^2}{n_B}\right)
$$

$$
t_{AB} = \frac{\left(\hat{A} - \hat{B}\right) - \left( \mu_A - \mu_B \right)}{\sqrt{\frac{S_A^2}{n_A} + \frac{S_B^2}{n_B}}}
$$

donde $t_{AB}$ representa la estadística $t$ de la diferencia entre las variablea aleatorias de interés, es decir, las medias muestrales en este caso. Una vez definida la estadística $t$ la cuál sigue una distribución una distribución t-student, debemos definir los grados de libertad de ésta distribución los cuáles no siguen la forma $df = n-1$ debido a que las muestras pueden ser de distinto tamaño y tener varianzas muestrales diferentes. La expresión para calcular los grados de libertad de dos grupos bajo una prueba $t$ es:

$$
df = \frac{\left(\frac{S_A^2}{n_A} + \frac{S_B^2}{n_B}\right)^2}{\frac{\left(\frac{S_A^2}{n_A}\right)^2}{n_A - 1} + \frac{\left(\frac{S_B^2}{n_B}\right)^2}{n_B - 1}}
$$


El t-test se basa en el cálculo de un valor t, que mide la diferencia entre las medias en relación con la variabilidad dentro de los grupos de datos. Una vez hallada la estadística $t$ y los grados de libertad, el valor t se compara con un valor crítico de la distribución t correspondiente al nivel de significancia ($\alpha$) y a los grados de libertad del test. Si el valor t calculado es mayor que el valor crítico, se rechaza la hipótesis nula, sugiriendo que hay una diferencia significativa entre las medias.



Para ilustrar los casos de uso de los t-tests, primero supongamos que nos encontramos ante una situación tipo **one-sample t-test** donde vamos a comparar la media de una muestra de nuestra población con un valor  hipotético que se asume previamente al test como verdadero (un valor de referencia). Harémos uso a continuación de nuestro conjunto de datos reales.

```python
# Parámetros de la población
variable = "total_likes_on_outofstation_checkin_received"
data = list(df_turismo[variable].values)
population_mean = np.round(np.mean(data),3)  # media poblacional
population_std  = np.round(np.std(data),3)   # desviación estándar poblacional   
n_samples = 10000                # número de muestras
sample_size = 50                 # tamaño de cada muestra

# Generamos una muestra
sample = random.sample(data, sample_size)
sample_mean = np.mean(sample)

# t test de un sample independiente
t_statistic, p_value = stats.ttest_1samp(sample, population_mean)

print(f"Media poblacional: {population_mean:.2f}")
print(f"Media muestral: {sample_mean:.2f}")
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretación
if p_value < 0.05:
    print(f"Rechazamos la hipótesis nula: La media de la muestra es significativamente diferente de {population_mean:.2f}.")
else:
    print(f"No se rechaza la hipótesis nula: No hay evidencia suficiente para decir que la media de la muestra es diferente de {population_mean:.2f}.")
```

Ya que los grupos de evaluación **no son independientes ya que representan la mísma población**, estamos incurriendo en un **paired-sample t-test** donde el  **numero de grados de libertad $df$ es igual a $n-1$** con $n$ el tamaño de la muestra poblacional. 

```python
# Comparación pareada para una mísma población 

# Variables numéricas y categóricas para segmentación
variables_num = list(df_turismo.drop(columns=['UserID']).select_dtypes(exclude=['object', int]).columns)

# Realizamos un independent-two-sample t-test para verificar si ambos grupos de interés son estadísticamente disímiles
data_antes = df_turismo[variables_num[0]]
noise = np.random.normal(loc=0, scale=2, size=len(data_antes))  # Ruido con media 0 y desviación estándar de 2
data_despues = data_antes + noise

t_statistic, p_value = stats.ttest_rel(data_antes, data_despues)

print(variables_num[0])
print(f"Media grupo A: {np.mean(data_antes):.2f}")
print(f"Media grupo B: {np.mean(data_despues):.2f}")
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretación
if p_value < 0.05:
    print("Rechazamos la hipótesis nula: Las medias de los dos grupos son significativamente diferentes y por ende, nuestra acción ha tenido un efecto en la población.")
else:
    print("No se rechaza la hipótesis nula: No hay evidencia suficiente para decir que las medias de los dos grupos son diferentes. No podemos decir que hemos tenido un impacto en la población.")
```

**<h2 id="section4" style="color: #007ACC;">4. Análisis de Varianza</h2>**

---
Análisis de Varianza (Analysis of Variance ANOVA)

---

En la sección aterior, vimos cómo estadísticas como la estadística $Z$ y la estadística $t$ nos permíten evaluar hipótesis entre dos pares de grupos o poblaciones tales como si una estimación de un parámetro a partir de una muestra pertenece a la población subyacente o comparar la similitud entre distribuciones de grupos/poblaciones a travéz de un estadístico de prueba. No obstante como podemos evidenciar, la más grande limitación de tanto la prueba de hipótesis por estadística $Z$ como $t$ es que solo se limitan a **dos grupos**. 

El análisis de varianza ANOVA, es una técnica estadística que se utiliza para comparar las medias de **tres o más grupos diferentes para determinar si al menos uno de ellos es significativamente diferente de los demás**. ANOVA es particularmente útil cuando se desea examinar si las diferencias observadas entre los grupos son lo suficientemente grandes como para concluir que no se deben al azar y corresponde una extensión de los t-tests a varios grupos.

#### **¿Que es un factor en ANOVA?**

Un factor dentro del contexto de ANOVA es una variable que permite **agrupar o segregar** los datos y permite comparar las medias entre diferéntes categorías, es decir, es una variable categórica que grácias a sus clases, permíte dividir los datos en múltiples conjuntos independientes. Esto quiere decir que el tipo de ANOVA que pueda realizarse dependerá de cuántas variables categóricas se tomen en cuenta para realizar un análisis de influencia sobre otra variable contínua.

#### **¿Cuál es la idea general de ANOVA**

EL análisis de varianza particiona la variabilidad total observada en los datos en dos componentes:

1. **Variabilidad entre grupos (Sum of Squares Between SSB)**: Mide la variabilidad debido a las diferencias entre las medias de los grupos.

2. **Variabilidad dentro de los grupos (Sum of Squares Within SSW)**: Mide la variabilidad dentro de los grupos, que es debida al azar o a diferencias dentro de cada grupo.

ANOVA se basa en el uso del **estadístico f** que compara la  **variabilidad entre grupos con la variabilidad dentro de los grupos**.

$$
f = \frac{\text{Varianza entre los grupos}}{\text{Varianza dentro de los grupos}}
$$

Si la variabilidad entre los grupos es grande en relación con la variabilidad dentro de los grupos, el valor de $f$ será grande, lo que sugiere que *al menos uno de los grupos tiene una media significativamente diferente*. Esto involucra la **no influencia o independencia entre los grupos**. En términos de la estadística $f$ y su relación con el testeo de hipótesis, podémos hacer las siguientes asociaciones:

- **Hipótesis**:
    - $H_0$ : las medias de los grupos provienen de la mísma población o, los grupos tienen influencia entre sí o, las distribuciones son similares.
    - $H_A$ : al menos una media de los grupos no proviene de la mísma poblacion o, no exíste influencia entre sí o, las distribuciones son distintas.

Con este panorama general, se plantean los siguientes escenarios, donde el primero ilustrado a continuación se interpreta como *al menos una de las medias no pertenece a la población o, las distribuciones son muy distintas entre sí*.

$$
\frac{\text{grande}}{\text{pequeña}} = \text{rechazamos } H_0 
$$

El siguiente caso se interpreta como *las medias son muy cercanas a la media poblacional, o es dificil distinguir entre distribuciones*.

$$
\frac{\text{similar}}{\text{similar}} = \text{fallamos en rechazar }H_0
$$

El siguiente caso se interpreta como *las medias son demasiado cercanas a la media poblacional, o las distribuciones son tan similares que éstas son prácticamente idénticas*:

$$
\frac{\text{pequeña}}{\text{grande}} = \text{fallamos en rechazar }H_0
$$

El análisis ANOVA posee las siguientes suposiciones:

1. **Independencia**: Las observaciones en cada grupo deben ser independiente entre sí.

2. **Normalidad**: Los datos dentro de cada grupo deben seguir una distribución normal.

3. **Homogeneidad de varianzas**: Los grupos deben tener varianzas similares (homocedasticidad)

#### **Distribución f**

El análisis de varianza se basa en la distribución **$F$ de Snedecor** (o simplemente distribución $F$). Esta distribución surge cuando se comparan dos estimaciones de varianza: una que mide la variabilidad entre grupos y otra que mide la variabilidad dentro de los grupos. La distribución $F$ es asimétrica y no puede tomar valores negativos, lo que refleja que siempre se está comparando una razón de varianzas (que son cuadradas y, por lo tanto, no negativas).

La distribución $F$ no sólo aparece en el análisis ANOVA sino también en las pruebas de hipótesis de igualdad de varianzas como Levene y Barlett o en regresión. Así pues, ANOVA se basa en ésta distribución para comparar la proporcion de las varianzas intergrupales e intragrupales. Algúnas de las propiedades de la distribucipon $F$ son:

1. **Aimetría positiva**: Es asimétrica hacia la derecha lo que significa que tiene una cola larga en el lado positivo.

2. **No-negativa**. Los valores de la distribución son siempre mayores o iguales acero ya que las varianzas y sus cocientes no pueden ser negativos.

3. **Grados de libertad**: La forma de la distribución $f$ depende de dos conjuntos de grados de libertad:
    - **m**: Grados de libertad asociados con el numerador (varianza del primer conjunto de datos o, de la varianza entre grupos).

    - **n**: Grados de libertad asociados con el denominador (varianza del segundo conjunto de datos o, de la varianza dentro de los grupos).

La función de densidad de probabilidad de la distribución $f$ viene dada por:


$$
f(x \mid m, n) = \frac{1}{x \, B\left(\frac{m}{2}, \frac{n}{2}\right)} \sqrt{\frac{(m x)^{m} \, n^{n}}{(m x + n)^{m + n}}}
$$

Donde  $B$ es la función beta dada por:

$$
B(x, y) = \int_0^1 t^{x-1} (1-t)^{y-1} \, dt
$$


```python
# Valores de x para graficar la PDF
x = np.linspace(0, 5, 500)

# Grados de libertad para diferentes distribuciones F
degrees_of_freedom = [(1,1),(2,1),(5, 10), (10, 15), (20, 25), (30, 40)]

plt.figure(figsize=(10, 6))

for m, n in degrees_of_freedom:
    y = stats.f.pdf(x, m, n)
    plt.plot(x, y, label=f'm={m}, n={n}')

plt.title('Distribución F para diferentes grados de libertad')
plt.xlabel('x')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

Los análisis de varianza son métricas de influencia o de "similaridad" intergrupales para más de dos grupos que nos permíten generar un análisis estadístico para casos muy complejos. Siendo éste tema tan amplio nos enfocaremos en solo cuatro casos del análisis ANOVA los cuáles provienen de las combinaciones siguientes:

1. **ANOVA por factor**: Corresponde a un específico tipo de ANOVA según la cantidad de factores implicados en el análisis. Siendo un facor una variable categórica que nos permite dividir en grupos  nuestros datos, **la cantidad de variables categóricas es la cantidad de factores involucrados en el análisis**, donde cada factor al poseer múltiples clases o categorías, implica que tendrémos tantos grupos en nuestros datos como combinatorias exístan entre las categorías disponibles en todos los factores.

2. **ANOVA por mediciones repetidas**: Corresponde al ANOVA usado para comparar las medias de diferentes condiciones o tratamientos, cuando las mediciones se realizan repetidamente sobre los mismos sujetos o unidades experimentales. 

Con lo anterior en mente, en el presente módulo verémos cuatro casos:

1. ANOVA con un factor sin repetición.
2. ANOVA con un factor con repetición.
3. ANOVA con dos factores sin repetición.
4. ANOVA con dos factores con repetición.

ANOVA de Un Factor o One-Way ANOVA sin Repetición

---

El ANOVA One-way sin repetición es la forma de ANOVA más sencilla que se utiliza para comparar las medias de tres o más grupos independientes para determinar si al menos una de ellas es significativamente diferente a las demás.  Así pues, un análisis ANOVA de un factor llevado acabo entre variables de un mísmo conjunto de datos extraídos a partir de la población, corresponde a evaluar la **influencia de una única variable categórica sobre una variable contínua**. 

#### **¿Cuándo usar ANOVA de un factor sin repetición?**

El análisis ANOVA  de un factor sin repetición es útil en situaciones donde:

1. Se quiere conocer si exíste influencia por parte de una variable categórica sobre una variable dependiente contínua sin repeticion de medidas, es decir, los miembros de cada grupo son distíntos y por ende independientes.

#### **Suposiciones del ANOVA de un factor sin repetición**

Las principales suposiciones del test ANOVA de un factor sin repetición son:

1. **Independencia de las observaciones**: Las observaciones en cada grupo son independientes unas de otras.
2. **Normalidad**: Los datos dentro de cada grupo deben estar normalmente distribuídos o, los resíduos del modelo (diferencias entre los valores observados y los valores predichos) siguen una distribución normal.
3. **Homocedasticidad**: La varianza de datos (o residuos en el caso de un modelo) es constante a través de los niveles de los factores.


#### **¿Cuáles son la Hipótesis del ANOVA de un factor sin repetición?**

El test ANOVA de un factor sin repetición al buscar la veracidad de influencia de una variable categórica sobre una numérica contínua tiene solo una hipótesis nula y una alternativa:

- **Hipótesis nulas $H_0$**:
    - No hay una diferencia significativa entre los grupos de factor y las medias de cada grupo no son estadísticamente diferentes.

- **Hipótesis nulas $H_A$**:
    - Exíste una diferencia significativa entre los grupos de factor y las medias de cada grupo son estadísticamente diferentes.

#### **Cálculo del ANOVA de un factor sin repetición**

Esto quiere  decir que el anova one-way va a tomar una única variable categórica para generar los grupos. Dentro del one-way se asume la siguiente descomposición de la varianza de los datos:

$$
\text{varianza total} = \text{varianza intergrupal} + \text{varianza intragrupal}
$$

Con lo anterior en mente, los pasos para calcular un anova one-way son los siguientes, donde tomarémos una variable **contínua** $X$ y la segmentaremos en grupos a partir de otra variable **categórica**:

1. **Definir las Hipótesis**:
   - **Hipótesis nula ($H_0$)**: Las medias de los grupos son iguales. $\mu_1$ = $\mu_2$ = $\mu_3$ = $\dots$ = $\mu_k$.
   - **Hipótesis alternativa ($H_A$)**: Al menos una de las medias es diferente.

2. **Cálculo de la Suma de Cuadrados**:

   - **Suma de Cuadrados entre Grupos (SSA)**: Representa la variabilidad entre las medias de los grupos.

     $$
     SSA = \sum_{i=1}^{k} n_i (\bar{X}_i - \bar{X}_\text{global})^2
     $$

     Donde:
     - $k$ es el número de grupos.
     - $n_i$ es el tamaño de la muestra en el grupo $i$.
     - $\bar{X}_i$ es la media del grupo $i$.
     - $\bar{X}_\text{global}$ es la media global de todas las observaciones.

   - **Suma de Cuadrados dentro de los Grupos (SSE)**: Representa la variabilidad dentro de cada grupo.

     $$
     SSE = \sum_{i=1}^{k} \sum_{j=1}^{n_i} (X_{ij} - \bar{X}_i)^2
     $$

     Donde:
     - $X_{ij}$ es la $j$-ésima observación en el grupo $i$.

   - **Suma de Cuadrados Total (SST)**: Es la variabilidad total en los datos.

     $$
     SST = SSA + SSE = \sum_{i=1}^{k} \sum_{j=1}^{n_i} (X_{ij} - \bar{X}_\text{global})^2
     $$

3. **Cálculo de los Grados de Libertad**:
   - **Grados de libertad entre grupos**: $df_{SSA} = k - 1$
   - **Grados de libertad dentro de los grupos**: $df_{SSE} = N - k$
     - Donde $N$ es el total de observaciones.

4. **Cálculo de la Media Cuadrática**:
   - **Media Cuadrática entre grupos (MSA)**: 
   $$
   MSA = \frac{SSA}{df_{SSA}} 
   $$

   - **Media Cuadrática dentro de los grupos (MSE)**: 
   $$
   MSE = \frac{SSE}{df_{SSE}} 
   $$

5. **Cálculo del Estadístico F**:
   - El estadístico $f$ compara la variabilidad entre los grupos con la variabilidad dentro de los grupos.

     $$
     f = \frac{MSA}{MSE} = \frac{\sigma_{SA}}{\sigma_{SE}}
     $$

     - Si $f$ es grande, sugiere que las diferencias entre las medias de los grupos no son solo debido al azar, lo que lleva a rechazar $H_0$.

6. **Comparación con el Valor Crítico**:
   - El valor $f$ obtenido se compara con el valor crítico de la distribución $f$ para un nivel de significancia dado $\alpha$. Si $f$ calculado es mayor que $f$ crítico, se rechaza la hipótesis nula $H_0$.



Para ilustrar el uso del ANOVA one-way de un factor sin repetición y cómo se espera evidenciar un efecto del factor en una variable contínua de interés, a continuación se presentan dos ejemplos que respectivamente muestran un caso de **no influencia del factor** y de **influencia del factor** sobre la variable aleatoria de interés.

```python
# Caso ANOVA one-way sin variabilidad entre grupos
# Generación de datos para tres grupos
np.random.seed(0)
group1 = np.random.normal(loc=70, scale=10, size=30)
group2 = np.random.normal(loc=72, scale=10, size=30)
group3 = np.random.normal(loc=73, scale=10, size=30)

# Creación de un DataFrame para los datos
df = pd.DataFrame({
    'Method_1': group1,
    'Method_2': group2,
    'Method_3': group3
})

# Realizar ANOVA de una vía
f_statistic, p_value = stats.f_oneway(df['Method_1'], df['Method_2'], df['Method_3'])

print(f"F-Statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretación
if p_value < 0.05:
    print("Rechazamos la hipótesis nula: Al menos un método tiene un rendimiento diferente.")
else:
    print("No se rechaza la hipótesis nula: No hay evidencia suficiente para decir que los métodos son diferentes.")

# Visualización de los datos
sns.boxplot(data=df)
plt.title('Comparación de Puntuaciones entre Métodos de Enseñanza')
plt.ylabel('Puntuaciones')
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

```python
# Caso ANOVA one-way con variabilidad entre grupos
# Simulación de datos
np.random.seed(42)

# Creación de un DataFrame con una variable categórica y una continua
data = {
    'Group': np.repeat(['A', 'B', 'C'], 30),
    'Score': np.concatenate([
        np.random.normal(55, 10, 30),  # Grupo A
        np.random.normal(46, 10, 30),  # Grupo B
        np.random.normal(60, 10, 30)   # Grupo C
    ])
}

df = pd.DataFrame(data)

# Visualizar los datos con un boxplot
sns.boxplot(x='Group', y='Score', data=df)
plt.title('Distribución de Puntuaciones por Grupo')
plt.xlabel('Grupo')
plt.ylabel('Puntuación')
plt.grid(linestyle='--', alpha=0.7)
plt.show()


# Separar los datos por grupo
group_A = df[df['Group'] == 'A']['Score']
group_B = df[df['Group'] == 'B']['Score']
group_C = df[df['Group'] == 'C']['Score']

# Realizar ANOVA One-Way
f_statistic, p_value = stats.f_oneway(group_A, group_B, group_C)

# Mostrar resultados
print(f"F-Statistic: {f_statistic}")
print(f"P-Value: {p_value}")

# Interpretación
alpha = 0.05
if p_value < alpha:
    print("Se rechaza la hipótesis nula: Hay diferencias significativas entre los grupos")
else:
    print("No se rechaza la hipótesis nula: No hay diferencias significativas entre los grupos")
```

ANOVA de Un Factor o One-Way ANOVA con Repetición

---

La variante del ANOVA One-Way llamada ANOVA One-Way con Medidas Repetidas (One-Way Repeated Measures ANOVA). Esta versión del ANOVA se utiliza cuando se realizan mediciones repetidas sobre los mismos sujetos o unidades experimentales bajo diferentes condiciones o en diferentes momentos.

- **ANOVA One-Way tradicional:** Se utiliza cuando se comparan las medias de varios grupos **independientes**. Por ejemplo, se podría comparar la presión arterial media de tres grupos diferentes de pacientes (tratamiento A, tratamiento B, tratamiento C).

- **ANOVA One-Way con Medidas Repetidas:** Se utiliza cuando los mismos sujetos son medidos varias veces bajo diferentes condiciones o en diferentes momentos. Por ejemplo, si un grupo de pacientes es sometido a un tratamiento en diferentes momentos, y se mide su presión arterial después de cada intervención. Aquí, los mismos pacientes son medidos repetidamente, por lo que la variabilidad dentro de los sujetos se tiene en cuenta.


#### **¿Cuándo usar ANOVA de un factor sin repetición?**

El análisis ANOVA  de un factor con repetición es útil en situaciones donde:

1. Se quiere conocer si una población ha cambiado su comportamiento a lo largo de diferentes mediciones o tiempos.

#### **Suposiciones del ANOVA de un factor sin repetición**

Las principales suposiciones del test ANOVA de un factor sin repetición son:

1. **Independencia de las observaciones**: Las observaciones en cada grupo son independientes unas de otras.
2. **Normalidad**: Los datos dentro de cada grupo deben estar normalmente distribuídos o, los resíduos del modelo (diferencias entre los valores observados y los valores predichos) siguen una distribución normal.
3. **Homocedasticidad**: La varianza de datos (o residuos en el caso de un modelo) es constante a través de los niveles de los factores.


#### **¿Cuáles son la Hipótesis del ANOVA de un factor sin repetición?**

El test ANOVA de un factor con repetición al buscar la veracidad a cerca del cambio de una variable numérica contínua tiene solo una hipótesis nula y una alternativa:

- **Hipótesis nulas $H_0$**:
    - No hay una diferencia significativa entre los momentos de medición y las medias de cada instancia de la población no son estadísticamente diferentes.

- **Hipótesis nulas $H_A$**:
    - Exíste una diferencia significativa entre los momentos de medición y las medias de cada instancia de la población son estadísticamente diferentes.


#### **Cálculo del ANOVA de un factor con repetición**

El one-way con medidas repetidas tiene en cuenta el **control de la variabilidad infividual** y la **variabilidad entre grupos o mediciones**. Esto quiere  decir que el ANOVA one-way con repetición multiples variables relacionadas entre si en la medida que cada una corresponde a un "tratamiento, estado o momento de medición de la población". Dentro del one-way con repetición se asume la siguiente descomposición de la varianza de los datos:

$$
\text{variabilidad de cada sujeto para cada tratamiento} = \text{variabilidad total} - \text{variabilidad entre estados/tratamientos}
- \text{variabilidad entre sujetos}
$$

Con lo anterior en mente, los pasos para calcular un anova one-way con repetición son los siguientes:

1. **Modelo Lineal**
El modelo matemático del ANOVA One-Way con Medidas Repetidas se puede expresar como:

$$
Y_{ij} = \mu + \tau_i + \pi_j + \epsilon_{ij}
$$

donde:
- $Y_{ij}$ es el valor observado para el sujeto $j$ en el tratamiento o condición $i$.
- $\mu$ es la media general.
- $\tau_i$ es el efecto del tratamiento $i$, que es el efecto fijo que queremos estudiar.
- $\pi_j$ es el efecto del sujeto $j$, que es un efecto aleatorio que controla las diferencias individuales entre sujetos.
- $\epsilon_{ij}$ es el término de error aleatorio asociado con la observación $Y_{ij}$.

2. **Hipótesis**
   - **Hipótesis nula ($H_0$)**: Las medias de las diferentes condiciones o tratamientos son iguales.
   $$
   H_0: \tau_1 = \tau_2 = \ldots = \tau_k
   $$
   - **Hipótesis alternativa ($H_A$)**: Al menos una de las medias es diferente.
   $$
   H_A: \text{Al menos una } \tau_i \text{ es diferente}
   $$

3. **Partición de la Suma de Cuadrados**
La variabilidad total en los datos se descompone en diferentes componentes:

    3.1. **Suma de Cuadrados Total (SST)**:
      $$
      SST = \sum_{i=1}^{k} \sum_{j=1}^{n} (Y_{ij} - \bar{Y})^2
      $$
      donde $\bar{Y}$ es la media general de todas las observaciones.

    3.2. **Suma de Cuadrados de Tratamientos (SSB)**: Representa la variabilidad debida a las diferencias entre las medias de los tratamientos.
    $$
    SSB = \sum_{i=1}^{k} n_i (\bar{Y}_i - \bar{Y})^2
    $$
    donde $\bar{Y}_i$ es la media del tratamiento $i$ y $n_i$ es el número de sujetos.

    3.3. **Suma de Cuadrados de Sujeto (SSS)**: Representa la variabilidad debida a las diferencias entre los sujetos.
    $$
    SSS = \sum_{j=1}^{n} k (\bar{Y}_j - \bar{Y})^2
    $$
    donde $\bar{Y}_j$ es la media de las mediciones para el sujeto $j$.

    3.4. **Suma de Cuadrados del Error (SSE)**: Representa la variabilidad no explicada dentro de cada sujeto para cada tratamiento.
    $$
    SSE = SST - SSB - SSS
    $$

4. **Grados de Libertad**
   - **Grados de libertad para los tratamientos ($df_B$)**: $ k - 1 $, donde $ k $ es el número de tratamientos o condiciones.
   - **Grados de libertad para los sujetos ($df_S$)**: $n - 1$, donde $n$ es el número de sujetos.
   - **Grados de libertad para el error ($df_E$)**: $(k - 1) \times (n - 1)$.

5. **Cuadrados Medios**
   - **Cuadrado Medio de Tratamientos (MSB)**:
   $$
   MSB = \frac{SSB}{df_B}
   $$
   - **Cuadrado Medio del Error (MSE)**:
   $$
   MSE = \frac{SSE}{df_E}
   $$

6. **Estadístico $f$**
El estadístico $f$ se calcula comparando la variabilidad explicada por los tratamientos con la variabilidad no explicada (error):

$$
F = \frac{MSB}{MSE}
$$


```python
# Cálculo de anova one way con medidas repetidas usando statsmodels
import statsmodels.api as sm
from statsmodels.stats.anova import AnovaRM

# Simulación de datos
np.random.seed(42)
n_subjects = 30
data = {
    'Subject': np.repeat(np.arange(1, n_subjects + 1), 3),
    'Condition': np.tile(['No Treatment', 'Treatment A', 'Treatment B'], n_subjects),
    'Score': np.concatenate([
        np.random.normal(75, 10, n_subjects),  # Sin tratamiento
        np.random.normal(80, 10, n_subjects),  # Tratamiento A
        np.random.normal(85, 10, n_subjects)   # Tratamiento B
    ])
}

df = pd.DataFrame(data)

# Realizar ANOVA One-Way con Medidas Repetidas
anova_rm = AnovaRM(df, 'Score', 'Subject', within=['Condition'])
res = anova_rm.fit()

# Mostrar resultados
print(res.summary())
```

ANOVA de Dos Factores o Two-Way ANOVA sin Repetición

---

El ANOVA de dos factores sin repetición (Two-Way ANOVA) es una extensión del análisis de varianza (ANOVA) de un factor, que permite evaluar el efecto de dos factores independientes en una variable dependiente continua. Matemáticamente, se utiliza para **descomponer la variabilidad total de los datos en componentes atribuibles a los dos factores y su interacción**. En éste sentido, el ANOVA permite examinar en nuestros datos lo siguiente:

1. Efecto individual de cada factor sobre una variable contínua (efectos principales).

2. Interacción entre los factores, es decir, cómo el efecto de un factor varía en función de los nivels del otro factor.

#### **¿Cuándo usar ANOVA de dos factores?**

El ANOVA de dos factores se usa en situaciones como:

- Exísten dos factores categóricos a investigar.
- Cada combinación de niveles de los factores tiene observaciones independientes (no hay medidas repetidas).
- Se quiere evaluar tanto los efectos principales como la interacción entre los factores.

#### **Suposiciones del ANOVA de dos factores**

1. **Independencia de las observaciones**: Las observaciones en cada grupo son independientes unas de otras.
2. **Normalidad**: Los datos dentro de cada grupo deben estar normalmente distribuídos o, los resíduos del modelo (diferencias entre los valores observados y los valores predichos) siguen una distribución normal.
3. **Homocedasticidad**: La varianza de datos (o residuos en el caso de un modelo) es constante a través de los niveles de los factores.
4. **Independencia factorial**: Los dos factores a usar son independientes entre si.

#### **¿Cuáles son la Hipótesis del ANOVA de dos factores?**

Debido a que el ANOVA de dos factores permíte evaluar tres aspectos de nuestros datos (influencia del factor uno, influencia del factor dos y la interacción entre los factores), éste tipo de test estadístico contiene tres hipótesis nulas y tres hipótesis alternativas:

- **Hipótesis nulas $H_0$**:
    - No hay una diferencia significativa entre los grupos del primer factor.
    - No hay una diferencia significativa entre los grupos del segúndo factor.
    - Un factor no tiene influencia sobre la influencia del otro factor sobre la variable contínua.

- **Hipótesis nulas $H_A$**:
    - Exíste una diferencia significativa entre los grupos del primer factor.
    - Exíste una diferencia significativa entre los grupos del segúndo factor.
    - Un factor tiene influencia sobre la influencia del otro factor sobre la variable contínua.

#### **Cálculo del ANOVA de dos factores sin repetición**

El ANOVA de dos factores sin repetición se basa en la siguiente descomposición de la varianza:

$$
\text{varianza total} = \text{var. explicada por factor A} + \text{var. explicada por factor B} + \text{var. explicada por la interacción A y B} + \text{var. de error}
$$

Esta descomposición de la varianza total de la variable contínua, la traducímos por notación término a término de la siguiente manera:

$$
SS_T = SS_A + SS_B + SS_{AB} + SS_{\text{err}}
$$

El test ANOVA de dos factores funciona bajo la lógica de que si la suma de los términos $SS_A + SS_B + SS_{AB}$ corresponde a la **mayor** contribución a la varianza total, se puede asegurar que los factores si tienen un efecto en la variable contínua, mientras que si el término de error $ SS_{\text{err}}$ contiene la mayor contribución a ala varianza total, se puede asegurar que los factores no tienen una influencia apreciable en la variable contínua. 

El procedimiento para calcular el ANOVA de dos factores es el siguiente:


1. **Modelo Matemático**

El modelo general para un ANOVA de dos factores sin repetición es:

$$
Y_{ijk} = \mu + \alpha_i + \beta_j + (\alpha\beta)_{ij} + \epsilon_{ijk}
$$

Donde:
- $Y_{ijk}$ es el valor observado de la variable dependiente para el $k$-ésimo sujeto en el nivel $i$ del factor $A$ y el nivel $j$ del factor $B$.
- $\mu$ es la media general de la población.
- $\alpha_i$ es el efecto del $i$-ésimo nivel del factor $A$.
- $\beta_j $ es el efecto del $j$-ésimo nivel del factor $B$.
- $(\alpha\beta)_{ij}$ es el efecto de interacción entre el $i$-ésimo nivel del factor $A$ y el $j$-ésimo nivel del factor $B$.
- $\epsilon_{ijk}$ es el término de error aleatorio, que se supone sigue una distribución normal $\mathcal{N}(0, \sigma^2)$.

2. **Descomposición de la Suma de Cuadrados**

El objetivo del ANOVA es dividir la variabilidad total observada en los datos en componentes atribuibles a los factores $A$ y $B$, a su interacción, y al error aleatorio.

- 2.1. **Suma de Cuadrados Total ($SS_{Total}$)**:
        
    $$
    SS_{Total} = \sum_{i=1}^{a} \sum_{j=1}^{b} \sum_{k=1}^{n} (Y_{ijk} - \bar{Y}_{G})^2
    $$

    Aquí, $\bar{Y}_{G}$  es la media global de todos los datos.

- 2.2. **Suma de Cuadrados para el Factor A ($SS_A$)**:
    
    $$
    SS_A = n \cdot b \sum_{i=1}^{a} (\bar{Y}_{i} - \bar{Y}_{G})^2
    $$

    Aquí, $\bar{Y}_{i}$ es la media del $i$-ésimo nivel del factor $A$.

- 2.3. **Suma de Cuadrados para el Factor B ($SS_B$)**:
    
    $$
    SS_B = n \cdot a \sum_{j=1}^{b} (\bar{Y}_{j} - \bar{Y}_{G})^2
    $$

    Aquí, $\bar{Y}_{j}$ es la media del $j$-ésimo nivel del factor $B$.

- 2.4. **Suma de Cuadrados para la Interacción $A × B$ ($SS_{AB}$)**: Para calcular la suma de cuadrados de la interacción, primero calculamos la suma de cuadrados entre los grupos, donde $\bar{Y}_{ij}$ es la media del grupo $ij$ producto de la combinación del nivel $i$ del factor $A$ y del nivel  $j$ del factor $B$.

    $$
    SS_{btw} = n \sum_{i=1}^{a} \sum_{j=1}^{b} (\bar{Y}_{ij} - \bar{Y}_{G})
    $$

    Finalmente, la a ésta suma de cuadrados se le resta la contribución de cada factor tal que:

    $$
    SS_{AB} = SS_{\text{btw}} - SS_A - SS_b
    $$

    La suma de los cuadrados para la interacción entre factores de manera genera, también puede expresarse como:
    
    $$
    SS_{AB} = n \sum_{i=1}^{a} \sum_{j=1}^{b} (\bar{Y}_{ij} - \bar{Y}_{i} - \bar{Y}_{j} + \bar{Y}_{G})^2
    $$

    Aquí, $\bar{Y}_{ij}$ es la media de las observaciones en la combinación del $i$-ésimo nivel del factor $A$ y el $j$-ésimo nivel del factor $B$.

    
- 2.5. **Suma de Cuadrados del Error ($SS_{Error}$)**:
  
    $$
    SS_{Error} = \sum_{i=1}^{a} \sum_{j=1}^{b} \sum_{k=1}^{n} (Y_{ijk} - \bar{Y}_{ij})^2
    $$

    Representa la variabilidad dentro de los grupos que no puede explicarse por los factores ni por su interacción.

3. **Grados de Libertad**

    Cada componente de la suma de cuadrados tiene un número asociado de grados de libertad ($df$):

    - **Grados de libertad del factor A**: $df_A = a - 1$
    - **Grados de libertad del factor B**: $df_B = b - 1$
    - **Grados de libertad de la interacción A × B**: $df_{AB} = (a - 1)(b - 1)$
    - **Grados de libertad del error**: $df_{Error} = ab(n - 1)$
    - **Grados de libertad total**: $df_{Total} = abn - 1$

4. **Cálculo de los Cuadrados Medios**

    Los cuadrados medios (Mean Squares, MS) se obtienen dividiendo las sumas de cuadrados por sus respectivos grados de libertad, lo cuál es equivalente a una varianza según cada uno de los criterios nombrados anteriormente:

    $$
    \sigma_A = \frac{SS_A}{df_A}, \quad \sigma_B = \frac{SS_B}{df_B}, \quad \sigma_{AB} = \frac{SS_{AB}}{df_{AB}}, \quad \sigma_{Error} = \frac{SS_{Error}}{df_{Error}}
    $$

5. **Cálculo de los Valores F**

    Cada efecto principal y la interacción se evalúan comparando el cuadrado medio del efecto con el cuadrado medio del error:

    $$
    f_A = \frac{\sigma_A}{\sigma_{Error}}, \quad f_B = \frac{\sigma_B}{\sigma_{Error}}, \quad f_{AB} = \frac{\sigma_{AB}}{\sigma_{Error}}
    $$

6. **Pruebas de Hipótesis**

- **Hipótesis nula para el factor A**: $H_0: \alpha_i = 0 $ para todos $i$.
- **Hipótesis nula para el factor B**: $H_0: \beta_j = 0$ para todos $j$.
- **Hipótesis nula para la interacción**: $H_0: (\alpha\beta)_{ij} = 0 $ para todos $ i, j $.

A continuación simulamos una situación muy sencilla:

- Factor A podría representar un tratamiento médico con dos niveles: A1 (sin tratamiento) y A2 (con tratamiento).
- Factor B podría representar un tipo de dieta con dos niveles: B1 (dieta regular) y B2 (dieta especial).
- Variable de respuesta: la variable continua podría ser una medida de un marcador biológico, como la presión arterial, después de aplicar el tratamiento y la dieta.

En éste sentido, el siguiente código genera unos datos sintéticos que intentan emular lo que ANOVA de dos factore busca capturar:
- `factor_A`: Representa el primer factor (tratamiento), donde A1 y A2 son los dos niveles.
    - A1: Sin tratamiento.
    - A2: Con tratamiento.

- `factor_B`: Representa el segundo factor (dieta), donde B1 y B2 son los dos niveles.
    - B1: Dieta regular.
    - B2: Dieta especial.
- `response`: Es la variable de respuesta que estamos midiendo. Se genera como una combinación de efectos del tratamiento, de la dieta, y de un ruido aleatorio (variabilidad que no está explicada por los factores).

Note que para el cálculo del ANOVA de dos factores hasta este momento se ha asumido que **todos los grupos tienen el mísmo número de individuos $n$** lo cuál si bien es una simplificación, facilita el análisis. En la práctica, no siempre es posible tener el mismo número de observaciones en cada grupo ya que las clases de los factores pueden estar desbalanceadas. En un diseño  ANOVA desbalanceado, los grupos pueden tener diferente número de observaciones pero esta agrega complejidad al modelo y su interpretación no es intuitiva.

```python
# Cálculo de estadística ANOVA sin repeticion mediante librerías estadísticas
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Generar datos simulados
np.random.seed(0)

# Variables
n = 30  # Número de observaciones por grupo
factor_A = np.repeat(['A1', 'A2'], n * 2)
factor_B = np.tile(np.repeat(['B1', 'B2'], n), 2)
data = 3 + 1.5 * (factor_A == 'A2') + 2 * (factor_B == 'B2') + 1.2 * (factor_A == 'A2') * (factor_B == 'B2')
data += np.random.normal(0, 1, len(data))

# Crear DataFrame
df = pd.DataFrame({
    'Factor_A': factor_A,
    'Factor_B': factor_B,
    'Response': data
})

# ANOVA de dos vías
model = ols('Response ~ Factor_A * Factor_B', data=df).fit()
#model = ols('Response ~ Factor_A + Factor_B + Factor_A:Factor_B', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Resultados
print(anova_table)

# Visualización
sns.boxplot(x='Factor_A', y='Response', hue='Factor_B', data=df)
plt.title('Boxplot de la respuesta para cada combinación de factores')
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

```python
# Cálculo manual de la estadística ANOVA de dos factores sin repetición

# Supongamos que tienes tus datos en un DataFrame de pandas
data = pd.DataFrame({
    'Factor1': ['A', 'A', 'B', 'B'] * 2,
    'Factor2': ['X', 'Y'] * 4,
    'Value': [10, 15, 20, 25, 12, 18, 22, 28]
})

def anova_two_way(df,factor1,factor2,valores):
    # Usaremos groupby y unstack para generar los promedios de cada grupo
    anova_data = data.groupby([factor1, factor2])[valores].mean().unstack()

    # Realiza el análisis ANOVA de dos factores con interacción
    # Extraemos las categorías y las dimensiones
    factor1 = anova_data.index
    factor2 = anova_data.columns
    n = len(factor1)
    m = len(factor2)

    # Hallamos las sumas de cuadrados
    grand_mean = anova_data.values.mean()
    ssTotal = ((data[valores] - grand_mean)**2).sum()

    ssA = m * ((anova_data.mean(axis=1) - grand_mean)**2).sum()
    ssB = n * ((anova_data.mean(axis=0) - grand_mean)**2).sum()

    ssAB = 0
    for i in range(n):
        for j in range(m):
            ssAB += (anova_data.iloc[i, j] - anova_data.iloc[i, :].mean() - anova_data.iloc[:, j].mean() + grand_mean)**2

    ssError = ssTotal - ssA - ssB - ssAB

     # Cálculo de los grados de libertad
    dfA = n - 1
    dfB = m - 1
    dfAB = (n - 1) * (m - 1)
    dfError = data.shape[0] - (n * m)

    # Cálculo de las varianzas por factor
    msA = ssA / dfA
    msB = ssB / dfB
    msAB = ssAB / dfAB
    msError = ssError / dfError

    # Cálculo de la estadística f
    F_A = msA / msError
    F_B = msB / msError
    F_AB = msAB / msError
     
    # Cálculo del valor p
    p_A = 1 - stats.f.cdf(F_A, dfA, dfError)
    p_B = 1 - stats.f.cdf(F_B, dfB, dfError)
    p_AB = 1 - stats.f.cdf(F_AB, dfAB, dfError)

    # Imprimimos los resultados
    print("ANOVA Results:")
    print(f"Factor A: F={F_A:.4f}, p={p_A:.4f}")
    print(f"Factor B: F={F_B:.4f}, p={p_B:.4f}")
    print(f"Interaction AB: F={F_AB:.4f}, p={p_AB:.4f}")
    
anova_two_way(data, 'Factor1', 'Factor2', 'Value')
```

---
ANOVA de Dos Factores o Two-Way ANOVA Con Repetición

---

El ANOVA de dos factores con repetición (Two-Way ANOVA with repetition) es una extensión del análisis de varianza dos factores donde  **un factor corresponde a una variable categórica la cuál genera un tipo de grupos** y exíste  **un segúndo factor pareado que incluye repetición**, es decir que el segúndo factor se constituye de los diferentes resultados obtenidos en dos momentos de medición hechos sobre la mísma población, teniendo en cuenta un factor categórico.

Esta técnica permite evaluar **el efecto de una división categórica a lo largo de mediciones repetidas efectuadas sobre una mísma población**, permitiendo **descomponer la variabilidad total de los datos en la contribución del factor de condición y el factor categórico, teniendo en cuenta un factor de error**. En éste sentido, el ANOVA permite examinar en nuestros datos lo siguiente:

1. Efecto del factor de repetición sobre la variable contínua dependiente.

2. Efecto del factor categórico (sin repetición) sobre la variable contínua dependiente.

2. Interacción entre los factores de repetición y el factor categórico sobre la variable contínua dependiente.

#### **¿Cuándo usar ANOVA de dos factores con repetición?**

El ANOVA de dos factores con repetición se usa en situaciones como:

- Exíste **un factor sin repetición** (variable categórica) que afecta a un **segúndo factor con repetición** (variable contínua), como por ejemplo ciando se quiere estudiar el efecto (segundo factor con repetición) de dos tratamientos diferentes (variable categórica o factor sin repetición) sobre pacientes y cada paciente recibe ambos tratamientos en diferentes momentos. 


- Cada combinación de niveles de los factores tiene observaciones dependientes, donde los mísmos sujetos son expuestos a todas las condiciones de uno de los factores, loque permite controlar la variabilidad entre sujetos.

- Se quiere evaluar tanto los efectos principales como la interacción entre los factores.

#### **Suposiciones del ANOVA de dos factores**

1. **Escala para el factor dependiente y categorización de los factores**: La variable dependiente debe ser contínua bajo una escala determinada (nivel de presión, salario, etc.) mientras que los factores implicados deben ser categóricos.

2. **Dependencia en los grupos del primer factor**: Las observaciones entre sujetos **NO son independientes**, aunque las observaciones  de cada sujeto pueden estar correlacionadas.

3. **Independencia en los grupos del segúndo factor**: Las observaciones entre grupos generados por la variable categórica **son independientes** entre sí.

4. **Normalidad**: Los datos dentro de cada grupo deben estar normalmente distribuídos o, los resíduos del modelo (diferencias entre los valores observados y los valores predichos) siguen una distribución normal.

5. **Homocedasticidad**: La varianza de datos (o residuos en el caso de un modelo) de los grupos formados por el facor **sin repetición** deben ser iguales o cercanas.

6. **Esfericidad (para el factor de repetición)**: Las varianzas de las diferencias entre todos los pares posibles de condiciones deben ser aproximadamente iguales.

#### **¿Cuáles son la Hipótesis del ANOVA de dos factores?**

Debido a que el ANOVA de dos factores con repetición permíte evaluar tres aspectos de nuestros datos (influencia del factor de repetición, influencia del factor categórico y la interacción entre la repetición y las categorías), éste tipo de test estadístico contiene tres hipótesis nulas y tres hipótesis alternativas:

- **Hipótesis nulas $H_0$**:
    - No hay una diferencia significativa entre los grupos del primer factor. Esto puede interpretarse como que no exíste una diferencia significativa entre **los instantes de medición**.
    - No hay una diferencia significativa entre los grupos del segúndo factor. Esto puede interpretarse como que la división categórica (ej. un conjunto de tratamientos) no tiene influencia.
    - Un factor no tiene influencia sobre la influencia del otro factor sobre la variable contínua. 

- **Hipótesis nulas $H_A$**:
    - Exíste una diferencia significativa entre los grupos del primer factor. Esto puede interpretarse como que exíste una diferencia significativa entre **los instantes de medición**.
    - Exíste una diferencia significativa entre los grupos del segúndo factor. Esto puede interpretarse como que la división categórica (ej. un conjunto de tratamientos) si tiene un efecto significativo.
    - Un factor tiene influencia sobre la influencia del otro factor sobre la variable contínua.

#### **Cálculo del ANOVA de dos factores con repetición**

El ANOVA de dos factores con repetición se basa en la siguiente descomposición de la varianza:

$$
\text{varianza total} = \text{var. explicada por factor S de repeticion} + \text{var. explicada por factor C categórico} + \text{var. explicada por la interacción S y C} + \text{var. de error}
$$

Esta descomposición de la varianza total de la variable contínua, la traducímos por notación término a término de la siguiente manera, donde la siguiente expresión hace referencia a la suma de los cuadrados de la diferencia en cada caso:

$$
SS_T = SS_C + SS_{T} + SS_{CS} + SS_{\text{Error}}
$$

El test ANOVA de dos factores con repetición funciona bajo la lógica de que si la suma de los términos $SS_C + SS_S + SS_{CS}$ corresponde a la **mayor** contribución a la varianza total, se puede asegurar que los factores si tienen un efecto en la variable contínua, mientras que si el término de error $ SS_{\text{Error}}$ contiene la mayor contribución a ala varianza total, se puede asegurar que los factores no tienen una influencia apreciable en la variable contínua. 

El procedimiento para calcular el ANOVA de dos factores es el siguiente:


1. **Modelo Matemático**

Suponiendo que tenemos un diseño de medidas repetidas de dos factores con un factor de condiciones o tratamientos $C$ con $k$ niveles y un segúndo factor de medida repetida $T$ con $m$ niveles de medición,
el modelo general para un ANOVA de dos factores sin repetición es:

$$
Y_{ijk} = \mu + \alpha_i + \beta_j + (\alpha \beta)_{ij} + S_k + \epsilon_{ijk}
$$

Donde:
- $Y_{ijk}$ es el avlor de la observación para el sujeto $k$, bajo el nivel $i$ del factor $C$ 
- $\mu$ es la media general de la población.
- $\alpha_i$ es el efecto del tratamiento o condición $i$ del factor $C$.
- $\beta_j$ es el efecto del nivel $j$ del factor de momentos $T$.
- $S_k $ es el efecto del sujeto $k$.
- $(\alpha \beta)_{ij}$ es la interacción entre los niveles $i$ y  $j$ de los factores de condición y momentos $C$ y $T$
- $\epsilon_{ijk}$ es el término de error aleatorio de la observación, que se supone sigue una distribución normal $\mathcal{N}(0, \sigma^2)$.

2. **Descomposición de la Suma de Cuadrados**

El objetivo del ANOVA de doble factor con repetición es dividir la variabilidad total observada en los datos en componentes atribuibles a los factores de repetición $T$ y el factor categórico $C$, a su interacción, y al error aleatorio.

- 2.1. **Suma de Cuadrados Total ($SS_{Total}$)**:
        
    $$
    SS_{Total} = \sum_{i=1}^{k} \sum_{j=1}^{m} \sum_{k=1}^{n} (Y_{ijk} - \bar{Y}_{G})^2
    $$

    Aquí, $\bar{Y}_{G}$  es la media global de todos los datos y $\bar{Y}_{ijk} es la observación del sujeto $k$ expuesto al tratamiento $i$ en el momento de medición $j$.

- 2.2. **Suma de Cuadrados Condicionales ($SS_C$)**:
    
    $$
    SS_C = n \cdot m \sum_{i=1}^{k} (\bar{Y}_{i} - \bar{Y}_{G})^2
    $$

    Aquí, $\bar{Y}_{i}$ es la media del $i$-ésma condición de $C$.

- 2.3. **Suma de Cuadrados Entre Situaciones/Tiempos ($SS_T$)**:
    
    $$
    SS_T = k \cdot n \sum_{j=1}^{m} (\bar{Y}_{j} - \bar{Y}_{G})^2
    $$

    Aquí, $\bar{Y}_{j}$ es la media del $j$-ésimo momento de medición.

- 2.4. **Suma de Cuadrados para la Interacción $C × T$ ($SS_{CT}$)**: Para calcular la suma de cuadrados de la interacción de repetición y la división categórica, primero calculamos la suma de cuadrados entre los grupos, donde $\bar{Y}_{ij}$ es la media del grupo $ij$ producto de la combinación del nivel $i$ del factor $C$ y del nivel  $j$ del factor $T$.
    
    $$
    SS_{CT} = n \sum_{i=1}^{k} \sum_{j=1}^{m} (\bar{Y}_{ij} - \bar{Y}_{i} - \bar{Y}_{j} + \bar{Y}_{G})^2
    $$

    Aquí, $\bar{Y}_{ij}$ es la media de las observaciones en la combinación del $i$-ésimo nivel del factor $C$ y el $j$-ésimo nivel del factor $T$.

    
- 2.5. **Suma de Cuadrados del Error ($SS_{Error}$)**:
  
    $$
    SS_{Error} = SST - SS_T -SS_C - SS_{CT}
    $$

    Representa la variabilidad dentro de los grupos que no puede explicarse por los factores ni por su interacción.

3. **Grados de Libertad**

    Cada componente de la suma de cuadrados tiene un número asociado de grados de libertad ($df$):

    - **Grados de libertad del factor C**: $df_C = k - 1$
    - **Grados de libertad del factor T**: $df_T = m - 1$
    - **Grados de libertad de la interacción C × T**: $df_{CT} = (k - 1) (m - 1)(n - 1)$
    - **Grados de libertad del error**: $df_{Error} = ab(n - 1)$
    - **Grados de libertad total**: $df_{Total} = abn - 1$

4. **Cálculo de los Cuadrados Medios**

    Los cuadrados medios (Mean Squares, MS) se obtienen dividiendo las sumas de cuadrados por sus respectivos grados de libertad, lo cuál es equivalente a una varianza según cada uno de los criterios nombrados anteriormente:

    $$
    \sigma_C = \frac{SS_C}{df_C}, \quad \sigma_B = \frac{SS_T}{df_T}, \quad \sigma_{CT} = \frac{SS_{CT}}{df_{CT}}, \quad \sigma_{Error} = \frac{SS_{Error}}{df_{Error}}
    $$

5. **Cálculo de los Valores F**

    Cada efecto principal y la interacción se evalúan comparando el cuadrado medio del efecto con el cuadrado medio del error:

    $$
    f_C = \frac{\sigma_C}{\sigma_{Error}}, \quad f_T = \frac{\sigma_T}{\sigma_{Error}}, \quad f_{CT} = \frac{\sigma_{CT}}{\sigma_{Error}}
    $$

6. **Pruebas de Hipótesis**

- **Hipótesis nula para el factor C**: $H_0: \alpha_i = 0 $ para todos $i$.
- **Hipótesis nula para el factor T**: $H_0: \beta_j = 0$ para todos $j$.
- **Hipótesis nula para la interacción**: $H_0: (\alpha\beta)_{ij} = 0 $ para todos $ i, j $.


```python
# Cálculo de estadística ANOVA con repeticion mediante librerías estadísticas
from statsmodels.stats.anova import AnovaRM

# Datos ficticios
np.random.seed(0)
data = {
    'Sujeto': np.repeat(np.arange(1, 11), 9),
    'Dieta': np.tile(np.repeat(['Dieta_1', 'Dieta_2', 'Dieta_3'], 3), 10),
    'Tiempo': np.tile(['T1', 'T2', 'T3'], 30),
    'Peso': np.random.normal(70, 5, 90)
}

df = pd.DataFrame(data)

# ANOVA de medidas repetidas factorial con interacción
aovrm = AnovaRM(df, 'Peso', 'Sujeto', within=['Dieta', 'Tiempo'])
res = aovrm.fit()

# Resultados
print(res)

# Graficar interacción
interaction_plot = pd.pivot_table(df, values='Peso', index='Tiempo', columns='Dieta')
interaction_plot.plot(marker='o')

plt.title('Interacción entre Dieta y Tiempo en la Pérdida de Peso')
plt.xlabel('Tiempo')
plt.ylabel('Peso Promedio')
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

Estimación e Inferencia

---


**La estimación** es un proceso fundamental en estadística que consiste en **inferir los valores de los parámetros desconocidos de una población a partir de una muestra de datos**. Estos parámetros pueden incluir la media, la varianza, la proporción, entre otros. La estimación nos permite hacer conjeturas informadas sobre la población sin tener que medir cada individuo en ella, lo que suele ser impráctico o imposible. 

En la presente sección, verémos una introducción a la estimación de parámetros más complejos que la media o la varianza, en concreto, observarémos como es posible mediante la teoría de la probabilidad deducir el comportamiento de la población a partir de la estimación de parámetros de modelos que pretenden ser la generalización del comportamiento bajo supuestos de su dinámica.

#### **Estimación Puntual y por Intervalos**

**La estimación puntual** es un tipo de estimación en la que se proporciona un único valor como estimado para un parámetro desconocido de la población. Este valor puntual se calcula a partir de los datos de la muestra. Por ejemplo, si queremos estimar la media de una población, podríamos utilizar la media muestral como una estimación puntual de la media poblacional.

Sin embargo, la estimación puntual no nos da información sobre la precisión de la estimación, es decir, cuánto puede variar este estimado del verdadero valor del parámetro poblacional.

Por otra parte, como vímos en la sección de pruebas de hipótesis, la **estimación por intervalos** es un método que estima el parámetro de dentro de un rango de valores en lugar de dar un único valor para la estimación. Este rango como vimos, se denomina como intervalo de confianza y proporciona una estimación más informativa, ya que refleja no solo el valor estimado del parámetro sino también la incertidumre asociada a dicha estimación.

#### **Maximum Likelihood Estimation (MLE)**

**Maximum Likelihood Estimation (MLE)** es un método de estimación puntual que se utiliza para estimar los parámetros de un modelo estadístico. El principio detrás de MLE es **encontrar los valores de los parámetros que maximizan la probabilidad de que los datos observados hayan sido generados por el modelo**.

En otras palabras, el MLE busca el valor del parámetro que hace que los datos observados sean "más probables" bajo el modelo estadístico asumido.

El MLE es un método muy popular en estadística debido a sus propiedades deseables, como la consistencia (a medida que el tamaño de la muestra aumenta, la estimación se aproxima al valor verdadero del parámetro) y la eficiencia (tiene la menor varianza posible entre los estimadores imparciales).

#### **Maximum A Posteriori Estimation (MAP)**

El **Maximum A Posteriori (MAP) Estimation** es un enfoque de estimación de parámetros en estadística bayesiana. A diferencia del Maximum Likelihood Estimation (MLE), que sólo considera la probabilidad de los datos dados los parámetros, MAP también incorpora una distribución previa sobre los parámetros. El objetivo del MAP es encontrar el valor de los parámetros que maximiza la distribución posterior, es decir, la probabilidad de los parámetros dados los datos observados.

#### **Relación con el Aprendizaje Automático (ML)**

**El aprendizaje automático (Machine Learning, ML)** utiliza extensivamente técnicas de estimación, incluido el MLE, para ajustar modelos a datos. En ML, el MLE se usa para ajustar modelos probabilísticos a los datos de entrenamiento, lo que implica encontrar los valores de los parámetros del modelo que mejor explican los datos observados.

Por ejemplo, en la regresión logística, un algoritmo de aprendizaje automático, se utiliza MLE para estimar los coeficientes de la regresión que maximizan la probabilidad de observar los datos de entrenamiento dados.

El machine learning como campo de estudio, se centra en el desarrollo de algoritmos que pueden aprender de datos y hacer predicciones o tomar decisiones basadas en esos datos mediante la estimación paramétrica no explicita, es decir, a través de un método de deducción y optimización de performance o explicabilidad que guia al algoritmo a encontrar la mejor respuesta sin que el programador, estadístico o usuario intervenga más alla que estipulando el método de búsqueda. Muchos algoritmos de machine learning, especialmente los de modelado probabilístico y los basados en funciones de pérdida, se basan en principios similares a los del MLE.

Ya que en las primeras secciones del presente módulo nos hemos dedicado a la estimación puntual y por intervalos, dedicarémos ésta sección a hablar de MLE y MLA.

**<h3 id="subsection51" style="color: #003366;">5.1. Maximum Likelihood Estimation</h3>**

---
Maximum Likelihood Estimation (MLE)

---

El **Maximum Likelihood Estimation (MLE)** es un método fundamental en estadística para estimar los parámetros de un modelo probabilístico. El objetivo del MLE es encontrar los valores de los parámetros que maximizan la **verosimilitud** de los datos observados, es decir, aquellos **parámetros que hacen que los datos observados sean más probables bajo el modelo**.

Supongamos que tenemos un conjunto de datos $\mathbf{X} = \{x_1, x_2, \dots, x_n\}$ y que estos datos provienen de una distribución paramétrica con función de densidad de probabilidad (pdf) $f(x; \theta) $, donde $\theta$ es un vector de parámetros desconocidos que queremos estimar.

La **verosimilitud** es la probabilidad de observar el conjunto de datos $\mathbf{X}$ dados los parámetros $\theta$. Si se asume que cada observación es independiente una de la otra, la función $f(x_i,\theta) da la probabilidad de que el dato puntual $x_i$ ha sido generado por el modelo, así pues si querémos tener un panorama a cerca de la probabilidad total de que dados unos parámetros se generen unas observaciones, la función de verosimilitud es el **producto de las probabilidades independientes de cada observación**:

$$
L(\theta; \mathbf{X}) = P(\mathbf{X} | \theta) = \prod_{i=1}^{n} f(x_i; \theta)
$$

El MLE busca los valores de $\theta$ que maximizan esta función de verosimilitud. Es decir, la combinación de parámetros que maximizan la probabilidad de que unos datos hallan sido observados dados un modelo, se encuentra mediante una optimización de la función (se pretende buscar un máximo dentro de la función de verosimilitud):

$$
\hat{\theta}_{\text{MLE}} = \text{argmax}_{\theta} \ L(\theta; \mathbf{X})
$$


Para simplificar los cálculos (usar una función objetivo más sencilla de manejar), en lugar de maximizar la verosimilitud directamente, a menudo se maximiza su logaritmo, conocido como la **log-verosimilitud**. Esto es posible porque el logaritmo es una función monótona creciente, lo que significa que maximizar la verosimilitud y maximizar la log-verosimilitud llevan a los mismos resultados.

$$
\ell(\theta; \mathbf{X}) = \log L(\theta; \mathbf{X}) = \sum_{i=1}^{n} \log f(x_i; \theta)
$$

El estimador MLE se encuentra entonces maximizando la log-verosimilitud:

$$
\hat{\theta}_{\text{MLE}} = \text{argmax}_{\theta} \ \ell(\theta; \mathbf{X})
$$

---
Estimación puntual de los Parámetros de una Distribución Normal

---

Consideremos el caso de estimar los parámetros de una **distribución normal** (gaussiana) $\mathcal{N}(\mu, \sigma^2)$, donde los parámetros a estimar son la media $\mu$ y la varianza $\sigma^2$ tal que $\theta = (\mu, \sigma^2)$.

Para empezar, ya que el modelo que queremos deducir es uno de forma normal, es decir, deseamos ajustar un modelo gaussiano a los datos observados, la función de densidad  a usar es la función de probabilidad para una distribución normal:

$$
f(x; \mu, \sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)
$$

Con esto en mente, asumiendo que **cada observación en los datos es independiente una de la otra, y que cada una de éstas ha sido generada a partir del modelo gaussiano**, la función de verosimilitud adquiere la forma:

$$
L(\mu, \sigma^2; \mathbf{X}) = \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x_i - \mu)^2}{2\sigma^2}\right)
$$

Simplificando, obtenemos:

$$
L(\mu, \sigma^2; \mathbf{X}) = \left(\frac{1}{\sqrt{2\pi\sigma^2}}\right)^n \exp\left(-\frac{1}{2\sigma^2} \sum_{i=1}^{n} (x_i - \mu)^2\right)
$$

Tomando el logaritmo de la función de verosimilitud eliminamos la productoria y la convertimos en una sumatoria gracias a las propiedades del logaritmo:

$$
\ell(\mu, \sigma^2; \mathbf{X}) = \log L(\mu, \sigma^2; \mathbf{X}) = -\frac{n}{2} \log(2\pi\sigma^2) - \frac{1}{2\sigma^2} \sum_{i=1}^{n} (x_i - \mu)^2
$$

Para encontrar los valores de $\mu$ y $\sigma^2$ que maximizan $ \ell(\mu, \sigma^2; \mathbf{X}) $, tomamos las derivadas parciales con respecto a $\mu$ y $\sigma^2$ y las igualamos a cero.

1. **Derivada con respecto a $ \mu $:**

$$
\frac{\partial \ell}{\partial \mu} = \frac{1}{\sigma^2} \sum_{i=1}^{n} (x_i - \mu) = 0
$$

Esto implica que:

$$
\sum_{i=1}^{n} (x_i - \mu) = 0
$$

$$
\hat{\mu}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

2. **Derivada con respecto a $\sigma^2$:**

$$
\frac{\partial \ell}{\partial \sigma^2} = -\frac{n}{2\sigma^2} + \frac{1}{2\sigma^4} \sum_{i=1}^{n} (x_i - \mu)^2 = 0
$$

Multiplicando ambos lados por $2\sigma^4$ y simplificando:

$$
\hat{\sigma}^2_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^{n} (x_i - \hat{\mu}_{\text{MLE}})^2
$$


```python
# Generamos datos simulados de una distribución normal con media 5 y desviación estándar 2
np.random.seed(0)
data = np.random.normal(5, 2, size=100)

# Estimación MLE de los parámetros
mu_mle = np.mean(data)
sigma_mle = np.sqrt(np.mean((data - mu_mle)**2))

print(f"MLE estimador de la media (mu): {mu_mle}")
print(f"MLE estimador de la desviación estándar (sigma): {sigma_mle}")

# Verificamos los resultados con la función de scipy
mu_scipy, sigma_scipy = stats.norm.fit(data)
print(f"Verificación con scipy: mu = {mu_scipy}, sigma = {sigma_scipy}")
```

---
Estimación de una Regresión Lineal con MLE

---

El MLE es una técnica probabilística que si bien asume la completa independencia de los datos, es lo suficientemente poderosa como para constituir un estimdor puntual de parámetros para una enorme variedad de modelos. En la presente subsección, ilustraremos por que el problema canonico dentro del Machine Learning, la regresión lineal, no solo está sustentado por las herramientas del cálculo diferencial sino que a su vez, por la teoría de la probabilidad y la estadística.

En este sentido, la lección más valiosa que podrémos ver en éste módulo (a parte del uso de pruebas de hipótesis) es el **origen de la función de error para los problemas de regresion dentro del machine learning**. Si bien esta función de error u objetivo proviene en un inicio del problema de regresión lineal, su definición cuadrática permíte ser aplicada a múltiples problemas y modelos dentro del aprendizaje automático. 

La formulación de la regresión lineal desde la probabilidad es la siguiente: suponiendo que tenemos un conjunto de datos $X = \{x_i,y_i\}_i^n$ donde $n$ es el número de datos, $x_i$ es el valor de la variable independiente para el dato $i$ (conocida) y $y_i$ el valor de la variable independiente del dato $i$ tal que el conjunto de datos $X$ en su colectivo siguen una tendencia lineal, podemos suponer que nuestra hipótesis es que dichos datos obedecen a un modelo lineal de la forma:


El **Maximum Likelihood Estimation (MLE)** es una técnica fundamental en estadística que se puede aplicar para estimar los parámetros de una regresión lineal. La regresión lineal es un modelo ampliamente utilizado para describir la relación entre una variable dependiente \( y \) y una o más variables independientes \( x_1, x_2, \dots, x_p \).

### 1. **Modelo de Regresión Lineal**

El modelo de regresión lineal simple, que involucra una sola variable independiente \( x \), se expresa como:

$$
y_i = \beta_0 + \beta_1 x_i + \epsilon_i
$$

Donde $b$ y $w$ son respectivamente los parámetros del modelo a estimar, siendo $b$ el intercepto con el eje y y $w$ la inclinación de la recta, nuestra hipótesis a demás se basa en el hecho de que el valor real para la variable dependiente $y_i$ corresponde a una combinación lineal de los parámetros anteriormente nombrados y la variable  de entrada $x_i$ más un término de error (o de desviación de la perfecta linealidad) $\epsilon_i$. 

Si se asume que los errores o desviaciones $\epsilon_i$ son **independientes** e **idénticamente distribuidos** según una distribución normal con media cero y varianza $\sigma^2$, es decir:

$$
\epsilon_i \sim \mathcal{N}(0, \sigma^2)
$$
$$
P(\epsilon) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{\epsilon^2}{2 \sigma^2}\right)
$$

La imágen a continuación, ilustra la formulación probabilística de la regresión lineal, donde se asume que la desviación del $y$ verdadero, se debe a causa de que este punto puede ser reproducido si se hace un muestreo aleatorio a partir de la distribución centrada en el modelo lineal. Note entonces que el modelo lineal es la **tendencia promedio de los datos**.

image.png

Con lo anterior en mente, si la distancia entre la estimación del modelo lineal para la entrada $x_i$ , $\hat{y}_i$, y el verdadero valor en los datos para la variable $y$ en el punto $i$, $y_i$, es $d_i = \hat{y}_i - y_i$, podémos traducir la probabilidad de producir el valor $y_i$ al rededor de la tendencia lineal central (donde las gaussianas están centradas) como la probabilidad de obtener el valor $\epsilon_i = d_i$ centrado en nuestro modelo lineal. Así pues, la probabilidad de obtener la $i$-ésima distancia entre los valores reales y las estimaciones es simplemente


$$
P(y_i|x_i,b,w) = P(d_i) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{d_i^2}{2 \sigma^2}\right)
$$

Donde en términos del modelo, se tiene:

$$
d_i = y_i - (x_i w + b)
$$

Si ahora consideramos todos los datos posibles, asumiendo la independencia de las observaciones, construimos la función de verosimilitud como 

$$
L(b,w;X) = P(Y|X,b,w) = \prod_{i=1}^{n} P(y_i|x_i,b,w)
$$

Como $y_i sigue una distribución normal, la función de densidad para cada \( y_i \) es:

$$
L(b, w, \sigma^2; \mathbf{y}) = \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(y_i - b - w x_i)^2}{2\sigma^2}\right)
$$

Para simplificar el proceso de maximización, tomamos el logaritmo de la función de verosimilitud para obtener la **log-verosimilitud**:

$$
\ell(b, w, \sigma^2; \mathbf{y}) = \log L(b, w, \sigma^2; \mathbf{y}) = -\frac{n}{2} \log(2\pi\sigma^2) - \frac{1}{2\sigma^2} \sum_{i=1}^{n} (y_i - b - w x_i)^2
$$


Ya que el objetivo es encontrar los valores de $b$, $w$ y $\sigma^2$  que maximicen $\ell(b, w, \sigma^2; \mathbf{y})$. Este problema se puede resolver tomando derivadas parciales de la log-verosimilitud con respecto a cada parámetro y luego igualando a cero para resolver el sistema de ecuaciones.

Maximicemos primero con respecto a $b$ y $w$. 

$$
\frac{\partial \ell}{\partial b} = \frac{1}{\sigma^2} \sum_{i=1}^{n} (y_i - b - w x_i) = 0
$$

$$
\frac{\partial \ell}{\partial w} = \frac{1}{\sigma^2} \sum_{i=1}^{n} x_i (y_i - b - w x_i) = 0
$$

Estas dos ecuaciones son conocidas como las **ecuaciones normales** para la regresión lineal, donde se asume un valor $\sigma^2 = 1$:

$$
\sum_{i=1}^{n} (y_i - \beta_0 - \beta_1 x_i) = 0
$$

$$
\sum_{i=1}^{n} x_i (y_i - \beta_0 - \beta_1 x_i) = 0
$$

Al resolver estas ecuaciones para $b$ y $w$, obtenemos las fórmulas conocidas para los estimadores de mínimos cuadrados ordinarios (OLS):

$$
\hat{w} = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}
$$

$$
\hat{b}_0 = \bar{y} - \hat{\beta}_1 \bar{x}
$$

Donde:
- $\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i $ es la media de $ x $,
- $\bar{y} = \frac{1}{n} \sum_{i=1}^{n} y_i $ es la media de $ y $.

En caso de que no se asuma un sigma igual a uno, para hallarlo basta simplemente con maximizar la función de log-verosimilitud con respecto a éste parámetro.

$$
\frac{\partial \ell}{\partial \sigma^2} = -\frac{n}{2\sigma^2} + \frac{1}{2\sigma^4} \sum_{i=1}^{n} (y_i - b - w x_i)^2 = 0
$$

Resolviendo para $ \sigma^2 $:

$$
\hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{\beta}_0 - \hat{\beta}_1 x_i)^2
$$

Lo cuál se puede rescribir simplemente como:

$$
\hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

Este estimador $\hat{\sigma}^2 $ es la **varianza residual** o **error cuadrático medio** la cuál es la función objetivo de la regresión lineal.

No obstante, hemos podido determinar la solución cerrada para los parámetros $b$ y  $w$ debido a la simplicidad del problema, esto quiere decir que no siempre será así, incluso en la regresión lineal múltiple. A pesar de ello, note que tanto la forma del MSE $\hat{\sigma}^2 $ como el segúndo término de la log-verosimulitud, indican que la verosimilitud es maximizada en la medida que la distancia cuadrática de los valores reales y los valores estimados para $y$ se reduzca.

El MLE aplicado a la regresión lineal simple produce los mismos estimadores que los obtenidos a través de los mínimos cuadrados ordinarios (OLS) como se puede ver en cualquier curso de cálculo diferencial, así, tenemos que los tres parámetros se definen como:

- **$\hat{b}$** y **$ \hat{w} $** son los estimadores de los coeficientes de regresión.
- **$\hat{\sigma}^2$** es el estimador de la varianza del término de error.


```python
# Regresión lineal simple por OLS
import statsmodels.api as sm

# Generamos datos simulados
np.random.seed(0)
n = 100
X = np.random.rand(n)
# La relación es lineal con algo de ruido
Y = 3 + 2 * X + np.random.randn(n)

# Añadimos un intercepto a X
X = sm.add_constant(X)

# Usamos el modelo OLS de statsmodels para obtener el MLE
model = sm.OLS(Y, X)
results = model.fit()

# Mostramos los resultados
print(results.summary())

# Valores MLE de los parámetros
print(f"Estimación de la constante (beta_0): {results.params[0]}")
print(f"Estimación del coeficiente (beta_1): {results.params[1]}")
```

Para determinar si los parámetros hallados en una regresión lineal son estadísticamente significativos, se utilizan **valores p** asociados a cada coeficiente del modelo. Estos valores p se calculan a partir de la distribución de **Student's t** y nos permiten hacer inferencias sobre si los coeficientes de la regresión son significativamente diferentes de cero. Para calcular el valor p, primero se determina el error estándar del parámetro hallado como:

$$
\text{SE}(\hat{\beta}_j) = \sqrt{\hat{\sigma}^2 \cdot (\mathbf{X}^T \mathbf{X})^{-1}_{jj}}
$$

Donde $(\mathbf{X}^T \mathbf{X})^{-1}_{jj}$ es el $j$-ésimo elemento diagonal de la matriz inversa del producto matricial $\mathbf{X}^T \mathbf{X}$.

Para calcular el estadístico $t$ para cada parámetro se determina la siguiente proporción:

$$
t_j = \frac{\hat{\beta}_j}{\text{SE}(\hat{\beta}_j)}
$$

Este estadístico t sigue una distribución t de Student con $n-k-1$ grados de libertad bajo la hipótesis nula $H_0 : \beta_j = 0$. Si el valor p de un coeficiente es menor que el nivel de significancia $\alpha$, comúnmente 0.05, rechazamos la hipótesis nula de que el coeficiente es igual a cero. Esto indica que **la variable asociada a ese coeficiente tiene un efecto significativo sobre la variable dependiente.** Si el valor p es mayor que $\alpha$, no podemos rechazar la hipótesis nula, lo que sugiere que la variable podría no tener un efecto significativo en el modelo.


```python
# Regresión lineal multivariable
import statsmodels.api as sm

# Generar datos de ejemplo
np.random.seed(0)
X = np.random.rand(100, 2)  # Dos variables explicativas
X = sm.add_constant(X)  # Añadir intercepto
beta = np.array([2, 3, 5])  # Coeficientes reales
y = X @ beta + np.random.normal(scale=1, size=100)  # Variable dependiente

# Ajustar el modelo de regresión
model = sm.OLS(y, X).fit()

# Mostrar los resultados
print(model.summary())
```

**<h3 id="subsection53" style="color: #003366;">5.3. Regresión Logística</h3>**

---
Estimación de una Regresión Logística con MLE

---

La **regresión logística** es una técnica utilizada para **modelar la probabilidad de un evento binario (tipo Bernoulli) en función de una o más variables explicativas**. La regresión logística se utiliza para predecir la probabilidad de que una observación pertenezca a una de dos categorías (por ejemplo, éxito o fracaso, 1 o 0). A diferencia de la regresión lineal, que predice valores continuos, la regresión logística predice probabilidades que están restringidas al intervalo [0, 1], lo que hace que éste modelo  sea considerado como un **modelo de clasificación**.

La regresión logística se basa en la función **sigmoide** o **logística**, que transforma una combinación lineal de variables independientes en una probabilidad:

$$
P(Y = 1 \mid \mathbf{X}) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_k x_k)}}
$$

Lo cuál es equivalente a:

$$
P(Y = 1 \mid \mathbf{X}) = \frac{1}{1 + \exp(-\mathbf{X}^\top \boldsymbol{\beta})} 
$$

donde $P(Y = 1 \mid \mathbf{X})$ es la probabilidad de que la variable dependiente $Y$ sea 1 dado el vector de características $\mathbf{X}$.  Por otra parte $\beta_0$ es el intercepto o bias,$\beta_1, \beta_2, \ldots, \beta_k$ son los coeficientes asociados a las variables explicativas $x_1, x_2, \ldots, x_k$. Así pues, nuestro problema se puede entender como:

$$
Y \mid \mathbf{X} \sim \text{Bernoulli}(p)
$$

Si consideramos las observaciones o datos, como independientes entre sí, la función de verosimilitud de  una sola observación es:

$$
\text{L}(\boldsymbol{\beta}) = p^{y} (1 - p)^{1 - y}
$$

Donde $p = \frac{1}{1 + \exp(-\mathbf{X}^\top \boldsymbol{\beta})}$ es la probabilidad de éxito (evento positivo). Por ende, la función de verosimilitud para todo el conjunto de datos es:

$$
\text{L}(\boldsymbol{\beta}) = \prod_{i=1}^{n} p_i^{y_i} (1 - p_i)^{1 - y_i} 
$$

Lo cuál es equivalente a:

$$
L(\beta) = \prod_{i=1}^{n} \left[ P(Y_i = 1 \mid \mathbf{X}_i)^{y_i} \times \left(1 - P(Y_i = 1 \mid \mathbf{X}_i)\right)^{1 - y_i} \right]
$$

Para simplificar el cálculo, se utiliza la log-verosimilitud, que es el logaritmo de la función de verosimilitud:

$$
\ell(\beta) = \log L(\beta) = \sum_{i=1}^{n} \left[y_i \log P(Y_i = 1 \mid \mathbf{X}_i) + (1 - y_i) \log \left(1 - P(Y_i = 1 \mid \mathbf{X}_i)\right) \right]
$$

Para encontrar los parámetros $\beta$ que maximizan la log-verosimilitud, se calculan las derivadas parciales de $\ell(\beta)$ con respecto a cada $\beta_j$ y se resuelven las ecuaciones resultantes:

$$
\frac{\partial \ell(\beta)}{\partial \beta_j} = \sum_{i=1}^{n} \left[y_i - P(Y_i = 1 \mid \mathbf{X}_i)\right] x_{ij} = 0
$$

siendo $x_{ij}$ es el valor de la variable $j$ para la observación $i$.

Note entonces que es gracias al principio de máxima verosimilitud que hemos hallado la **función de costo de un problema de clasificación binaria**  una vez consideramos, que maximizar la verosimilitud es equivalente a minimizar su negativo. En este sentido, la función de error de un problema de clasificación binaria es denominada como **Entropía Cruzada** y viene dad por:

$$
E(\beta) = - \ell(\beta)
$$


```python
# Regresión logística multivariable
import statsmodels.api as sm

# Generar datos de ejemplo
np.random.seed(0)
X = np.random.rand(100, 2)  # Dos características
X = sm.add_constant(X)  # Añadir intercepto
beta = np.array([1, 2, 3])  # Coeficientes reales
logits = X @ beta + np.random.normal(scale=1, size=100)  # Valores logit
p = 1 / (1 + np.exp(-logits))  # Probabilidades
y = np.random.binomial(1, p)  # Variable dependiente binaria

# Ajustar el modelo de regresión logística
model = sm.Logit(y, X).fit()

# Mostrar los resultados
print(model.summary())
```
