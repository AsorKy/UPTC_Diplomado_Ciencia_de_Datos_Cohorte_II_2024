Conceptos Bsicos de Estadstica Inferencial

---

La **estadstica inferencial** es una rama de la estadstica que se enfoca en hacer inferencias, predicciones o generalizaciones sobre una poblacin a partir de una muestra de datos. A diferencia de la estadstica descriptiva, que se limita a describir y resumir los datos disponibles, la estadstica inferencial va un paso ms all, utilizando los datos de la muestra para hacer afirmaciones o tomar decisiones sobre la poblacin de la que provienen esos datos.

La estadstica inferencial es esencial en la investigacin cientfica, ya que permite a los investigadores y profesionales tomar decisiones informadas basadas en datos. En lugar de limitarse a describir lo que ha ocurrido, la estadstica inferencial permite predecir lo que podra ocurrir en el futuro y evaluar la validez de las afirmaciones sobre una poblacin ms amplia. Esto tiene aplicaciones en diversas reas, incluyendo la medicina, la psicologa, la economa, las ciencias sociales, la ingeniera, y muchas ms.

Por ejemplo, en la investigacin mdica, la estadstica inferencial es vital para determinar si un nuevo tratamiento es efectivo y si los resultados obtenidos en un grupo de pacientes pueden generalizarse a toda la poblacin. En la industria, se utiliza para controlar la calidad, optimizar procesos y predecir la demanda de productos.

Dentro de los objetivos o tareas de la estadstica inferencial encontramos:

1. **Estimacin de Parmetros:** Uno de los objetivos principales de la estadstica inferencial es estimar los parmetros desconocidos de una poblacin, como la media, la varianza o la proporcin. Estas estimaciones se realizan a partir de los estadsticos calculados en una muestra. Por ejemplo, si se desea conocer la media de una poblacin, pero solo se tiene acceso a una muestra, se utiliza la media muestral como una estimacin de la media poblacional.

2. **Pruebas de Hiptesis:** Otro objetivo clave es la realizacin de pruebas de hiptesis, que permiten a los investigadores evaluar si existe suficiente evidencia en los datos muestrales para aceptar o rechazar una hiptesis sobre la poblacin. Por ejemplo, si un cientfico quiere saber si un nuevo medicamento es ms efectivo que un tratamiento estndar, puede usar una prueba de hiptesis para determinar si los resultados observados en un grupo de pacientes son lo suficientemente convincentes como para generalizar esa conclusin a todos los pacientes.

3. **Prediccin:** La estadstica inferencial tambin se utiliza para hacer predicciones sobre futuros eventos o comportamientos basados en los datos actuales. Por ejemplo, un modelo de regresin puede predecir el rendimiento acadmico de un estudiante en funcin de su rendimiento pasado y otras variables relacionadas.

4. **Generalizacin:** A partir de una muestra representativa, se pueden hacer generalizaciones sobre una poblacin entera. Esto es crucial en muchas reas de la ciencia, la economa, la medicina, y otros campos donde no es prctico o posible observar a toda la poblacin.

### **Fundamentos de la Estadstica Inferencial:**

A lo largo del mdulo, nos oscuparemos de diferenetes aspectos de la estadstica inferencial que nos capacitarn con las herramientas bsicas necesarias para llevar a cabo estimaciones de parmetros, pruebas de hiptesis, predicciones y generalizaciones de una poblacin a partir de muestras. As pues, para poder comenzar es crucial conocer los siguientes conceptos:

- **Muestras y Poblaciones:** La estadstica inferencial se basa en la relacin entre una muestra y su poblacin correspondiente. Una poblacin es el conjunto completo de elementos o individuos que son objeto de estudio, mientras que una muestra es un subconjunto representativo de esa poblacin. La clave para la inferencia es que *la muestra debe ser representativa de la poblacin para que las conclusiones sean vlidas*.

- **Sesgo y Precisin:** El sesgo se refiere a *cualquier error sistemtico que puede distorsionar los resultados*. Por ejemplo, si una muestra no es representativa de la poblacin (sesgo de seleccin), las inferencias hechas a partir de esa muestra pueden ser incorrectas. La precisin se refiere a *la consistencia o reproducibilidad de los resultados de la inferencia*. Se mejora aumentando el tamao de la muestra y minimizando las fuentes de error.

- **Error Muestral:** Cuando se estima un parmetro poblacional usando un estadstico muestral, siempre habr una diferencia entre el valor estimado y el valor verdadero del parmetro, conocida como error muestral. Este error es inevitable, pero se puede medir y manejar dentro de un rango aceptable usando el concepto de intervalos de confianza.

### **Tipos de Estadstica Inferencial:**

Los tipos de estadstica inferencial son:

1. **Inferencia Paramtrica:** Supone que los datos provienen de una poblacin que sigue una distribucin conocida (como la distribucin normal) y utiliza esta suposicin para hacer inferencias. Ejemplos incluyen la estimacin de la media poblacional y la realizacin de pruebas de hiptesis para medias y proporciones.

2. **Inferencia No Paramtrica:** No requiere suposiciones estrictas sobre la forma de la distribucin de la poblacin y es ms flexible, pero generalmente menos poderosa que la inferencia paramtrica. Ejemplos incluyen la prueba de rangos con signo de Wilcoxon y la prueba de chi-cuadrado.

3. **Mtodos Bayesianos:** Esta es una forma de inferencia que combina datos muestrales con informacin previa (a priori) usando el teorema de Bayes. Es particularmente til cuando se dispone de informacin previa slida o cuando se realizan inferencias en situaciones complejas o con datos limitados.

### **Relacin entre la Estadstica Inferencial y la Teora de la Probabilidad**

La estadstica inferencial y la teora de la probabilidad estn ntimamente relacionadas, ya que la estadstica inferencial se basa en los principios de la probabilidad para realizar inferencias y tomar decisiones sobre una poblacin a partir de una muestra. A continuacin, se describe la relacin entre estas dos disciplinas:

- **1. Fundamento Terico:** La teora de la probabilidad proporciona el marco matemtico que permite modelar la incertidumbre y el azar, conceptos esenciales en la estadstica inferencial. En esencia, la probabilidad cuantifica la incertidumbre asociada a los resultados de eventos aleatorios. La estadstica inferencial, por su parte, *utiliza esta teora para hacer inferencias sobre una poblacin a partir de una muestra de datos y los expresa como probabilidades que dan cuenta de la certidumbre de la estimacin o hiptesis*.

- **2. Distribuciones Muestrales:** La estadstica inferencial utiliza distribuciones muestrales, que son distribuciones de probabilidad de estadsticos (como la media o la proporcin) obtenidos a partir de mltiples muestras de una poblacin. Estas distribuciones muestrales son fundamentales para realizar estimaciones y pruebas de hiptesis ya que modelan el comportamiento del fenmeno sobre el cual se pretende inferir.

- **3. Estimacin e Intervalos de Confianza:** Los intervalos de confianza, una herramienta inferencial importante los cules estn construidos utilizando distribuciones de probabilidad. Un intervalo de confianza para una media, por ejemplo, se basa en la distribucin de probabilidad de la media muestral, que es derivada de la teora de la probabilidad. La probabilidad juega un papel crucial en la interpretacin de estos intervalos, que indican el *rango de valores en el que, con un cierto nivel de confianza (por ejemplo, 95%), se espera que se encuentre el parmetro poblacional*.

- **4. Pruebas de Hiptesis:** Al formular una hiptesis nula y una hiptesis alternativa, *se utiliza la probabilidad para determinar la probabilidad de observar un resultado tan extremo como el observado, o ms extremo, bajo la hiptesis nula*. El p-valor, una medida crucial en las pruebas de hiptesis, es una probabilidad calculada utilizando la distribucin de probabilidad del estadstico de prueba.

- **5. Regresin y Modelos Probabilsticos:** La regresin y otros modelos estadsticos son construidos bajo supuestos probabilsticos. Por ejemplo, en la regresin lineal simple, se asume que el trmino de error tiene una distribucin normal con media cero y varianza constante. Este supuesto probabilstico es fundamental para hacer inferencias sobre los coeficientes de regresin y para construir intervalos de confianza y realizar pruebas de hiptesis sobre estos coeficientes.


### **Conceptos Fundamentales de la Estadstica Inferencial**

- **Poblacin y Muestra**: 
  - **Poblacin**: Conjunto completo de todos los elementos que se desea estudiar.
  - **Muestra**: Subconjunto de la poblacin seleccionado para el estudio.

- **Parmetros y Estadsticos**:
  - **Parmetro**: Medida descriptiva calculada a partir de toda la poblacin (por ejemplo, la media poblacional).
  - **Estadstico**: Medida descriptiva calculada a partir de una muestra (por ejemplo, la media muestral).

- **Distribucin de Frecuencia y de Probabilidad**: Organizacin de datos que muestra la frecuencia (nmero de veces) de ocurrencia de cada valor o rango de valores en un conjunto de datos. Por otra parte, la distribucin de probabilidad siendo en el caso discreto, la versin normalizada de la distribucin de frecuencia y en el caso contnuo, la funcin de densidad normalizada de probabilidad, describe cmo se distribuyen los valores de una variable aleatoria de manera que asigna probabilidades de ocurrencia a cada valore de la variable dependiendo de su presencia en el espacio muestral.

- **Distribucin Muestral:** Es la distribucin de un estadstico como la media o la proporcin, obteniendo mltiples muestras de una poblacin. Es una forma de distribucin que recopila la informacin extrada a partir de las muestras de una poblacin empleada para efectuar inferencias y tomar desiciones sobre la poblacin de orgen.

- **Medidas de Resumen**: Mtricas numricas que compactan propiedades de la distribucin de frecuencia y evalan diferentes aspectos de la msma.
  - **Medidas de Tendencia Central**: Valores que representan un punto central o tpico en un conjunto de datos. Estas medidas ayudan a identificar la ubicacin central de los datos y proporcionan una indicacin del "centro" de la distribucin. Entre ellas encontramos la Media, mediana y moda.
  - **Medidas de Dispersin**: Grado de variabilidad o dispersin de los datos en relacin con una medida de tendencia central. Estas medidas ayudan a entender la dispersin de los datos y a identificar la consistencia de los mismos. Entre ellas encontramos el Rango, varianza y desviacin estndar.
  - **Medidas de Posicin Relativa**: Las medidas de posicin relativa indican la posicin de un valor en relacin con el resto del conjunto de datos. Estas medidas incluyen percentiles, cuartiles y deciles.
  - **Medidas de Forma**: Las medidas de forma describen la forma de la distribucin de los datos mediante aspectos tales como la Asimetra y curtosis.

- **Estimacin:** Corresponde al proceso de inferir el valor de un parmetro poblacional a partir de un estadstico muestral. Exsten dos tipos de estimacin:
  - **Estimacin puntual:** Proporciona el nico valor como estimacin del parmetro (por ejemplo, la media meustral como estimacin de la media poblacional).

  - **Estimacin por Intervalo:** Proporciona un rango de valores dentro del cual se espera que se encuentre el parmetro poblacinal, con un cierto nivel de confianza.

- **Pruebas de Hiptesis:** Las pruebas de hiptesis son procedimientos utilizados para tomar desiciones sobre los parmetros poblacionales basados en datos muestrales (o desiciones sobre la poblacin msma). Implica la formulacin de dos hiptesis:
  - **Hiptesis nula ($H_0$):** Es la hiptesis que se asume como **verdadera** a menos que la evidencia en contra de ella sea fuerte.
  - **Hiptesis alternativa ($H_1$):** Es la **hiptesis que se desea probar** y que se acepta si la hiptesis es rechazada.
  - **p-valor:** Es la probabilidad de obtener un resultado **tan extremo o ms extremo** si la hiptesis nula es verdadera. Un p-valor bajo indica que la hiptesis nula es improbable.
  - **Errores Tipo I y Tipo II:** Un error Tipo I ocurre cuando se rechaza incorrectamente la hiptesis nula, mientras que un error Tipo II ocurre cuando no se rechaza la hiptesis nula cuando esta es falsa.

- **Pruebas**:
  - **Pruebas Paramtricas**: Son pruebas de hiptesis que suponen que los datos provienen de una poblacin que sigue una distribucin de probabilidad conocida (como la distribucin normal).
  - **Pruebas No Paramtricas:** No requieren suposiciones estrictas sobre la distribucin de la poblacin y se utilizan cuando las condiciones para las pruebas paramtricas no se cumplen.

- **Inferencia Bayesiana:** La inferencia bayesiana es un enfoque de la estadstica inferencial que incorpora la probabilidad previa sobre un parmetro junto con los datos muestrales para actualizar las creencias sobre ese parmetro. Utiliza el Teorema de Bayes para combinar informacin previa y la probabilidad de la muestra.

Mecansmos de Muestreo

---

Los mecansmos de muestreo son aquellas maneras en las que se toman muestras de cierto tamao a partir de una poblacin mediante un enfoque o estrategia de extraccin de datos. As pues, encontramos tres tipos de mecanismos:

1. **Muestreo Aleatorio:** Corresponde a la extraccin de $m$ mustras a partir de una poblacin de tamao $n$ de manera completamente aleatoria tal que $m<<n$.
2. **Muestreo Sistemtico:** Mtodo de muestreo probabilstico en el que se seleccionan elementos de una poblacin a intervalos regulares. Consiste en seleccionar cada $k$ elementos de la poblacin, siendo $k$ una constante predeterminada.
3. **Muestreo Estratificado:** Mtodo de muestreo probabilstico que consiste en dividir la poblacin en distintos subgrupos o estratos en funcin de determinadas caractersticas. El objetivo es garantizar que cada subgrupo est bien representado en la muestra para obtener estimaciones ms exactas y precisas de los parmetros de la poblacin. El muestreo estratificado es ventajoso porque garantiza que cada subgrupo o estrato est adecuadamente representado en la muestra, lo que permite obtener estimaciones ms precisas de los parmetros de poblacin de cada subgrupo. Este mtodo es especialmente til cuando existen diferencias o variaciones sustanciales entre los subgrupos de la poblacin. Ayuda a reducir la variabilidad del muestreo y puede proporcionar resultados ms fiables y precisos en comparacin con otras tcnicas de muestreo.

```python
# Definicin de mapeos para 'preferred_device'

device_mapping= {'iOS and Android': 'Mobile',
                'iOS' : 'Mobile',
                'ANDROID' : 'Mobile',
                'Android': 'Mobile',
                'Android OS' : 'Mobile',
                'Other': 'Mobile',
                'Others' : 'Mobile',
                'Tab' : 'Mobile'}

df_turismo['preferred_device'] = df_turismo['preferred_device'].replace(device_mapping)

# Definicin de mapeos para 'yearly_avg_Outstation_checkins'
df_turismo['yearly_avg_Outstation_checkins']= df_turismo['yearly_avg_Outstation_checkins'].replace('*',np.nan)
df_turismo['yearly_avg_Outstation_checkins'] = pd.to_numeric(df_turismo['yearly_avg_Outstation_checkins'], errors='coerce', downcast='integer')
df_turismo['yearly_avg_Outstation_checkins'].unique()

# Definicin de mapeos para 'following_company_page'
page_mapping= {'1': 'Yes',
               '0' : 'No',
               'Yeso' : 'Yes',
               1: 'Yes',
               0: 'No'}
df_turismo['following_company_page'] = df_turismo['following_company_page'].replace(page_mapping)

# Definicin de mapeos para 'member_in_family'
df_turismo['member_in_family']= df_turismo['member_in_family'].replace('Three',3)
df_turismo['member_in_family'] = pd.to_numeric(df_turismo['member_in_family'], errors='coerce', downcast='integer')

# Definicin de mapeos para 'working_flag'
df_turismo['working_flag']= df_turismo['working_flag'].replace('0','No')
```

```python
# Remplazamos los posibles strings 'nan' y 'NaN' por el valor ausente que pandas reconoce por defecto el cual es np.nan
df_turismo = df_turismo.applymap(lambda x: np.nan if x == 'nan' else x)
df_turismo = df_turismo.applymap(lambda x: np.nan if x == 'NaN' else x)

# Imputacin para variables categricas
cat_columns = df_turismo.select_dtypes(exclude=['int64', 'float64']).columns
for i in cat_columns:
    df_turismo[i].fillna(df_turismo[i].mode()[0], inplace = True)
    
# Imputacin para variables numricas
num_columns = df_turismo.select_dtypes(exclude=['object']).columns
for column in num_columns:
    median_value = df_turismo[column].median()
    df_turismo[column].fillna(median_value, inplace=True)
```

```python
# Funcin para ejecutar un muestreo sistemtico 

def sys_sampling(dataframe, step, start=0):
  indexes = np.arange(start, len(dataframe), step=step)
  systematic_sample = dataframe.iloc[indexes]
  return systematic_sample
```

A continuacin, se define una funcin de muestreo estratificado donde se especifican las columnas bajo las cules se establece la estratificacin, los valores de estratificacin y las proporciones requeridas para cada valor de estratificacin.

```python
def stratified_data(data, columnas_estratificacion, valores_estrat, prop_estrat, random_state=None):
    
    df_estrat = pd.DataFrame(columns = data.columns) # Creamos un data frame vaco con los nombres de las columnas del dataframe
    # Creamos la columna estratificada
    data['estratificado'] = data[columnas_estratificacion].astype(str).agg('_'.join, axis=1)
    
    pos = -1
    for i in range(len(valores_estrat)): # iteracin sobre los valores estratificados
        pos += 1
        if pos == len(valores_estrat) - 1: 
            ratio_len = len(data) - len(df_estrat) # si es la iteracin final calcula el nmero de valores de salida tenga el mismo nmero de filas que de entrada
        else:
            ratio_len = int(len(data) * prop_estrat[i]) # calcula el nmero de filas segn la proporcin deseada

        df_filtrado = data[data['estratificado'] ==valores_estrat[i]] # filtra los datos de origen segn los valores seleccionados en la estratificacin de datos
        df_temp = df_filtrado.sample(replace=True, n=ratio_len, random_state=random_state) # haz un sample de los datos filtrados usando la ratio que hemos calculado
        
        df_estrat = pd.concat([df_estrat, df_temp]) # junta las tablas de sample con la estratificada para producir el resultado final
        
    return df_estrat # Return the stratified, re-sampled data   
```

Ley de Grandes Nmeros

---

La Ley de los Grandes Nmeros (LGN) es un principio fundamental en la teora de la probabilidad y la estadstica que describe el comportamiento del promedio de una gran cantidad de observaciones independientes e idnticamente distribuidas. Segn esta ley, a medida que el tamao de la muestra aumenta, la media aritmtica de las observaciones se aproxima cada vez ms al valor esperado o promedio de la poblacin de la que se toma la muestra.

En otras palabras, cuanto mayor sea el nmero de repeticiones de un experimento aleatorio, ms cerca estar la media muestral de la media poblacional verdadera.

### **Tipos de la Ley de los Grandes Nmeros**

Hay dos versiones principales de la Ley de los Grandes Nmeros:

1. **Ley Dbil de los Grandes Nmeros (WDLL):**
   - Establece que, para un tamao de muestra suficientemente grande, la media muestral converge en probabilidad al valor esperado de la variable aleatoria. Esto significa que la probabilidad de que la media muestral difiera del valor esperado por ms de un margen pequeo tiende a cero a medida que el tamao de la muestra aumenta.
   
   - Formalmente, si $X_1, X_2, \dots, X_n$ son variables aleatorias independientes e idnticamente distribuidas con valor esperado  $E(X_i) = \mu$, entonces:
     $$
     \lim_{n \to \infty} P\left(\left|\frac{1}{n} \sum_{i=1}^{n} X_i - \mu\right| > \epsilon\right) = 0 \text{ para cualquier } \epsilon > 0.
     $$

2. **Ley Fuerte de los Grandes Nmeros (SLLN):**
   - Establece que la media muestral converge casi seguramente (o con probabilidad 1) al valor esperado de la variable aleatoria a medida que el tamao de la muestra tiende a infinito. Esto es un resultado ms fuerte que la ley dbil porque implica que la convergencia no solo es probable, sino que ocurrir con certeza en casi todos los casos.
   
   - Formalmente, si $X_1, X_2, \dots, X_n$ son variables aleatorias independientes e idnticamente distribuidas con valor esperado $E(X_i) = \mu$, entonces:
     $$
     P\left(\lim_{n \to \infty} \frac{1}{n} \sum_{i=1}^{n} X_i = \mu\right) = 1.
     $$

### **Propiedades de la Ley de los Grandes Nmeros**

Para que la ley de los grandes nmeros sea vlida, debe satisfacerse:

1. **Independencia muestral:** Las muestras generadas a partir de una poblacin deben ser tomadas aleatoriamente.
2. **Significancia de tamao:** El tamao de la muestra debe ser estadsticamente significativa para dar razn del parmetro a estimar.
3. **Independencia observacional:** Cada observacin de la poblacin debe ser independiente de las otras.

Algnas de las propiedades de la ley de grandes nmeros son:

1. **Convergencia al Valor Esperado:** La LGN asegura que la media muestral converge al valor esperado de la distribucin subyacente cuando el tamao de la muestra es suficientemente grande.

2. **Independencia y Distribucin Idntica:** La LGN requiere que las *variables aleatorias sean independientes y sigan la misma distribucin*. Si estas condiciones no se cumplen, la ley puede no ser aplicable.

3. **Tamao de Muestra:** Aunque la LGN garantiza la convergencia, no especifica la rapidez con que esto ocurre. En la prctica, el tamao de muestra necesario para que la media muestral sea un buen estimador de la media poblacional puede variar segn la distribucin de los datos.

4. **Aplicabilidad General:** La LGN es aplicable a una amplia gama de distribuciones, incluyendo aquellas con una varianza infinita, siempre que el valor esperado exista y sea finito.

Teorema del Limite Central

---

El **Teorema del Lmite Central (TLC)** es uno de los resultados ms fundamentales en estadstica y teora de probabilidades. Este teorema establece que, bajo ciertas condiciones, la distribucin de la suma (o media) de un gran nmero de variables aleatorias independientes e identicamente distribuidas (i.i.d.) tiende a ser aproximadamente normal, independientemente de la distribucin original de las variables.

Si $X_1, X_2, ..., X_n$ son variables aleatorias independientes e identicamente distribuidas con media $\mu$ y varianza $\sigma^2$, entonces la distribucin de la suma $S_n = X_1 + X_2 + ... + X_n$ tiende a una distribucin normal con media $N\mu$ y varianza $N\sigma^2$ cuando $N$ es grande.

Para la media muestral $\bar{X} = \frac{S_n}{n} $, la distribucin se aproxima a una distribucin normal con media $\mu$ y varianza $\frac{\sigma^2}{n}$. con esto en mente, se tienen las siguientes propiedades:

1. **Convergencia a la Normalidad**: Independientemente de la forma de la distribucin original, la distribucin de la media muestral se aproximar a una normal si el tamao de la muestra es suficientemente grande.
2. **Tamao de la Muestra**: Aunque el TLC se aplica a muestras grandes, en la prctica, una muestra de tamao 30, 50 o ms suele ser suficiente para que la distribucin de la media muestral se aproxime a una normal.
3. **Aplicacin Prctica**: El TLC permite hacer inferencias sobre la media de poblaciones grandes, incluso cuando la distribucin de la poblacin es desconocida.


El ejemplo ms simple del teorema del lmite central, consiste en generar un numero $N$ determinado de muestras de tamap $n$ creadas a partir del muestro uniforme y aleatorio para las cuales calculamos la media. Cuando graficamos la distribucin de stas medias obtenemos una distribucin normal.

Otro ejemplo del TLC, corresponde a la distribucin Binomial de $N$ experimentos de Bernoulli con probabilidad de exito $p$.
```python
# Definir los posibles resultados de un lanzamiento de moneda
outcomes = ['H', 'T']  # H para "Cara" (Heads), T para "Cruz" (Tails)

# Generar el espacio muestral para tres lanzamientos
sample_space = list(product(outcomes, repeat=10))

# Definir la variable aleatoria X como el nmero de "caras" (H) en tres lanzamientos
def count_heads(trial):
    return trial.count('H')

# Calcular la funcin de probabilidad (PMF)
results = [count_heads(trial) for trial in sample_space]
mu, std = np.round(np.mean(results),3), np.round(np.std(results),3)
pmf = Counter(results)

# Normalizar las frecuencias para obtener probabilidades
total_outcomes = len(sample_space)
for x in pmf:
    pmf[x] /= total_outcomes

# Generamos el grfico de la distribucin PMF
print('Valor esperado np = ', {10 * 0.5})
print('Varianza np(1-p) = ', {10 * 0.5*(1-0.5)})
graficacion_TLC(pmf, mu, std)
```

Intervalos de Confianza

---

Un **intervalo de confianza** es un rango de valores que se utiliza para *estimar un parmetro poblacional desconocido con un nivel de certeza especificado*. En otras palabras, proporciona un rango dentro del cual es probable que se encuentre el valor verdadero del parmetro basado en una muestra de datos. Nos permiten:

1. **Estimacin de Parmetros**: Los intervalos de confianza permiten establecer el rango de posibles valores de parmetros poblacionales, como la media o la proporcin, con un grado de certeza. En lugar de proporcionar un nico valor estimado, ofrecen un rango que refleja la incertidumbre de la estimacin.

2. **Toma de Decisiones**: Son tiles en la toma de decisiones al proporcionar una medida de la precisin de las estimaciones. Por ejemplo, en un estudio clnico, un intervalo de confianza para la diferencia en eficacia entre dos tratamientos ayuda a determinar si la diferencia observada es estadsticamente significativa y relevante.

3. **Evaluacin de Resultados**: Permiten evaluar la variabilidad y la precisin de los resultados obtenidos a partir de una muestra. 

4. **Comunicacin de Resultados**: Facilitan la comunicacin de la incertidumbre y la fiabilidad de las estimaciones a otros profesionales y al pblico en general.

Exsten diferentes enfoques para hacer el clculo de intervalos de confianza, dentro de los cules encontramos:

1. **Mtodos Basados en la Distribucin Normal**: Cuando la muestra es grande o la distribucin subyacente es normal, se puede usar la distribucin normal estndar para calcular el intervalo de confianza. Se utilizan valores crticos de la distribucin normal para determinar el rango.

2. **Mtodos Basados en la Distribucin $t$ de Student**: Para muestras pequeas y cuando la desviacin estndar poblacional no es conocida, se emplea la distribucin $t$ de Student. La distribucin $t$ es similar a la normal pero tiene colas ms anchas, lo que permite captar la mayor incertidumbre en las estimaciones de pequeas muestras.

3. **Mtodos de Bootstrap**: Tcnicas de remuestreo como el bootstrap pueden ser usadas para estimar intervalos de confianza sin asumir una distribucin especfica para los datos. El mtodo bootstrap involucra la creacin de mltiples muestras a partir de la muestra original y el clculo del intervalo basado en estos datos simulados.

4. **Mtodos No Paramtricos**: En casos donde no se puede asumir una distribucin especfica para los datos, se pueden utilizar mtodos no paramtricos, como los intervalos de confianza de percentiles, para estimar la variabilidad del parmetro poblacional.

5. **Mtodos Bayesianos**: En la inferencia bayesiana, los intervalos de confianza se basan en distribuciones a posteriori y se denominan intervalos de credibilidad. Estos intervalos reflejan la probabilidad de que el parmetro poblacional se encuentre dentro del intervalo dado los datos observados.

Distribucin $Z$

---

La **distribucin Z**, tambin conocida como **distribucin normal estndar**, es una distribucin normal con una media de 0 y una desviacin estndar de 1. Es una forma de la distribucin normal que se utiliza para *simplificar el anlisis de datos y realizar comparaciones entre diferentes conjuntos de datos*. La distribucin $Z$ se calcula estandarizando una variable aleatoria normal. Esto se hace transformando los datos de una distribucin normal con media $\mu$ y desviacin estndar $\sigma$ a una distribucin normal estndar con media 0 y desviacin estndar 1. La frmula para convertir una puntuacin $X$ en una puntuacin $Z$ es:

$$
Z = \frac{X - \mu}{\sigma}
$$

1. **Comparacin de Datos**: Permite comparar datos provenientes de diferentes distribuciones normales al convertirlas a una escala comn.

2. **Pruebas de Hiptesis**: Se utiliza para calcular valores crticos en pruebas de hiptesis, como la prueba $Z$ para medias y proporciones.

3. **Intervalos de Confianza**: Se emplea para calcular intervalos de confianza para la media cuando la desviacin estndar de la poblacin es conocida.

4. **Puntuaciones Estndar**: Permite determinar cuntas desviaciones estndar un dato est alejado de la media, lo que es til en anlisis estadsticos y en la deteccin de valores atpicos.

$$

$$

El **Error Estndar** es un concepto fundamental en la estadstica inferencial, y su comprensin es clave para interpretar correctamente los resultados de las estimaciones basadas en muestras. El Error Estndar es una medida de la **precisin de una estimacin estadstica**. Especficamente, cuando hablamos del **Error Estndar de la Media (EEM)**, nos referimos a la dispersin o variabilidad de las medias muestrales alrededor de la media verdadera de la poblacin. En otras palabras, el Error Estndar cuantifica *cunto vara la media de una muestra particular en relacin con la media de la poblacin de la que proviene*. El Error Estndar de la Media se relaciona con la Desviacin Estndar de la siguiente manera:

$$
\text{EEM} = \frac{\sigma}{\sqrt{n}} 
$$

Donde $\sigma$ es la desviacin estndar de la poblacin y $n$ es el nmero de observaciones de la muestra poblacional. No obstante ya que la desviacin estndar de la poblacin rara vez se conoce, el error estndar de la media suele estimarse como la desviacin estndar de la muestra dividida por la raiz cuadrada del tamao de la muestra.

$$
\text{EEM} = \frac{\sigma_{\text{muestra}}}{\sqrt{n}} 
$$

El Error Estndar indica cmo varan las medias muestrales alrededor de la media poblacional (o un estimado de ella). Un EEM pequeo sugiere que la media muestral est muy cerca de la media poblacional, lo que implica que nuestra muestra es una buena representacin de la poblacin. Por el contrario, un EEM alto sugiere mayor variabilidad y menos precisin en nuestra estimacin.

El **tamao de la muestra  determina la precisin de la estimacin** ya que a medida que aumenta el tamao de la muestra, el Error Estndar disminuye, lo que significa que las medias muestrales se vuelven ms cercanas a la media verdadera de la poblacin. Esto se debe a que ms datos proporcionan una estimacin ms precisa. El error estndar puede ser empleado para:

- **Construccin de Intervalos de Confianza:** Se utiliza para determinar cun confiable es una estimacin, ayudando a construir intervalos dentro de los cuales es probable que se encuentre el parmetro poblacional.

- **Pruebas de Hiptesis:** Se utiliza para evaluar la significancia estadstica de los resultados de una muestra, comparando la media muestral con la media poblacional.


1.**Construccin Intervalos de Confianza**: Se utiliza para determinar cun confiable es una estimacin, ayudando a construir intervalos dentro de los cuales es probable que se encuentre el parmetro poblacional. El intervalo de confianza para la media poblacional, cuando se conoce el Error Estndar, se calcula de la siguiente manera:

$$
IC = \bar{x} \pm Z_{\alpha/2} \cdot \text{EEM} 
$$

Donde:
- $ \bar{x} $ es la media muestral.
- $ Z_{\alpha/2} $ es el valor crtico de la distribucin normal estndar correspondiente al nivel de confianza deseado **(por ejemplo, 1.96 para un 95% de confianza)**.
- $ \text{EEM} $ es el Error Estndar de la Media.


2. **Pruebas de Hiptesis**: Las pruebas de hiptesis son un procedimiento estadstico para evaluar si una afirmacin sobre un parmetro poblacional es compatible con los datos muestrales. El Error Estndar se utiliza para determinar si la diferencia observada es estadsticamente significativa.

   - **Formulacin de Hiptesis**:
      - **Hiptesis nula $H_0$**: No hay efecto o diferencia $ \mu = \mu_0 $.
      - **Hiptesis alternativa $H_a$**: Hay un efecto o diferencia $ \mu \neq \mu_0$.

   - **Clculo del Estadstico de Prueba**:
      El estadstico de prueba se basa en la diferencia entre la media muestral y el valor hipottico bajo $H_0$:
      $$
      Z = \frac{\bar{x} - \mu_0}{\text{EEM}}
      $$

   - **Determinacin del P-valor**:
      - Se calcula el p-valor correspondiente al estadstico $Z$ obtenido.
      - Si el p-valor es menor que el nivel de significancia $\alpha$, se rechaza $H_0$.


Para empezar a ilustrar estos conceptos, primero generemos un experimento aleatorio donde supondrmos que nuestros datos siguen una distribucin normal,luego, realizarmos un estimativo de la media muestral a partir de los datos del experimento y generarmos los intervalos de confianza primero a partir del EEM. El error estndar de la media (es decir, el error debido a la estimacin de la media poblacional a partir de las medias muestrales) es la desviacin estndar de todas las posibles muestras de un tamao dado escogidos de esa poblacin. Especficamente, *el error estndar de la media se refiere a la desviacin estndar de la distribucin de la media muestral*. En trminos sencillos, indica cunto se espera que la media muestral vare de una muestra a otra si se tomaran mltiples muestras de la misma poblacin.

De sta manera, ya que exste una equivalencia entre los *percentiles* de la distribucin normal y la distribucin estandar $Z$, sabmos que $Z_{\alpha/2}$ = 1.96 es el valor crtico de $Z$ tal que *contiene el 95% de ala acumulacin de los datos **al rededor de la media de la distribucin***. As, los **mrgenes de error** son los valores de la variable aleatoria (estimacin del parmetro poblacional) que delimtan el intervalo de confianza tanto por izquierda como por derecha de la media de la estimacin poblacional.

$$
ME = Z_{\alpha / 2} \frac{\sigma}{\sqrt{n}}
$$


$$
IC = \bar{x} \pm ME = \bar{x} \pm Z_{\alpha/2} \cdot \text{EEM} 
$$


```python
# Parmetros de la poblacin
population_mean = 100   # Media poblacional
population_std = 15     # Desviacin estndar poblacional
n_samples = 10000        # Nmero de muestras
sample_size = 50        # Tamao de cada muestra

# Generacin de mltiples muestras de la poblacin
sample_means = []
for _ in range(n_samples):
    sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
    sample_means.append(np.mean(sample))

# Clculo del error estndar de la media
eem = population_std / np.sqrt(sample_size)
eem_sample = np.std(sample_means, ddof=1) / np.sqrt(sample_size)
print(f"Error Estndar de la Media (EEM polacional): {eem:.2f}")
print(f"Error Estndar de la Media (EEM muestral): {eem_sample:.2f}")
print(f"Desviacin Estndar Poblacional:", population_std)
print(f"Desviacin Estndar Muestral: {np.std(sample_means, ddof=1):.2f}")
print(f"Media Muestral: {np.mean(sample_means):.2f}")
print(f"Media Poblacional:", population_mean)

# Visualizacin de la distribucin de las medias muestrales

graficar_eem(sample_means,population_mean, eem_sample )
```

Para calcular intervalos de confianza para la estimacin de un parmetro seguimos los siguientes pasos:

1. Estimamos el parmetro poblacional (como por ejemplo, la media muestral).
2. Calculamos el error estandar por medio de la poblacin o de la muestra.
3. Especificamos un valor de confianza, el cul es la probabilidad de que en el intervalo de confianza construido, exsta el valor verdadero poblacional. Comnmente se utilizan los niveles de 90%, 95% o 99%.
4. Especificamos un valor  $Z_{\alpha}$ que refleje el nivel de confianza esperado, el cul se obtiene a partir de la distribucin normal estndar. Por ejemplo, para un 95% de confianza, $Z_{\alpha}$ es aproximadamente 1.96.
5. Calculamos los intervalos de confianza usando la relacin: 

$$
IC = \bar{x} \pm Z_{\alpha/2} \cdot \text{EEM} 
$$

```python
# Parmetros de la poblacin
population_mean = 100   # Media poblacional
population_std = 15     # Desviacin estndar poblacional
n_samples = 10000        # Nmero de muestras
sample_size = 50        # Tamao de cada muestra

# Generacin de mltiples muestras de la poblacin
sample_means = []
for _ in range(n_samples):
    sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
    sample_means.append(np.mean(sample))

# Estimacin del parmetro poblacional
mean_sample = np.mean(sample_means)

# Clculo del error estndar de la media (usamos eem poblacional para el IC)
eem = population_std / np.sqrt(sample_size)
eem_sample = np.std(sample_means)/np.sqrt(sample_size)

# Intervalo de confianza del 95%
confianza = 0.95
Z_critical = stats.norm.ppf(1 - (1 - confianza) / 2)  # Clculo del valor crtico Z
margen_error = Z_critical * eem  # Usamos eem basado en la desviacin estndar poblacional
intervalo_confianza = (mean_sample - margen_error, mean_sample + margen_error)

# Clculo de ocnfianza al 95% mediante scipy.stats
intervalos_stats = stats.norm.interval(confidence=0.95, loc=mean_sample, scale=np.std(sample_means))

# Imprimimos los resultados
print(f"Media muestral: {mean_sample:.2f}")
print(f"Z critico: {Z_critical:.2f}")
print(f"Intervalo de confianza del 95%: {intervalo_confianza}")
print(f"Intervalo de confianza del 95% (stats): {intervalos_stats}")
print(f"Margen de error: {margen_error:.2f}")

# Visualizacin de la distribucin de las medias muestrales
graficar_ID_normal(sample_means, population_mean, intervalo_confianza)
```

El **nivel de confianza** refleja el **grado de certeza asociado con un intervalo de confianza**. El nivel de confianza, a menudo expresado como $1 - \alpha$, representa la proporcin de intervalos de confianza que, en un nmero infinito de muestras, contendran el verdadero valor del parmetro poblacional. En otras palabras, es **la probabilidad de que el intervalo de confianza calculado a partir de una muestra incluya el valor verdadero del parmetro en cuestin**.

1. **Nivel de Confianza (1 - $$)**: Si se construyen mltiples intervalos de confianza con un nivel de confianza del 95% ($$ = 0.05), se espera que el 95% de estos intervalos contengan el valor verdadero del parmetro poblacional. Por lo tanto, un nivel de confianza del 95% implica que existe una probabilidad del 95% de que el intervalo de confianza calculado a partir de una muestra contenga el verdadero valor del parmetro.

2. **Error de Confianza o Significancia ($$)**: El nivel de confianza es complementario al nivel de **significancia** ($$), que es la probabilidad de que el intervalo de confianza **no** contenga el valor verdadero del parmetro. Por ejemplo, un nivel de confianza del 95% corresponde a un nivel de significancia del 5% ($$ = 0.05), lo que significa que existe una probabilidad del 5% de que el intervalo no contenga el valor verdadero.

La diferencia es algo sutil, no obstante considere lo siguiente: el parmetro poblacional a estimar **es fijo y no posee una distribucin de probabilidad, solo es desconocido** y su relacin con la distribucin muestral es simplemente que *este valor puede o no estar dentro de los intervalos de confianza*, por otra parte **el estadstico muestral SI tiene una distribucin de probabilidad**. Esto quiere decir que **el nivel de confianza se relaciona con la taza de xito que el intervalo de confianza tiene para capturar/contener el parmetro poblacional**, mas no representa la *probabilidad de que un intervalo de confianza contenga el parmetro poblacional*.

Hasta este punto, hemos realizado mltiples estimaciones puntuales asumiendo el conocimiento de la desviacin estndar de nuestra poblacin, lo cual no es muchas veces acertado ya que en casos de la vida real dcho parmetro no es de nuestro conocimiento. Para afrontar ste problema usamos en lugar de la distribucin $Z$ (la cul hace alucin a la poblacin), la distribucin **t de Student**, comnmente conocida como distribucin t. Es una distribucin simtrica y en forma de campana, similar a la distribucin normal pero con colas ms gruesas lo cla significa que hay una *mayor probabilidad de obtener valores extremos en comparacin con la distribucin normal*.

La distribucin $t$ depende de un parmetro llamado **grados de libertad (df)**, que generalmente se define como el *tamao de la muestra menos uno* ($df = n - 1$). A medida que los grados de libertad aumentan, la distribucin $t$ se asemeja cada vez ms a la distribucin normal estndar. Concrtamente, la distribucin $t$ al igual que la distribucin $Z$, es una distribucin muestral que se genera a partir de una "estandarizacin" de la variable aleatoria, solo que a diferencia de la distribucin $Z$, se emplea dirctamente la varianza muestral $\text{s}^2$ de modo que sta distribucin describe la densidad de probabilidad de la variable aleatoria siguiente:

$$
t = \frac{\hat{x} - \mu} {\text{s}/\sqrt{n}}
$$

siendo $t$ el t-valor, $\hat{x}$ la media muestral, $\mu$ la media poblacional, $s$ la desviacin estndar muestral y $n$el tamao muestral. La distribucin $t$ es til tanto en csos donde el tamao de la muestra es pequeo ($n<30$) o cuando la desviacin estndar poblacional es desconocda. De manera anloga ala construccin de intervalos de confianza con base a la variable $Z$, cuando se desconoce la desviacin estndar de la poblacin, un intervalo de confianza se genera a partir de la distribucin t-student de la forma:

$$
IC = \bar{x} \pm t_{\alpha / 2} \frac{s}{\sqrt{n}}
$$


No obstante, la funcin de densidad de probabilidad de una distribucin t-student  es muy diferente en su escritura a la distribucin $Z$. Recordemos que si bien la distribucin $Z$ es una *distribucin normal estndar que modela la variable aleatoria $Z$*, as msmo, a pesar de que nosotros experimentalmente definamos la variable $t$ con la relacin anterior, la funcin de densidad de probabilidad para una variable $t$ viene dada por la siguiente expresin, donde $\nu$ es el nmero de grados de libertad:

$$
f(t|\nu) = \frac{\Gamma\left(\frac{\nu + 1}{2}\right)}{\sqrt{\nu \pi} \ \Gamma\left(\frac{\nu}{2}\right)} \left(1 + \frac{t^2}{\nu}\right)^{-\frac{\nu + 1}{2}}
$$

Donde:
- $\Gamma$ es la funcin Gamma, que generaliza el concepto de factorial a nmeros reales y complejos.
- $\nu$ representa los grados de libertad.
- $t$ es el valor de la variable aleatoria $t$.


La funcin gamma $\Gamma(x)$ es una generalizacin del concepto de factorial que se aplica no solo a enteros positivos, sino tambin a nmeros reales y complejos (excepto los enteros negativos y el cero). La funcin Gamma es definida mediante la siguiente integral impropia:

$$
\Gamma(x) = \int_0^\infty t^{x-1} e^{-t} \, dt
$$

Donde:
- $x$ es un nmero real o complejo.
- $t$ es la variable de integracin.

Algnas de sus propiedades son:

1. **Relacin con el Factorial**: Para un nmero entero positivo $n$, la funcin Gamma est relacionada con el factorial de la siguiente manera:
   $$
   \Gamma(n) = (n-1)!
   $$
   Por ejemplo:
   $$
   \Gamma(5) = 4! = 24
   $$

2. **Desplazamiento**: Existe una propiedad importante llamada la "propiedad de desplazamiento":
   $$
   \Gamma(x+1) = x \cdot \Gamma(x)
   $$
   Esta propiedad es til para derivar valores de la funcin Gamma para nmeros no enteros.

3. **Valores Especiales**: Algunos valores especiales de la funcin Gamma incluyen:
   $$
   \Gamma\left(\frac{1}{2}\right) = \sqrt{\pi}
   $$
   $$
   \Gamma(1) = 1
   $$

A continuacin, se mustra manualmente como generar una PDF para una variable $t$, no obstante el mdulo `scipy` nos permite realizar stos clculos directamente.

```python
# Funcin para hallar la PDF de la distribucin t
from scipy.special import gamma

def t_distribution_pdf(t, nu):
    numerator = gamma((nu + 1) / 2)
    denominator = np.sqrt(nu * np.pi) * gamma(nu / 2)
    return (numerator / denominator) * (1 + (t ** 2) / nu) ** (- (nu + 1) / 2)

# Parmetros
nu_values = list(range(1, 11))  # Grados de libertad
t_values = np.linspace(-5, 5, 1000)  # Valores de t

# Graficar la distribucin t
plt.figure(figsize=(8, 4))
for degree in nu_values:
    # Clculo de la PDF para cada valor de t
    pdf_values = t_distribution_pdf(t_values, degree)
    plt.plot(t_values, pdf_values, label=f't-distribution (df={degree})')

plt.title("Distribucin t de Student")
plt.xlabel("t")
plt.ylabel("Densidad de Probabilidad")
plt.legend()
plt.grid(True)
plt.show()
```

Como se observa en el plot anterior, una peculiaridad de la distribucin $t$ es que si bien posee una forma general especificada por su PDF, sta al depender de los grados de libertad y o de la desviacin estndar, posee una forma nica para cada tamao muestral y por ende, es necesario calcular la distribucin cada vez que se tenga un tamao de muestra dado ya que sto nos permitir calcular intervalos de confianza acertados para nuestro caso particular.

Para ilustrar el uso de la distribucin $t$ para generar intervalos de confianza, vamos a generar dos experimentos aleatorios: el primero, tomar solo una muestra de tamao $n$ de la distribucin poblacional, calcular la desviacin estndar muestral a partir de una nica muestra y calcular los intervalos de confianza. El segndo experimento, generar $N$ muestras de tamao $n$, hallar la desviacin estndar muestral de cada muestra y estimar la varianza poblacional como un promedio de las varianzas muestrales para posteriormente calcular el intervalo de confianza. En generl se siguen los siguientes procedimientos:

1. Clculo de la media y desviacin estndar muestral.
2. Definicin de los grados de libertad como $df = n -1$.
3. Especificacin del nivel de confianza y significancia.
4. Clculo del error estndar.
5. Clculo del valor crtico de $t$ teniendo en cuenta los grados de libertad.
6. Clculo del mrgen de error.
7. Clculo del intervalo de confianza.

```python
# Parmetros de la poblacin
population_mean = 100   # Media poblacional
population_std = 15     # Desviacin estndar poblacional
sample_size = 50        # Tamao de cada muestra

# Generacin de una nica muestra de la poblacin
sample = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)
mean_sample = np.mean(sample)
sample_std = np.std(sample, ddof=1)

# Grados de libertad
df = sample_size - 1

# Nivel de confianza
confidence_level = 0.95
alpha = 1 - confidence_level

# Error estndar de la media (calculado para la muestra especfica)
sem = sample_std / np.sqrt(sample_size)

# Valor crtico t para el nivel de confianza
t_critical = stats.t.ppf(1 - alpha / 2, df)

# Margen de error
margen_error = t_critical * sem

# Intervalo de confianza
intervalo_confianza = (mean_sample - margen_error , mean_sample + margen_error )

# Imprimimos los resultados
print(f"Media muestral: {mean_sample:.2f}")
print(f"t critico: {t_critical:.2f}")
print(f"Intervalo de confianza del 95%: {intervalo_confianza}")
print(f"Margen de error: {margen_error:.2f}")
```

Intervalos de COnfianza para Proporciones

Los intervalos de confianza para proporciones se utilizan para estimar el rango dentro del cual se espera que est la verdadera proporcin de una poblacin, basndose en una muestra. Este intervalo proporciona una estimacin de la precisin de la proporcin muestral y se calcula usando la distribucin normal, especialmente cuando el tamao de la muestra es grande. 

Una proporcin poblacional $\hat(p)$ se calcula como el *nmero de xitos (eventos de inters)* dividido por el tamao de la muestra. 
$$
\hat{p} = \frac{x}{n}
$$
donde $x$ es el nmero de xitos y $n$ es el tamao de la muestra. En este sentido, el **error estndar de la proporcin** usado para medir la variabilidad de la proporcin muestral, se calcula como:

$$
SE = \sqrt{\frac{\hat{p} (1 - \hat{p})}{n}}
$$

De sta manera, los intervalos de confianza se calculan mediante la **distribucin normal estndar** segn el nivel de confianza dado, de modo que

$$
ICP = \hat{p} \pm Z_{\alpha / 2} SE
$$


```python
# Parmetros de la muestra
n = 200   # Tamao de la muestra
x = 120   # Nmero de xitos (personas que prefieren el producto)

# Proporcin muestral
p_hat = x / n

# Error estndar de la proporcin
SE = np.sqrt(p_hat * (1 - p_hat) / n)

# Nivel de confianza
confidence_level = 0.95
alpha = 1 - confidence_level
Z = stats.norm.ppf(1 - alpha / 2)  # Valor crtico Z para el intervalo de confianza

# Margen de error
margin_of_error = Z * SE

# Intervalo de confianza
confidence_interval = (p_hat - margin_of_error, p_hat + margin_of_error)

# Resultados
print(f"Proporcin muestral: {p_hat:.2f}")
print(f"Error estndar de la proporcin: {SE:.2f}")
print(f"Valor crtico Z: {Z:.2f}")
print(f"Margen de error: {margin_of_error:.2f}")
print(f"Intervalo de confianza del 95%: {confidence_interval}")
```

**<h2 id="section3" style="color: #007ACC;">3. Prueba de Hiptesis</h2>**

---
Prueba de Hiptesis

---

En el mundo de la estadstica, tomar decisiones informadas basadas en datos es fundamental para la investigacin cientfica, el anlisis de negocios y la toma de decisiones en general. Las pruebas de hiptesis son herramientas poderosas que nos permiten **evaluar si los datos que hemos recolectado apoyan o refutan una afirmacin especfica sobre una poblacin o un proceso**. Estas pruebas proporcionan un marco sistemtico para tomar decisiones basadas en evidencia, en lugar de en suposiciones o intuiciones.

Por ejemplo, un investigador puede querer saber si un nuevo medicamento es ms efectivo que el tratamiento estndar. Utilizando una prueba de hiptesis, puede determinar si las diferencias observadas en los resultados entre el nuevo medicamento y el tratamiento estndar son lo suficientemente grandes como para no ser atribuibles al azar. De manera similar, en un entorno empresarial, las pruebas de hiptesis pueden ayudar a evaluar si una nueva estrategia de marketing ha llevado a un aumento significativo en las ventas.

#### **Componentes Principales de una Prueba de Hiptesis**

Una prueba de hiptesis se basa en varios componentes clave tles como:

1. **Hiptesis Nula ($H_0$)**: Es la afirmacin que se pone a prueba. Generalmente, esta hiptesis representa una situacin de no efecto o no diferencia, y se asume como verdadera hasta que los datos sugieren lo contrario. Por ejemplo, podra ser que no hay diferencia en la eficacia entre dos tratamientos o entre la medicin de una media muestral y una media poblacional.

2. **Hiptesis Alternativa ($H_1$ o $H_A$)**: Es la afirmacin que se considera si la hiptesis nula es rechazada. Representa una diferencia o efecto que el investigador est tratando de encontrar. En el ejemplo del medicamento, la hiptesis alternativa podra ser que el nuevo medicamento es ms efectivo que el tratamiento estndar, o que la medicin de un parmetro poblacional como la media, si pose diferencias sustanciales entre el valor estimado a partir de la muestra y el valor poblacional real. 

3. **Nivel de Significancia ($\alpha$)**: Es el umbral para tomar una decisin sobre la hiptesis nula. Representa la probabilidad de cometer un error tipo I, es decir, rechazar la hiptesis nula cuando en realidad es cierta. Comnmente, se usa un nivel de significancia del 5% ($\alpha = 0.05$).

4. **Estadstico de Prueba**: Es una medida calculada a partir de los datos que se utiliza para tomar decisiones sobre la hiptesis nula. Dependiendo del tipo de prueba y los datos, el estadstico de prueba puede tomar diversas formas tles como el valor $t$ de la distribucin $t$-student.

5. **Valor p**: Es la **probabilidad de observar un estadstico de prueba al menos tan extremo como el observado**, bajo la suposicin de que la hiptesis nula es cierta. El valor p se compara con el nivel de significancia para tomar una decisin.

6. **Decisin**: Basado en el valor $p$ y el nivel de significancia, se decide si se rechaza o no la hiptesis nula. Si el valor $p$ es menor que $\alpha$, se rechaza la hiptesis nula en favor de la hiptesis alternativa. 


En el proceso de decisin basados en pruebas de hiptesis exste una sutileza conceptual. En el momento en que se rechaza la hiptesis nula $H_0$ **se acepta la validez de la hiptesis alternativa $H_A$**, por el contrario, al fallar en rechazar la hiptesis alternativa $H_A$ **no significa que $H_0$ sea cierta**, sino que se est concluyendo que **no se tiene suficiente evidencia** para asumir la veracidad de $H_A$.

#### **Cmo Calcular las Pruebas de Hiptesis**

El clculo de una prueba de hiptesis involucra los siguientes pasos:

1. **Formular las Hiptesis**: Definir claramente las hiptesis nula y alternativa.
2. **Seleccionar el Nivel de Significancia**: Establecer el umbral para decidir si rechazar la hiptesis nula.
3. **Recoger y Analizar los Datos**: Realizar el experimento o recolectar los datos necesarios.
4. **Calcular el Estadstico de Prueba**: Utilizar los datos para calcular un estadstico que refleje la diferencia entre las hiptesis.
5. **Obtener el Valor p**: Calcular la probabilidad asociada con el estadstico de prueba.
6. **Tomar una Decisin**: Comparar el valor p con el nivel de significancia y decidir si se rechaza la hiptesis nula.

---
Tipos de errores

---

Un tipo de error dentro del contexto de toma de decisiones en pruebas de hiptesis, corresponden a rechazos incorrectos de la hiptesis nula o la hiptesis alternativa. Segn sea el caso tenemos los siguiente tipos de errores:

- **Error tipo I (falso positivo - error $\alpha$)**: Corresponde al error cometido durante el proceso de evaluacin de la evidencia y toma de decisin, donde la hiptesis nula **$H_0$** es rechazada y la hiptesis alternativa **H_A** es tomada como cierta siendo la hiptesis nula **$H_0$** la opcin correcta a considerar como verdadera.

- **Error tipo II (falso negativo - error $\beta$)**: Corresponde al error cometido durante el proceso de evaluacin de la evidencia y toma de decisin, en donde la hiptesis alternativa **$H_A$** es rechazada y en su lugar, la hiptesis nula **$H_0$** es considerada como verdadera siendo la hiptesis alternativa **$H_A$** la opcin correcta a considerar como verdadera.



image.png

Note entonces que los errores tipo I y tipo II no tienen el msmo impacto sobre el problema en cuestin. Si consideramos que en general, $H_A$ es la hiptesis que deseamos poner a prueba mientras que  $H_0$ representa la hiptesis base, o status-quo, considerar la hiptesis alternativa suele tener peores repercuciones ya que es aquella hiptesis que contiene una consecuencia debido a que "aprueba" una desicin a tomar. Por ejemplo, en el caso mdico, si $H_A$ corresponde a "validar que un medicamento es mejor que otro y no posee efectos secundarios", un error tipo I generara percances de salud a la poblacin objetivo de dicho medicamento. En el caso cientfico, $H_A$ hace alucin a validar que cierta hiptesis sobre un fenmeno o sistemasea cierta, as un error tipo I implicara falsamente dar validez a una teora.

Con esto en mente, si consideramos que una toma de decisin puede ejecutarse mltiples veces segn una situacin lo requiera, pero dicha decisin siempre estar basada en la evidencia que respalde a la hiptesis detrs de la msma, cabra preguntarse **cul es la probabilidad ms alta  de errores tipo I que permitiramos en esta toma de decisiones?** o, en el caso opuesto, **cul es la probabilidad ms alta  de errores tipo II que permitiramos en esta toma de decisiones?**.

En este sentido, formulamos las siguientes definiciones:

1. **Significancia $\alpha$**: Probabilidad mxima de cometer un **error tipo I** tolerada en nuestro experimento. Corresponde al **humbral para rechazar a $H_0$ (o no)**, dada una evidencia que soporte a $H_A$. Tamben puede interpretarse como la probabilidad de rechazar $H_0$ cuando en realidad es verdadera.

$$
\alpha = \text{max} P(\text{Error tipo I})
$$

2. **Potencia de Prueba**: Probabilidad mxima de cometer un **error tipo II** tolerada en nuestro experimento. Corresponde al **humbral para rechazar $H_A$(o no)**. Tambin puede interpretarse como la probabilidad de rechazar $H_0$ cuando esta es falsa, es decir, es la capacidad de la prueba de detectar un efecto real.

$$
1- \beta
$$

**<h3 id="subsection31" style="color: #003366;">3.1. Tipos de Prueba de Hiptesis</h3>**

---
Tipos de Prueba de Hiptesis

---

En las pruebas de hiptesis, el objetivo es decidir si los datos observados proporcionan suficiente evidencia para rechazar una hiptesis nula (\(H_0\)) a favor de una hiptesis alternativa (\(H_1\)). Dependiendo de la naturaleza de la hiptesis alternativa, las pruebas estadsticas se pueden clasificar en tres tipos:

1. **Right-Tailed Test (Prueba de Cola Derecha)**
2. **Left-Tailed Test (Prueba de Cola Izquierda)**
3. **Two-Sided Test (Prueba de Dos Colas)**

Antes de definir los tipos de test estadsticos de significancia de hiptesis, definmos los siguientes conceptos:

  - **p-valor**: Es la **probabilidad de que, asumiendo $H_0$ es verdadera, el estadstico de prueba tome un valor tan extremo o ms extremo que el valor observado**, donde un estadstico de prueba puede ser por ejemplo la media muestral. En este sentido, el $\text{p-avlor}$ es el **rea de la distribucin del estadstico de prueba ms alla de un valor crtico**, es decir, es el **cuantil $1 - \alpha$**.
  - **Valor Crtico:** Corresponden a aquellos valores del estadstico (variable aleatoria) a partir de los cules se rechaza la hiptesis nula, es decir, corresponden a aqueelos valores donde $\text{p-value}=\alpha$.

#### 1. **Right-Tailed Test (Prueba de Cola Derecha)**

En una prueba de cola derecha, la hiptesis alternativa ($H_A$) propone que el parmetro de inters es **mayor que el valor especificado en la hiptesis nula** ($H_0$). En otras palabras, estamos interesados en *saber si el valor observado es significativamente mayor que el valor bajo $H_0$*. Tal que en el caso de que nuestro parmetro poblacional de inters es la media muestral, tenemos:

- **Hiptesis:**
  - $H_0$: $\mu = \mu_0$
  - $H_A$: $\mu > \mu_0$

En ste escenario, buscamos entonces verificar la hiptesis alternativa $H_A$ mediante la siguiente probabilidad, donde si $\text{p-value}$ < $\alpha$, rechazamos la hiptesis nula $H_0$  en favor de la hiptesis $H_A$ alternativa. 

$$
\text{p-value} = P(T(X) > t | H_0)
$$

Donde:
  - $T(X)$: estadstica de prueba de una variable aleatoria.
  - $t$: valor observado para la estadstica.
  - $H_0$: hiptesis nula.

Note entonces siendo $\hat{X}$ el estadstico muestral, $\mu = a$ es el valor del parmetro que respalda a $H_0$ mientras que $\mu = b$ es el valor del parmetro poblacional que respalda a $H_A$. 
$$
\text{p-value} = P(\hat{X} > b | \mu = a)
$$

#### 2. **Left-Tailed Test (Prueba de Cola Izquierda)**

En una prueba de cola izquierda, la hiptesis alternativa ($H_A$) propone que el parmetro de inters es **menor que el valor especificado en la hiptesis nula** ($H_0$). Aqu, estamos interesados en *saber si el valor observado es significativamente menor que el valor bajo $H_0$*.

- **Hiptesis:**
  - $H_0$: $\mu = \mu_0$
  - $H_A$: $\mu < \mu_0$

Para pruebas de cola izquierda, ya que se busca conocer si un valor del estadstico de prueba es estadsticamente menor como para rechazar la hiptesis nula, el valor de desicin de significancia $\text{p-value}$ se expresa como:

$$
\text{p-value} = P(T(X) < t | H_0)
$$

#### 3. **Two-Sided Test (Prueba de Dos Colas)**

En una prueba de dos colas, la hiptesis alternativa ($H_A$) propone que el parmetro de inters es **diferente (ya sea mayor o menor) al valor especificado en la hiptesis nula** ($H_0$). Aqu, estamos interesados en *saber si el valor observado es significativamente diferente (en cualquier direccin) del valor bajo* $H_0$.

- **Hiptesis:**
  - $H_0$: $\mu = \mu_0$
  - $H_A$: $\mu \neq \mu_0$

Debido a que en un test de dos colas, se busca una variacin estadsticamente significativa bien sea por incremento o decremento, el valor de decisin $\text{p-value}$ se expresa como:

$$
\text{p-value} = P(|T(X) t_0| > |t - t_0| | H_0)
$$

donde $t_0$ es el valor del estadstico que apoya a  $H_0$ mientras que $t$ es el valor observado, el cul apoya a $H_A$.


Para empezar a ilustrar estos conceptos, supongamos la siguiente situacin: en un estudio hecho hace unos aos, se determin que la media poblacional posea un valor de $\mu = 100$. Aos despus, otro investigador lleg a nosotros reportando que realiz una medicin sobre la poblacin recientemente y observ que la media aument, de hecho, sta adquiri un valor de $\mu = 106$. Si en este ejemplo se asume que la varianza poblacional **es conocida y no ha cambiado**, es la diferencia que el investigador encontr en las medias lo suficientemente significativa estadsticamente para asegurar que la media poblacional aumento?.

Para resolver sta incgnica, ejecutamos los siguientes procedimientos:

1. Determinamos el valor del parmetro y del estadstico muestral.
2. Ya que la varianza es conocida, calculamos con ella el error estndar.
3. Determinamos el $\text{p-valor}$ como el cuantil $1-\alpha$.
4. Determinamos la significancia estadstica segn la proporcin entre $\text{p-valor}$  y  $\alpha$.

Note a dems, que acontinuacin para nuestro ejemplo, asumimos la normalidad de la distribucin muestral gracias al teorema del lmite central.

```python
# Parmetros conocidos
population_mean = 100  # Media poblacional bajo H0
population_std = 15    # Desviacin estndar poblacional (conocida)
sample_size = 30       # Tamao de la muestra

# Generamos una muestra a partir de la distribucin poblacional
np.random.seed(0)
data = np.random.normal(population_mean, population_std, sample_size)

# Calculamos el estadstico de prueba y el error estandar
sample_mean = np.mean(data)
sem = population_std / np.sqrt(sample_size)

# Calculamos el valor crtico para right-tailed test
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha)  # Z crtico ( cuantil 1- alpha )
critical_value_right = population_mean + z_critical * sem  

# Calculamos el valor crtico para left-tailed test
alpha = 0.05 
z_critical_left = stats.norm.ppf(alpha)  # Z crtico ( cuantil alpha )
critical_value_left = population_mean + z_critical_left * sem

# Calculamos el valor crtico para two-sided test
z_critical_two_sided = stats.norm.ppf(1 - alpha / 2)
critical_value_lower = population_mean - z_critical_two_sided * sem
critical_value_upper = population_mean + z_critical_two_sided * sem

# Grfico right-tail test
graficar_right_normal_test(90, 110, population_mean, sample_mean, sem, critical_value_right)
# Grfico left-tail test
graficar_left_normal_test(90, 110, population_mean, sample_mean, sem, critical_value_left)
# Grfico two-tail test    
graficar_two_normal_test(90, 110, population_mean, sample_mean, sem, critical_value_lower, critical_value_upper)
```

En el ejemplo anterior, nos es fcil entender que est sucediendo: cuando se realiza una distribucin muestral de un estadstico para estimar un parmetro, por el teorema del lmite central tenemos que la distribucin muestral sigue una distribucin normal. sto nos permte formular una hiptesis nula $H_0$ que se entiende como que **no exste una diferencia significativa entre el estadstico muestral y el parmetro poblacional** y a su vez, una hiptesis alternativa $H_A$ que se entiende como **exste una diferencia significativa entre el estadstico y el parmetro en algna direccin**. As, mediante el error estndar y los percentiles de la distribucin normal, nos es posible encontrar la regin de valores de la variable aleatoria (media muestral) los cules representan casos extremos que permten evidenciar diferencias estadsticamente significativas entre el parmetro y la estimacin desde la mustra.

Por otra parte, si empleamos dirctamente la distribucin $Z$, el procedimiento para determinar las regiones de rechazo de la hiptesis nula $H_0$ ser ms fcil ya que los valores crticos corresponden directamente a los percentiles relacionados con la significancia establecida $\alpha = 0.05$. A continuacin, modelaremos el msmo ejemplo anterior pero sta vez haciendo uso directo de la distribucin $Z$.

**<h3 id="subsection32" style="color: #003366;">3.2. Poder de una Prueba y Error Tipo II</h3>**

---
Poder de una Prueba, Error Tipo II y la Distribucin para $H_A$

---

Hasta ste punto, hemos realizado diferentes ilustraciones a cerca de cmo poner a prueba una hiptesis con vsta a una evidencia disponible. No obstante, en secciones anteriores hablabamos a cerca de cmo muchas veces el error tipo I puede ser ms peligroso debido a que la hiptesis alternativa $H_A$ es aquella que nos incita a tomar una accin debido a que es la hiptesis que deseamos poner a prueba. As pues, nuestras discuciones han estado basadas o referenciadas con respecto al error tipo I y por ende, nos hemos concentrado en realizar la prueba de hiptesis considerando la evidencia **en contra** de $H_0$ gracias a que a travs de nuestras pruebas de hiptesis deseamos saber si exste suficiente evidencia **a favor** de  $H_A$.

sto quiere decir, que las distribuciones muestrales que hemos empleado para demostrar el uso del testeo de hiptesis hasta ste punto, corresponden a la evidencia que **sustenta a $H_0$** y por ende, la probabilidad que est contenida dentro de los valores crticos corresponde a **la evidencia disponible a favor de $H_0$ y en contra de $H_A$**. Sin embargo, si bien esta metodologa nos permite **establecer el rango de valores da partir de los cules rechazamos a H_0 y consideraremos H_A como verdadera**, en la presente subseccin vermos que exste una perspectiva complementaria a sta de modo tal que ya no nos preguntarmos a cerca de la significancia estadstica de la hiptesis nula sino que dirigirmos nuestra atencin a **establecer el rango de valores a partir de los cules rechazarmos H_A debido a que no contamos con la suficiente evidencia para soportarla**.

$$
\alpha = P(\text{Rechazamos} \text{ } H_0 | H_0 \text{ } \text{es Verdadera}) = \text{probabilidad tolerada de faltos positivos}
$$

$$
\beta = P(\text{ NO Rechazamos} \text{ } H_0 | H_0 \text{ } \text{es Falsa}) = \text{probabilidad tolerada de faltos negativos}
$$

As, $\beta$ se calcula como la probabilidad de que la estimacin del parmetro caiga dentro de la regin de **no rechazo de la hiptesis nula H_0** definida por:

$$
\text{Zona de aceptacin } H_0 = \left[Z_{\text{H}_{(0,-)}}, Z_{\text{H}_{(0,+)}}\right]
$$

donde $Z_{\text{H}_{0,-/+}}$ es el lmite inferior (-) y superior (+) de la regin de aceptacin de $H_0$ definida por la distribucin de hiptesis bajo $H_0$. Es entonces $\beta$ la probabilidad acumulada de la distribucin bajo la hiptesis $H_A$ evaluada en sta regin. Por ende, la probabilidad $\beta$ **NO depende de la mustra sino del nivel de significancia**.

$$
\beta = P(T(X) < t | H_A) 
$$

El **poder de un test** es la  **probabilidad de rechazar correctamente H_0 cuando en realidad es falsa** (probabilidad de verdadero negativo). Es decir, es la capacidad de un test para *detextar un efecto cuando ste realmente exste* y viene dado por:

$$
P(\text{Rechazamos} \text{ } H_0 | H_0 \text{ } \text{es Falsa}) =  1 - \beta
$$

En ste sentido, el poder del test es **inversamente proporcional al error tipo II** donde un alto $1-\beta$ significa que hay una *baja probabilidad de cometer un error tipo II*. El nivel de significancia $\alpha$ siendo la probabilidad de cometer un error tipo I, se relaciona con el poder de la prueba en la medida de que si disminuimos $\alpha$ (lo que significa que queremos ser ms cautelosos para evitar un error tipo I), podramos estar aumentando $\beta$ lo que disminuye el poder del test. Por otro lado, si aumentamos el poder del test, incrementamos la probabilidad de cometer un error tipo I.

- **Distribucin H_0 (en azul)**:
    - Representa la distribucin de la media muestral (o estadstico) bajo la hiptesis nula.
    - Las regiones sombreadas en azul corresponden a la **regin de rezhado de H_0** y corresponden a la probabilidad del error tipo I $\alpha$.
- **Distribucin H_A (en naranja)**:
    - Representa la distribucin de la media muestral bajo la hiptesis alternativa.
    - La regin sombreada en naranja entre los valores crticos es la **probabilidad de error tipo II (\beta)**
- **Poder del Test**:
    - La porcin de la distribucin $H_A$ que caue fuera de los valores crticos (es decir, donde rechazaramos $H_0$) es el poder del test $1-\beta$.
    - Un test con un poder alto tendr una menor rea sombreada en naranja (menor $\beta$).

```python
# Parmetros de la distribucin bajo la hiptesis nula (H0)
mu_0 = 100  # Media bajo H0
sigma = 15  # Desviacin estndar (conocida)

# Parmetros de la distribucin bajo la hiptesis alternativa (H1)
mu_1 = 110  # Media bajo H1 (efecto real)
sample_size = 30

# Nivel de significancia y valor crtico para un test de dos colas
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha / 2)

# Calculamos el valor crtico en la distribucin bajo H0
critical_value_left = mu_0 - z_critical * (sigma / np.sqrt(sample_size))
critical_value_right = mu_0 + z_critical * (sigma / np.sqrt(sample_size))

# Generamos datos para las distribuciones bajo H0 y H1
x = np.linspace(90, 120, 1000)
y_H0 = stats.norm.pdf(x, mu_0, sigma / np.sqrt(sample_size))
y_H1 = stats.norm.pdf(x, mu_1, sigma / np.sqrt(sample_size))

# Grfico
plt.figure(figsize=(10, 6))
plt.plot(x, y_H0, label='Distribucin bajo H0', color='blue')
plt.plot(x, y_H1, label='Distribucin bajo H1', color='orange')

# Sombrear la regin de error tipo I ()
plt.fill_between(x, 0, y_H0, where=(x <= critical_value_left) | (x >= critical_value_right), color='blue', alpha=0.3, label='Error Tipo I ()')

# Sombrear la regin de error tipo II ()
plt.fill_between(x, 0, y_H1, where=(x > critical_value_left) & (x < critical_value_right), color='orange', alpha=0.3, label='Error Tipo II ()')

# Sombrear la regin de poder del test (1 - )
plt.fill_between(x, 0, y_H1, where=(x <= critical_value_left) | (x >= critical_value_right), color='green', alpha=0.3, label='Poder del Test (1 - )')

# Lneas verticales para valores crticos y medias
plt.axvline(critical_value_left, color='red', linestyle='dashed', label='Valor Crtico Izquierdo (Lmite inferior de NO rechazo H_0)')
plt.axvline(critical_value_right, color='red', linestyle='dashed', label='Valor Crtico Derecho (Lmite superior rechazo H_0)')
plt.axvline(mu_1, color='green', linestyle='dashed', label=f'Media bajo H1 = {mu_1}')

# Ttulo y etiquetas
plt.title('Visualizacin del Poder del Test con Regiones de Error Tipo I y II')
plt.xlabel('Valor')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

Interpretacin de Resultados

En algnas circunstancias, la correcta interpretacin de las pruebas de hiptesis puede ser algo confusa, as pues resumamos a continuacin las correctas interpretaciones de los elementos y resultados obtenidos en las discuciones mencionadas hasta ahora.

Los pasos necesarios para generar una prueba de hiptesis son:

1. **Definir las hiptesis**:
    - Definir la hiptesis (baseline) $H_0$
    - Definir la hiptesis alternativa a probar $H_A$
2. **Definir el tipo de prueba**:
    - Definir un estadstico de prueba.
    - Decidir el nivel de significancia $\alpha$
3. **Computar la estadstica de inters**
    - Definir el estadstico en la muestra.
    - Definir el valor $\text{p-value}$.
4. **Definir una conclusin**
    - Si el $\text{p-value}$ es menor a $\alpha$, rechazamos $H_0$

Las correctas interpretaciones de los elementos involucrados en una prueba de hiptesis son:

- **Error tipo I**:
    Se rechaza la hiptesis nula $H_0$ como cierta y se escoge $H_A$ siendo sta falsa (falso positivo).
- **Error tipo II**:
    Se **falla** en rechazar $H_0$ siendo sta falsa y se descarta $H_A$ como verdadera (falso negativo).
- **Significancia $\alpha$**:
    Mxima probabilidad admitida de cometer un error tipo I.
- **\text{p-value}**:
    El valor p  **NO** representa la probabilidad de que $H_0$ sea cierta, en su lugar, representa la probabilidad de *observar la data registrada al azar*.
- **Conclusiones en las pruebas de hiptesis**:
    - Al rechazar $H_0$ **se asume que  $H_A$ es verdadera**.
    - Al NO rechazar $H_0$ NO se asume que $H_0$ sea verdadera, solo se **asume que no hay suficiente evidencia para validar $H_A$**

Clculo de Intervalos de Confianza a Datos Reales

---

A continuacin generaremos los intervalos de confianza para la estimacin de la media de una variable de nuestro conjunto de datos real donde emplearemos el error estandar calculado a partir de la desviacin estndar poblacional estimada por medio del proceso de muestreo. El primer caso, corresponde a una prueba de hiptesis generada sobre un conjunto de 50 muestras.

```python
# Parmetros de la poblacin
variable = "total_likes_on_outofstation_checkin_received"
data = list(df_turismo[variable].values)
population_mean = np.round(np.mean(data),3)  # media poblacional
population_std  = np.round(np.std(data),3)   # desviacin estndar poblacional   
n_samples = 10000                # nmero de muestras
sample_size = 50                 # tamao de cada muestra

# Generacin de mltiples muestras de la poblacin
sample_means = []
sample_variance = []
for _ in range(n_samples):
    sample = random.sample(data, sample_size)
    sample_means.append(np.mean(sample))
    sample_variance.append(np.std(sample))

# Estimacin del parmetro poblacional
mean_sample = np.mean(sample_means)
variance_sample = np.mean(sample_variance)

# Clculo del error estndar de la media (usamos eem poblacional para el IC)
eem = population_std / np.sqrt(sample_size)
eem_sample = variance_sample/np.sqrt(sample_size)

# Intervalo de confianza del 95%
confianza = 0.95
Z_critical = stats.norm.ppf(1 - (1 - confianza) / 2)  # Clculo del valor crtico Z
margen_error = Z_critical * eem_sample  # Usamos eem basado en la desviacin estndar poblacional
intervalo_confianza = (mean_sample - margen_error, mean_sample + margen_error)

# Clculo de ocnfianza al 95% mediante scipy.stats
intervalos_stats = stats.norm.interval(confidence=0.95, loc=mean_sample, scale=np.std(sample_means))

# Imprimimos los resultados
print(f"Media poblacional: {population_mean:.2f}")
print(f"Media muestral: {mean_sample:.2f}")
print(f"Error Estndar: {eem_sample:.2f}")
print(f"Z critico: {Z_critical:.2f}")
print(f"Intervalo de confianza del 95%: {intervalo_confianza}")
print(f"Intervalo de confianza del 95% (stats): {intervalos_stats}")

# Visualizacin de la distribucin de las medias muestrales
graficar_ID_normal(sample_means, population_mean, intervalo_confianza)


# Validamos nuestra hiptesis de que nuestra estimacin es acertada

# Hiptesis nula H0: mu = 50
mu_0 = population_mean

# Estadstico de prueba Z
Z = (mean_sample - mu_0) / eem_sample
p_value = 2 * (1 - stats.norm.cdf(np.abs(Z)))  # Prueba de dos colas

print(f"Estadstico de prueba Z: {Z:.2f}")
print(f"P-valor: {p_value:.4f}")
print(f"Discrepancia de la estiamcin:", np.abs(population_mean - mean_sample))

# Decisin
alpha = 0.05
if p_value < alpha:
    print("Rechazamos la hiptesis nula.")
else:
    print("No podemos rechazar la hiptesis nula.")
```

Ahora, notemos que en el ejemplo anterior usamos un promedio de las medias muestrales para calcular nuestros intervalos de confianza y testear nuestra hiptesis. No obstante es posible generar stos msmos clculos usando solo una muestra.

```python
# Parmetros de la poblacin
variable = "total_likes_on_outofstation_checkin_received"
data = list(df_turismo[variable].values)
sample_size = 50 
sample = random.sample(data, sample_size)

# Estadsticos muestrales
sample_mean = np.mean(sample)
sample_std  = np.std(sample, ddof=1)

# Nivel de confianza
confianza = 0.95
alfa = 1 - confianza

# Grados de libertad
df_grados = sample_size - 1

# Estadstico t crtico
t_critical = stats.t.ppf(1 - alfa/2, df_grados)

# Error estndar de la media
eem = sample_std / np.sqrt(sample_size)

# Margen de error
margen_error = t_critical * eem

# Intervalo de confianza
intervalo_confianza = (sample_mean - margen_error, sample_mean + margen_error)

print(f"Media poblacional: {np.mean(data):.2f}")
print(f"Media muestral: {sample_mean:.2f}")
print(f"Margen de Error: {margen_error:.2f}")
print(f"Intervalo de confianza del 95%: {intervalo_confianza}")
```

En resumen podemos observar que *asumimos la normalidad en la distribucin de los estimativos miestrales* fundamentada en el teorema del lmite central. Luego, empleamos la distribucin $Z$ para mediante el usi de sus **percentiles**, determinar los niveles de confianza (acumulacin percentil de datos) dentro de nuestro espacio de muestras. De hecho, vemos las siguientes conclusiones:

1. **Intervalos de Confianza:** Para generar los intervalos de confianza, primero  *calculamos el error esperado del parmetro poblacional para cada muestra (EEM)*, luego, determinamos el *nivel de confianza en el que esperamos que nuestros datos caigan* mediante fijar un porcentaje de acumulacion de datos o *percentil* para luego, determinar el *valor $Z$* correspondiente. Una vez determinado el nivel de confianza, hallamos las fronteras de confianza como  el  *producto entre el parmetro estimado mas/menos el producto entre el valor de confianza $Z$ y el error esperado*.

2. **Prueba de Hiptesis:** En el contexto de la estimacin paramtrica, nuestra hipotesis a corroborar es *si nuestra estimacin es vlida*. Para ello, calculamos el *valor Z generado a partir de la discrepancia entre la estimacin y el valor real dividido por el EEM*, luego, usamos ese valor $Z$ para calcular *el  p-valor, el cual es la probabilidad de que nuestra hipotesis sea correcta*.

**<h3 id="subsection34" style="color: #003366;">3.4. t-tests</h3>**

---
t-tests y Comparaciones Entre Muestras/Poblaciones

---

Los **t-tests** son un conjunto de pruebas estadsticas que se utilizan para *comparar las medias de uno o dos grupos de datos*. El objetivo principal de un t-test es determinar si *existe una diferencia significativa entre las medias de los grupos comparados*, lo cual puede ayudar a *inferir si esas diferencias se deben a variabilidad aleatoria o si son lo suficientemente grandes como para ser consideradas estadsticamente significativas*.

Un t-test es un tipo de prueba de hiptesis que utiliza la distribucin t de Student para determinar si hay diferencias significativas entre las medias de una o dos poblaciones. Las pruebas de hiptesis t-test asumen que los dos conjuntos de datos son:

- Todos los elementos dentro de las muestras de los dos grupos son **distintos**, es decir, no exste un elemento que pertenezca simultneamente a dos grupos.
- Cada elemento en ambos samples son **independientes**.
- Las poblaciones estn **normalmente** distribuidas.

Dentro de los tipos de t-tests encontramos:

1. **t-Test para Una Muestra (One-sample t-test)**: Determina si la media de una sola muestra difiere significativamente de un valor conocido o hipottico.

2. **t-Test para Muestras Independientes (Independent two-sample t-test)**: Compara las medias de dos grupos independientes para ver si hay diferencias significativas.

3. **t-Test para Muestras Pareadas (Paired-sample t-test)**: Compara las medias de dos medidas relacionadas (como antes y despus de un tratamiento) dentro del mismo grupo.

#### **Fundamentos Matemticos del t-Test**

Supongamos que deseamos comparar dos poblaciones a travs de una muestra extrada de cada grupo y determinar si las poblaciones son distntas a travs del anlisis de las muestras generadas a partir  de cada una de ella. En ste sentido, tomarmos como el estadstico de prueba la media muestral de cada grupo, donde al desconocer la varianza de ambas poblaciones, la nica informacin a cerca de la variabilidad de los datos con la que contmos es la varianza muestral. Con base a ste propsito podmos generar los siguientes tres escenarios para comprobar la diferencia entre las poblaciones a travs de su media:

- **Hiptesis (caso mayor que):**
  - $H_0$: $\mu_A = \mu_B$
  - $H_A$: $\mu_A > \mu_B$

- **Hiptesis (caso menor que):**
  - $H_0$: $\mu_A = \mu_B$
  - $H_A$: $\mu_A < \mu_B$

- **Hiptesis (caso diferente de):**
  - $H_0$: $\mu_A = \mu_B$
  - $H_A$: $\mu_A \neq \mu_B$

Si cada grupo poblacional $A$ y $B$ siguen cada uno una distribucin normal tal que $A \sim \mathcal{N}(\mu_A, \sigma_A^2)$ y $B \sim \mathcal{N}(\mu_B, \sigma_B^2)$ respectivamente, observaremos las diferencias de las poblaciones a travs de las diferencias de sus medias muestrales. Debido a que sto significa generar una combinacin lineal de dos distribuciones gausianas, se tiene que:

$$
\hat{A} - \hat{B} \sim \mathcal{N}\left(\mu_A - \mu_B, \frac{\sigma_A^2}{n_A} + \frac{\sigma_B^2}{n_B}\right)
$$

Ahora, si quisieramos ejecutar una prueba de hiptesis sera necesario generar una estadstica $Z$, no obstante, debido a que no conocemos la varianza poblacional tenmos que usar la varianza muestral y por ende, usar la distribucin t-student.

$$
\hat{A} - \hat{B} \sim \mathcal{N}\left(\mu_A - \mu_B, \frac{S_A^2}{n_A} + \frac{S_B^2}{n_B}\right)
$$

$$
t_{AB} = \frac{\left(\hat{A} - \hat{B}\right) - \left( \mu_A - \mu_B \right)}{\sqrt{\frac{S_A^2}{n_A} + \frac{S_B^2}{n_B}}}
$$

donde $t_{AB}$ representa la estadstica $t$ de la diferencia entre las variablea aleatorias de inters, es decir, las medias muestrales en este caso. Una vez definida la estadstica $t$ la cul sigue una distribucin una distribucin t-student, debemos definir los grados de libertad de sta distribucin los cules no siguen la forma $df = n-1$ debido a que las muestras pueden ser de distinto tamao y tener varianzas muestrales diferentes. La expresin para calcular los grados de libertad de dos grupos bajo una prueba $t$ es:

$$
df = \frac{\left(\frac{S_A^2}{n_A} + \frac{S_B^2}{n_B}\right)^2}{\frac{\left(\frac{S_A^2}{n_A}\right)^2}{n_A - 1} + \frac{\left(\frac{S_B^2}{n_B}\right)^2}{n_B - 1}}
$$


El t-test se basa en el clculo de un valor t, que mide la diferencia entre las medias en relacin con la variabilidad dentro de los grupos de datos. Una vez hallada la estadstica $t$ y los grados de libertad, el valor t se compara con un valor crtico de la distribucin t correspondiente al nivel de significancia ($\alpha$) y a los grados de libertad del test. Si el valor t calculado es mayor que el valor crtico, se rechaza la hiptesis nula, sugiriendo que hay una diferencia significativa entre las medias.



Para ilustrar los casos de uso de los t-tests, primero supongamos que nos encontramos ante una situacin tipo **one-sample t-test** donde vamos a comparar la media de una muestra de nuestra poblacin con un valor  hipottico que se asume previamente al test como verdadero (un valor de referencia). Harmos uso a continuacin de nuestro conjunto de datos reales.

```python
# Parmetros de la poblacin
variable = "total_likes_on_outofstation_checkin_received"
data = list(df_turismo[variable].values)
population_mean = np.round(np.mean(data),3)  # media poblacional
population_std  = np.round(np.std(data),3)   # desviacin estndar poblacional   
n_samples = 10000                # nmero de muestras
sample_size = 50                 # tamao de cada muestra

# Generamos una muestra
sample = random.sample(data, sample_size)
sample_mean = np.mean(sample)

# t test de un sample independiente
t_statistic, p_value = stats.ttest_1samp(sample, population_mean)

print(f"Media poblacional: {population_mean:.2f}")
print(f"Media muestral: {sample_mean:.2f}")
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretacin
if p_value < 0.05:
    print(f"Rechazamos la hiptesis nula: La media de la muestra es significativamente diferente de {population_mean:.2f}.")
else:
    print(f"No se rechaza la hiptesis nula: No hay evidencia suficiente para decir que la media de la muestra es diferente de {population_mean:.2f}.")
```

---
Anlisis de Varianza (Analysis of Variance ANOVA)

---

En la seccin aterior, vimos cmo estadsticas como la estadstica $Z$ y la estadstica $t$ nos permten evaluar hiptesis entre dos pares de grupos o poblaciones tales como si una estimacin de un parmetro a partir de una muestra pertenece a la poblacin subyacente o comparar la similitud entre distribuciones de grupos/poblaciones a travz de un estadstico de prueba. No obstante como podemos evidenciar, la ms grande limitacin de tanto la prueba de hiptesis por estadstica $Z$ como $t$ es que solo se limitan a **dos grupos**. 

El anlisis de varianza ANOVA, es una tcnica estadstica que se utiliza para comparar las medias de **tres o ms grupos diferentes para determinar si al menos uno de ellos es significativamente diferente de los dems**. ANOVA es particularmente til cuando se desea examinar si las diferencias observadas entre los grupos son lo suficientemente grandes como para concluir que no se deben al azar y corresponde una extensin de los t-tests a varios grupos.

#### **Que es un factor en ANOVA?**

Un factor dentro del contexto de ANOVA es una variable que permite **agrupar o segregar** los datos y permite comparar las medias entre diferntes categoras, es decir, es una variable categrica que grcias a sus clases, permte dividir los datos en mltiples conjuntos independientes. Esto quiere decir que el tipo de ANOVA que pueda realizarse depender de cuntas variables categricas se tomen en cuenta para realizar un anlisis de influencia sobre otra variable contnua.

#### **Cul es la idea general de ANOVA**

EL anlisis de varianza particiona la variabilidad total observada en los datos en dos componentes:

1. **Variabilidad entre grupos (Sum of Squares Between SSB)**: Mide la variabilidad debido a las diferencias entre las medias de los grupos.

2. **Variabilidad dentro de los grupos (Sum of Squares Within SSW)**: Mide la variabilidad dentro de los grupos, que es debida al azar o a diferencias dentro de cada grupo.

ANOVA se basa en el uso del **estadstico f** que compara la  **variabilidad entre grupos con la variabilidad dentro de los grupos**.

$$
f = \frac{\text{Varianza entre los grupos}}{\text{Varianza dentro de los grupos}}
$$

Si la variabilidad entre los grupos es grande en relacin con la variabilidad dentro de los grupos, el valor de $f$ ser grande, lo que sugiere que *al menos uno de los grupos tiene una media significativamente diferente*. Esto involucra la **no influencia o independencia entre los grupos**. En trminos de la estadstica $f$ y su relacin con el testeo de hiptesis, podmos hacer las siguientes asociaciones:

- **Hiptesis**:
    - $H_0$ : las medias de los grupos provienen de la msma poblacin o, los grupos tienen influencia entre s o, las distribuciones son similares.
    - $H_A$ : al menos una media de los grupos no proviene de la msma poblacion o, no exste influencia entre s o, las distribuciones son distintas.

Con este panorama general, se plantean los siguientes escenarios, donde el primero ilustrado a continuacin se interpreta como *al menos una de las medias no pertenece a la poblacin o, las distribuciones son muy distintas entre s*.

$$
\frac{\text{grande}}{\text{pequea}} = \text{rechazamos } H_0 
$$

El siguiente caso se interpreta como *las medias son muy cercanas a la media poblacional, o es dificil distinguir entre distribuciones*.

$$
\frac{\text{similar}}{\text{similar}} = \text{fallamos en rechazar }H_0
$$

El siguiente caso se interpreta como *las medias son demasiado cercanas a la media poblacional, o las distribuciones son tan similares que stas son prcticamente idnticas*:

$$
\frac{\text{pequea}}{\text{grande}} = \text{fallamos en rechazar }H_0
$$

El anlisis ANOVA posee las siguientes suposiciones:

1. **Independencia**: Las observaciones en cada grupo deben ser independiente entre s.

2. **Normalidad**: Los datos dentro de cada grupo deben seguir una distribucin normal.

3. **Homogeneidad de varianzas**: Los grupos deben tener varianzas similares (homocedasticidad)

#### **Distribucin f**

El anlisis de varianza se basa en la distribucin **$F$ de Snedecor** (o simplemente distribucin $F$). Esta distribucin surge cuando se comparan dos estimaciones de varianza: una que mide la variabilidad entre grupos y otra que mide la variabilidad dentro de los grupos. La distribucin $F$ es asimtrica y no puede tomar valores negativos, lo que refleja que siempre se est comparando una razn de varianzas (que son cuadradas y, por lo tanto, no negativas).

La distribucin $F$ no slo aparece en el anlisis ANOVA sino tambin en las pruebas de hiptesis de igualdad de varianzas como Levene y Barlett o en regresin. As pues, ANOVA se basa en sta distribucin para comparar la proporcion de las varianzas intergrupales e intragrupales. Algnas de las propiedades de la distribucipon $F$ son:

1. **Aimetra positiva**: Es asimtrica hacia la derecha lo que significa que tiene una cola larga en el lado positivo.

2. **No-negativa**. Los valores de la distribucin son siempre mayores o iguales acero ya que las varianzas y sus cocientes no pueden ser negativos.

3. **Grados de libertad**: La forma de la distribucin $f$ depende de dos conjuntos de grados de libertad:
    - **m**: Grados de libertad asociados con el numerador (varianza del primer conjunto de datos o, de la varianza entre grupos).

    - **n**: Grados de libertad asociados con el denominador (varianza del segundo conjunto de datos o, de la varianza dentro de los grupos).

La funcin de densidad de probabilidad de la distribucin $f$ viene dada por:


$$
f(x \mid m, n) = \frac{1}{x \, B\left(\frac{m}{2}, \frac{n}{2}\right)} \sqrt{\frac{(m x)^{m} \, n^{n}}{(m x + n)^{m + n}}}
$$

Donde  $B$ es la funcin beta dada por:

$$
B(x, y) = \int_0^1 t^{x-1} (1-t)^{y-1} \, dt
$$


```python
# Valores de x para graficar la PDF
x = np.linspace(0, 5, 500)

# Grados de libertad para diferentes distribuciones F
degrees_of_freedom = [(1,1),(2,1),(5, 10), (10, 15), (20, 25), (30, 40)]

plt.figure(figsize=(10, 6))

for m, n in degrees_of_freedom:
    y = stats.f.pdf(x, m, n)
    plt.plot(x, y, label=f'm={m}, n={n}')

plt.title('Distribucin F para diferentes grados de libertad')
plt.xlabel('x')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

ANOVA de Un Factor o One-Way ANOVA sin Repeticin

---

El ANOVA One-way sin repeticin es la forma de ANOVA ms sencilla que se utiliza para comparar las medias de tres o ms grupos independientes para determinar si al menos una de ellas es significativamente diferente a las dems.  As pues, un anlisis ANOVA de un factor llevado acabo entre variables de un msmo conjunto de datos extrados a partir de la poblacin, corresponde a evaluar la **influencia de una nica variable categrica sobre una variable contnua**. 

#### **Cundo usar ANOVA de un factor sin repeticin?**

El anlisis ANOVA  de un factor sin repeticin es til en situaciones donde:

1. Se quiere conocer si exste influencia por parte de una variable categrica sobre una variable dependiente contnua sin repeticion de medidas, es decir, los miembros de cada grupo son distntos y por ende independientes.

#### **Suposiciones del ANOVA de un factor sin repeticin**

Las principales suposiciones del test ANOVA de un factor sin repeticin son:

1. **Independencia de las observaciones**: Las observaciones en cada grupo son independientes unas de otras.
2. **Normalidad**: Los datos dentro de cada grupo deben estar normalmente distribudos o, los resduos del modelo (diferencias entre los valores observados y los valores predichos) siguen una distribucin normal.
3. **Homocedasticidad**: La varianza de datos (o residuos en el caso de un modelo) es constante a travs de los niveles de los factores.


#### **Cules son la Hiptesis del ANOVA de un factor sin repeticin?**

El test ANOVA de un factor sin repeticin al buscar la veracidad de influencia de una variable categrica sobre una numrica contnua tiene solo una hiptesis nula y una alternativa:

- **Hiptesis nulas $H_0$**:
    - No hay una diferencia significativa entre los grupos de factor y las medias de cada grupo no son estadsticamente diferentes.

- **Hiptesis nulas $H_A$**:
    - Exste una diferencia significativa entre los grupos de factor y las medias de cada grupo son estadsticamente diferentes.

#### **Clculo del ANOVA de un factor sin repeticin**

Esto quiere  decir que el anova one-way va a tomar una nica variable categrica para generar los grupos. Dentro del one-way se asume la siguiente descomposicin de la varianza de los datos:

$$
\text{varianza total} = \text{varianza intergrupal} + \text{varianza intragrupal}
$$

Con lo anterior en mente, los pasos para calcular un anova one-way son los siguientes, donde tomarmos una variable **contnua** $X$ y la segmentaremos en grupos a partir de otra variable **categrica**:

1. **Definir las Hiptesis**:
   - **Hiptesis nula ($H_0$)**: Las medias de los grupos son iguales. $\mu_1$ = $\mu_2$ = $\mu_3$ = $\dots$ = $\mu_k$.
   - **Hiptesis alternativa ($H_A$)**: Al menos una de las medias es diferente.

2. **Clculo de la Suma de Cuadrados**:

   - **Suma de Cuadrados entre Grupos (SSA)**: Representa la variabilidad entre las medias de los grupos.

     $$
     SSA = \sum_{i=1}^{k} n_i (\bar{X}_i - \bar{X}_\text{global})^2
     $$

     Donde:
     - $k$ es el nmero de grupos.
     - $n_i$ es el tamao de la muestra en el grupo $i$.
     - $\bar{X}_i$ es la media del grupo $i$.
     - $\bar{X}_\text{global}$ es la media global de todas las observaciones.

   - **Suma de Cuadrados dentro de los Grupos (SSE)**: Representa la variabilidad dentro de cada grupo.

     $$
     SSE = \sum_{i=1}^{k} \sum_{j=1}^{n_i} (X_{ij} - \bar{X}_i)^2
     $$

     Donde:
     - $X_{ij}$ es la $j$-sima observacin en el grupo $i$.

   - **Suma de Cuadrados Total (SST)**: Es la variabilidad total en los datos.

     $$
     SST = SSA + SSE = \sum_{i=1}^{k} \sum_{j=1}^{n_i} (X_{ij} - \bar{X}_\text{global})^2
     $$

3. **Clculo de los Grados de Libertad**:
   - **Grados de libertad entre grupos**: $df_{SSA} = k - 1$
   - **Grados de libertad dentro de los grupos**: $df_{SSE} = N - k$
     - Donde $N$ es el total de observaciones.

4. **Clculo de la Media Cuadrtica**:
   - **Media Cuadrtica entre grupos (MSA)**: 
   $$
   MSA = \frac{SSA}{df_{SSA}} 
   $$

   - **Media Cuadrtica dentro de los grupos (MSE)**: 
   $$
   MSE = \frac{SSE}{df_{SSE}} 
   $$

5. **Clculo del Estadstico F**:
   - El estadstico $f$ compara la variabilidad entre los grupos con la variabilidad dentro de los grupos.

     $$
     f = \frac{MSA}{MSE} = \frac{\sigma_{SA}}{\sigma_{SE}}
     $$

     - Si $f$ es grande, sugiere que las diferencias entre las medias de los grupos no son solo debido al azar, lo que lleva a rechazar $H_0$.

6. **Comparacin con el Valor Crtico**:
   - El valor $f$ obtenido se compara con el valor crtico de la distribucin $f$ para un nivel de significancia dado $\alpha$. Si $f$ calculado es mayor que $f$ crtico, se rechaza la hiptesis nula $H_0$.



Para ilustrar el uso del ANOVA one-way de un factor sin repeticin y cmo se espera evidenciar un efecto del factor en una variable contnua de inters, a continuacin se presentan dos ejemplos que respectivamente muestran un caso de **no influencia del factor** y de **influencia del factor** sobre la variable aleatoria de inters.

```python
# Caso ANOVA one-way sin variabilidad entre grupos
# Generacin de datos para tres grupos
np.random.seed(0)
group1 = np.random.normal(loc=70, scale=10, size=30)
group2 = np.random.normal(loc=72, scale=10, size=30)
group3 = np.random.normal(loc=73, scale=10, size=30)

# Creacin de un DataFrame para los datos
df = pd.DataFrame({
    'Method_1': group1,
    'Method_2': group2,
    'Method_3': group3
})

# Realizar ANOVA de una va
f_statistic, p_value = stats.f_oneway(df['Method_1'], df['Method_2'], df['Method_3'])

print(f"F-Statistic: {f_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretacin
if p_value < 0.05:
    print("Rechazamos la hiptesis nula: Al menos un mtodo tiene un rendimiento diferente.")
else:
    print("No se rechaza la hiptesis nula: No hay evidencia suficiente para decir que los mtodos son diferentes.")

# Visualizacin de los datos
sns.boxplot(data=df)
plt.title('Comparacin de Puntuaciones entre Mtodos de Enseanza')
plt.ylabel('Puntuaciones')
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

```python
# Caso ANOVA one-way con variabilidad entre grupos
# Simulacin de datos
np.random.seed(42)

# Creacin de un DataFrame con una variable categrica y una continua
data = {
    'Group': np.repeat(['A', 'B', 'C'], 30),
    'Score': np.concatenate([
        np.random.normal(55, 10, 30),  # Grupo A
        np.random.normal(46, 10, 30),  # Grupo B
        np.random.normal(60, 10, 30)   # Grupo C
    ])
}

df = pd.DataFrame(data)

# Visualizar los datos con un boxplot
sns.boxplot(x='Group', y='Score', data=df)
plt.title('Distribucin de Puntuaciones por Grupo')
plt.xlabel('Grupo')
plt.ylabel('Puntuacin')
plt.grid(linestyle='--', alpha=0.7)
plt.show()


# Separar los datos por grupo
group_A = df[df['Group'] == 'A']['Score']
group_B = df[df['Group'] == 'B']['Score']
group_C = df[df['Group'] == 'C']['Score']

# Realizar ANOVA One-Way
f_statistic, p_value = stats.f_oneway(group_A, group_B, group_C)

# Mostrar resultados
print(f"F-Statistic: {f_statistic}")
print(f"P-Value: {p_value}")

# Interpretacin
alpha = 0.05
if p_value < alpha:
    print("Se rechaza la hiptesis nula: Hay diferencias significativas entre los grupos")
else:
    print("No se rechaza la hiptesis nula: No hay diferencias significativas entre los grupos")
```

ANOVA de Un Factor o One-Way ANOVA con Repeticin

---

La variante del ANOVA One-Way llamada ANOVA One-Way con Medidas Repetidas (One-Way Repeated Measures ANOVA). Esta versin del ANOVA se utiliza cuando se realizan mediciones repetidas sobre los mismos sujetos o unidades experimentales bajo diferentes condiciones o en diferentes momentos.

- **ANOVA One-Way tradicional:** Se utiliza cuando se comparan las medias de varios grupos **independientes**. Por ejemplo, se podra comparar la presin arterial media de tres grupos diferentes de pacientes (tratamiento A, tratamiento B, tratamiento C).

- **ANOVA One-Way con Medidas Repetidas:** Se utiliza cuando los mismos sujetos son medidos varias veces bajo diferentes condiciones o en diferentes momentos. Por ejemplo, si un grupo de pacientes es sometido a un tratamiento en diferentes momentos, y se mide su presin arterial despus de cada intervencin. Aqu, los mismos pacientes son medidos repetidamente, por lo que la variabilidad dentro de los sujetos se tiene en cuenta.


#### **Cundo usar ANOVA de un factor sin repeticin?**

El anlisis ANOVA  de un factor con repeticin es til en situaciones donde:

1. Se quiere conocer si una poblacin ha cambiado su comportamiento a lo largo de diferentes mediciones o tiempos.

#### **Suposiciones del ANOVA de un factor sin repeticin**

Las principales suposiciones del test ANOVA de un factor sin repeticin son:

1. **Independencia de las observaciones**: Las observaciones en cada grupo son independientes unas de otras.
2. **Normalidad**: Los datos dentro de cada grupo deben estar normalmente distribudos o, los resduos del modelo (diferencias entre los valores observados y los valores predichos) siguen una distribucin normal.
3. **Homocedasticidad**: La varianza de datos (o residuos en el caso de un modelo) es constante a travs de los niveles de los factores.


#### **Cules son la Hiptesis del ANOVA de un factor sin repeticin?**

El test ANOVA de un factor con repeticin al buscar la veracidad a cerca del cambio de una variable numrica contnua tiene solo una hiptesis nula y una alternativa:

- **Hiptesis nulas $H_0$**:
    - No hay una diferencia significativa entre los momentos de medicin y las medias de cada instancia de la poblacin no son estadsticamente diferentes.

- **Hiptesis nulas $H_A$**:
    - Exste una diferencia significativa entre los momentos de medicin y las medias de cada instancia de la poblacin son estadsticamente diferentes.


#### **Clculo del ANOVA de un factor con repeticin**

El one-way con medidas repetidas tiene en cuenta el **control de la variabilidad infividual** y la **variabilidad entre grupos o mediciones**. Esto quiere  decir que el ANOVA one-way con repeticin multiples variables relacionadas entre si en la medida que cada una corresponde a un "tratamiento, estado o momento de medicin de la poblacin". Dentro del one-way con repeticin se asume la siguiente descomposicin de la varianza de los datos:

$$
\text{variabilidad de cada sujeto para cada tratamiento} = \text{variabilidad total} - \text{variabilidad entre estados/tratamientos}
- \text{variabilidad entre sujetos}
$$

Con lo anterior en mente, los pasos para calcular un anova one-way con repeticin son los siguientes:

1. **Modelo Lineal**
El modelo matemtico del ANOVA One-Way con Medidas Repetidas se puede expresar como:

$$
Y_{ij} = \mu + \tau_i + \pi_j + \epsilon_{ij}
$$

donde:
- $Y_{ij}$ es el valor observado para el sujeto $j$ en el tratamiento o condicin $i$.
- $\mu$ es la media general.
- $\tau_i$ es el efecto del tratamiento $i$, que es el efecto fijo que queremos estudiar.
- $\pi_j$ es el efecto del sujeto $j$, que es un efecto aleatorio que controla las diferencias individuales entre sujetos.
- $\epsilon_{ij}$ es el trmino de error aleatorio asociado con la observacin $Y_{ij}$.

2. **Hiptesis**
   - **Hiptesis nula ($H_0$)**: Las medias de las diferentes condiciones o tratamientos son iguales.
   $$
   H_0: \tau_1 = \tau_2 = \ldots = \tau_k
   $$
   - **Hiptesis alternativa ($H_A$)**: Al menos una de las medias es diferente.
   $$
   H_A: \text{Al menos una } \tau_i \text{ es diferente}
   $$

3. **Particin de la Suma de Cuadrados**
La variabilidad total en los datos se descompone en diferentes componentes:

    3.1. **Suma de Cuadrados Total (SST)**:
      $$
      SST = \sum_{i=1}^{k} \sum_{j=1}^{n} (Y_{ij} - \bar{Y})^2
      $$
      donde $\bar{Y}$ es la media general de todas las observaciones.

    3.2. **Suma de Cuadrados de Tratamientos (SSB)**: Representa la variabilidad debida a las diferencias entre las medias de los tratamientos.
    $$
    SSB = \sum_{i=1}^{k} n_i (\bar{Y}_i - \bar{Y})^2
    $$
    donde $\bar{Y}_i$ es la media del tratamiento $i$ y $n_i$ es el nmero de sujetos.

    3.3. **Suma de Cuadrados de Sujeto (SSS)**: Representa la variabilidad debida a las diferencias entre los sujetos.
    $$
    SSS = \sum_{j=1}^{n} k (\bar{Y}_j - \bar{Y})^2
    $$
    donde $\bar{Y}_j$ es la media de las mediciones para el sujeto $j$.

    3.4. **Suma de Cuadrados del Error (SSE)**: Representa la variabilidad no explicada dentro de cada sujeto para cada tratamiento.
    $$
    SSE = SST - SSB - SSS
    $$

4. **Grados de Libertad**
   - **Grados de libertad para los tratamientos ($df_B$)**: $ k - 1 $, donde $ k $ es el nmero de tratamientos o condiciones.
   - **Grados de libertad para los sujetos ($df_S$)**: $n - 1$, donde $n$ es el nmero de sujetos.
   - **Grados de libertad para el error ($df_E$)**: $(k - 1) \times (n - 1)$.

5. **Cuadrados Medios**
   - **Cuadrado Medio de Tratamientos (MSB)**:
   $$
   MSB = \frac{SSB}{df_B}
   $$
   - **Cuadrado Medio del Error (MSE)**:
   $$
   MSE = \frac{SSE}{df_E}
   $$

6. **Estadstico $f$**
El estadstico $f$ se calcula comparando la variabilidad explicada por los tratamientos con la variabilidad no explicada (error):

$$
F = \frac{MSB}{MSE}
$$


```python
# Clculo de anova one way con medidas repetidas usando statsmodels
import statsmodels.api as sm
from statsmodels.stats.anova import AnovaRM

# Simulacin de datos
np.random.seed(42)
n_subjects = 30
data = {
    'Subject': np.repeat(np.arange(1, n_subjects + 1), 3),
    'Condition': np.tile(['No Treatment', 'Treatment A', 'Treatment B'], n_subjects),
    'Score': np.concatenate([
        np.random.normal(75, 10, n_subjects),  # Sin tratamiento
        np.random.normal(80, 10, n_subjects),  # Tratamiento A
        np.random.normal(85, 10, n_subjects)   # Tratamiento B
    ])
}

df = pd.DataFrame(data)

# Realizar ANOVA One-Way con Medidas Repetidas
anova_rm = AnovaRM(df, 'Score', 'Subject', within=['Condition'])
res = anova_rm.fit()

# Mostrar resultados
print(res.summary())
```

ANOVA de Dos Factores o Two-Way ANOVA sin Repeticin

---

El ANOVA de dos factores sin repeticin (Two-Way ANOVA) es una extensin del anlisis de varianza (ANOVA) de un factor, que permite evaluar el efecto de dos factores independientes en una variable dependiente continua. Matemticamente, se utiliza para **descomponer la variabilidad total de los datos en componentes atribuibles a los dos factores y su interaccin**. En ste sentido, el ANOVA permite examinar en nuestros datos lo siguiente:

1. Efecto individual de cada factor sobre una variable contnua (efectos principales).

2. Interaccin entre los factores, es decir, cmo el efecto de un factor vara en funcin de los nivels del otro factor.

#### **Cundo usar ANOVA de dos factores?**

El ANOVA de dos factores se usa en situaciones como:

- Exsten dos factores categricos a investigar.
- Cada combinacin de niveles de los factores tiene observaciones independientes (no hay medidas repetidas).
- Se quiere evaluar tanto los efectos principales como la interaccin entre los factores.

#### **Suposiciones del ANOVA de dos factores**

1. **Independencia de las observaciones**: Las observaciones en cada grupo son independientes unas de otras.
2. **Normalidad**: Los datos dentro de cada grupo deben estar normalmente distribudos o, los resduos del modelo (diferencias entre los valores observados y los valores predichos) siguen una distribucin normal.
3. **Homocedasticidad**: La varianza de datos (o residuos en el caso de un modelo) es constante a travs de los niveles de los factores.
4. **Independencia factorial**: Los dos factores a usar son independientes entre si.

#### **Cules son la Hiptesis del ANOVA de dos factores?**

Debido a que el ANOVA de dos factores permte evaluar tres aspectos de nuestros datos (influencia del factor uno, influencia del factor dos y la interaccin entre los factores), ste tipo de test estadstico contiene tres hiptesis nulas y tres hiptesis alternativas:

- **Hiptesis nulas $H_0$**:
    - No hay una diferencia significativa entre los grupos del primer factor.
    - No hay una diferencia significativa entre los grupos del segndo factor.
    - Un factor no tiene influencia sobre la influencia del otro factor sobre la variable contnua.

- **Hiptesis nulas $H_A$**:
    - Exste una diferencia significativa entre los grupos del primer factor.
    - Exste una diferencia significativa entre los grupos del segndo factor.
    - Un factor tiene influencia sobre la influencia del otro factor sobre la variable contnua.

#### **Clculo del ANOVA de dos factores sin repeticin**

El ANOVA de dos factores sin repeticin se basa en la siguiente descomposicin de la varianza:

$$
\text{varianza total} = \text{var. explicada por factor A} + \text{var. explicada por factor B} + \text{var. explicada por la interaccin A y B} + \text{var. de error}
$$

Esta descomposicin de la varianza total de la variable contnua, la traducmos por notacin trmino a trmino de la siguiente manera:

$$
SS_T = SS_A + SS_B + SS_{AB} + SS_{\text{err}}
$$

El test ANOVA de dos factores funciona bajo la lgica de que si la suma de los trminos $SS_A + SS_B + SS_{AB}$ corresponde a la **mayor** contribucin a la varianza total, se puede asegurar que los factores si tienen un efecto en la variable contnua, mientras que si el trmino de error $ SS_{\text{err}}$ contiene la mayor contribucin a ala varianza total, se puede asegurar que los factores no tienen una influencia apreciable en la variable contnua. 

El procedimiento para calcular el ANOVA de dos factores es el siguiente:


1. **Modelo Matemtico**

El modelo general para un ANOVA de dos factores sin repeticin es:

$$
Y_{ijk} = \mu + \alpha_i + \beta_j + (\alpha\beta)_{ij} + \epsilon_{ijk}
$$

Donde:
- $Y_{ijk}$ es el valor observado de la variable dependiente para el $k$-simo sujeto en el nivel $i$ del factor $A$ y el nivel $j$ del factor $B$.
- $\mu$ es la media general de la poblacin.
- $\alpha_i$ es el efecto del $i$-simo nivel del factor $A$.
- $\beta_j $ es el efecto del $j$-simo nivel del factor $B$.
- $(\alpha\beta)_{ij}$ es el efecto de interaccin entre el $i$-simo nivel del factor $A$ y el $j$-simo nivel del factor $B$.
- $\epsilon_{ijk}$ es el trmino de error aleatorio, que se supone sigue una distribucin normal $\mathcal{N}(0, \sigma^2)$.

2. **Descomposicin de la Suma de Cuadrados**

El objetivo del ANOVA es dividir la variabilidad total observada en los datos en componentes atribuibles a los factores $A$ y $B$, a su interaccin, y al error aleatorio.

- 2.1. **Suma de Cuadrados Total ($SS_{Total}$)**:
        
    $$
    SS_{Total} = \sum_{i=1}^{a} \sum_{j=1}^{b} \sum_{k=1}^{n} (Y_{ijk} - \bar{Y}_{G})^2
    $$

    Aqu, $\bar{Y}_{G}$  es la media global de todos los datos.

- 2.2. **Suma de Cuadrados para el Factor A ($SS_A$)**:
    
    $$
    SS_A = n \cdot b \sum_{i=1}^{a} (\bar{Y}_{i} - \bar{Y}_{G})^2
    $$

    Aqu, $\bar{Y}_{i}$ es la media del $i$-simo nivel del factor $A$.

- 2.3. **Suma de Cuadrados para el Factor B ($SS_B$)**:
    
    $$
    SS_B = n \cdot a \sum_{j=1}^{b} (\bar{Y}_{j} - \bar{Y}_{G})^2
    $$

    Aqu, $\bar{Y}_{j}$ es la media del $j$-simo nivel del factor $B$.

- 2.4. **Suma de Cuadrados para la Interaccin $A  B$ ($SS_{AB}$)**: Para calcular la suma de cuadrados de la interaccin, primero calculamos la suma de cuadrados entre los grupos, donde $\bar{Y}_{ij}$ es la media del grupo $ij$ producto de la combinacin del nivel $i$ del factor $A$ y del nivel  $j$ del factor $B$.

    $$
    SS_{btw} = n \sum_{i=1}^{a} \sum_{j=1}^{b} (\bar{Y}_{ij} - \bar{Y}_{G})
    $$

    Finalmente, la a sta suma de cuadrados se le resta la contribucin de cada factor tal que:

    $$
    SS_{AB} = SS_{\text{btw}} - SS_A - SS_b
    $$

    La suma de los cuadrados para la interaccin entre factores de manera genera, tambin puede expresarse como:
    
    $$
    SS_{AB} = n \sum_{i=1}^{a} \sum_{j=1}^{b} (\bar{Y}_{ij} - \bar{Y}_{i} - \bar{Y}_{j} + \bar{Y}_{G})^2
    $$

    Aqu, $\bar{Y}_{ij}$ es la media de las observaciones en la combinacin del $i$-simo nivel del factor $A$ y el $j$-simo nivel del factor $B$.

    
- 2.5. **Suma de Cuadrados del Error ($SS_{Error}$)**:
  
    $$
    SS_{Error} = \sum_{i=1}^{a} \sum_{j=1}^{b} \sum_{k=1}^{n} (Y_{ijk} - \bar{Y}_{ij})^2
    $$

    Representa la variabilidad dentro de los grupos que no puede explicarse por los factores ni por su interaccin.

3. **Grados de Libertad**

    Cada componente de la suma de cuadrados tiene un nmero asociado de grados de libertad ($df$):

    - **Grados de libertad del factor A**: $df_A = a - 1$
    - **Grados de libertad del factor B**: $df_B = b - 1$
    - **Grados de libertad de la interaccin A  B**: $df_{AB} = (a - 1)(b - 1)$
    - **Grados de libertad del error**: $df_{Error} = ab(n - 1)$
    - **Grados de libertad total**: $df_{Total} = abn - 1$

4. **Clculo de los Cuadrados Medios**

    Los cuadrados medios (Mean Squares, MS) se obtienen dividiendo las sumas de cuadrados por sus respectivos grados de libertad, lo cul es equivalente a una varianza segn cada uno de los criterios nombrados anteriormente:

    $$
    \sigma_A = \frac{SS_A}{df_A}, \quad \sigma_B = \frac{SS_B}{df_B}, \quad \sigma_{AB} = \frac{SS_{AB}}{df_{AB}}, \quad \sigma_{Error} = \frac{SS_{Error}}{df_{Error}}
    $$

5. **Clculo de los Valores F**

    Cada efecto principal y la interaccin se evalan comparando el cuadrado medio del efecto con el cuadrado medio del error:

    $$
    f_A = \frac{\sigma_A}{\sigma_{Error}}, \quad f_B = \frac{\sigma_B}{\sigma_{Error}}, \quad f_{AB} = \frac{\sigma_{AB}}{\sigma_{Error}}
    $$

6. **Pruebas de Hiptesis**

- **Hiptesis nula para el factor A**: $H_0: \alpha_i = 0 $ para todos $i$.
- **Hiptesis nula para el factor B**: $H_0: \beta_j = 0$ para todos $j$.
- **Hiptesis nula para la interaccin**: $H_0: (\alpha\beta)_{ij} = 0 $ para todos $ i, j $.

```python
# Clculo de estadstica ANOVA sin repeticion mediante libreras estadsticas
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Generar datos simulados
np.random.seed(0)

# Variables
n = 30  # Nmero de observaciones por grupo
factor_A = np.repeat(['A1', 'A2'], n * 2)
factor_B = np.tile(np.repeat(['B1', 'B2'], n), 2)
data = 3 + 1.5 * (factor_A == 'A2') + 2 * (factor_B == 'B2') + 1.2 * (factor_A == 'A2') * (factor_B == 'B2')
data += np.random.normal(0, 1, len(data))

# Crear DataFrame
df = pd.DataFrame({
    'Factor_A': factor_A,
    'Factor_B': factor_B,
    'Response': data
})

# ANOVA de dos vas
model = ols('Response ~ Factor_A * Factor_B', data=df).fit()
#model = ols('Response ~ Factor_A + Factor_B + Factor_A:Factor_B', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Resultados
print(anova_table)

# Visualizacin
sns.boxplot(x='Factor_A', y='Response', hue='Factor_B', data=df)
plt.title('Boxplot de la respuesta para cada combinacin de factores')
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

```python
# Clculo manual de la estadstica ANOVA de dos factores sin repeticin

# Supongamos que tienes tus datos en un DataFrame de pandas
data = pd.DataFrame({
    'Factor1': ['A', 'A', 'B', 'B'] * 2,
    'Factor2': ['X', 'Y'] * 4,
    'Value': [10, 15, 20, 25, 12, 18, 22, 28]
})

def anova_two_way(df,factor1,factor2,valores):
    # Usaremos groupby y unstack para generar los promedios de cada grupo
    anova_data = data.groupby([factor1, factor2])[valores].mean().unstack()

    # Realiza el anlisis ANOVA de dos factores con interaccin
    # Extraemos las categoras y las dimensiones
    factor1 = anova_data.index
    factor2 = anova_data.columns
    n = len(factor1)
    m = len(factor2)

    # Hallamos las sumas de cuadrados
    grand_mean = anova_data.values.mean()
    ssTotal = ((data[valores] - grand_mean)**2).sum()

    ssA = m * ((anova_data.mean(axis=1) - grand_mean)**2).sum()
    ssB = n * ((anova_data.mean(axis=0) - grand_mean)**2).sum()

    ssAB = 0
    for i in range(n):
        for j in range(m):
            ssAB += (anova_data.iloc[i, j] - anova_data.iloc[i, :].mean() - anova_data.iloc[:, j].mean() + grand_mean)**2

    ssError = ssTotal - ssA - ssB - ssAB

     # Clculo de los grados de libertad
    dfA = n - 1
    dfB = m - 1
    dfAB = (n - 1) * (m - 1)
    dfError = data.shape[0] - (n * m)

    # Clculo de las varianzas por factor
    msA = ssA / dfA
    msB = ssB / dfB
    msAB = ssAB / dfAB
    msError = ssError / dfError

    # Clculo de la estadstica f
    F_A = msA / msError
    F_B = msB / msError
    F_AB = msAB / msError
     
    # Clculo del valor p
    p_A = 1 - stats.f.cdf(F_A, dfA, dfError)
    p_B = 1 - stats.f.cdf(F_B, dfB, dfError)
    p_AB = 1 - stats.f.cdf(F_AB, dfAB, dfError)

    # Imprimimos los resultados
    print("ANOVA Results:")
    print(f"Factor A: F={F_A:.4f}, p={p_A:.4f}")
    print(f"Factor B: F={F_B:.4f}, p={p_B:.4f}")
    print(f"Interaction AB: F={F_AB:.4f}, p={p_AB:.4f}")
    
anova_two_way(data, 'Factor1', 'Factor2', 'Value')
```

---
ANOVA de Dos Factores o Two-Way ANOVA Con Repeticin

---

El ANOVA de dos factores con repeticin (Two-Way ANOVA with repetition) es una extensin del anlisis de varianza dos factores donde  **un factor corresponde a una variable categrica la cul genera un tipo de grupos** y exste  **un segndo factor pareado que incluye repeticin**, es decir que el segndo factor se constituye de los diferentes resultados obtenidos en dos momentos de medicin hechos sobre la msma poblacin, teniendo en cuenta un factor categrico.

Esta tcnica permite evaluar **el efecto de una divisin categrica a lo largo de mediciones repetidas efectuadas sobre una msma poblacin**, permitiendo **descomponer la variabilidad total de los datos en la contribucin del factor de condicin y el factor categrico, teniendo en cuenta un factor de error**. En ste sentido, el ANOVA permite examinar en nuestros datos lo siguiente:

1. Efecto del factor de repeticin sobre la variable contnua dependiente.

2. Efecto del factor categrico (sin repeticin) sobre la variable contnua dependiente.

2. Interaccin entre los factores de repeticin y el factor categrico sobre la variable contnua dependiente.

#### **Cundo usar ANOVA de dos factores con repeticin?**

El ANOVA de dos factores con repeticin se usa en situaciones como:

- Exste **un factor sin repeticin** (variable categrica) que afecta a un **segndo factor con repeticin** (variable contnua), como por ejemplo ciando se quiere estudiar el efecto (segundo factor con repeticin) de dos tratamientos diferentes (variable categrica o factor sin repeticin) sobre pacientes y cada paciente recibe ambos tratamientos en diferentes momentos. 


- Cada combinacin de niveles de los factores tiene observaciones dependientes, donde los msmos sujetos son expuestos a todas las condiciones de uno de los factores, loque permite controlar la variabilidad entre sujetos.

- Se quiere evaluar tanto los efectos principales como la interaccin entre los factores.

#### **Suposiciones del ANOVA de dos factores**

1. **Escala para el factor dependiente y categorizacin de los factores**: La variable dependiente debe ser contnua bajo una escala determinada (nivel de presin, salario, etc.) mientras que los factores implicados deben ser categricos.

2. **Dependencia en los grupos del primer factor**: Las observaciones entre sujetos **NO son independientes**, aunque las observaciones  de cada sujeto pueden estar correlacionadas.

3. **Independencia en los grupos del segndo factor**: Las observaciones entre grupos generados por la variable categrica **son independientes** entre s.

4. **Normalidad**: Los datos dentro de cada grupo deben estar normalmente distribudos o, los resduos del modelo (diferencias entre los valores observados y los valores predichos) siguen una distribucin normal.

5. **Homocedasticidad**: La varianza de datos (o residuos en el caso de un modelo) de los grupos formados por el facor **sin repeticin** deben ser iguales o cercanas.

6. **Esfericidad (para el factor de repeticin)**: Las varianzas de las diferencias entre todos los pares posibles de condiciones deben ser aproximadamente iguales.

#### **Cules son la Hiptesis del ANOVA de dos factores?**

Debido a que el ANOVA de dos factores con repeticin permte evaluar tres aspectos de nuestros datos (influencia del factor de repeticin, influencia del factor categrico y la interaccin entre la repeticin y las categoras), ste tipo de test estadstico contiene tres hiptesis nulas y tres hiptesis alternativas:

- **Hiptesis nulas $H_0$**:
    - No hay una diferencia significativa entre los grupos del primer factor. Esto puede interpretarse como que no exste una diferencia significativa entre **los instantes de medicin**.
    - No hay una diferencia significativa entre los grupos del segndo factor. Esto puede interpretarse como que la divisin categrica (ej. un conjunto de tratamientos) no tiene influencia.
    - Un factor no tiene influencia sobre la influencia del otro factor sobre la variable contnua. 

- **Hiptesis nulas $H_A$**:
    - Exste una diferencia significativa entre los grupos del primer factor. Esto puede interpretarse como que exste una diferencia significativa entre **los instantes de medicin**.
    - Exste una diferencia significativa entre los grupos del segndo factor. Esto puede interpretarse como que la divisin categrica (ej. un conjunto de tratamientos) si tiene un efecto significativo.
    - Un factor tiene influencia sobre la influencia del otro factor sobre la variable contnua.

#### **Clculo del ANOVA de dos factores con repeticin**

El ANOVA de dos factores con repeticin se basa en la siguiente descomposicin de la varianza:

$$
\text{varianza total} = \text{var. explicada por factor S de repeticion} + \text{var. explicada por factor C categrico} + \text{var. explicada por la interaccin S y C} + \text{var. de error}
$$

Esta descomposicin de la varianza total de la variable contnua, la traducmos por notacin trmino a trmino de la siguiente manera, donde la siguiente expresin hace referencia a la suma de los cuadrados de la diferencia en cada caso:

$$
SS_T = SS_C + SS_{T} + SS_{CS} + SS_{\text{Error}}
$$

El test ANOVA de dos factores con repeticin funciona bajo la lgica de que si la suma de los trminos $SS_C + SS_S + SS_{CS}$ corresponde a la **mayor** contribucin a la varianza total, se puede asegurar que los factores si tienen un efecto en la variable contnua, mientras que si el trmino de error $ SS_{\text{Error}}$ contiene la mayor contribucin a ala varianza total, se puede asegurar que los factores no tienen una influencia apreciable en la variable contnua. 

El procedimiento para calcular el ANOVA de dos factores es el siguiente:


1. **Modelo Matemtico**

Suponiendo que tenemos un diseo de medidas repetidas de dos factores con un factor de condiciones o tratamientos $C$ con $k$ niveles y un segndo factor de medida repetida $T$ con $m$ niveles de medicin,
el modelo general para un ANOVA de dos factores sin repeticin es:

$$
Y_{ijk} = \mu + \alpha_i + \beta_j + (\alpha \beta)_{ij} + S_k + \epsilon_{ijk}
$$

Donde:
- $Y_{ijk}$ es el avlor de la observacin para el sujeto $k$, bajo el nivel $i$ del factor $C$ 
- $\mu$ es la media general de la poblacin.
- $\alpha_i$ es el efecto del tratamiento o condicin $i$ del factor $C$.
- $\beta_j$ es el efecto del nivel $j$ del factor de momentos $T$.
- $S_k $ es el efecto del sujeto $k$.
- $(\alpha \beta)_{ij}$ es la interaccin entre los niveles $i$ y  $j$ de los factores de condicin y momentos $C$ y $T$
- $\epsilon_{ijk}$ es el trmino de error aleatorio de la observacin, que se supone sigue una distribucin normal $\mathcal{N}(0, \sigma^2)$.

2. **Descomposicin de la Suma de Cuadrados**

El objetivo del ANOVA de doble factor con repeticin es dividir la variabilidad total observada en los datos en componentes atribuibles a los factores de repeticin $T$ y el factor categrico $C$, a su interaccin, y al error aleatorio.

- 2.1. **Suma de Cuadrados Total ($SS_{Total}$)**:
        
    $$
    SS_{Total} = \sum_{i=1}^{k} \sum_{j=1}^{m} \sum_{k=1}^{n} (Y_{ijk} - \bar{Y}_{G})^2
    $$

    Aqu, $\bar{Y}_{G}$  es la media global de todos los datos y $\bar{Y}_{ijk} es la observacin del sujeto $k$ expuesto al tratamiento $i$ en el momento de medicin $j$.

- 2.2. **Suma de Cuadrados Condicionales ($SS_C$)**:
    
    $$
    SS_C = n \cdot m \sum_{i=1}^{k} (\bar{Y}_{i} - \bar{Y}_{G})^2
    $$

    Aqu, $\bar{Y}_{i}$ es la media del $i$-sma condicin de $C$.

- 2.3. **Suma de Cuadrados Entre Situaciones/Tiempos ($SS_T$)**:
    
    $$
    SS_T = k \cdot n \sum_{j=1}^{m} (\bar{Y}_{j} - \bar{Y}_{G})^2
    $$

    Aqu, $\bar{Y}_{j}$ es la media del $j$-simo momento de medicin.

- 2.4. **Suma de Cuadrados para la Interaccin $C  T$ ($SS_{CT}$)**: Para calcular la suma de cuadrados de la interaccin de repeticin y la divisin categrica, primero calculamos la suma de cuadrados entre los grupos, donde $\bar{Y}_{ij}$ es la media del grupo $ij$ producto de la combinacin del nivel $i$ del factor $C$ y del nivel  $j$ del factor $T$.
    
    $$
    SS_{CT} = n \sum_{i=1}^{k} \sum_{j=1}^{m} (\bar{Y}_{ij} - \bar{Y}_{i} - \bar{Y}_{j} + \bar{Y}_{G})^2
    $$

    Aqu, $\bar{Y}_{ij}$ es la media de las observaciones en la combinacin del $i$-simo nivel del factor $C$ y el $j$-simo nivel del factor $T$.

    
- 2.5. **Suma de Cuadrados del Error ($SS_{Error}$)**:
  
    $$
    SS_{Error} = SST - SS_T -SS_C - SS_{CT}
    $$

    Representa la variabilidad dentro de los grupos que no puede explicarse por los factores ni por su interaccin.

3. **Grados de Libertad**

    Cada componente de la suma de cuadrados tiene un nmero asociado de grados de libertad ($df$):

    - **Grados de libertad del factor C**: $df_C = k - 1$
    - **Grados de libertad del factor T**: $df_T = m - 1$
    - **Grados de libertad de la interaccin C  T**: $df_{CT} = (k - 1) (m - 1)(n - 1)$
    - **Grados de libertad del error**: $df_{Error} = ab(n - 1)$
    - **Grados de libertad total**: $df_{Total} = abn - 1$

4. **Clculo de los Cuadrados Medios**

    Los cuadrados medios (Mean Squares, MS) se obtienen dividiendo las sumas de cuadrados por sus respectivos grados de libertad, lo cul es equivalente a una varianza segn cada uno de los criterios nombrados anteriormente:

    $$
    \sigma_C = \frac{SS_C}{df_C}, \quad \sigma_B = \frac{SS_T}{df_T}, \quad \sigma_{CT} = \frac{SS_{CT}}{df_{CT}}, \quad \sigma_{Error} = \frac{SS_{Error}}{df_{Error}}
    $$

5. **Clculo de los Valores F**

    Cada efecto principal y la interaccin se evalan comparando el cuadrado medio del efecto con el cuadrado medio del error:

    $$
    f_C = \frac{\sigma_C}{\sigma_{Error}}, \quad f_T = \frac{\sigma_T}{\sigma_{Error}}, \quad f_{CT} = \frac{\sigma_{CT}}{\sigma_{Error}}
    $$

6. **Pruebas de Hiptesis**

- **Hiptesis nula para el factor C**: $H_0: \alpha_i = 0 $ para todos $i$.
- **Hiptesis nula para el factor T**: $H_0: \beta_j = 0$ para todos $j$.
- **Hiptesis nula para la interaccin**: $H_0: (\alpha\beta)_{ij} = 0 $ para todos $ i, j $.


```python
# Clculo de estadstica ANOVA con repeticion mediante libreras estadsticas
from statsmodels.stats.anova import AnovaRM

# Datos ficticios
np.random.seed(0)
data = {
    'Sujeto': np.repeat(np.arange(1, 11), 9),
    'Dieta': np.tile(np.repeat(['Dieta_1', 'Dieta_2', 'Dieta_3'], 3), 10),
    'Tiempo': np.tile(['T1', 'T2', 'T3'], 30),
    'Peso': np.random.normal(70, 5, 90)
}

df = pd.DataFrame(data)

# ANOVA de medidas repetidas factorial con interaccin
aovrm = AnovaRM(df, 'Peso', 'Sujeto', within=['Dieta', 'Tiempo'])
res = aovrm.fit()

# Resultados
print(res)

# Graficar interaccin
interaction_plot = pd.pivot_table(df, values='Peso', index='Tiempo', columns='Dieta')
interaction_plot.plot(marker='o')

plt.title('Interaccin entre Dieta y Tiempo en la Prdida de Peso')
plt.xlabel('Tiempo')
plt.ylabel('Peso Promedio')
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

Estimacin e Inferencia

---


**La estimacin** es un proceso fundamental en estadstica que consiste en **inferir los valores de los parmetros desconocidos de una poblacin a partir de una muestra de datos**. Estos parmetros pueden incluir la media, la varianza, la proporcin, entre otros. La estimacin nos permite hacer conjeturas informadas sobre la poblacin sin tener que medir cada individuo en ella, lo que suele ser imprctico o imposible. 

En la presente seccin, vermos una introduccin a la estimacin de parmetros ms complejos que la media o la varianza, en concreto, observarmos como es posible mediante la teora de la probabilidad deducir el comportamiento de la poblacin a partir de la estimacin de parmetros de modelos que pretenden ser la generalizacin del comportamiento bajo supuestos de su dinmica.

#### **Estimacin Puntual y por Intervalos**

**La estimacin puntual** es un tipo de estimacin en la que se proporciona un nico valor como estimado para un parmetro desconocido de la poblacin. Este valor puntual se calcula a partir de los datos de la muestra. Por ejemplo, si queremos estimar la media de una poblacin, podramos utilizar la media muestral como una estimacin puntual de la media poblacional.

Sin embargo, la estimacin puntual no nos da informacin sobre la precisin de la estimacin, es decir, cunto puede variar este estimado del verdadero valor del parmetro poblacional.

Por otra parte, como vmos en la seccin de pruebas de hiptesis, la **estimacin por intervalos** es un mtodo que estima el parmetro de dentro de un rango de valores en lugar de dar un nico valor para la estimacin. Este rango como vimos, se denomina como intervalo de confianza y proporciona una estimacin ms informativa, ya que refleja no solo el valor estimado del parmetro sino tambin la incertidumre asociada a dicha estimacin.

#### **Maximum Likelihood Estimation (MLE)**

**Maximum Likelihood Estimation (MLE)** es un mtodo de estimacin puntual que se utiliza para estimar los parmetros de un modelo estadstico. El principio detrs de MLE es **encontrar los valores de los parmetros que maximizan la probabilidad de que los datos observados hayan sido generados por el modelo**.

En otras palabras, el MLE busca el valor del parmetro que hace que los datos observados sean "ms probables" bajo el modelo estadstico asumido.

El MLE es un mtodo muy popular en estadstica debido a sus propiedades deseables, como la consistencia (a medida que el tamao de la muestra aumenta, la estimacin se aproxima al valor verdadero del parmetro) y la eficiencia (tiene la menor varianza posible entre los estimadores imparciales).

#### **Maximum A Posteriori Estimation (MAP)**

El **Maximum A Posteriori (MAP) Estimation** es un enfoque de estimacin de parmetros en estadstica bayesiana. A diferencia del Maximum Likelihood Estimation (MLE), que slo considera la probabilidad de los datos dados los parmetros, MAP tambin incorpora una distribucin previa sobre los parmetros. El objetivo del MAP es encontrar el valor de los parmetros que maximiza la distribucin posterior, es decir, la probabilidad de los parmetros dados los datos observados.

#### **Relacin con el Aprendizaje Automtico (ML)**

**El aprendizaje automtico (Machine Learning, ML)** utiliza extensivamente tcnicas de estimacin, incluido el MLE, para ajustar modelos a datos. En ML, el MLE se usa para ajustar modelos probabilsticos a los datos de entrenamiento, lo que implica encontrar los valores de los parmetros del modelo que mejor explican los datos observados.

**<h3 id="subsection51" style="color: #003366;">5.1. Maximum Likelihood Estimation</h3>**

---
Maximum Likelihood Estimation (MLE)

---

El **Maximum Likelihood Estimation (MLE)** es un mtodo fundamental en estadstica para estimar los parmetros de un modelo probabilstico. El objetivo del MLE es encontrar los valores de los parmetros que maximizan la **verosimilitud** de los datos observados, es decir, aquellos **parmetros que hacen que los datos observados sean ms probables bajo el modelo**.

Supongamos que tenemos un conjunto de datos $\mathbf{X} = \{x_1, x_2, \dots, x_n\}$ y que estos datos provienen de una distribucin paramtrica con funcin de densidad de probabilidad (pdf) $f(x; \theta) $, donde $\theta$ es un vector de parmetros desconocidos que queremos estimar.

La **verosimilitud** es la probabilidad de observar el conjunto de datos $\mathbf{X}$ dados los parmetros $\theta$. Si se asume que cada observacin es independiente una de la otra, la funcin $f(x_i,\theta) da la probabilidad de que el dato puntual $x_i$ ha sido generado por el modelo, as pues si quermos tener un panorama a cerca de la probabilidad total de que dados unos parmetros se generen unas observaciones, la funcin de verosimilitud es el **producto de las probabilidades independientes de cada observacin**:

$$
L(\theta; \mathbf{X}) = P(\mathbf{X} | \theta) = \prod_{i=1}^{n} f(x_i; \theta)
$$

El MLE busca los valores de $\theta$ que maximizan esta funcin de verosimilitud. Es decir, la combinacin de parmetros que maximizan la probabilidad de que unos datos hallan sido observados dados un modelo, se encuentra mediante una optimizacin de la funcin (se pretende buscar un mximo dentro de la funcin de verosimilitud):

$$
\hat{\theta}_{\text{MLE}} = \text{argmax}_{\theta} \ L(\theta; \mathbf{X})
$$


Para simplificar los clculos (usar una funcin objetivo ms sencilla de manejar), en lugar de maximizar la verosimilitud directamente, a menudo se maximiza su logaritmo, conocido como la **log-verosimilitud**. Esto es posible porque el logaritmo es una funcin montona creciente, lo que significa que maximizar la verosimilitud y maximizar la log-verosimilitud llevan a los mismos resultados.

$$
\ell(\theta; \mathbf{X}) = \log L(\theta; \mathbf{X}) = \sum_{i=1}^{n} \log f(x_i; \theta)
$$

El estimador MLE se encuentra entonces maximizando la log-verosimilitud:

$$
\hat{\theta}_{\text{MLE}} = \text{argmax}_{\theta} \ \ell(\theta; \mathbf{X})
$$

---
Estimacin puntual de los Parmetros de una Distribucin Normal

---

Consideremos el caso de estimar los parmetros de una **distribucin normal** (gaussiana) $\mathcal{N}(\mu, \sigma^2)$, donde los parmetros a estimar son la media $\mu$ y la varianza $\sigma^2$ tal que $\theta = (\mu, \sigma^2)$.

Para empezar, ya que el modelo que queremos deducir es uno de forma normal, es decir, deseamos ajustar un modelo gaussiano a los datos observados, la funcin de densidad  a usar es la funcin de probabilidad para una distribucin normal:

$$
f(x; \mu, \sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)
$$

Con esto en mente, asumiendo que **cada observacin en los datos es independiente una de la otra, y que cada una de stas ha sido generada a partir del modelo gaussiano**, la funcin de verosimilitud adquiere la forma:

$$
L(\mu, \sigma^2; \mathbf{X}) = \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x_i - \mu)^2}{2\sigma^2}\right)
$$

Simplificando, obtenemos:

$$
L(\mu, \sigma^2; \mathbf{X}) = \left(\frac{1}{\sqrt{2\pi\sigma^2}}\right)^n \exp\left(-\frac{1}{2\sigma^2} \sum_{i=1}^{n} (x_i - \mu)^2\right)
$$

Tomando el logaritmo de la funcin de verosimilitud eliminamos la productoria y la convertimos en una sumatoria gracias a las propiedades del logaritmo:

$$
\ell(\mu, \sigma^2; \mathbf{X}) = \log L(\mu, \sigma^2; \mathbf{X}) = -\frac{n}{2} \log(2\pi\sigma^2) - \frac{1}{2\sigma^2} \sum_{i=1}^{n} (x_i - \mu)^2
$$

Para encontrar los valores de $\mu$ y $\sigma^2$ que maximizan $ \ell(\mu, \sigma^2; \mathbf{X}) $, tomamos las derivadas parciales con respecto a $\mu$ y $\sigma^2$ y las igualamos a cero.

1. **Derivada con respecto a $ \mu $:**

$$
\frac{\partial \ell}{\partial \mu} = \frac{1}{\sigma^2} \sum_{i=1}^{n} (x_i - \mu) = 0
$$

Esto implica que:

$$
\sum_{i=1}^{n} (x_i - \mu) = 0
$$

$$
\hat{\mu}_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^{n} x_i
$$

2. **Derivada con respecto a $\sigma^2$:**

$$
\frac{\partial \ell}{\partial \sigma^2} = -\frac{n}{2\sigma^2} + \frac{1}{2\sigma^4} \sum_{i=1}^{n} (x_i - \mu)^2 = 0
$$

Multiplicando ambos lados por $2\sigma^4$ y simplificando:

$$
\hat{\sigma}^2_{\text{MLE}} = \frac{1}{n} \sum_{i=1}^{n} (x_i - \hat{\mu}_{\text{MLE}})^2
$$


```python
# Generamos datos simulados de una distribucin normal con media 5 y desviacin estndar 2
np.random.seed(0)
data = np.random.normal(5, 2, size=100)

# Estimacin MLE de los parmetros
mu_mle = np.mean(data)
sigma_mle = np.sqrt(np.mean((data - mu_mle)**2))

print(f"MLE estimador de la media (mu): {mu_mle}")
print(f"MLE estimador de la desviacin estndar (sigma): {sigma_mle}")

# Verificamos los resultados con la funcin de scipy
mu_scipy, sigma_scipy = stats.norm.fit(data)
print(f"Verificacin con scipy: mu = {mu_scipy}, sigma = {sigma_scipy}")
```

Estimacin de una Regresin Lineal con MLE

---

El MLE es una tcnica probabilstica que si bien asume la completa independencia de los datos, es lo suficientemente poderosa como para constituir un estimdor puntual de parmetros para una enorme variedad de modelos. En la presente subseccin, ilustraremos por que el problema canonico dentro del Machine Learning, la regresin lineal, no solo est sustentado por las herramientas del clculo diferencial sino que a su vez, por la teora de la probabilidad y la estadstica.

En este sentido, la leccin ms valiosa que podrmos ver en ste mdulo (a parte del uso de pruebas de hiptesis) es el **origen de la funcin de error para los problemas de regresion dentro del machine learning**. Si bien esta funcin de error u objetivo proviene en un inicio del problema de regresin lineal, su definicin cuadrtica permte ser aplicada a mltiples problemas y modelos dentro del aprendizaje automtico. 

La formulacin de la regresin lineal desde la probabilidad es la siguiente: suponiendo que tenemos un conjunto de datos $X = \{x_i,y_i\}_i^n$ donde $n$ es el nmero de datos, $x_i$ es el valor de la variable independiente para el dato $i$ (conocida) y $y_i$ el valor de la variable independiente del dato $i$ tal que el conjunto de datos $X$ en su colectivo siguen una tendencia lineal, podemos suponer que nuestra hiptesis es que dichos datos obedecen a un modelo lineal de la forma:


El **Maximum Likelihood Estimation (MLE)** es una tcnica fundamental en estadstica que se puede aplicar para estimar los parmetros de una regresin lineal. La regresin lineal es un modelo ampliamente utilizado para describir la relacin entre una variable dependiente \( y \) y una o ms variables independientes \( x_1, x_2, \dots, x_p \).

### 1. **Modelo de Regresin Lineal**

El modelo de regresin lineal simple, que involucra una sola variable independiente \( x \), se expresa como:

$$
y_i = \beta_0 + \beta_1 x_i + \epsilon_i
$$

Donde $b$ y $w$ son respectivamente los parmetros del modelo a estimar, siendo $b$ el intercepto con el eje y y $w$ la inclinacin de la recta, nuestra hiptesis a dems se basa en el hecho de que el valor real para la variable dependiente $y_i$ corresponde a una combinacin lineal de los parmetros anteriormente nombrados y la variable  de entrada $x_i$ ms un trmino de error (o de desviacin de la perfecta linealidad) $\epsilon_i$. 

Si se asume que los errores o desviaciones $\epsilon_i$ son **independientes** e **idnticamente distribuidos** segn una distribucin normal con media cero y varianza $\sigma^2$, es decir:

$$
\epsilon_i \sim \mathcal{N}(0, \sigma^2)
$$
$$
P(\epsilon) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{\epsilon^2}{2 \sigma^2}\right)
$$

La imgen a continuacin, ilustra la formulacin probabilstica de la regresin lineal, donde se asume que la desviacin del $y$ verdadero, se debe a causa de que este punto puede ser reproducido si se hace un muestreo aleatorio a partir de la distribucin centrada en el modelo lineal. Note entonces que el modelo lineal es la **tendencia promedio de los datos**.

image.png

Con lo anterior en mente, si la distancia entre la estimacin del modelo lineal para la entrada $x_i$ , $\hat{y}_i$, y el verdadero valor en los datos para la variable $y$ en el punto $i$, $y_i$, es $d_i = \hat{y}_i - y_i$, podmos traducir la probabilidad de producir el valor $y_i$ al rededor de la tendencia lineal central (donde las gaussianas estn centradas) como la probabilidad de obtener el valor $\epsilon_i = d_i$ centrado en nuestro modelo lineal. As pues, la probabilidad de obtener la $i$-sima distancia entre los valores reales y las estimaciones es simplemente


$$
P(y_i|x_i,b,w) = P(d_i) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{d_i^2}{2 \sigma^2}\right)
$$

Donde en trminos del modelo, se tiene:

$$
d_i = y_i - (x_i w + b)
$$

Si ahora consideramos todos los datos posibles, asumiendo la independencia de las observaciones, construimos la funcin de verosimilitud como 

$$
L(b,w;X) = P(Y|X,b,w) = \prod_{i=1}^{n} P(y_i|x_i,b,w)
$$

Como $y_i sigue una distribucin normal, la funcin de densidad para cada \( y_i \) es:

$$
L(b, w, \sigma^2; \mathbf{y}) = \prod_{i=1}^{n} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(y_i - b - w x_i)^2}{2\sigma^2}\right)
$$

Para simplificar el proceso de maximizacin, tomamos el logaritmo de la funcin de verosimilitud para obtener la **log-verosimilitud**:

$$
\ell(b, w, \sigma^2; \mathbf{y}) = \log L(b, w, \sigma^2; \mathbf{y}) = -\frac{n}{2} \log(2\pi\sigma^2) - \frac{1}{2\sigma^2} \sum_{i=1}^{n} (y_i - b - w x_i)^2
$$


Ya que el objetivo es encontrar los valores de $b$, $w$ y $\sigma^2$  que maximicen $\ell(b, w, \sigma^2; \mathbf{y})$. Este problema se puede resolver tomando derivadas parciales de la log-verosimilitud con respecto a cada parmetro y luego igualando a cero para resolver el sistema de ecuaciones.

Maximicemos primero con respecto a $b$ y $w$. 

$$
\frac{\partial \ell}{\partial b} = \frac{1}{\sigma^2} \sum_{i=1}^{n} (y_i - b - w x_i) = 0
$$

$$
\frac{\partial \ell}{\partial w} = \frac{1}{\sigma^2} \sum_{i=1}^{n} x_i (y_i - b - w x_i) = 0
$$

Estas dos ecuaciones son conocidas como las **ecuaciones normales** para la regresin lineal, donde se asume un valor $\sigma^2 = 1$:

$$
\sum_{i=1}^{n} (y_i - \beta_0 - \beta_1 x_i) = 0
$$

$$
\sum_{i=1}^{n} x_i (y_i - \beta_0 - \beta_1 x_i) = 0
$$

Al resolver estas ecuaciones para $b$ y $w$, obtenemos las frmulas conocidas para los estimadores de mnimos cuadrados ordinarios (OLS):

$$
\hat{w} = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}
$$

$$
\hat{b}_0 = \bar{y} - \hat{\beta}_1 \bar{x}
$$

Donde:
- $\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i $ es la media de $ x $,
- $\bar{y} = \frac{1}{n} \sum_{i=1}^{n} y_i $ es la media de $ y $.

En caso de que no se asuma un sigma igual a uno, para hallarlo basta simplemente con maximizar la funcin de log-verosimilitud con respecto a ste parmetro.

$$
\frac{\partial \ell}{\partial \sigma^2} = -\frac{n}{2\sigma^2} + \frac{1}{2\sigma^4} \sum_{i=1}^{n} (y_i - b - w x_i)^2 = 0
$$

Resolviendo para $ \sigma^2 $:

$$
\hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{\beta}_0 - \hat{\beta}_1 x_i)^2
$$

Lo cul se puede rescribir simplemente como:

$$
\hat{\sigma}^2 = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

Este estimador $\hat{\sigma}^2 $ es la **varianza residual** o **error cuadrtico medio** la cul es la funcin objetivo de la regresin lineal.

No obstante, hemos podido determinar la solucin cerrada para los parmetros $b$ y  $w$ debido a la simplicidad del problema, esto quiere decir que no siempre ser as, incluso en la regresin lineal mltiple. A pesar de ello, note que tanto la forma del MSE $\hat{\sigma}^2 $ como el segndo trmino de la log-verosimulitud, indican que la verosimilitud es maximizada en la medida que la distancia cuadrtica de los valores reales y los valores estimados para $y$ se reduzca.

El MLE aplicado a la regresin lineal simple produce los mismos estimadores que los obtenidos a travs de los mnimos cuadrados ordinarios (OLS) como se puede ver en cualquier curso de clculo diferencial, as, tenemos que los tres parmetros se definen como:

- **$\hat{b}$** y **$ \hat{w} $** son los estimadores de los coeficientes de regresin.
- **$\hat{\sigma}^2$** es el estimador de la varianza del trmino de error.


```python
# Regresin lineal simple por OLS
import statsmodels.api as sm

# Generamos datos simulados
np.random.seed(0)
n = 100
X = np.random.rand(n)
# La relacin es lineal con algo de ruido
Y = 3 + 2 * X + np.random.randn(n)

# Aadimos un intercepto a X
X = sm.add_constant(X)

# Usamos el modelo OLS de statsmodels para obtener el MLE
model = sm.OLS(Y, X)
results = model.fit()

# Mostramos los resultados
print(results.summary())

# Valores MLE de los parmetros
print(f"Estimacin de la constante (beta_0): {results.params[0]}")
print(f"Estimacin del coeficiente (beta_1): {results.params[1]}")
```

Para determinar si los parmetros hallados en una regresin lineal son estadsticamente significativos, se utilizan **valores p** asociados a cada coeficiente del modelo. Estos valores p se calculan a partir de la distribucin de **Student's t** y nos permiten hacer inferencias sobre si los coeficientes de la regresin son significativamente diferentes de cero. Para calcular el valor p, primero se determina el error estndar del parmetro hallado como:

$$
\text{SE}(\hat{\beta}_j) = \sqrt{\hat{\sigma}^2 \cdot (\mathbf{X}^T \mathbf{X})^{-1}_{jj}}
$$

Donde $(\mathbf{X}^T \mathbf{X})^{-1}_{jj}$ es el $j$-simo elemento diagonal de la matriz inversa del producto matricial $\mathbf{X}^T \mathbf{X}$.

Para calcular el estadstico $t$ para cada parmetro se determina la siguiente proporcin:

$$
t_j = \frac{\hat{\beta}_j}{\text{SE}(\hat{\beta}_j)}
$$

Este estadstico t sigue una distribucin t de Student con $n-k-1$ grados de libertad bajo la hiptesis nula $H_0 : \beta_j = 0$. Si el valor p de un coeficiente es menor que el nivel de significancia $\alpha$, comnmente 0.05, rechazamos la hiptesis nula de que el coeficiente es igual a cero. Esto indica que **la variable asociada a ese coeficiente tiene un efecto significativo sobre la variable dependiente.** Si el valor p es mayor que $\alpha$, no podemos rechazar la hiptesis nula, lo que sugiere que la variable podra no tener un efecto significativo en el modelo.


```python
# Regresin lineal multivariable
import statsmodels.api as sm

# Generar datos de ejemplo
np.random.seed(0)
X = np.random.rand(100, 2)  # Dos variables explicativas
X = sm.add_constant(X)  # Aadir intercepto
beta = np.array([2, 3, 5])  # Coeficientes reales
y = X @ beta + np.random.normal(scale=1, size=100)  # Variable dependiente

# Ajustar el modelo de regresin
model = sm.OLS(y, X).fit()

# Mostrar los resultados
print(model.summary())
```

**<h3 id="subsection53" style="color: #003366;">5.3. Regresin Logstica</h3>**

---
Estimacin de una Regresin Logstica con MLE

---

La **regresin logstica** es una tcnica utilizada para **modelar la probabilidad de un evento binario (tipo Bernoulli) en funcin de una o ms variables explicativas**. La regresin logstica se utiliza para predecir la probabilidad de que una observacin pertenezca a una de dos categoras (por ejemplo, xito o fracaso, 1 o 0). A diferencia de la regresin lineal, que predice valores continuos, la regresin logstica predice probabilidades que estn restringidas al intervalo [0, 1], lo que hace que ste modelo  sea considerado como un **modelo de clasificacin**.

La regresin logstica se basa en la funcin **sigmoide** o **logstica**, que transforma una combinacin lineal de variables independientes en una probabilidad:

$$
P(Y = 1 \mid \mathbf{X}) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x_1 + \beta_2 x_2 + \cdots + \beta_k x_k)}}
$$

Lo cul es equivalente a:

$$
P(Y = 1 \mid \mathbf{X}) = \frac{1}{1 + \exp(-\mathbf{X}^\top \boldsymbol{\beta})} 
$$

donde $P(Y = 1 \mid \mathbf{X})$ es la probabilidad de que la variable dependiente $Y$ sea 1 dado el vector de caractersticas $\mathbf{X}$.  Por otra parte $\beta_0$ es el intercepto o bias,$\beta_1, \beta_2, \ldots, \beta_k$ son los coeficientes asociados a las variables explicativas $x_1, x_2, \ldots, x_k$. As pues, nuestro problema se puede entender como:

$$
Y \mid \mathbf{X} \sim \text{Bernoulli}(p)
$$

Si consideramos las observaciones o datos, como independientes entre s, la funcin de verosimilitud de  una sola observacin es:

$$
\text{L}(\boldsymbol{\beta}) = p^{y} (1 - p)^{1 - y}
$$

Donde $p = \frac{1}{1 + \exp(-\mathbf{X}^\top \boldsymbol{\beta})}$ es la probabilidad de xito (evento positivo). Por ende, la funcin de verosimilitud para todo el conjunto de datos es:

$$
\text{L}(\boldsymbol{\beta}) = \prod_{i=1}^{n} p_i^{y_i} (1 - p_i)^{1 - y_i} 
$$

Lo cul es equivalente a:

$$
L(\beta) = \prod_{i=1}^{n} \left[ P(Y_i = 1 \mid \mathbf{X}_i)^{y_i} \times \left(1 - P(Y_i = 1 \mid \mathbf{X}_i)\right)^{1 - y_i} \right]
$$

Para simplificar el clculo, se utiliza la log-verosimilitud, que es el logaritmo de la funcin de verosimilitud:

$$
\ell(\beta) = \log L(\beta) = \sum_{i=1}^{n} \left[y_i \log P(Y_i = 1 \mid \mathbf{X}_i) + (1 - y_i) \log \left(1 - P(Y_i = 1 \mid \mathbf{X}_i)\right) \right]
$$

Para encontrar los parmetros $\beta$ que maximizan la log-verosimilitud, se calculan las derivadas parciales de $\ell(\beta)$ con respecto a cada $\beta_j$ y se resuelven las ecuaciones resultantes:

$$
\frac{\partial \ell(\beta)}{\partial \beta_j} = \sum_{i=1}^{n} \left[y_i - P(Y_i = 1 \mid \mathbf{X}_i)\right] x_{ij} = 0
$$

siendo $x_{ij}$ es el valor de la variable $j$ para la observacin $i$.

Note entonces que es gracias al principio de mxima verosimilitud que hemos hallado la **funcin de costo de un problema de clasificacin binaria**  una vez consideramos, que maximizar la verosimilitud es equivalente a minimizar su negativo. En este sentido, la funcin de error de un problema de clasificacin binaria es denominada como **Entropa Cruzada** y viene dad por:

$$
E(\beta) = - \ell(\beta)
$$


```python
# Regresin logstica multivariable
import statsmodels.api as sm

# Generar datos de ejemplo
np.random.seed(0)
X = np.random.rand(100, 2)  # Dos caractersticas
X = sm.add_constant(X)  # Aadir intercepto
beta = np.array([1, 2, 3])  # Coeficientes reales
logits = X @ beta + np.random.normal(scale=1, size=100)  # Valores logit
p = 1 / (1 + np.exp(-logits))  # Probabilidades
y = np.random.binomial(1, p)  # Variable dependiente binaria

# Ajustar el modelo de regresin logstica
model = sm.Logit(y, X).fit()

# Mostrar los resultados
print(model.summary())
```
