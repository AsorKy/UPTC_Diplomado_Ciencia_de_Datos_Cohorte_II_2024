# Distribuciones Discretas

Dado que nuestro experimento aleatorio implica una clasificacion binaria (ecologica o no ecologica), el experimento aleatorio corresponde a un experimento tipo Bernoulli. Por otra parte, ya que realizamos un sampleo de 15 muestras independientes entre si mediante el msmo experimento, necesitamos usar una distribucin Binomial en donde  N=15 (total de muestras), k=10 (nmero de casos exitosos) y p=0.79, siendo esta la probabilidad de que se clasifique a un hogar como ecolgico.

````python
from scipy.stats import binom

# Parmetros
n = 15  # nmero de hogares
p = 0.79  # probabilidad de que un hogar sea ecolgicamente sostenible

# Calculamos la probabilidad de que un mximo de 10 hogares sean sostenibles
k = 10
prob = binom.cdf(k, n, p)
prob
````

La distribucin geomtrica y la distribucin binomial negativa (tambin llamada distribucin geomtrica negativa) estn relacionadas, pero modelan diferentes escenarios de ensayos aleatorios.

### Distribucin Geomtrica

**Definicin:**
La distribucin geomtrica modela el nmero de ensayos necesarios hasta obtener el primer xito en una serie de ensayos de Bernoulli independientes (ensayos con dos posibles resultados: xito o fracaso) con una probabilidad de xito constante \( p \).

**Caractersticas:**
- Variable aleatoria: \( X \) (nmero de ensayos hasta el primer xito).
- La probabilidad de que el primer xito ocurra en el \( k \)-simo ensayo es:

\[ P(X = k) = (1 - p)^{k-1} \cdot p \]

**Esperanza (valor esperado):**

\[ E(X) = \frac{1}{p} \]

### Distribucin Binomial Negativa

**Definicin:**
La distribucin binomial negativa generaliza la distribucin geomtrica y modela el nmero de ensayos necesarios para obtener un nmero fijo \( r \) de xitos. Es decir, se usa cuando queremos saber cuntos ensayos se necesitan para obtener \( r \) xitos, en vez de solo el primer xito.

**Caractersticas:**
- Variable aleatoria: \( Y \) (nmero de ensayos hasta obtener \( r \) xitos).
- La probabilidad de que el \( r \)-simo xito ocurra en el \( k \)-simo ensayo es:

\[ P(Y = k) = \binom{k-1}{r-1} p^r (1 - p)^{k-r} \]

donde \( \binom{k-1}{r-1} \) es un coeficiente binomial que cuenta el nmero de maneras de distribuir \( k-1 \) ensayos entre \( r-1 \) xitos y \( k-r \) fracasos.

**Esperanza (valor esperado):**

\[ E(Y) = \frac{r}{p} \]

### Relacin entre ambas distribuciones:

- **Distribucin Geomtrica** es un caso especial de la **Distribucin Binomial Negativa** donde \( r = 1 \). Es decir, cuando se busca el nmero de ensayos hasta el primer xito, la distribucin binomial negativa se reduce a la distribucin geomtrica.
- Ambas distribuciones asumen que los ensayos son independientes y que la probabilidad de xito \( p \) es constante en cada ensayo.

Para resolver este problema, usaremos la distribucin binomial negativa, que se utiliza para modelar el nmero de ensayos necesarios para obtener un nmero fijo de xitos en una serie de ensayos de Bernoulli independientes.

En este caso, sabemos que:
- \( r = 3 \) (queremos encontrar el tercer hogar clasificado como ecolgicamente sostenible).
- De los primeros cinco hogares seleccionados, dos son ecolgicamente sostenibles.
- Queremos encontrar la probabilidad de que el noveno hogar revisado sea el tercero clasificado como ecolgicamente sostenible.

Este problema puede ser abordado usando la funcin de probabilidad de la distribucin binomial negativa condicional.

La frmula para la probabilidad de que el \( k \)-simo ensayo sea el \( r \)-simo xito es:

\[ P(Y = k) = \binom{k-1}{r-1} p^r (1 - p)^{k-r} \]

Donde:
- \( Y \) es el nmero de ensayos hasta el \( r \)-simo xito.
- \( \binom{k-1}{r-1} \) es un coeficiente binomial.
- \( p \) es la probabilidad de xito.
- \( k \) es el nmero de ensayos.

En este caso:
- \( p = 0.79 \)
- \( r = 3 \)
- \( k = 9 \) (noveno hogar)

Dado que ya tenemos dos xitos en los primeros cinco ensayos, necesitamos un xito ms en los siguientes cuatro ensayos. Por lo tanto, debemos encontrar la probabilidad de que el tercer xito ocurra en el noveno ensayo, dado que ya hemos tenido dos xitos en los primeros cinco ensayos.

Vamos a calcular esto en Python:

```python
import math

# Parmetros
p = 0.79  # Probabilidad de xito
r = 3     # Nmero de xitos deseados
k = 9     # Nmero de ensayos

# Funcin para calcular el coeficiente binomial
def binomial_coefficient(n, k):
    return math.comb(n, k)

# Calcular la probabilidad usando la frmula de la distribucin binomial negativa
probabilidad = binomial_coefficient(k-1, r-1) * (p**r) * ((1 - p)**(k-r))
print(f"La probabilidad de que el noveno hogar revisado sea el tercero clasificado como ecolgicamente sostenible es: {probabilidad:.4f}")
```

Este cdigo calcula y muestra la probabilidad de que el noveno hogar revisado sea el tercero clasificado como ecolgicamente sostenible, utilizando la frmula de la distribucin binomial negativa.

````python
import math

# Parmetros
p = 0.79  # Probabilidad de xito
r = 3 #2     # Nmero de xitos deseados
k = 9 #3     # Nmero de ensayos

# Funcin para calcular el coeficiente binomial
def binomial_coefficient(n, k):
    return math.comb(n, k)

# Calcular la probabilidad usando la frmula de la distribucin binomial negativa
probabilidad = binomial_coefficient(k-1, r-1) * (p**r) * ((1 - p)**(k-r))
print(f"La probabilidad de que el noveno hogar revisado sea el tercero clasificado como ecolgicamente sostenible es: {probabilidad:.4f}")
````

No obstante este resultado es INCORRECTO, ya que tenemos una informacin extra que corresponde a conocer que de los cinco primeros hogares dos son ecologicamente sostenibles.

Para resolver este problema, vamos a usar la distribucin binomial negativa. En este caso, queremos calcular la probabilidad de que el noveno hogar revisado sea el tercero clasificado como ecolgicamente sostenible, dado que ya se han encontrado dos hogares ecolgicamente sostenibles en los primeros cinco hogares revisados.

### Definicin del Problema

- \( p \): Probabilidad de que un hogar sea ecolgicamente sostenible = 0.79.
- \( q \): Probabilidad de que un hogar no sea ecolgicamente sostenible = 1 - p = 0.21.
- \( k \): Nmero de xitos requeridos (en este caso, el tercer hogar ecolgicamente sostenible).
- \( n \): Posicin en la que ocurre el \( k \)-simo xito (en este caso, el noveno hogar).

Queremos encontrar la probabilidad de que el noveno hogar sea el tercer hogar ecolgicamente sostenible. Para esto, usamos la frmula de la distribucin binomial negativa:

\[ P(X = n) = \binom{n-1}{k-1} \cdot p^k \cdot q^{n-k} \]

Donde:
- \(\binom{n-1}{k-1}\) es el coeficiente binomial que representa las combinaciones de \(n-1\) lugares para colocar los \(k-1\) xitos.
- \(p^k\) es la probabilidad de que haya exactamente \(k\) xitos.
- \(q^{n-k}\) es la probabilidad de que haya \(n-k\) fracasos.

### Aplicacin de la Frmula

En este caso, \(k = 3\) (tercer hogar ecolgicamente sostenible) y \(n = 9\) (noveno hogar revisado).

\[ P(X = 9) = \binom{9-1}{3-1} \cdot p^3 \cdot q^{9-3} \]

\[ P(X = 9) = \binom{8}{2} \cdot (0.79)^3 \cdot (0.21)^6 \]

Calculamos los componentes:

1. Coeficiente binomial:
\[ \binom{8}{2} = \frac{8!}{2!(8-2)!} = \frac{8 \cdot 7}{2 \cdot 1} = 28 \]

2. Probabilidad de xito:
\[ (0.79)^3 \approx 0.493 \]

3. Probabilidad de fracaso:
\[ (0.21)^6 \approx 0.000340 \]

Multiplicamos estos valores:

\[ P(X = 9) = 28 \cdot 0.493 \cdot 0.000340 \approx 0.0047 \]

Por lo tanto, la probabilidad de que el noveno hogar revisado sea el tercero clasificado como ecolgicamente sostenible es aproximadamente \(0.0047\), o \(0.47\%\).

````python
from math import comb

# Datos del problema
p = 0.79  # Probabilidad de xito (ecolgicamente sostenible)
q = 1 - p  # Probabilidad de fracaso
k = 3  # Nmero de xitos requeridos
n = 9  # Posicin en la que ocurre el k-simo xito

# Clculo de la probabilidad usando la frmula de la distribucin binomial negativa
probabilidad = comb(n-1, k-1) * (p**k) * (q**(n-k))

print(f"La probabilidad de que el noveno hogar revisado sea el tercero clasificado como ecolgicamente sostenible es aproximadamente {probabilidad:.4f}")
````

# Distribucin de Poisson

````python
import numpy as np
import matplotlib.pyplot as plt

def generar_distribucion_poisson(lambda_param, t, num_samples):
    # Parmetro de la distribucin de Poisson ajustado por el tiempo
    lambda_t = lambda_param * t
    
    # Generar muestras de la distribucin de Poisson
    samples = np.random.poisson(lambda_t, num_samples)
    
    # Visualizar la distribucin
    plt.hist(samples, bins=range(min(samples), max(samples) + 1), alpha=0.75, edgecolor='black')
    plt.title('Distribucin de Poisson')
    plt.xlabel('Nmero de eventos')
    plt.ylabel('Frecuencia')
    plt.show()

# Parmetros de ejemplo
lambda_param = 0.75  # tasa media de ocurrencias por unidad de tiempo
t = 5  # intervalo de tiempo
num_samples = 1000  # nmero de muestras

# Generar y visualizar la distribucin de Poisson
generar_distribucion_poisson(lambda_param, t, num_samples)
````
