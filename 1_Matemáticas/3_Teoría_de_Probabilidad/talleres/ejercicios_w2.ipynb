# Distribuciones Discretas

Dado que nuestro experimento aleatorio implica una clasificacion binaria (ecologica o no ecologica), el experimento aleatorio corresponde a un experimento tipo Bernoulli. Por otra parte, ya que realizamos un sampleo de 15 muestras independientes entre si mediante el mísmo experimento, necesitamos usar una distribución Binomial en donde  N=15 (total de muestras), k=10 (número de casos exitosos) y p=0.79, siendo esta la probabilidad de que se clasifique a un hogar como ecológico.

````python
from scipy.stats import binom

# Parámetros
n = 15  # número de hogares
p = 0.79  # probabilidad de que un hogar sea ecológicamente sostenible

# Calculamos la probabilidad de que un máximo de 10 hogares sean sostenibles
k = 10
prob = binom.cdf(k, n, p)
prob
````

La distribución geométrica y la distribución binomial negativa (también llamada distribución geométrica negativa) están relacionadas, pero modelan diferentes escenarios de ensayos aleatorios.

### Distribución Geométrica

**Definición:**
La distribución geométrica modela el número de ensayos necesarios hasta obtener el primer éxito en una serie de ensayos de Bernoulli independientes (ensayos con dos posibles resultados: éxito o fracaso) con una probabilidad de éxito constante \( p \).

**Características:**
- Variable aleatoria: \( X \) (número de ensayos hasta el primer éxito).
- La probabilidad de que el primer éxito ocurra en el \( k \)-ésimo ensayo es:

\[ P(X = k) = (1 - p)^{k-1} \cdot p \]

**Esperanza (valor esperado):**

\[ E(X) = \frac{1}{p} \]

### Distribución Binomial Negativa

**Definición:**
La distribución binomial negativa generaliza la distribución geométrica y modela el número de ensayos necesarios para obtener un número fijo \( r \) de éxitos. Es decir, se usa cuando queremos saber cuántos ensayos se necesitan para obtener \( r \) éxitos, en vez de solo el primer éxito.

**Características:**
- Variable aleatoria: \( Y \) (número de ensayos hasta obtener \( r \) éxitos).
- La probabilidad de que el \( r \)-ésimo éxito ocurra en el \( k \)-ésimo ensayo es:

\[ P(Y = k) = \binom{k-1}{r-1} p^r (1 - p)^{k-r} \]

donde \( \binom{k-1}{r-1} \) es un coeficiente binomial que cuenta el número de maneras de distribuir \( k-1 \) ensayos entre \( r-1 \) éxitos y \( k-r \) fracasos.

**Esperanza (valor esperado):**

\[ E(Y) = \frac{r}{p} \]

### Relación entre ambas distribuciones:

- **Distribución Geométrica** es un caso especial de la **Distribución Binomial Negativa** donde \( r = 1 \). Es decir, cuando se busca el número de ensayos hasta el primer éxito, la distribución binomial negativa se reduce a la distribución geométrica.
- Ambas distribuciones asumen que los ensayos son independientes y que la probabilidad de éxito \( p \) es constante en cada ensayo.

Para resolver este problema, usaremos la distribución binomial negativa, que se utiliza para modelar el número de ensayos necesarios para obtener un número fijo de éxitos en una serie de ensayos de Bernoulli independientes.

En este caso, sabemos que:
- \( r = 3 \) (queremos encontrar el tercer hogar clasificado como ecológicamente sostenible).
- De los primeros cinco hogares seleccionados, dos son ecológicamente sostenibles.
- Queremos encontrar la probabilidad de que el noveno hogar revisado sea el tercero clasificado como ecológicamente sostenible.

Este problema puede ser abordado usando la función de probabilidad de la distribución binomial negativa condicional.

La fórmula para la probabilidad de que el \( k \)-ésimo ensayo sea el \( r \)-ésimo éxito es:

\[ P(Y = k) = \binom{k-1}{r-1} p^r (1 - p)^{k-r} \]

Donde:
- \( Y \) es el número de ensayos hasta el \( r \)-ésimo éxito.
- \( \binom{k-1}{r-1} \) es un coeficiente binomial.
- \( p \) es la probabilidad de éxito.
- \( k \) es el número de ensayos.

En este caso:
- \( p = 0.79 \)
- \( r = 3 \)
- \( k = 9 \) (noveno hogar)

Dado que ya tenemos dos éxitos en los primeros cinco ensayos, necesitamos un éxito más en los siguientes cuatro ensayos. Por lo tanto, debemos encontrar la probabilidad de que el tercer éxito ocurra en el noveno ensayo, dado que ya hemos tenido dos éxitos en los primeros cinco ensayos.

Vamos a calcular esto en Python:

```python
import math

# Parámetros
p = 0.79  # Probabilidad de éxito
r = 3     # Número de éxitos deseados
k = 9     # Número de ensayos

# Función para calcular el coeficiente binomial
def binomial_coefficient(n, k):
    return math.comb(n, k)

# Calcular la probabilidad usando la fórmula de la distribución binomial negativa
probabilidad = binomial_coefficient(k-1, r-1) * (p**r) * ((1 - p)**(k-r))
print(f"La probabilidad de que el noveno hogar revisado sea el tercero clasificado como ecológicamente sostenible es: {probabilidad:.4f}")
```

Este código calcula y muestra la probabilidad de que el noveno hogar revisado sea el tercero clasificado como ecológicamente sostenible, utilizando la fórmula de la distribución binomial negativa.

````python
import math

# Parámetros
p = 0.79  # Probabilidad de éxito
r = 3 #2     # Número de éxitos deseados
k = 9 #3     # Número de ensayos

# Función para calcular el coeficiente binomial
def binomial_coefficient(n, k):
    return math.comb(n, k)

# Calcular la probabilidad usando la fórmula de la distribución binomial negativa
probabilidad = binomial_coefficient(k-1, r-1) * (p**r) * ((1 - p)**(k-r))
print(f"La probabilidad de que el noveno hogar revisado sea el tercero clasificado como ecológicamente sostenible es: {probabilidad:.4f}")
````

No obstante este resultado es INCORRECTO, ya que tenemos una información extra que corresponde a conocer que de los cinco primeros hogares dos son ecologicamente sostenibles.

Para resolver este problema, vamos a usar la distribución binomial negativa. En este caso, queremos calcular la probabilidad de que el noveno hogar revisado sea el tercero clasificado como ecológicamente sostenible, dado que ya se han encontrado dos hogares ecológicamente sostenibles en los primeros cinco hogares revisados.

### Definición del Problema

- \( p \): Probabilidad de que un hogar sea ecológicamente sostenible = 0.79.
- \( q \): Probabilidad de que un hogar no sea ecológicamente sostenible = 1 - p = 0.21.
- \( k \): Número de éxitos requeridos (en este caso, el tercer hogar ecológicamente sostenible).
- \( n \): Posición en la que ocurre el \( k \)-ésimo éxito (en este caso, el noveno hogar).

Queremos encontrar la probabilidad de que el noveno hogar sea el tercer hogar ecológicamente sostenible. Para esto, usamos la fórmula de la distribución binomial negativa:

\[ P(X = n) = \binom{n-1}{k-1} \cdot p^k \cdot q^{n-k} \]

Donde:
- \(\binom{n-1}{k-1}\) es el coeficiente binomial que representa las combinaciones de \(n-1\) lugares para colocar los \(k-1\) éxitos.
- \(p^k\) es la probabilidad de que haya exactamente \(k\) éxitos.
- \(q^{n-k}\) es la probabilidad de que haya \(n-k\) fracasos.

### Aplicación de la Fórmula

En este caso, \(k = 3\) (tercer hogar ecológicamente sostenible) y \(n = 9\) (noveno hogar revisado).

\[ P(X = 9) = \binom{9-1}{3-1} \cdot p^3 \cdot q^{9-3} \]

\[ P(X = 9) = \binom{8}{2} \cdot (0.79)^3 \cdot (0.21)^6 \]

Calculamos los componentes:

1. Coeficiente binomial:
\[ \binom{8}{2} = \frac{8!}{2!(8-2)!} = \frac{8 \cdot 7}{2 \cdot 1} = 28 \]

2. Probabilidad de éxito:
\[ (0.79)^3 \approx 0.493 \]

3. Probabilidad de fracaso:
\[ (0.21)^6 \approx 0.000340 \]

Multiplicamos estos valores:

\[ P(X = 9) = 28 \cdot 0.493 \cdot 0.000340 \approx 0.0047 \]

Por lo tanto, la probabilidad de que el noveno hogar revisado sea el tercero clasificado como ecológicamente sostenible es aproximadamente \(0.0047\), o \(0.47\%\).

````python
from math import comb

# Datos del problema
p = 0.79  # Probabilidad de éxito (ecológicamente sostenible)
q = 1 - p  # Probabilidad de fracaso
k = 3  # Número de éxitos requeridos
n = 9  # Posición en la que ocurre el k-ésimo éxito

# Cálculo de la probabilidad usando la fórmula de la distribución binomial negativa
probabilidad = comb(n-1, k-1) * (p**k) * (q**(n-k))

print(f"La probabilidad de que el noveno hogar revisado sea el tercero clasificado como ecológicamente sostenible es aproximadamente {probabilidad:.4f}")
````

# Distribución de Poisson

````python
import numpy as np
import matplotlib.pyplot as plt

def generar_distribucion_poisson(lambda_param, t, num_samples):
    # Parámetro de la distribución de Poisson ajustado por el tiempo
    lambda_t = lambda_param * t
    
    # Generar muestras de la distribución de Poisson
    samples = np.random.poisson(lambda_t, num_samples)
    
    # Visualizar la distribución
    plt.hist(samples, bins=range(min(samples), max(samples) + 1), alpha=0.75, edgecolor='black')
    plt.title('Distribución de Poisson')
    plt.xlabel('Número de eventos')
    plt.ylabel('Frecuencia')
    plt.show()

# Parámetros de ejemplo
lambda_param = 0.75  # tasa media de ocurrencias por unidad de tiempo
t = 5  # intervalo de tiempo
num_samples = 1000  # número de muestras

# Generar y visualizar la distribución de Poisson
generar_distribucion_poisson(lambda_param, t, num_samples)
````
