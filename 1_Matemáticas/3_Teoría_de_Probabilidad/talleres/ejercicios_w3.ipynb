* Punto 1:

Pregunta 1
La distribuci√≥n normal como parte de las distribuciones continuas de mayor aplicaci√≥n, describe el comportamiento de gran cantidad de experimentos aleatorios desarrollados en espacios muestrales grandes, generalmente con un n√∫mero de ensayos superior a 30. Raz√≥n por la cual, por medio de la consideraci√≥n de un caso particular en el cual los resultados siguen una distribuci√≥n normal, se pretende que logres afianzar los conocimientos necesarios a la hora de estandarizar una distribuci√≥n normal y realizar los c√°lculos asociados a las probabilidades de dicho evento.    

EL GRE √≥‚ÄØGraduate‚ÄØRecord‚ÄØExamination‚ÄØes un examen estandarizado requerido en el proceso de admisi√≥n de un gran n√∫mero de postgrados en todo el‚ÄØmundo.‚ÄØ Este examen eval√∫a la‚ÄØcapacidad de escritura y‚ÄØredacci√≥n, el conocimiento en‚ÄØmatem√°ticas b√°sicas (aritm√©tica, √°lgebra, problemas matem√°ticos y geometr√≠a), matem√°ticas avanzadas (estad√≠stica, probabilidad, combinaciones y permutaciones) e interpretaci√≥n y an√°lisis de datos.  

El resultado del test para un estudiante seleccionado al azar es una variable aleatoria que sigue una distribuci√≥n Normal de media igual a 152‚ÄØpuntos‚ÄØy desviaci√≥n est√°ndar igual a 6 puntos. 


¬øQu√© porcentaje‚ÄØde personas‚ÄØobtienen‚ÄØuna‚ÄØcalificaci√≥n inferior o igual‚ÄØa‚ÄØ152 puntos en el GRE?   

```python
import scipy.stats as stats

# Par√°metros de la distribuci√≥n normal
media = 152
sigma = 6

# Calculamos la probabilidad de obtener una calificaci√≥n inferior o igual a 152

p = stats.norm.cdf(152, loc=media, scale=sigma)
porcentaje = p * 100
print('El porcentaje de calificaciones a 152 es =', porcentaje)
```

¬øCu√°l es la probabilidad de que el tiempo entre dos llamadas consecutivas sea menor a 10 minutos?   

```python
import math 

# Parametros de la distribuci√≥n exponencial
media = 3
lambda_ = 1 / media 

# Tiempo para el calculo de p
tiempo = 10

# Calculo de P 
p_cdf = 1 - math.exp(-lambda_ * tiempo)
print(f"La probabilidad de que el tiempo entre dos llamadas consecutivas sea menor a 10 minutos es: {p_cdf:.4f}")
```

Si la √∫ltima llamada que entr√≥ al call-center fue hace 6 minutos, ¬øcu√°l es la probabilidad de que la siguiente llamada entre al call-center despu√©s de 8 minutos?

* Punto 6

¬øCu√°l es la probabilidad de que en los siguientes 5 minutos entren 3 llamadas al call-center?   

```python
import math

# Par√°metro de la distribuci√≥n de Poisson
lambda_poisson = 5 / 3

# N√∫mero de llamadas
k = 3

# C√°lculo de la probabilidad usando la f√≥rmula de Poisson
probabilidad = (math.exp(-lambda_poisson) * (lambda_poisson ** k)) / math.factorial(k)
print(f"La probabilidad de que entren exactamente 3 llamadas en los siguientes 5 minutos es: {probabilidad:.4f}")
```
