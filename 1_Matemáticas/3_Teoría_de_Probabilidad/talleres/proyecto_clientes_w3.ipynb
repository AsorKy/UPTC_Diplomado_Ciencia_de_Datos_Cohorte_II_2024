
# Caso de aplicación – Fuga de clientes

Considerando la importancia actual de las telecomunicaciones en el desarrollo de las actividades humanas, los clientes de las empresas de telecomunicaciones buscan un servicio eficiente que soporte las diferentes actividades que realizan, con llamadas, datos, servicio de internet, mensajes de texto, y servicio al cliente. Al contratar un paquete de telecomunicaciones con una empresa, los clientes esperan obtener una buena relación costo/beneficio, de tal forma que el servicio que recibe el cliente compense el gasto total mensual del mismo. Así las cosas, si los clientes sufren ciertas fallas en el servicio o en la calidad de la atención que reciben, se ven incentivados a cancelar la suscripción y contratar servicios de telecomunicaciones con otra empresa. El archivo de datos Fuga de Clientes, recopila la información de 289 clientes y el uso que cada uno de ellos hace de su plan de telecomunicaciones. Las variables que contiene el archivo de datos se encuentran descritas a continuación:

### Variables 

* ID Cliente: número de identificación del cliente en la red.
* Edad en la red: tiempo que ha pasado, en meses, desde que el cliente ingresó a la red.
* Gasto total al mes: gasto total del cliente en la red, en el mes de agosto de 2017.
* Gasto en SMS por SMS: el gasto total en SMS (Short Message Service) por cliente obtenido a través del servicio de SMS utilizado por el suscriptor.
* Gasto de datos por SMS: gasto total de datos/Internet de un cliente obtenido a través del servicio de SMS utilizado por el suscriptor. 
* Consumo total de datos: cantidad de datos/Internet, en KB, consumida por el cliente durante el mes de agosto de 2017.
* Velocidad promedio de carga: velocidad promedio de carga (tiempo que tarda la información en llegar desde el celular a internet), en megabytes por segundo MB/S.  
* Velocidad promedio de descarga: velocidad promedio de descarga (tiempo que tarda la información en llegar al celular desde internet), en megabytes por segundo MB/S.   
* Llamadas únicas: número total de llamadas únicas realizadas durante el mes de agosto de 2017. 
* Llamadas caídas: número total de llamadas caídas durante el mes de agosto de 2017.
* Duración promedio llamada: duración promedio, en segundos, de las llamadas realizadas durante el mes de agosto de 2017.
* Duración promedio de las llamadas al mismo operador: duración promedio, en segundos, de las llamadas realizadas al mismo operador durante el mes de agosto de 2017. 
* Duración promedio de las llamadas a otros operadores: duración promedio, en segundos, de las llamadas realizadas a operadores diferentes al del cliente durante el mes de agosto de 2017.  
* Gasto llamadas mismo operador: gasto total en llamadas del mismo operador.
* Gasto llamadas otro operador: gasto total del cliente en llamadas realizadas a otros operadores.
* Llamadas por quejas al mes: número de llamadas por quejas realizadas por el cliente durante el mes de agosto de 2017.
* Tipo de red preferida: tipo de red preferida por el cliente.
* Competidor favorito: compañía favorita a la cual se iría el cliente.
* Cancelación de la suscripción: 1, si el cliente se fue de la red (cambió de operador), 0 de lo contrario.

Queremos encontrar el valor de x tal que P(X>x) = 0.352, esto es equivalente a hallar P(X<=x) = 1 -P(X>x). Utilizaremos la función iversa del CDF para encontrar dicho valor.

```python
import scipy.stats as stats
import numpy as np

media = 20
sigma = np.sqrt(0.25 ** 2)

p = 1 - 0.3525
# Calcular el valor de x usando la función inversa de la CDF (PPF)
x = stats.norm.ppf(p, loc=media, scale=sigma)
print(f"El valor de x tal que P(X > x) = 0.3525 es: {x:.4f}")
```

```python
import scipy.stats as stats

# Parámetros de la distribución normal
media = 20
desviacion_estandar = 0.25

# Probabilidad acumulada
probabilidad = 1 - 0.3525

# Calcular el valor de Z correspondiente a la probabilidad acumulada
z = stats.norm.ppf(probabilidad)

# Desestandarizar para encontrar el valor de x
x = z * desviacion_estandar + media
print(f"El valor de x tal que P(X > x) = 0.3525 es: {x:.4f}")
```
