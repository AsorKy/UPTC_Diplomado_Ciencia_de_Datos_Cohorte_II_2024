```python
from math import comb

def probabilidad_binomial(n, k, p):
    """
    Calcula la probabilidad de que ocurran exactamente k √©xitos
    en una distribuci√≥n binomial con n pruebas y probabilidad p de √©xito.
    
    :param n: N√∫mero de pruebas
    :param k: N√∫mero de √©xitos
    :param p: Probabilidad de √©xito
    :return: Probabilidad de que ocurran exactamente k √©xitos
    """
    return comb(n, k) * (p**k) * ((1 - p)**(n - k))

# Datos del problema
n = 12  # N√∫mero de frascos en una caja
p = 0.079  # Probabilidad de que un frasco no sea apto
k_max = 3  # N√∫mero m√°ximo de frascos no aptos

# Calcular la probabilidad de encontrar a lo sumo 3 frascos no aptos
probabilidad_total = sum(probabilidad_binomial(n, k, p) for k in range(k_max + 1))

print(f"La probabilidad de que se encuentren a lo sumo {k_max} frascos no aptos en una caja es aproximadamente {probabilidad_total:.4f}")
```

```python
import math

def probabilidad_poisson(lambda_, k):
    """
    Calcula la probabilidad de que ocurran exactamente k eventos
    en un intervalo, dado un promedio lambda.
    
    :param lambda_: N√∫mero esperado de eventos en el intervalo
    :param k: N√∫mero de eventos observados
    :return: Probabilidad de que ocurran exactamente k eventos
    """
    return (lambda_**k * math.exp(-lambda_)) / math.factorial(k)

# Datos del problema
lambda_dia = 8  # N√∫mero esperado de fallas en un d√≠a
k_max = 5  # N√∫mero m√°ximo de fallas a considerar

# Calcular la probabilidad de tener a lo sumo 5 fallas
probabilidad_total = sum(probabilidad_poisson(lambda_dia, k) for k in range(k_max + 1))

print(f"La probabilidad de que la m√°quina presente a lo sumo {k_max} fallas en un d√≠a es aproximadamente {probabilidad_total:.4f}")
```

```python
import math

def probabilidad_poisson(lambda_, k):
    """
    Calcula la probabilidad de que ocurran exactamente k eventos
    en un intervalo, dado un promedio lambda.
    
    :param lambda_: N√∫mero esperado de eventos en el intervalo
    :param k: N√∫mero de eventos observados
    :return: Probabilidad de que ocurran exactamente k eventos
    """
    return (lambda_**k * math.exp(-lambda_)) / math.factorial(k)

# Datos del problema
lambda_dia = 8  # N√∫mero esperado de fallas en un d√≠a
k_max = 5  # N√∫mero m√°ximo de fallas a considerar

# Calcular la probabilidad de tener a lo sumo 5 fallas
probabilidad_max_5_fallas = sum(probabilidad_poisson(lambda_dia, k) for k in range(k_max + 1))

# Calcular la probabilidad de tener al menos 6 fallas
probabilidad_min_6_fallas = 1 - probabilidad_max_5_fallas

print(f"La probabilidad de que la m√°quina presente al menos 6 fallas en un d√≠a es aproximadamente {probabilidad_min_6_fallas:.4f}")
```
