Variables Aleatorias Discretas

---

Una **variable aleatoria discreta** es una variable que puede tomar un conjunto finito o contable de valores posibles, cada uno con una cierta probabilidad asociada. Estas variables son el resultado de experimentos aleatorios donde los resultados pueden ser contados claramente, como el número de veces que un evento ocurre en una serie de ensayos o el valor que toma una tirada de dado.

Una variable aleatoria discreta se caracteriza por su **función de probabilidad** (también conocida como función de masa de probabilidad, **PMF**), que asigna a cada valor posible de la variable una probabilidad. La importancia y utilidad de las variables aleatorias consíste en que nos permíten modelar enteramente el experimento aleatorio y nos facilita el entender las propiedades del sistema de interés.

Para ilustrar que es una variable aleatoria, observemos los siguientes ejemplos donde teniendo una moneda justa, nuestr variable aleatoria se definirá como *"X = Número de caras que resultan de lanzar una moneda N veces"*. 

```python
# Definir los posibles resultados de un lanzamiento de moneda
outcomes = ['H', 'T']  # H para "Cara" (Heads), T para "Cruz" (Tails)

# Generar el espacio muestral para tres lanzamientos
sample_space = list(product(outcomes, repeat=3))

# Definir la variable aleatoria X como el número de "caras" (H) en tres lanzamientos
def count_heads(trial):
    return trial.count('H')

# Calcular la función de probabilidad (PMF)
pmf = Counter(count_heads(trial) for trial in sample_space)

# Normalizar las frecuencias para obtener probabilidades
total_outcomes = len(sample_space)
for x in pmf:
    pmf[x] /= total_outcomes

# Mostrar la función de probabilidad (PMF)
print("Variable Aleatoria X: Número de Caras en Tres Lanzamientos de Moneda")
print("Espacio muestral = ", sample_space)

for outcome, probability in sorted(pmf.items()):
    print(f"P(X = {outcome}) = {probability:.2f}")
    
# Generamos el gráfico de la distribución PMF
graficar_distribucion(list(range(1, len(pmf) + 1)), pmf.values(), title='Probabilidad de # caras en tres lanzamientos', x_label='X', y_label='P(X)')
```

```python
# Definir los posibles resultados de un lanzamiento de moneda
outcomes = ['H', 'T']  # H para "Cara" (Heads), T para "Cruz" (Tails)

# Generar el espacio muestral para tres lanzamientos
sample_space = list(product(outcomes, repeat=10))

# Definir la variable aleatoria X como el número de "caras" (H) en tres lanzamientos
def count_heads(trial):
    return trial.count('H')

# Calcular la función de probabilidad (PMF)
pmf = Counter(count_heads(trial) for trial in sample_space)

# Normalizar las frecuencias para obtener probabilidades
total_outcomes = len(sample_space)
for x in pmf:
    pmf[x] /= total_outcomes
    
# Generamos el gráfico de la distribución PMF
graficar_distribucion(list(range(1, len(pmf) + 1)), pmf.values(), title='Probabilidad de # caras en tres lanzamientos', x_label='X', y_label='P(X)')
```

Note que las ateriores distribuciones para nuestra variable aleatoria han sido generadas a partir de generar el espacio muestral de la variable, por otra parte, es posible llegar a los mismos resultados si realizamos el experimento aleatorio como tal (lanzando una moneda múltiples veces) y deducir la distribución PMF a partir del conteo de los resultados generados por el experimento.

```python
# Definir la función para simular el lanzamiento de la moneda
def simulate_coin_tosses(n_tosses):
    outcomes = ['H', 'T']
    return [random.choice(outcomes) for _ in range(n_tosses)]

# Definir la función para contar las "caras" en un experimento
def count_heads(tosses):
    return tosses.count('H')

# Número de experimentos y lanzamientos por experimento
num_experiments = 1000
n_tosses = 10

# Realizar los experimentos y contar las "caras"
results = [count_heads(simulate_coin_tosses(n_tosses)) for _ in range(num_experiments)]

# Calcular la función de probabilidad (PMF) basada en los resultados de los experimentos
pmf = Counter(results)
total_experiments = num_experiments

# Normalizar para obtener probabilidades
for x in pmf:
    pmf[x] /= total_experiments

# Mostrar la función de probabilidad (PMF)
print("Variable Aleatoria X: Número de Caras en Tres Lanzamientos de Moneda (1000 Experimentos)")
for outcome, probability in sorted(pmf.items()):
    print(f"P(X = {outcome}) = {probability:.2f}")


# Generamos el gráfico de la distribución PMF
graficar_distribucion(pmf.keys(), pmf.values(), x_label='X', y_label='P(X)', title='Probabilidad de # caras en tres lanzamientos')
```

Valores Esperados y Varianzas

---

Teniendo en cuenta que la distribución de probabilidad de una variable aleatoria $X$ viene dada por.

$$
g_X (x_i) = P(X = x_i)
$$

El **valor esperado** de una variable aleatoria discreta es una medida de la centralidad de su distribución de probabilidad. Se calcula como la suma de todos los posibles valores de la variable aleatoria, cada uno multiplicado por su probabilidad. Esencialmente, representa la media ponderada de los valores de la variable aleatoria, la cual es equivalente al *promedio* $\mu$ de la variable.

Matemáticamente, para una variable aleatoria discreta $X$ con posibles valores $x_i$ y probabilidades asociadas $P(x_i)$, el valor esperado $E(X)$ se define como:

$$
E(X) = \sum_{i} x_i P(x_i) = \mu _{X}
$$

Por otra parte, la **varianza** de una variable aleatoria discreta es una medida de la dispersión de sus valores respecto al valor esperado. Indica cuánto varían los valores de la variable aleatoria en torno a su media. La varianza se calcula como la suma de los cuadrados de las diferencias entre cada valor y el valor esperado, ponderada por la probabilidad de cada valor.

Matemáticamente, para una variable aleatoria discreta $X$, la varianza $\text{Var}(X)$ se define como:

$$
\text{Var}[X] = E[(X - E[X])^2]
$$

Esto se puede expandir como:

$$
\text{Var}[X] = \sum_{i} (x_i - E[X])^2 P(x_i)
$$

Donde la **variación estándard** es una métrica de la dispersión de la variable definida en las mísmas unidades de ésta, y corresponde simplemente a la raiz cuadrada de la varianza:

$$
\sigma_X = \sqrt{\text{Var}[X]}
$$


Para ilustrar estos conceptos, calcularemos el valor esperado y la varianza de la variable $X$ la cual corresponde al número de caras posibles en 10 lanzamientos de una moneda.

```python
# Simulación de 1000 experimentos con 10 lanzamientos de moneda
num_experiments = 1000
n_tosses = 10

# Función para simular lanzamientos de moneda y contar "caras"
def simulate_coin_tosses(n_tosses):
    outcomes = ['H', 'T']
    return [random.choice(outcomes) for _ in range(n_tosses)]

def count_heads(tosses):
    return tosses.count('H')

# Realizar experimentos y contar resultados
results = [count_heads(simulate_coin_tosses(n_tosses)) for _ in range(num_experiments)]

# Calcular la PMF
pmf = Counter(results)
total_experiments = num_experiments
for x in pmf:
    pmf[x] /= total_experiments

# Cálculo del valor esperado (E(X))
expected_value = sum(x * p for x, p in pmf.items())

# Cálculo de la varianza (Var(X))
variance = sum((x - expected_value) ** 2 * p for x, p in pmf.items())

# Resultados
print(f"Valor Esperado (E(X)): {expected_value:.2f}")
print(f"Varianza (Var(X)): {variance:.2f}")
```

---
Distribución de Probabilidad Discreta

---

Una **distribución de probabilidad discreta** es una función que asigna probabilidades a los valores posibles de una variable aleatoria discreta. Una variable aleatoria discreta es aquella que puede tomar un número finito o infinitamente contable de valores específicos, a diferencia de una variable aleatoria continua que puede tomar cualquier valor en un intervalo continuo. Las distribuciones discretas poseen las siguientes características:

1. **Lista de Valores Posibles**: La distribución especifica todos los valores posibles que puede tomar la variable aleatoria.

2. **Probabilidades Asociadas**: A cada valor posible de la variable aleatoria se le asigna una probabilidad, que indica la frecuencia con la que se espera que ocurra ese valor en un gran número de experimentos.

3. **Suma de Probabilidades**: La suma de todas las probabilidades en una distribución de probabilidad discreta es igual a 1.

Matemáticamente, si $X$ es una variable aleatoria discreta con valores posibles $x_1, x_2, \ldots, x_n$ y sus respectivas probabilidades $p(x_1), p(x_2), \ldots, p(x_n)$, entonces la distribución de probabilidad se define como:

$$
P(X = x_i) = p(x_i)
$$


Dentro de las distribuciones de probabilidad discretas más importantes encontramos:

1. **Distribución de Bernoulli**: Describe la probabilidad de éxito de un único ensayo de Bernoulli.

2. **Distribución Binomial**: Describe el número de éxitos en una serie de ensayos independientes de Bernoulli.

3. **Distribución Binomial Negativa**: Describe el número de ensayos necesarios para obtener un número fijo $r$ de éxitos en una serie de ensayos de Bernoulli independientes.

4. **Distribución de Poisson**: Describe el número de eventos que ocurren en un intervalo fijo de tiempo o espacio.

5. **Distribución Geométrica**: Describe el número de ensayos necesarios para obtener el primer éxito en una serie de ensayos de Bernoulli.

6. **Distribución Uniforme Discreta**: Cada valor en el conjunto tiene la misma probabilidad de ocurrir.


Supongamos que estamos lanzando un dado justo. La variable aleatoria $X$ representa el número que aparece en el dado. Los valores posibles de $X$  son 1, 2, 3, 4, 5 y 6. La distribución de probabilidad discreta para este caso es:

$$
P(X = x) =
\begin{cases}
\frac{1}{6} & \text{si } x = 1, 2, 3, 4, 5, 6 \\
0 & \text{en caso contrario}
\end{cases}
$$

En este ejemplo, cada valor tiene una probabilidad de $\frac{1}{6}$, y la suma de todas las probabilidades es 1.

Las distribuciones de probabilidad son fundamentales en estadística y probabilidad porque proporcionan una forma estructurada de describir la incertidumbre y la variabilidad en eventos y datos. Su importancia y utilidad abarcan varias áreas clave:

**1. Descripción y Modelado de Fenómenos Aleatorios**: Las distribuciones de probabilidad permiten modelar fenómenos aleatorios y describir la distribución de datos observados. Por ejemplo, en la vida real, podemos usar distribuciones para modelar cosas como el número de llamadas que recibe un centro de atención al cliente en una hora (distribución de Poisson) o el tiempo entre llegadas de autobuses (distribución exponencial).

**2. Cálculo de Probabilidades**: Permiten calcular la probabilidad de que ocurra un evento específico. Por ejemplo, con una distribución normal, podemos calcular la probabilidad de que una variable aleatoria caiga dentro de un rango determinado, lo que es útil en muchas aplicaciones, como el control de calidad o la gestión de riesgos financieros.

**3. Toma de Decisiones Informadas**: Las distribuciones de probabilidad son esenciales para la toma de decisiones bajo incertidumbre. En economía, finanzas, y muchos otros campos, la probabilidad de eventos futuros se utiliza para tomar decisiones que minimicen riesgos y maximicen beneficios.

**4. Estimación y Predicción**: Permiten hacer inferencias sobre poblaciones con base en muestras. En estadística inferencial, las distribuciones de probabilidad ayudan a estimar parámetros desconocidos de una población (como la media o la varianza) y a realizar predicciones.

**5. Inferencia Estadística y Pruebas de Hipótesis**: Las distribuciones de probabilidad son la base para realizar pruebas de hipótesis. Por ejemplo, se pueden utilizar para determinar si una muestra proviene de una población con una media específica o si dos muestras difieren significativamente.

**6. Medición de la Incertidumbre**: Ofrecen una manera de cuantificar la incertidumbre. La varianza y la desviación estándar, derivadas de una distribución de probabilidad, son medidas comunes de la dispersión y la incertidumbre en los datos.

**7. Simulación y Modelado Computacional**: Las distribuciones de probabilidad son cruciales en simulaciones y modelos computacionales, que se utilizan para predecir y entender sistemas complejos. Por ejemplo, en la ingeniería y las ciencias naturales, las simulaciones de Monte Carlo usan distribuciones de probabilidad para modelar y analizar sistemas que son demasiado complejos para resolver de forma analítica.

**8. Comunicación de Resultados**: Las distribuciones de probabilidad facilitan la comunicación de resultados estadísticos. Al describir la forma de una distribución (simetría, curtosis, etc.), se puede dar una idea clara de cómo se distribuyen los datos y qué esperar de ellos.

Distribución de Probabilidad Acumulada Discreta

---

Una **distribución de probabilidad acumulada** (CDF, por sus siglas en inglés) es una función que describe la probabilidad de que una variable aleatoria tome un valor menor o igual a un cierto número. Formalmente, para una variable aleatoria $X$, la CDF $F(x)$ se define como:

$$
F(x) = P(X \leq x)
$$

donde $P(X \leq x)$ representa la probabilidad de que $X$ tome un valor menor o igual a $x$. Este tipo de distribuciones tienen las siguientes características:

1. **Monótona no Decreciente**: La CDF siempre aumenta o se mantiene constante a medida que aumenta $x$.

2. **Rango de Valores**: La CDF toma valores en el intervalo [0, 1]. Es decir, $F(x)$.

3. **Punto de Inicio y Fin**: Para una variable aleatoria continua, $F(x)$ se aproxima a 0 cuando $x$ tiende a $-\infty$ y se aproxima a 1 cuando $x$ tiende a $\infty$.

4. **Uso en Probabilidad**: La CDF se utiliza para determinar la probabilidad de que una *variable aleatoria esté dentro de un cierto rango*. Por ejemplo, $ P(a \leq X \leq b) = F(b) - F(a) $.

5. **Aplicaciones**: La CDF es útil en estadística y probabilidad para resumir y describir la distribución de una variable aleatoria. También se utiliza en pruebas de hipótesis y análisis de datos.

En términos matemáticos, la distribución de probabilidad acumulada se determina de la siguiente manera, donde $R_X$ es el rango de la variable aleatoria:

$$
F(x) = \sum _{x_i \in R_X, x_i <= x } g_{X} (x_i)



```python
# Definir la función para simular el lanzamiento de la moneda
def simulate_coin_tosses(n_tosses):
    outcomes = ['H', 'T']
    return [random.choice(outcomes) for _ in range(n_tosses)]

# Definir la función para contar las "caras" en un experimento
def count_heads(tosses):
    return tosses.count('H')

# Número de experimentos y lanzamientos por experimento
num_experiments = 1000
n_tosses = 10

# Realizar los experimentos y contar las "caras"
results = [count_heads(simulate_coin_tosses(n_tosses)) for _ in range(num_experiments)]

# Calcular la función de probabilidad (PMF) basada en los resultados de los experimentos
pmf = Counter(results)
total_experiments = num_experiments

# Normalizar para obtener probabilidades
for x in pmf:
    pmf[x] /= total_experiments
    
# Calculamos la distribución acumulada CDF
cumm_prob = 0
cdf = {}
for x in sorted(pmf.keys()):
    cumm_prob += pmf[x]
    cdf[x] = cumm_prob
    

# Mostrar la función de probabilidad (PMF)
print("Variable Aleatoria X: Número de Caras en Tres Lanzamientos de Moneda (1000 Experimentos)")
for outcome, probability in sorted(cdf.items()):
    print(f"P(X = {outcome}) = {probability:.2f}")


# Generamos el gráfico de la distribución PMF
graficar_distribucion(cdf.keys(), cdf.values(), x_label='X', y_label='P(X <= x)', title=f'CDF para una moneda lanzada {n_tosses}')
```

$$
P(X > x) = 1 - P(X <= x) = 1 - F(x)
$$

Note que al contrario de la CDF, la CCDF (por sus siglas en inglés) es una función de probabilidad monotónica decreciente.

---
Distribución de Bernoulli

---

**Definición**: La distribución de Bernoulli es una distribución de probabilidad discreta que describe el resultado de *un solo experimento con dos posibles resultados*: éxito (generalmente representado por 1) y fracaso (generalmente representado por 0). Es la distribución más simple y básica en probabilidad y estadística.

**Parámetros**:
- **\( p \)**: La probabilidad de éxito (el evento que se define como 1).
- **\( 1 - p \)**: La probabilidad de fracaso (el evento que se define como 0).

**Función de Masa de Probabilidad (PMF)**:
$$
P(X = x) = 
\begin{cases} 
p & \text{si } x = 1 \\
1 - p & \text{si } x = 0 
\end{cases}
$$

La distribución de Bernoulli nos permite **modelar de experimentos binarios** los cuáles son aquellos experimentos que tienen exactamente dos posibles resultados. Por ejemplo, lanzar una moneda (cara o cruz), verificar si un producto pasa o falla una prueba de calidad, etc. Esta distribución corresponde la base para distribuciones más complejas tales como la distribución binomial. Un experimento binomial es simplemente una serie de experimentos de Bernoulli.

Algunas de las pripiedades más importantes de ésta distribución son:

**1. Media (Valor Esperado)**: La media o valor esperado de una variable aleatoria $X$ con distribución de Bernoulli, donde $p$ es la probabilidad de éxito, se calcula como:
$$
\mathbb{E}[X] = p 
$$

**2. Varianza**: Para la distribución de Bernoulli, la varianza se calcula como:
$$
\text{Var}(X) = p \cdot (1 - p) 
$$

**3. Desviación Estándar**: La desviación estándar para la distribución de Bernoulli, se calcula como:
$$
\sigma = \sqrt{p \cdot (1 - p)} 
$$

**6. Función de Distribución Acumulada (CDF)**: La función de distribución acumulada para la distribución de Bernoulli es:
$$
F(x) =
\begin{cases} 
0 & \text{si } x < 0 \\
1 - p & \text{si } 0 \leq x < 1 \\
1 & \text{si } x \geq 1 
\end{cases}
$$

7. **Propiedad de Independencia**: La distribución de Bernoulli es una distribución binomial especial con $n = 1$. La suma de $n$ variables aleatorias independientes con distribución de Bernoulli (con la misma probabilidad $p$) sigue una distribución binomial con parámetros $n$ y $p$.

```python
# Parámetro de la distribución de Bernoulli
p = 0.3

# Definir los posibles resultados
x_values = [0, 1]
pmf_values = [1 - p, p]

# Imprimir los resultados de la PMF
print(f"PMF para p={p}:")
for x, prob in zip(x_values, pmf_values):
    print(f"P(X = {x}) = {prob:.2f}")
    
# Generamos el gráfico de la distribución PMF
graficar_distribucion(['Fracaso (0)', 'Exito (1)'], pmf_values, x_label='X', y_label='P(X)', title='Distribución de Bernoulli')

```

**<h3 id="subsection123" style="color: #003366;">1.2.3. Distribucion Binomial</h3>**

---
Distribución Binomial

---

La distribución binomial es una distribución de probabilidad discreta que describe el *número de éxitos en una secuencia fija de experimentos independientes*, cada uno de los cuales tiene dos resultados posibles (éxito o fracaso). Es una extensión natural de la distribución de Bernoulli a múltiples ensayos. Si una variable aleatoria $X$ sigue una distribución Binomial, se denota como $ X \sim \text{Binomial}(n, p) $, donde:
- $ n $ es el número de ensayos (o pruebas) independientes.
- $ p $ es la probabilidad de éxito en cada ensayo.
- $ 1 - p $ es la probabilidad de fracaso en cada ensayo.

La función de masa de probabilidad (PMF) de una distribución Binomial se define como:

$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k} 
$$

donde $ k $ es el número de éxitos deseados $0 ≤  k  ≤  n $. Dentro de sus propiedades estadísticas encontramos:

**1. Media (Valor Esperado)**: La media o valor esperado de una variable aleatoria Binomial $ X $ es:
$$
   \mathbb{E}[X] = n \cdot p 
$$

**2. Varianza**: La varianza de una variable aleatoria Binomial $ X $ es:
$$
   \text{Var}(X) = n \cdot p \cdot (1 - p) 
$$

**3. Desviación Estándar**: La desviación estándar es la raíz cuadrada de la varianza:
$$
   \sigma = \sqrt{n \cdot p \cdot (1 - p)} 
$$

Pero cabe preguntarse, ¿por que la fórmula matemática de ésta distribución tiene esta forma?, para ello exploraremos los siguientes conceptos:

**Coeficiente Binomial**

El coeficiente binomial, **\binom{n}{k}**, es una parte fundamental de la fórmula de la distribución binomial. Representa el *número de maneras diferentes de seleccionar $ k $ éxitos (o elementos) de un total de $ n $ ensayos*, es decir, mide la cantidad de *combinaciones de tamaño $k$ para un para $n$ lanzamientos*.El coeficiente binomial se calcula como:

$$
\binom{n}{k} = \frac{n!}{k! \cdot (n - k)!} 
$$

donde $n!$ (n factorial) es el producto de todos los enteros positivos hasta $n$ y representa la cantidad de *permutaciones ordinarias posibles dentro de $n$ lanzamientos* (una permutación ordinaria es aquella secuencia ordenada de elementos que tiene el mísmo tamaño al espacio muestral, es decir $p(r=n;n)= n!$), $k!$ es el factorial de $k$, y $(n - k)!$ es el factorial de $( n - k )$.

**Justificación Matemática**:

- **Permutaciones Totales Posibles**: Para $n$ elementos, hay $n!$ maneras de permutarlos ordinariamente.
- **Selección de $k$ Éxitos**: De estas $ n! $ permutaciones, queremos seleccionar solo aquellos en los que hay exactamente $ k $ son éxitos. 
- **División por Permutaciones Internas**:
  - Dentro de los $ k $ éxitos, las permutaciones entre ellos no cambian el resultado final (es decir, son indistinguibles).
  - Dentro de los $( n - k )$ fracasos, las permutaciones también son indistinguibles.
- **Ajuste**: Por lo tanto, el número de maneras de seleccionar $ k $ éxitos (o cualquier subconjunto de tamaño $ k $) de $ n $ es:
  $ \frac{n!}{k! \cdot (n - k)!} $

Para una secuencia específica con exáctamente $k$ éxitos y $n-k$ fracasos, la probabilidad de que ocurran dicho número de éxitos es igual al *producto de las probabilidades de ocurrencia de cada ensayo posible* ya que cada evento es *independiente uno del otro*. Así, si tenemos $k$ éxitos, la probabilidad de esto es igual a $p^k$ mientras que la probabilidad de tener  $(n-k)$ es $p^{n-k}$. Así, la proabilidad de describir un escenario con $k$ ensayos y $n-k$ fracasos es simplemente $p^k(1-p)^{n-k}$. 

Dado que hay \binom{n}{k} secuencias diferentes que pueden resultar en exactamente $k$ éxitos, la probabilidad total de tener exactamente $k$ éxitos es el producto de las probabilidades de todas esas secuencias.

Note entonces, que la distribución de Bernoulli puede ser recuperada a partir de la distribución Binomial si $n = k =1$

$$
P(X = 1) =  p^1 (1 - p)^{n - 1} 
$$


```python
# Definición de una distribución binomial
def binomial(k,n,p):
    binom = factorial(n) / (factorial(k) * factorial(n-k))
    pmf = binom * (p**k) * (1-p)**(n-k)
    return pmf

# Calcular la PMF usando nuestra función
n = 10  # Número de ensayos
p = 0.3  # Probabilidad de éxito
x = np.arange(0, n+1)
pmf = [binomial(k,n,p) for k in x]

# Generamos el gráfico de la distribución PMF
graficar_distribucion(x, pmf, x_label='X', y_label='P(X)', 
                      title=f'Distribución Binomial para k y {n} Lanzamientos de Moneda')
```

```python
# Parámetros de la distribución binomial
n = 10  # Número de ensayos
p = 0.3  # Probabilidad de éxito

# Calcular la PMF usando scipy
x = np.arange(0, n+1)
pmf = stats.binom.pmf(x, n, p)

# Calcular y mostrar el coeficiente binomial para un valor específico k
k = 3
coef_binomial = comb(n, k)
print(f"Coeficiente Binomial C({n}, {k}) = {coef_binomial}")

# Ejemplo de PMF manual para un valor específico k
pmf_k = stats.binom.pmf(k, n, p)
print(f"Probabilidad de obtener exactamente {k} éxitos: {pmf_k:.3f}")

# Generamos el gráfico de la distribución PMF
graficar_distribucion(x, pmf, x_label='X', y_label='P(X)', 
                      title=f'Distribución Binomial para k y {n} Lanzamientos de Moneda')
```

**<h3 id="subsection123" style="color: #003366;">1.2.3. Distribucion Binomial Negativa</h3>**

---
Distribución Binomial Negativa

---

La distribución binomial negativa es una distribución de probabilidad discreta que generaliza la distribución Binomial. Es útil para modelar el *número de ensayos necesarios hasta que ocurran un número fijo de éxitos en una serie de experimentos de Bernoulli independientes*. Vamos a desglosar su definición, uso, casos de utilidad y justificación matemática en detalle. La distribución binomial negativa describe el número de fracasos (ensayos fallidos) que ocurren antes de alcanzar un número fijo $r$ de éxitos en una secuencia de experimentos de Bernoulli independientes, donde cada experimento tiene una probabilidad $p$ de éxito. Posee los siguientes parámetros

1. **$ r $**: Número de éxitos deseados.
2. **$p $**: La probabilidad de éxito en cada ensayo.

El número de fracasos $ k $ sigue una distribución Binomial negativa si la probabilidad de obtener exactamente $ k $ fracasos antes de lograr $ r $ éxitos está dada por la función de masa de probabilidad:

$$
 P(X = k) = \binom{k + r - 1}{r - 1} (1 - p)^k p^r 
$$

donde $ \binom{k + r - 1}{r - 1} $ es el coeficiente binomial que representa el número de formas en que se pueden distribuir los $ k $ fracasos entre los $ r $ éxitos.

**Justificación Matemática**

La función de masa de probabilidad (PMF) de la distribución Binomial negativa puede derivarse a partir de una serie de experimentos de Bernoulli independientes. La probabilidad de que se necesiten exactamente $k$ fracasos antes de obtener $ r $ éxitos se puede calcular considerando las siguientes observaciones:

- **Número Total de Ensayos**: Para obtener $ r $ éxitos, necesitamos realizar $ k + r $ ensayos en total.
- **Posiciones de los Éxitos**: Elegimos $ r $ posiciones para los éxitos entre los $ k + r $ ensayos. El resto de los ensayos serán fracasos.
- **Probabilidad de Éxitos y Fracasos**: Cada uno de estos $ k + r $ ensayos tiene una probabilidad $ p $ de éxito y $ 1 - p $ de fracaso.

El número de formas en que los $ k $ fracasos pueden ser distribuidos en los $ k + r - 1 $ ensayos previos a la obtención del $ r $-ésimo éxito es dado por $ \binom{k + r - 1}{r - 1} $. Luego, multiplicamos esto por la probabilidad de tener exactamente $ k $ fracasos y $ r $ éxitos en los $ k + r $ ensayos. 


**Propiedades Estadísticas de la Distribución de Geométrica**

Algunas de las propiedades estadísticas más importantes de la distribución Binomial negativa son:


**1. Esperanza (Media):**
$$
E[X] = \frac{r(1 - p)}{p} 
$$

**2. Varianza:**
$$
\text{Var}(X) = \frac{r(1 - p)}{p^2} 
$$

**Casos de Uso y Aplicaciones**

Algunas de las aplicaciones de la distribución Binomial Negativa son:

- **Tiempo hasta un Número Fijo de Éxitos**: La distribución binomial negativa es útil para modelar situaciones en las que estamos interesados en el número de intentos necesarios hasta alcanzar un número específico de éxitos. Por ejemplo, el número de fracasos antes de lograr una cantidad determinada de ventas en un modelo de ventas.

- **Modelado de Datos Contables**: En finanzas y contabilidad, puede ser útil para modelar el número de errores o problemas que ocurren antes de alcanzar un objetivo de calidad.

- **Número de Pruebas Clínicas**: Puede ser usada para modelar el número de pruebas fallidas hasta que se encuentren un número fijo de casos positivos en estudios médicos.


- **Análisis de Campañas Publicitarias**: Permite modelar el número de intentos publicitarios necesarios para lograr una cantidad fija de conversiones o ventas.


```python
# Definición de una distribución binomial negativa
def binomial_negativa(k, r, p):
    binom = factorial(k + r - 1) / (factorial(r - 1) * factorial(k))
    pmf = binom * (p**r) * ((1 - p)**k)
    return pmf

# Calcular la PMF usando nuestra función
r = 5  # Número de ensayos
k_max = 20 # Número máximo de fracasos
p = 0.4  # Probabilidad de éxito
k = np.arange(0, k_max + 1)
pmf = [binomial_negativa(k_val,r,p) for k_val in k]

# Generamos el gráfico de la distribución PMF
graficar_distribucion(k, pmf, x_label='X', y_label='P(X)', 
                      title=f'Distribución Negativa Binomial para \n k fracasos y  {r} éxitos con probabilidad {p}')
```

```python
# Parámetros de la distribución binomial negativa
r = 5  # Número de ensayos
k_max = 20 # Número máximo de fracasos
p = 0.4  # Probabilidad de éxito

# Calcular la PMF usando scipy
k = np.arange(0, k_max + 1)
pmf = stats.nbinom.pmf(k, r, p)

# Calcular y mostrar el coeficiente binomial para un valor específico k

coef_binomial = comb(k_max + r - 1, r - 1)
print(f"Coeficiente Binomial C({k_max + r - 1}, {r - 1}) = {coef_binomial}")

# Ejemplo de PMF manual para un valor específico k
pmf_k = stats.nbinom.pmf(k_max, r, p)
print(f"Probabilidad de obtener exactamente {r} éxitos: {pmf_k:.3f}")

# Generamos el gráfico de la distribución PMF
graficar_distribucion(k, pmf, x_label='X', y_label='P(X)', 
                      title=f'Distribución Negativa Binomial para \n k fracasos y  r éxitos con probabilidad p')
```

Distribución Geométrica

---

La distribución geométrica es una distribución de probabilidad que modela el *número de ensayos necesarios para obtener el primer éxito en una serie de ensayos independientes*, cada uno con la misma probabilidad de éxito. Es una distribución discreta que se aplica en situaciones donde el resultado de interés es el número de intentos hasta que ocurre el primer éxito.

- **Variable Aleatoria**: La variable aleatoria $X$ representa el número de ensayos necesarios para obtener el primer éxito.
- **Parámetro**: $p$ es la probabilidad de éxito en cada ensayo.
- **Función de Masa de Probabilidad (PMF)**: La probabilidad de que el primer éxito ocurra en el $k$-ésimo ensayo está dada por:
$$
  P(X = k) = (1 - p)^{k - 1} \cdot p
$$
  donde $k$ es el número de ensayos hasta el primer éxito.


**Propiedades Estadísticas de la Distribucuión Geométrica**
Alguna de las propiedades estadísticas más importantes de la distribución geométrica son:

1. **Esperanza (Media)**: 
$$
E[X] = \frac{1}{p}
$$

2. **Varianza**:
$$
\text{Var}[X] = \frac{1 - p}{p^2}
$$

**Casos de Uso y Aplicaciones**

La distribución geométrica es útil en una variedad de contextos, tales como:
- **Modelos de espera**: En situaciones donde se espera un resultado exitoso en una serie de intentos, como en juegos de azar o pruebas de productos.
- **Simulación de eventos**: Para modelar el número de intentos necesarios antes de que ocurra un evento específico en experimentos repetidos.

```python
# Definición de una distribución binomial negativa
def geometrica(k, p):
    pmf =  (p) * ((1 - p)**(k-1))
    return pmf

# Calcular la PMF usando nuestra función
k_max = 20 # Número máximo de fracasos
p = 0.3  # Probabilidad de éxito
k = np.arange(1, k_max + 1)
pmf = [geometrica(k_val, p) for k_val in k]

# Generamos el gráfico de la distribución PMF
graficar_distribucion(k, pmf, x_label='X', y_label='P(X)', 
                      title=f'Distribución Geométrica con probabilidad de éxito {p}')
```

### <!-- Subsection 1.2.5 -->
**<h3 id="subsection125" style="color: #003366;">1.2.5. Distribucion de Poisson</h3>**

---
Distribución de Poisson

---

Un **proceso de Poisson** es un modelo estocástico que describe eventos que ocurren de manera aleatoria e independiente en el tiempo o el espacio. Es ampliamente utilizado en la teoría de colas, telecomunicaciones, economía y otros campos para modelar la llegada de eventos en un intervalo de tiempo o región espacial.

Una distribución de Poisson es una distribución de probabilidad discreta que describe el *número de eventos que ocurren en un intervalo de tiempo $t$ (o de espacio) fijo*. Dichos eventos son de tipo Bernoulli en donde la situación puede modelarse bajo la premisa sencilla de *"ocurre un evento"* o *"el evento no ocurrio"*. Un proceso de Poisson posee las siguientes características:


1. **Eventos Aleatorios e Independientes**: Los eventos ocurren de manera aleatoria e independiente unos de otros.
2. **Estacionariedad**: La tasa a la que ocurren los eventos es constante en el tiempo. Es decir, la probabilidad de que ocurra un evento en un intervalo de tiempo dado solo depende de la longitud del intervalo a considerar mas no del momento específico. El contínuo *t* puede ser dividido en partes iguales $\delta t$ lo suficientemente pequeñas como para que a lo sumo, uno de los eventos puntuales tenga lugar en dicho intervalo de tiempo. 
3. **No Acumulación**: No hay dos eventos que ocurran exactamente al mismo tiempo.
4. **Taza de ocurrencia**: La probabilidad de que un evento puntual tenga lugar en un intervalo $\delta t$ depende de la longitud del intervalo y de una constante $\lambda > 0$ que caracteríza al proceso, donde esta constante se denomina como la *taza promedio de eventos por unidad de tiempo*. 

$$
P(\text{Ocurrencia de un evento en intervalo de tiempo delta t}) = \lambda \delta t
$$

Note entonces que la probabilidad de ocurrencia de un evento no depende de la ubicación del intervalo $\delta t$, por ende, podemos decir que las llegadas o eventos están uniformemente distribuidas en el tiempo.

El proceso de Poisson caracterizado por una única tasa de eventos promedio $\lambda$, representa el número promedio de eventos por unidad de tiempo. Si $N(t)$ es el número de eventos que ocurren en el intervalo de tiempo $[0, t]$, entonces $N(t)$ sigue una distribución de Poisson con parámetro $\lambda t$. Este modelo probabilístico es equivalente a modelar la premisa de *obtener exactamente $k$ éxitos en los $N$ experimentos independientes de Bernoulli que consisten en observar si ha habido o no una llegada del proceso en el intervalo de tiempo de longitud $t/N$*

La función de masa de probabilidad (PMF) para la variable *$N(t) = X$ = número de llegadas en el intervalo de tiempo $t$*, es:

$$
g_{X}(k;\lambda, t) = P(X = k) = \frac{(\lambda t)^k e^{-\lambda t}}{k!}
$$

donde:
- $k$ es el número de eventos.
- $\lambda$ es la tasa de eventos por unidad de tiempo.
- $t$ es el intervalo de tiempo considerado.


**Propiedades Estadísticas de la Distribución de Poisson**

Dentro de las propiedades estadísticas más importantes del proceso de Poisson encontramos:

1. **Esperanza (Media)**: La esperanza matemática es $\lambda t$. Es el número promedio de eventos en el intervalo de tiempo $t$.

$$
E[X] = \lambda t
$$

2. **Varianza**: La varianza es también $\lambda t$, lo que significa que la media y la varianza son iguales.

$$
\text{Var}[X] = \lambda t
$$

3. **Desviación Estándar**: La desviación estándar es:

$$
\sigma _{X} = \sqrt{\lambda t}
$$



**Casos de Uso y Aplicaciones**

1. **Teoría de Colas**: Modelar la llegada de clientes a una tienda, llamadas a un centro de atención telefónica, o paquetes a un servidor.
2. **Telecomunicaciones**: Modelar la llegada de mensajes o llamadas en una red de comunicación.
3. **Economía y Finanzas**: Modelar la ocurrencia de eventos financieros como cambios de precios, ventas de activos, etc.
4. **Ciencias Naturales**: Modelar la aparición de mutaciones genéticas, llegadas de partículas, entre otros.


Para ilustrar la aplicación de la distribución de Poisson, modelaremos la cantidad de llamadas que llegan a un centro de atención al cliente en un intervalo de tiempo de $t = 60$ minutos, con una tasa promedio de $\lambda = 3$ llamadas por minuto. Utilizando una distribución de Poisson, calculamos la probabilidad de recibir un cierto número de llamadas en ese intervalo de tiempo.

```python
# Definimos la función de Poisson
def poisson(k, lambda_, t):
    return (((lambda_ * t)**k) * np.exp(-lambda_*t)) / (factorial(k))

# Parámetros del problema
lambda_rate = 0.4  # Tasa de llamadas por minuto
t = 20  # Intervalo de tiempo en minutos

k_values = np.arange(0,16)
pmf = [poisson(k_val, lambda_rate, t) for k_val in k_values]


# Numero esperado de llamadas
print('Número esperado de llamadas en t = ', lambda_rate * t)
print('Varianza del número de llamadas en t = ', lambda_rate * t)

# Generamos el gráfico de la distribución PMF
graficar_distribucion(k_values, pmf, x_label='X = número de llamadas', y_label='P(X)', 
                      title=f'Distribución Poisson con taza {lambda_rate} para t = {t}')
```

```python
# Parámetros del problema
lambda_rate = 0.4  # Tasa de llamadas por minuto
t = 20  # Intervalo de tiempo en minutos
lambda_t = lambda_rate * t

k_values = np.arange(0,16)
pmf = stats.poisson.pmf(k_values, lambda_t)

# Numero esperado de llamadas
print('Número esperado de llamadas en t = ', lambda_t)
print('Varianza del número de llamadas en t = ', lambda_t)

# Generamos el gráfico de la distribución PMF
graficar_distribucion(k_values, pmf, x_label='X = número de llamadas k', y_label='P(X = k)', 
                      title=f'Distribución Poisson con taza {lambda_rate} para t = {t}')
```

Para un análisis de eventos raros, la distribución de Poisson es especialmente útil debido a su capacidad para modelar eventos que ocurren con una baja probabilidad en un gran espacio de tiempo o espacio. Vamos a simular un caso donde analizamos la ocurrencia de un evento raro, como por ejemplo, la llegada de llamadas de emergencia a una línea especial que, en promedio, recibe 1 llamada por semana.

Vamos a suponer que estamos interesados en el número de semanas en las que se recibieron 0, 1, 2 o más llamadas. Utilizaremos una tasa promedio de $\lambda = 1$ llamada por semana y simularemos las probabilidades para los valores de llamadas recibidas.

Variables Aleatorias Contínuas

---

Las **variables aleatorias continuas** son aquellas que pueden tomar un número infinito de valores dentro de un rango continuo. A diferencia de las variables aleatorias discretas, que se asocian con conteos de eventos y pueden tomar valores específicos y separados, las variables aleatorias continuas están asociadas con mediciones y pueden tomar cualquier valor real dentro de un intervalo.

**Características de las Variables Aleatorias Continuas**

1. **Infinitos Valores Posibles**: Una variable aleatoria continua puede tomar un número infinito de valores dentro de un rango determinado. Por ejemplo, la temperatura en una ciudad o la altura de una persona pueden tomar cualquier valor dentro de un intervalo continuo dado.

2. **Función de Densidad de Probabilidad (PDF)**: En lugar de una función de masa de probabilidad (PMF) como en el caso de las variables discretas, las variables continuas se describen mediante una **función de densidad de probabilidad (PDF)**. La PDF, denotada como $f(X)$, **no** da la probabilidad de que la variable tome un valor específico (lo cual es técnicamente cero para variables continuas), sino que describe la *densidad de probabilidad por unidad de medida*. El hecho de que la probabilidad de un valor específico sea igual a cero se debe a que si tomamos la lógica de las variables discretas, la probabilidad se calcula como casos favorables para el valor de la variable dividido por el número de casos posibles lo cuál, para variables contínuas, es infinito.

3. **Probabilidad como Área Bajo la Curva**: La probabilidad de que una variable aleatoria continua tome un valor dentro de un intervalo específico se calcula como el área bajo la curva de la PDF en ese intervalo. Formalmente, si $X$ es una variable aleatoria continua, entonces la probabilidad de que $X$ esté entre $a$ y $b$ es:

   $$
   P(a \leq X \leq b) = \int_{a}^{b} f(x) \, dx
   $$

4. **Función de Distribución Acumulada (CDF)**: Similar a las variables discretas, las variables continuas también tienen una **función de distribución acumulada (CDF)**, que da la probabilidad de que la variable aleatoria tome un valor menor o igual a $x$. La CDF de una variable aleatoria continua $X$ se define como:

   $$
   F(x) = P(X \leq x) = \int_{-\infty}^{x} f(t) \, dt
   $$

5. **Normalización de la probabilidad**: Al igual que toda probabilidad, la función de densidad de probabilidad es una función completamente normalizada, es decir:
   $$
   \int_{-\infty}^{\infty} f_X(x) \, dx = 1

Para ilustrar los conceptos anteriores, observemos cómo se comporta una variable aleatoria contínua normalmente distribuída (tipo de variable que estudiaremos más a detalle en subsecciones siguientes) e intentaremos calcular la probabilidad de un rango de valores a partir de la PDF y la CDF.

```python
# Definimos la función de densidad de probabilidad (PDF) para una distribución normal
mu, sigma = 0, 1  # media y desviación estándar
dist_normal = stats.norm(mu, sigma)

# Generamos valores para la variable aleatoria continua
x = np.linspace(-4, 4, 1000)
pdf = dist_normal.pdf(x)

# Graficar la función de densidad de probabilidad (PDF)
graficar_densidad(x, pdf, title='Función de Densidad de Probabilidad (PDF) de la Distribución Normal',
                  label='PDF de la Distribución Normal',
                  x_label='X', y_label='Densidad de Probabilidad')

# Calcular la probabilidad de que la variable esté en un intervalo específico
a, b = -1, 1
probabilidad_intervalo = dist_normal.cdf(b) - dist_normal.cdf(a)
print(f"La probabilidad de que X esté entre {a} y {b} es {probabilidad_intervalo:.4f}")

# Graficar el área bajo la curva para el intervalo [a, b]
x_intervalo = np.linspace(a, b, 1000)
pdf_values_intervalo = dist_normal.pdf(x_intervalo)

# Graficar la función de densidad de probabilidad (PDF) en un intervalo
graficar_densidad_intervalo(x, pdf, x_intervalo, pdf_values_intervalo,
                            label='PDF de la distribución normal',
                            title='Área bajo la curva para el intervalo [a,b]',
                            x_label='X', y_label='Densidad de Probabilidad')

```

Valores Esperados y Varianzas

---


El **valor esperado** (también conocido como esperanza matemática o media) de una variable aleatoria continua $X$ se define como:

$$
E[X] = \int_{-\infty}^{\infty} x f_X(x) \, dx = \mu _{X}
$$

Por otra parte, la **varianza** de una variable aleatoria continua $X$ se calcula mediante la siguiente expresión, la cuál puede expresarse en términos de la función de densidad de probabilidad f_X{x}:

$$
\text{Var}[X] = E[(X - E[X])^2] = \int_{-\infty}^{\infty} (x - E[X])^2 f_X(x) \, dx 
$$

La **desviación estándar** es la raíz cuadrada de la varianza y proporciona una medida de la dispersión en las mismas unidades que la variable aleatoria, la cuál se calcula simplemente como:

$$
\sigma = \sqrt{\text{Var}[X]}
$$

```python
# Cálculo del valor esperado y la varianza de una variable contínua
mu = 8
sigma = 6.7
normal = stats.norm(mu,sigma)

# Calculamos el valor esperado
expected_value, _ = integrate.quad(lambda x: x * normal.pdf(x), -np.inf, np.inf)

# Calculamos la varianza
variance, _ = integrate.quad(lambda x: (x - expected_value)**2 * normal.pdf(x), -np.inf, np.inf)

#Calculamos la desviación estandard
sigma_X = np.sqrt(variance)

# Imprimimos los valores
print(f'El valor esperado de nuestra variable X es = {round(expected_value,3)}')
print(f'La varianza de nuestra variable X es = {round(variance,3)}')
print(f'La desviación estandard de nuestra variable X es = {round(sigma_X,3)}')
```

---
Distribución de Densidad Acumulada

---

La función de densidad acumulada (CDF, por sus siglas en inglés) de una variable aleatoria continua es una función que describe la probabilidad de que la variable aleatoria tome un *valor menor o igual a un cierto valor*. Es una herramienta fundamental en estadística y probabilidad para analizar la distribución de probabilidades de variables continuas. Para una variable aleatoria continua $X$, la función de densidad acumulada $F(x)$ está definida como:

$$
F(x) = P(X \leq x) = \int_{-\infty}^{x} f_X(t) \, dt 
$$

donde $f_X(t)$ es la función de densidad de probabilidad (PDF, por sus siglas en inglés) de $X$. Algunas propiedades de la CDF son:

1. **No decreciente**: La CDF es una función no decreciente, es decir, si $a \leq b$, entonces $F(a) \leq F(b)$.
2. **Límites**:
   - $\lim_{x \to -\infty} F(x) = 0$
   - $\lim_{x \to \infty} F(x) = 1$
3. **Rango**: El valor de la CDF está en el rango [0, 1].
4. **Continuidad**: Para una variable aleatoria continua, la CDF es una función continua.

Una de las principales utilidades de la CDF es que nos permíte realizar el cálculo de probabilidades para una variable como:

$$
P(a \leq X \leq b) = F_X(b) - F_X(a)


```python
# Cálculo de la probabilidad acumulada para una distribución normal
import scipy.integrate as integrate

# Parámetros de la distribución normal
mu = 8
sigma = 6.7
normal = stats.norm(mu, sigma)

# Definimos una función para hallar la CDF
def cdf(funcion, valores_x, a):
    return [integrate.quad(lambda x: funcion.pdf(x), -np.inf, x_val)[0] for x_val in valores_x if x_val <= a]

# Generamos los valores de la distribución
x = np.linspace(-4 * sigma + mu, 4 * sigma + mu, 300)

# Limitar el valor máximo para la CDF
a = 4 * sigma + mu 
cdf_vals = cdf(normal, x, a)

# Graficar la función de densidad de probabilidad (CDF)
graficar_densidad(x, cdf_vals, title='Función de Densidad de Acumulada (CDF) de la Distribución Normal',
                  label='CDF de la Distribución Normal',
                  x_label='X', y_label='Densidad de Probabilidad Acumulada')
```

Dentro de las aplicaciones de cualquier CDF encontramos el *cálculo de percentiles*. Para calcular percentiles utilizando la función de densidad acumulada (CDF) de una distribución, se usa la relación entre la CDF y la función de distribución inversa (también conocida como la función cuantil o percentil).

El **percentil** $p$ (donde $ 0 < p < 1 $) es el valor de la variable aleatoria tal que una fracción $ p $ de la distribución está por debajo de ese valor. Es decir, el percentil $ p $ es el valor $ x_p $ tal que $ F(x_p) = p $. 

Por otra parte, la **Función Cuantil (Inversa de la CDF)** denotada como $ F^{-1}(p) $, es la inversa de la CDF. Para un valor dado $ p $, $ F^{-1}(p) $ devuelve el valor de $ x $ tal que la probabilidad acumulada es $ p $. Matemáticamente, se define como:

$$
F^{-1}(p) = x_p \quad \text{donde} \quad F(x_p) = p
$$

Así pues, para calcular el percentil $p$ de una distribución dada, se sigue el siguiente procedimiento:

1. **Identificar el valor de $p$**: Decidir el percentil deseado. Por ejemplo, para el percentil 90, $p = 0.90$.

2. **Calcular la inversa de la CDF**: Usar la función cuantil $ F^{-1}(p) $ para encontrar el valor de $ x $ correspondiente. Este valor es el percentil buscado.

En este sentido, algunas de las aplicaciones del cálculo de percentiles mediante el uso de la CDF son:

- **Evaluación de Pruebas y Clasificación**: Los percentiles son usados para clasificar puntuaciones en pruebas estandarizadas, determinando cuántas personas están por debajo de una cierta puntuación.
- **Riesgo Financiero**: En finanzas, los percentiles (como el Value at Risk, VaR) se utilizan para medir el riesgo de inversiones.
- **Control de Calidad**: En manufactura, los percentiles ayudan a determinar límites de tolerancia para asegurar la calidad de los productos.

Para ilustrar el cálculo de cuartíles mediante el uso de la CDF, consideremos el caso de la distribución normal la cuál hemos usado como nuestra distribución contínua de ejemplo hasta ahora. Concretamente, la CDF de la distribución normal la cual es simplemente su integral viene dada por:

$$
F(x) = \frac{1}{2} \left[ 1 + \text{erf}\left(\frac{x - \mu}{\sigma \sqrt{2}}\right) \right]
$$

donde $\text{erf}()$ es la *función error* dada por:

$$
\text{erf}(x) = \frac{2}{\sqrt{\pi}} \int_0^x e^{-t^2} \, dt
$$

Para encontrar el percentil $p$ realizamos la igualdad $p = F(x)$ y despejamos el valor de $x$.

$$
p = \frac{1}{2} \left[ 1 + \text{erf}\left(\frac{x_p - \mu}{\sigma \sqrt{2}}\right) \right]
$$

Despejando $x_p$:

$$
2p - 1 = \text{erf}\left(\frac{x_p - \mu}{\sigma \sqrt{2}}\right)
$$

$$
\frac{x_p - \mu}{\sigma \sqrt{2}} = \text{erfinv}(2p - 1)
$$

$$
x_p = \mu + \sigma \sqrt{2} \, \text{erfinv}(2p - 1)
$$

Este $x_p$ es el valor del percentil $p$ para la distribución normal con parámetros $\mu$ y $\sigma$.


Para ilustar su uso, calcularemos a continuación el valor de $x_p$ mediante el uso de la función error inversa, valor el cuál corresponde al valor de $X$ que representa el percentil $p$ para una distribución normal.

```python
# Cálculo del percentil 90% para una distribución normal
from scipy.special import erf, erfinv
from scipy.stats import norm

def percentile_error_function(percentile, mu=0, sigma=1):
    x_p = mu + sigma * (np.sqrt(2) * erfinv(2*percentile - 1))
    return x_p

# Hallamos el percentile de interés usando la función error inversa
mu = 8
sigma = 6.7
percentile = 0.9
normal = stats.norm(mu, sigma)

x_p_value = percentile_error_function(percentile, mu, sigma)
p_value = normal.cdf(x_p_value)

print(f'El valor de X correspondiente al percentile {percentile*100} es =', round(x_p_value,3))
print(f'Si evaluamos x_p={round(x_p_value,3)} dentro de la CDF normal, obtenemos el percentile = ', p_value*100)
```

**Estimación Numérica de los Percentiles**

Notemos que el calculo de los percentiles es realmente sencillo una vez podemos hallar $F^{-1}(p) = x_p$ en el caso de estar interesados en el valor de $X$ que representa el percentil $p$ de nuestros datos. Sin embargo, muchas veces no tenemos conocimiento de la forma de la función inversa de la CDF, por ende, nos es posible deducir el valor $x_p$ de manera numérica si conocemos cuál es nuestra función CDF a través de diferentes métodos de optimización tales como el *Método Newton-Raphson*.

El método de Newton-Raphson es una técnica iterativa para encontrar raíces de una función (sus ceros), que en el contexto de la distribución normal se utiliza para calcular el percentil correspondiente a un valor de probabilidad acumulada. El método de Newton-Raphson se basa en una aproximación lineal de la función en torno a una estimación inicial:

1. **Definición del Problema**:
   - Queremos encontrar el valor $x$ tal que la función de distribución acumulada (CDF) de la distribución normal con media $\mu$ y desviación estándar $\sigma$ sea igual a un valor $p$, es decir, queremos resolver $F(x) = p $.

2. **Función Objetivo**:
   - Definimos una función $g(x)$ que representa la diferencia entre la CDF y el valor deseado $p$:
     $$
     g(x) = F(x) - p
     $$
   - Queremos encontrar la raíz de esta función, es decir, el valor de $x$ para el cual $g(x) = 0$.

3. **Derivada de la Función**:
   - La derivada de $g(x)$ con respecto a $x$ es la función de densidad de probabilidad (PDF) de la distribuciónde interés ya que dentro del contexto de $g(x)$, $p$ es una constante y al ser CDF = $F(x)$ la integral de la PDF = $f(x)$, la derivada de la CDF es la mísma PDF:
     $$
     g'(x) = f(x) = \text{PDF}(x)
     $$

4. **Iteración de Newton-Raphson**:
   - El método utiliza la fórmula de actualización:
     $$
     x_{n+1} = x_n - \frac{g(x_n)}{g'(x_n)} = x_n - \frac{g(x_n)}{f(x_n)}
     $$
   - Aquí, $x_n$ es la estimación actual del percentil, y $x_{n+1}$ es la nueva estimación basada en la aproximación lineal de la función en $x_n$.

5. **Proceso Iterativo**:
   - Se inicia con una estimación inicial $x_0$ (a menudo se toma la media $\mu$ como una buena estimación inicial).
   - Se actualiza la estimación usando la fórmula de Newton-Raphson hasta que la diferencia entre estimaciones sucesivas sea menor que una tolerancia predefinida, o hasta alcanzar un número máximo de iteraciones.


```python
#Estimación numérica de los percentiles de una distribución normal
import numpy as np
from scipy.special import erf, erfinv
from scipy.stats import norm

# Definimos la CDF  y la PDF  de la distribución
def normal_cdf(x, mu=0, sigma=1):
    return 0.5 * (1 + erf((x - mu) / (sigma * np.sqrt(2))))

def normal_pdf(x, mu=0, sigma=1):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Definimos el método Newton-Raphson para hallar los ceros de la función objetivo
def newton_raphson_cdf(p, mu=0, sigma=1, x0=0, tol=1e-6, max_iter=100):
    
    # Definimos la funcion objetivo
    def func(x):
        return normal_cdf(x, mu, sigma) - p
    
    # Definimos la derivada de la función objetivo
    def d_func(x):
        return normal_pdf(x, mu, sigma)
    
    x = x0
    for _ in range(max_iter):
        fx = func(x)
        dfx = d_func(x)
        if abs(dfx) < 1e-10:  # Evitar división por cero
            return x
        x_new = x - fx / dfx
        if abs(x_new - x) < tol:
            return x_new
        x = x_new
    return x

# Parámetros de la distribución normal
mu = 8
sigma = 6.7

# Percentil deseado (por ejemplo, el percentil 90)
percentile = 0.90

# Calcular el valor del percentil usando Newton-Raphson
result = newton_raphson_cdf(percentile, mu, sigma, x0=mu)
print(f'El valor de X correspondiente al percentile {percentile*100} es = {result:.2f}')
```

Distribución Uniforme

---

La distribución de probabilidad uniforme es la distribución de probabilidad contínua más elemental dentro de las distribuciones de variables aleatorias contínuas. Corres ponde a la función de densidad de probabilidad definida en los rangos $R = [a,b]$ donde todos los posibles valores de $X$ tienen el mísmo valor de probabilidad. 

$$
f_X(x) = \frac{1}{b-a}
$$

para todo $a \leq x \leq b$, fuera de éste intervalo, la PDF es nula. Por otra parte, la CDF de la distribución contínua viene dada por:

$$
F_X(x) = \int_a^x \frac{1}{b - a} \, dt = \frac{x-a}{b-a}
$$

$$
F_X(x) =
\begin{cases} 
p & \text{si } x < a \\
(x-a)f_X(x) & \text{si } a\leq x < b \\
1 & \text{si } b\leq x
\end{cases}
$$

**Propiedades Estadísticas de la Distribución de Uniforme**

Dentro de las propiedades estadísticas más importantes de la distribución uniforme encontramos:

1. **Esperanza (Media)**:q

$$
E[X] = \frac{a+b}{2}
$$

2. **Varianza**: La varianza es también $\lambda t$, lo que significa que la media y la varianza son iguales.

$$
\text{Var}[X] = \frac{(b-a)^2}{12}
$$

3. **Desviación Estándar**: 

$$
\sigma _{X} = \sqrt{\frac{(b-a)^2}{12}}
$$



**Casos de Uso y Aplicaciones**

1. **Generación de Números Aleatorios:** La distribución uniforme se utiliza para generar números aleatorios en simulaciones y experimentos de Monte Carlo.
2. **Sistemas de Control:** Se usa en sistemas de control y diseño experimental para representar incertidumbre y variabilidad.
3. **Pruebas Estadísticas:** En algunas pruebas estadísticas, la distribución uniforme puede ser utilizada para evaluar la aleatoriedad y la independencia de los datos.
4. **Modelado de Ruido:** En algunos modelos de ruido y señal, la distribución uniforme puede ser usada para representar errores o variaciones aleatorias.

```python
# Parámetros de la distribución uniforme
a = 0
b = 10

# Crear una distribución uniforme
dist = stats.uniform(loc=a, scale=b - a)

# Generar valores aleatorios
random_samples = dist.rvs(size=1000)

# Calcular PDF y CDF
x = np.linspace(a - 1, b + 1, 1000)
pdf = dist.pdf(x)
cdf = dist.cdf(x)

# Estadísticas principales
mean = dist.mean()
variance = dist.var()
std_dev = dist.std()

print(f"Media: {mean}")
print(f"Varianza: {variance}")
print(f"Desviación Estándar: {std_dev}")

# Graficar la función de densidad de probabilidad (PDF)
graficar_densidad(x, pdf, title='PDF de la distribución Uniforme',
                  label='PDF de la Distribución Normal',
                  x_label='X', y_label='Densidad de Probabilidad f(x)')


# Graficar la función de densidad de probabilidad (CDF)
graficar_densidad(x, cdf, title='CDF de la distribución Uniforme',
                  label='CDF de la Distribución Normal',
                  x_label='X', y_label='Densidad de Probabilidad Acumulada F(x)')
```

---
Distribución Normal

---

La distribución normal $X \sim \mathcal{N}(\mu, \sigma^2)$, también conocida como distribución gaussiana, es una distribución de probabilidad continua que se utiliza ampliamente en estadística, matemáticas, y muchas otras disciplinas. Su forma es conocida como la campana de Gauss debido a su gráfica característica. La función de densidad de probabilidad (PDF) de la distribución normal se define como:

$$
f_X(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp \left( -\frac{(x - \mu)^2}{2 \sigma^2} \right) 
$$

donde:
- $\mu$ es la media de la distribución y caracteriza la ubiacción del **centro de la distribución normal**.
- $\sigma$ es la desviación estándar (la raíz cuadrada de la varianza $\sigma^2$) la cuál caracteriza **el ancho de la distribución normal**.

Dentro de las características más importantes de la distirbución normal encontramos las siguientes:

1. **Forma de la Distribución**: La gráfica de la distribución normal es simétrica y tiene forma de campana. *La media, la mediana y la moda de la distribución normal son todas iguales y se encuentran en el centro de la distribución*.

2. **Media $\mu$**: Es el valor alrededor del cual la distribución está centrada. Es la medida del "centro" de la distribución.

3. **Desviación Estándar $\sigma$**: Mide la dispersión o extensión de la distribución. Una desviación estándar más grande significa que la distribución es más ancha y plana, mientras que una desviación estándar más pequeña significa que la distribución es más estrecha y alta.

4. **Propiedades de la Distribución relacionadas a los percentiles**:
   - **Simetría**: La distribución es perfectamente simétrica alrededor de la media $\mu$.
   - **Área bajo la Curva**: El total del área bajo la curva de la función de densidad es 1, lo que representa una probabilidad total del 100%.
   - **68-95-99.7 Regla**: Aproximadamente el 68% de los datos están dentro de la primera desviación estándar de la media, el 95% dentro de 2 desviaciones estándar, y el 99.7% dentro de 3 desviaciones estándar. Esta regla es también conocida como *la regla empírica*. Esta regla establece entonces que el 99% porciento de los valores de la distribución se encuentran entre las primeras tres desviaciones estandar a partir de la media:
   $$
   P(\mu - 3\sigma  \leq x \leq \mu + 3\sigma) = 0.997

5. **Propiedad de la Suma**: La suma de variables aleatorias independientes que siguen una distribución normal también sigue una distribución normal. Esta propiedad es fundamental en la teoría de probabilidad y en la inferencia estadística.

6. **Propiedad de la Normalidad**: Muchos procesos naturales y fenómenos tienden a aproximarse a una distribución normal debido al teorema central del límite. Esto se debe a que muchas variables aleatorias independientes y suficientemente grandes tienden a seguir una distribución normal, independientemente de la forma de la distribución original.

**Casos de Uso y Aplicaciones**

1. **Teorema Central del Límite**: El teorema central del límite establece que la suma (o promedio) de un gran número de variables aleatorias independientes, con una media y varianza finitas, tiende a seguir una distribución normal, independientemente de la distribución original. Esto justifica el uso de la distribución normal en muchos métodos estadísticos.

2. **Inferencia Estadística**: La distribución normal es la base de muchos métodos de inferencia estadística, incluyendo intervalos de confianza y pruebas de hipótesis. Se utilizan para hacer inferencias sobre parámetros poblacionales basados en muestras.

3. **Modelado de Datos**: En el modelado estadístico y la regresión, a menudo se asume que los errores de los modelos siguen una distribución normal para simplificar el análisis y los cálculos, lo que permite aplicar métodos de optimización como el método de máxima verosimilitud.

4. **Control de Calidad**: En el control de calidad, la distribución normal se usa para modelar variaciones en procesos de fabricación y para establecer límites de control.

5. **Economía y Finanzas**: En economía y finanzas, la distribución normal se usa para modelar retornos de activos y para realizar simulaciones y análisis de riesgo.


```python
# Definimos la función de densidad de probabilidad (PDF) para una distribución normal
def normal_pdf(x, mu=0, sigma=1):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Generamos valores para la variable aleatoria continua
mu, sigma = 8, 6.5 
x = np.linspace(mu-4*sigma, mu+4*sigma, 300)
pdf = normal_pdf(x,mu,sigma)

# Graficar la función de densidad de probabilidad (PDF)
graficar_densidad(x, pdf, title='Función de Densidad de Probabilidad (PDF) de la Distribución Normal',
                  label='PDF de la Distribución Normal',
                  x_label='X', y_label='Densidad de Probabilidad')
```

Distribución Estandar

---

La estandarización de una variable es un proceso que transforma una variable aleatoria en una variable estandarizada, que tiene una *media de 0 y una desviación estándar de 1*. Este proceso *convierte las unidades originales de la variable a una escala común, facilitando la comparación y el análisis* Para estandarizar una variable $X$ con media $\mu$ y desviación estándar $\sigma$, se usa la siguiente fórmula:

$$
Z = \frac{X - \mu}{\sigma}
$$

Donde:
- $X$ es el valor original de la variable.
- $\mu$ es la media de la variable.
- $\sigma$ es la desviación estándar de la variable.
- $Z$ es el valor estandarizado o puntaje $Z$.

**Propiedades de la Variable Estandarizada**

- **Media**: La media de los valores estandarizados es 0.
- **Desviación Estándar**: La desviación estándar de los valores estandarizados es 1.
- **Distribución**: Si la variable original $X$ sigue una distribución normal $ \mathcal{N}(\mu, \sigma^2) $, entonces la variable estandarizada $Z$ sigue una distribución normal estándar $ \mathcal{N}(0, 1) $.

**Distribución Normal Estándar**

La distribución normal estándar es una distribución normal con una media de 0 y una desviación estándar de 1. Se denota como $ \mathcal{N}(0, 1) $. La función de densidad de probabilidad para una distribución normal estándar se define como:

$$
f(z) = \frac{1}{\sqrt{2 \pi}} \exp \left( -\frac{z^2}{2} \right) 
$$

Donde $ z $ es un valor de la variable estandarizada.

**Importancia en Ciencia y Análisis de Datos**

**1. Comparación y Normalización**: La estandarización es crucial en ciencia y análisis de datos para normalizar diferentes variables que pueden tener unidades o escalas diferentes. Esto permite comparar variables de manera equitativa y mejora el rendimiento de algoritmos que dependen de la magnitud de los datos, como los algoritmos de aprendizaje automático que utilizan distancias, como K-means o máquinas de soporte vectorial (SVM).

**2. Identificación de Valores Atípicos**: La interpretación del valor $Z$ corresponde al ***número de desviaciones estandar \sigma que una observación $x$ está por encima o por debajo de la media***. Así, considerando la regla 68-95-99 tenemos que un valor $x$ se clasifica como:
  * $Z$ > 0 : $x$ se encuentra por encima de la media (donde la media representa la ubicación del 50% de los datos).
  * $Z$ < 0 : $x$ se encuentra por debajo de la media.
  * $Z$ > 2 o $Z$ > 3 : $x$ puede ser considerado como un valor atípico dependiendo del humbral del contexto, no obstante el valor más utilizado es  $Z$>3 ya que ésto corresponde a que la observación se encuentra tres desviaciones estandar por encíma o por debajo de la media, lo que es equivalente a que es un dato que no corresponde con la tendencia del 99.7% de los datos.

**Cálculo de Probabilidades con la Distribución Normal Estándar**

Gracias a la perfección y simetría de la distribución normal, es posible determinar la probabilidad de un valor de la en la distribución original a partir de la partir de la PDF de la distribución estandar mediante su CDF como hemos visto anteriormente. Esto, devido a que la integral de la distribución estandar es mucho más fácil de calcular. La CDF proporciona la probabilidad acumulada de que la variable aleatoria tome un valor menor o igual a un valor específico. Para calcular la probabilidad de que un valor estandarizado \(Z\) sea menor o igual a un cierto valor \(z\), se usa la CDF:
$$
   P(X \leq x) = P(Z \leq z) = \Phi(z)
$$

   Donde $\Phi(z)$ es la CDF de la distribución normal estándar.

Por otra parte, se tiene que gracias a la simetría de la distribución estandar:

$$
  P(Z \leq -z) = 1 - P(Z \leq z)
$$

Donde $P(Z \leq z)$ es el **percentil** de interés de la variable aleatoria.

Note a continuación, que el rango de la distribución normal estandar se encuentra entre -4 y 4, lo que simboliza el número de desviaciones estandar a partir de la media cero.

```python
# Definimos la función de densidad de probabilidad (PDF) para una distribución normal
def normal_pdf(x, mu=0, sigma=1):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Generamos valores para la variable aleatoria continua
mu, sigma = 0,1
x = np.linspace(mu-4*sigma, mu+4*sigma, 300)
pdf = normal_pdf(x,mu,sigma)

# Graficar la función de densidad de probabilidad (PDF)
graficar_densidad(x, pdf, title='Función de Densidad de Probabilidad (PDF) de la Distribución Estandar',
                  label='PDF de la Distribución Normal Estandar',
                  x_label='X', y_label='Densidad de Probabilidad')
```

Distribución Exponencial

---


La **distribución exponencial** es una distribución de probabilidad continua que describe *la probabilidad del tiempo entre eventos en un proceso de Poisson*, donde los eventos ocurren de manera continua, no simultánea e independiente a una tasa constante $\lambda$.

La distribución exponencial se utiliza para *modelar el tiempo hasta el primer evento en un proceso de Poisson*. Un proceso de Poisson es un proceso estocástico en el que los eventos ocurren con una tasa constante $\lambda$ en tiempo continuo, y los eventos son independientes entre sí. Por ejemplo, puede representar el tiempo hasta la próxima llamada telefónica que recibe un centro de atención al cliente o el tiempo de vida de un componente antes de que falle.

Note entonces, que a pesar de que el tipo de procesos por detrás de la distribución de Poisson y la distribución Exponencial son el mísmo, cada una modela un aspecto diferente del fenómeno:

* **Poisson:** Modela la probabilidad de obtener $k$ éxitos de eventos independientes de Bernoulli en un intervalo de tiempo definido.

* **Exponencial:** Modela la probabilidad de evidenciar un tiempo $t$ entre que transcurre un evento de Bernoulli y la siguiente lledada de éste tipo de evento en un proceso de Poisson con parámetro $\lambda$.

La función de densidad de probabilidad exponencial, modela pues la variable  aleatoria $X_E$ la cuál da cuenta del tiempo entre eventos de Bernoulli en un proceso de Poisson, así, esta función de densidad viene dada por:

$$
f_X(t) = \lambda e^{-\lambda t} \quad \text{para } t \geq 0
$$

Donde:

- $f_X(x)$ es la probabilidad de que el tiempo hasta el siguiente evento esté en el intervalo $[t, t + dt) $.
- $\lambda e^{-\lambda t} $ es la expresión que define la densidad de probabilidad.

**Obtención Matemática a partir de la Distribución de Poisson**

La relación entre la distribución exponencial y el proceso de Poisson es fundamental. El tiempo entre dos eventos consecutivos en un proceso de Poisson sigue una distribución exponencial. 

Supongamos que los eventos de un proceso de Poisson ocurren a una tasa promedio $\lambda$ por unidad de tiempo. Si $X$ es una variable aleatoria que representa el tiempo hasta el próximo evento, la probabilidad de que no ocurra ningún evento en un intervalo de tiempo $ t $ es:

$$
P(X > t) = P(\text{ningún evento en } [0, t]) 
$$

$$
P(X > t ) = 1 - F_X(t|\lambda)
$$

Dado que los eventos son independientes, la probabilidad de ningún evento en $[0, t]$ es la probabilidad de cero eventos en un proceso de Poisson con parámetro $\lambda t$. Esto se puede expresar como:

$$
P(X > t) = P(X = 0) = g_{X_p}(0|\lambda, t) = e^{-\lambda t}
$$

Donde aquí, $X$ es la variable aleatoria exponencial (tiempos de llegada) y $X_p$ es la variable de Poisson (número de llegadas en $t$). A demás, aca hemos empleado la notación $g_{X_p}(0|\lambda, t)$ para denotar la densidad de probabilidad del proceso de Poisson dados un $t$ y un parémetro $\lambda$. Tambíen, el lado derecho de la anterior igualdad se explica debido a:

$$
g_{X_p}(0|\lambda, t) = \frac{(\lambda t)^{0}e^{-\lambda t}}{0!} = e^{-\lambda t}
$$

ya que el factorial de cero es uno y todo número elevado a la cero es iguala uno. Si $X$ es el número de eventos en el intervalo $[0, t] $ mientras que $X_P$ denota el número de llegadas del proceso de Poisson dado un $\lambda t$, la función de distribución acumulativa (CDF) de $X$ es entonces:

$$
F_X(t) = 1 - P(X > t) = 1 - e^{-\lambda t} 
$$

para todo $t > 0$. La función de densidad de probabilidad (PDF) de $ X $ es la derivada de la CDF:

$$
f_X(t) = \frac{d}{dt} F_X(t) = \frac{d}{dt} \left( 1 - e^{-\lambda t} \right) = \lambda e^{-\lambda t}
$$

Así, hemos obtenido la PDF de la distribución exponencial a partir de la distribución de Poisson. Por otra parte, es evidente ver que la distribución CDF de la distribución exponencial es siplemente:

$$
F_X(x) = 1 - e^{-\lambda t}
$$

**Propiedades Estadísticas de la Distribución Exponencial**

Dentro de las propiedades estadísticas más importantes de la distribución Exponencial encontramos:

1. **Valor Esperado E[X]**: Tiempo promedio entre eventos de Bernoulli en un proceso de Poisson.
$$
   E[X] = \frac{1}{\lambda}
$$

2. **Varianza \( \text{Var}(X) \)**: Dispersión de los tiempos entre eventos de Bernoulli en un proceso de Poisson.
$$
   \text{Var}(X) = \frac{1}{\lambda^2} 
$$

3. **Memoria sin memoria**:
   La propiedad de falta de memoria indica que la distribución exponencial es la única distribución continua con la propiedad de que:

$$
   P(X > s + t \mid X > s) = P(X > t)
$$
   Esto significa que el tiempo adicional esperado hasta el siguiente evento no depende del tiempo ya transcurrido.

**Casos de Uso y Aplicaciones**

La distribución exponencial es ampliamente utilizada en diversas áreas como:

- **Teoría de colas**: Modela el tiempo entre llegadas de clientes o llamadas.
- **Fiabilidad**: Modela el tiempo de vida de productos o sistemas antes de fallar.
- **Economía**: Modela el tiempo hasta el próximo cambio de precios o el tiempo hasta el próximo default en finanzas.
- **Física y Química**: Modela el tiempo que tarda un átomo radioactivo en desintegrarse donde la constante de desintegración está relacionada con la media de la distribución.
- **Biología**: Modela el tiempo de vida de organizmos especialmente en contextos de riesgo de muerte. También en genética, modela el tiempo entre ventos de mutación genética.
- **Ciencias de la Computación**: Modela el tráfico de red en donde se intenta determinar el tiempo de llegadas entre paquetes de datos. También es utilizada para en alguno algoritmos avanzados, utilizar tiempos de espera distribuidos exponencialmente para decidir cuándo generar ciertas operaciones.

---
Propiedades de los Valores Esperados, las Varianzas y las Covarianzas

---

Hemos visto hasta ahora, que dentro de las propiedades globales más importantes de una distribución de probabilidad encontramos el *valor esperado* y *la varianza*. En el caso del valor esperado tenemos el caso discreto y contínuo respectívamente dados por:

$$
E[X] = \sum_{i} x_i \cdot p_i
$$
$$
E[X] = \int_{-\infty}^{\infty} x f_X(x) \, dx
$$

Con éstas definiciones, podémos definir las siguientes propiedades:

**1. Escalabilidad:** Si $c$ es una constante, se cumple tanto para el caso contínuo como para el caso discreto la siguiente igualdad:

$$
cE[X] = E[cX]
$$

**2. Potenciación de una variable aleatoria:** El valor esperado de una variable elevada a la $n$-esima potencia o también llamado como el *n-ésimo momento* se define para el caso discreto y contínuo respectivamente como:

$$
E[X^n] = \sum_{i} x_i^n \cdot p_i
$$
$$
E[X^n] = \int_{-\infty}^{\infty} x^n f_X(x) \, dx
$$

**3. Linealidad del valor esperado:** Para dos variables aleatorias $X$ y $Y$ con constantes $a$ y $b$, el valor esperado de una combinación lineal es:

$$
E[aX + bY] = aE[X] + bE[Y]
$$

ya que desde la definición del valor esperado vemos lo siguiente:
Usando la definición de valor esperado para combinaciones lineales:

$$
E[aX + bY] = \sum_{i} (a x_i + b y_i) \cdot p_i = a \sum_{i} x_i \cdot p_i + b \sum_{i} y_i \cdot p_i = aE[X] + bE[Y]
$$

De manera similar, se tiene que para el caso contínuo se cumple:

$$
E[aX + bY] = \int_{-\infty}^{\infty} (a x + b y) f_{XY}(x, y) \, dx \, dy = aE[X] + bE[Y]
$$


Por otra parte, cuando hablamos de la varianza de la variable aleatoria, definimos la varianza discreta así como la contínua respectivamente de la siguiente manera.

$$
\text{Var}(X) = \sum_{i} (x_i - E[X] )^2 f_X(x_i)
$$

$$
\text{Var}(X) = \int_{-\infty}^{\infty} (x - E[X] )^2 f_X(x) \, dx
$$

Así, mediante la igualdad $(x - \mu _X)^2 = x^2 - 2x\mu _X + \mu _X^2$ donde $\mu _X = \mathbb{E}[X]$,  podemos rescribir la varianza en términos del valor esperado como:

$$
\begin{align*}
\text{Var}(x) &= \mathbb{E}[(x - \mu_X)^2] \\
&= \mathbb{E}[x^2 - 2x\mu_X + (\mu_X)^2] \\
&= \mathbb{E}[x^2] - 2\mathbb{E}[x]\mu_X + (\mu_X)^2 \\
&= \mathbb{E}[x^2] - 2\mu_X \mu_X + (\mu_X)^2 \\
&= \mathbb{E}[x^2] - (\mathbb{E}[x])^2
\end{align*}
$$

Con lo anterior en mente, algúnas de las propiedades más importantes de la varianza son:

**1. Escalabilidad de la Varianza:** Si $k$ es una constante, se cumple tanto para el caso contínuo como para el caso discreto la siguiente igualdad:

$$
Var[kX] = k^2 Var[X]
$$

**2. Varianza de una Combinación Lineal:** Para dos variables aleatorias $ X $ y $ Y $ y constantes $ a $ y $ b $, la varianza de una combinación lineal es:

$$
\begin{align*}
\text{Var}(aX + bY) &= E[(aX + bY - E[aX + bY])^2] = E[(a(X - E[X]) + b(Y - E[Y]))^2] \\
&= a^2E[(X - E[X])^2] + b^2E[(Y - E[Y])^2] + 2abE[(X - E[X])(Y - E[Y])] \\
&= a^2 \text{Var}(X) + b^2 \text{Var}(Y) + 2ab \text{Cov}(X, Y) \\
\end{align*}
$$

Donde la **covarianza** es una medida de la relación entre dos variables aleatorias. Específicamente, cuantifica cómo varían conjuntamente estas variables con respecto a sus medias. Si la covarianza es positiva, indica que las variables tienden a aumentar o disminuir juntas. Si es negativa, indica que cuando una variable aumenta, la otra tiende a disminuir. Una covarianza de cero indica que no hay una relación lineal entre las variables.

Para dos variables aleatorias discretas $X$ e $Y$, la covarianza $\text{Cov}(X, Y)$ se define como:

$$
\text{Cov}(X, Y) = E[(X - E[X])(Y - E[Y])]
$$

Esto se puede expresar también como:

$$
\text{Cov}(X, Y) = \sum_{x_i, y_j} (x_i - E[X])(y_j - E[Y]) P(X = x_i, Y = y_j)
$$

donde $P(X = x_i, Y = y_j)$ es la probabilidad conjunta de que $X$ tome el valor $x_i$ e $Y$ tome el valor $y_j$. Para dos variables aleatorias continuas $ X $ e $ Y $ con función de densidad conjunta $ f_{XY}(x, y) $, la covarianza $ \text{Cov}(X, Y) $ se define como:

$$
\text{Cov}(X, Y) = E[(X - E[X])(Y - E[Y])]
$$

Esto se puede expresar también como:

$$
\text{Cov}(X, Y) = \int_{-\infty}^{\infty} \int_{-\infty}^{\infty} (x - E[X])(y - E[Y]) f_{XY}(x, y) \, dx \, dy
$$

donde \( f_{XY}(x, y) \) es la función de densidad conjunta de \( X \) e \( Y \).

Algúnas de las propiedades de la covarianza son:

1. **Simetría**:
   $$
   \text{Cov}(X, Y) = \text{Cov}(Y, X)
   $$

2. **Linealidad**: Si $a$ y $b$ son constantes,
   $$
   \text{Cov}(aX + b, Y) = a \, \text{Cov}(X, Y)
   $$

3. **Independencia**: Si $ X $ e $ Y $ son independientes, entonces
   $$
   \text{Cov}(X, Y) = 0
   $$

   Sin embargo, una covarianza de 0 ***no necesariamente implica independencia***, ya que puede haber una relación no lineal entre las variables.

**Interpretación de la Covarianza**

- **Covarianza positiva**: Indica que a medida que una variable aumenta, la otra también tiende a aumentar. 
- **Covarianza negativa**: Indica que a medida que una variable aumenta, la otra tiende a disminuir.
- **Covarianza cero**: Indica que no hay una relación lineal clara entre las variables.

Así pues, vémos que de manera general, una función del estadístico principal (valor esperado) puede escribirse de la siguiente forma, donde su tomamos $u(x) = (x -\mu_X)^2$ recuperamos la definición de la varianza:

$$
E[u(X)] = \sum_{i} u(X) f_X(x)
$$

$$
E[u(X)] = \int_{-\infty}^{\infty} u(X) f_X(x) \, dx
$$

**Escalabilidad del Valor Esperado**

```python
# Definimos los valores y probabilidades de X
valores_X = np.array([1, 8, 5])
probabilidades_X = np.array([0.2, 0.5, 0.3])

# Cálculo del valor esperado de X
E_X = np.sum(valores_X * probabilidades_X)
print(f"Valor esperado E[X]: {E_X}")

# Definimos una constante
a = 3

# Cálculo del valor esperado de a * X
E_aX = np.sum(a * valores_X * probabilidades_X)
print(f"Valor esperado E[aX]: {E_aX}")

# Verificación de la propiedad: E[aX] = a * E[X]
a_E_X = a * E_X
print(f"a * E[X]: {a_E_X}")

# Verificamos que E[aX] es igual a a * E[X]
print(f"¿E[aX] == a * E[X]? {np.isclose(E_aX, a_E_X)}")
```

**N-ésima potencia de una variable aleatoria**

```python
# Definimos los valores y probabilidades de X
valores_X = np.array([1, 2, 3])
probabilidades_X = np.array([0.2, 0.5, 0.3])

# Definimos el exponente n
n = 2

# Cálculo del valor esperado de X^n
E_X_n = np.sum((valores_X ** n) * probabilidades_X)
print(f"Valor esperado E[X^{n}]: {E_X_n}")
```

**Linealidad del Valor Esperado**

```python
# Definimos los valores y probabilidades de X
valores_X = np.array([1, 2, 3])
probabilidades_X = np.array([0.3, 0.4, 0.3])

# Definimos los valores y probabilidades de Y
valores_Y = np.array([4, 5, 6])
probabilidades_Y = np.array([0.2, 0.5, 0.3])

# Constantes
a = 2
b = 3

# Cálculo del valor esperado de X y Y
E_X = np.sum(valores_X * probabilidades_X)
E_Y = np.sum(valores_Y * probabilidades_Y)

# Cálculo del valor esperado de aX + bY
E_aX_bY = a * E_X + b * E_Y

# Verificación de la propiedad de linealidad
# Primero calculamos aE[X] + bE[Y]
valor_esperado_linealidad = a * E_X + b * E_Y

print(f"Valor esperado E[X]: {E_X}")
print(f"Valor esperado E[Y]: {E_Y}")
print(f"Valor esperado E[aX + bY]: {E_aX_bY}")
print(f"Verificación de la linealidad: {np.isclose(E_aX_bY, valor_esperado_linealidad)}")
```

**Correspondencia entre el Valor Esperado y la Varianza**

```python
# Definimos los valores y probabilidades de X
valores_X = np.array([1, 2, 3])
probabilidades_X = np.array([0.2, 0.5, 0.3])

# Cálculo del valor esperado E[X]
E_X = np.sum(valores_X * probabilidades_X)

# Cálculo del valor esperado E[X^2]
E_X2 = np.sum((valores_X ** 2) * probabilidades_X)

# Cálculo de la varianza Var(X)
varianza_X = E_X2 - (E_X ** 2)

# Cálculo de la varianza considerando las probabilidades
var_X_correcta = np.sum(probabilidades_X * (valores_X - E_X)**2)

print(f"Valor esperado E[X]: {E_X}")
print(f"Valor esperado E[X^2]: {E_X2}")
print(f"Varianza Var(X): {varianza_X}")
print(f"Varianza Var(X) tradicional (correcta): {var_X_correcta}")
```

**Escalabilidad de la Varianza**

```python
# Definimos los valores y probabilidades de X
valores_X = np.array([1, 2, 3])
probabilidades_X = np.array([0.2, 0.5, 0.3])

# Cálculo del valor esperado E[X]
E_X = np.sum(valores_X * probabilidades_X)

# Cálculo de la varianza Var(X)
varianza_X = np.sum(probabilidades_X * (valores_X - E_X)**2)

# Definimos una constante a
a = 3

# Cálculo del valor esperado E[aX]
E_aX = a * E_X

# Cálculo de la varianza Var(aX)
varianza_aX = np.sum(probabilidades_X * ((a * valores_X) - E_aX)**2)

# También podemos usar la propiedad de escalabilidad directamente
varianza_aX_teorica = (a ** 2) * varianza_X

print(f"Valor esperado E[X]: {E_X}")
print(f"Varianza Var(X): {varianza_X}")
print(f"Valor esperado E[aX]: {E_aX}")
print(f"Varianza Var(aX) (cálculo directo): {varianza_aX}")
print(f"Varianza Var(aX) (usando propiedad): {varianza_aX_teorica}")
```

**Calculo de la Covarianza Entre Dos Variables**

```python
# Definimos los valores y probabilidades de X e Y
valores_X = np.array([1, 2, 3])
valores_Y = np.array([4, 5, 6])
probabilidades_conjuntas = np.array([0.1, 0.4, 0.5])  # P(X, Y)

# Cálculo de los valores esperados E[X] y E[Y]
E_X = np.sum(valores_X * probabilidades_conjuntas)
E_Y = np.sum(valores_Y * probabilidades_conjuntas)

# Cálculo de la covarianza Cov(X, Y)
covarianza_XY = np.sum(probabilidades_conjuntas * (valores_X - E_X) * (valores_Y - E_Y))

print(f"Valor esperado E[X]: {E_X}")
print(f"Valor esperado E[Y]: {E_Y}")
print(f"Covarianza Cov(X, Y): {covarianza_XY}")
```

**Varianza de una Combinación Lineal de Variables Aleatorias**

```python
# Definimos los valores y probabilidades de X e Y
valores_X = np.array([1, 2, 3])
valores_Y = np.array([4, 5, 6])
probabilidades_conjuntas = np.array([0.1, 0.4, 0.5])  # P(X, Y)

# Definimos los coeficientes de la combinación lineal
a = 2
b = 3

# Cálculo de los valores esperados E[X] y E[Y]
E_X = np.sum(valores_X * probabilidades_conjuntas)
E_Y = np.sum(valores_Y * probabilidades_conjuntas)

# Cálculo de la varianza de X
varianza_X = np.sum(probabilidades_conjuntas * (valores_X - E_X) ** 2)

# Cálculo de la varianza de Y
varianza_Y = np.sum(probabilidades_conjuntas * (valores_Y - E_Y) ** 2)

# Cálculo de la covarianza Cov(X, Y)
covarianza_XY = np.sum(probabilidades_conjuntas * (valores_X - E_X) * (valores_Y - E_Y))

# Cálculo de la varianza de la combinación lineal Z = aX + bY
varianza_Z = a**2 * varianza_X + b**2 * varianza_Y + 2 * a * b * covarianza_XY

print(f"Valor esperado E[X]: {E_X}")
print(f"Valor esperado E[Y]: {E_Y}")
print(f"Varianza Var(X): {varianza_X}")
print(f"Varianza Var(Y): {varianza_Y}")
print(f"Covarianza Cov(X, Y): {covarianza_XY}")
print(f"Varianza de Z = {a}X + {b}Y: {varianza_Z}")
```

---
Sampleo de Distribuciones Discretas

---

Para distribuciones discretas, como una distribución de probabilidad de una variable aleatoria discreta, el proceso para generar muestras que repliquen la distribución original es el siguiente:

1. **Definir la Distribución**:
   - Especifica los valores posibles que la variable aleatoria puede tomar y sus probabilidades asociadas. La suma de todas las probabilidades debe ser 1.

2. **Construir la CDF (Función de Distribución Acumulada)**:
   - La CDF para una distribución discreta se construye sumando las probabilidades acumuladas hasta cada punto. Si $ p(x_i) $ es la probabilidad de cada valor $ x_i $, entonces:
     $$
     \text{CDF}(x_i) = \sum_{j \leq i} p(x_j)
     $$

3. **Generar una Muestra Aleatoria**:
   - Genera un número aleatorio uniforme $u$ entre 0 y 1.
   - Usa este número para determinar el valor de la variable aleatoria. Encuentra el primer valor $ x_i $ para el cual $ \text{CDF}(x_i) $ es mayor o igual a $ u $.

4. **Repetir**:
   - Repite el proceso para obtener tantas muestras como se desee.





A continuación, se ilustra el protocolo para generar samples a partir de una distribución de probabilidad dada.

```python
# Definimos la distribución discreta
valores = [1, 2, 3, 4]
probabilidades = [0.1, 0.3, 0.4, 0.2]  # Probabilidades de cada valor

# Creamos una función de sampleo a partir de probabilidades y clases dadas
def sample_discreta(pmf, valores, num_samples):
    samples = []
    cdf = np.cumsum(pmf)
    for _ in range(num_samples):
        r = np.random.uniform(0, 1)  # Genera un número aleatorio uniformemente distribuido
        # Encuentra el índice adecuado 
        idx = 0
        while idx < len(cdf) and r > cdf[idx]:
            idx += 1
        samples.append(valores[idx])
    return samples

# Generar muestras
num_samples = 10000
samples = sample_discreta(probabilidades, valores, num_samples)

# Imprimir las primeras muestras generadas
print("Muestras generadas (discretas):", samples[:10])

# Calcular probabilidades empíricas
frecuencias = Counter(samples)
total_muestras = len(samples)
probabilidades_reconstruidas = {k: v / total_muestras for k, v in frecuencias.items()}
probabilidades_reconstruidas = dict(sorted(probabilidades_reconstruidas.items()))

# Observamos los resultados generados
print("Probabilidades originales:", probabilidades_reconstruidas)
print("Probabilidades reconstruidas:", probabilidades_reconstruidas)

# Graficamos el experimento
comparador_bar(valores, probabilidades, probabilidades_reconstruidas.values(), 
               'Generación de muestras a partir de una distribución', 'Distribución Original', 'Distribución Experimental')
```

Sampleo de Distribuciones Contínuas

---

Para distribuciones contínuas, el proceso para generar muestras que repliquen la distribución original es el siguiente:

1. **Definir la Densidad de Probabilidad**:
   - Especifica la función de densidad de probabilidad a partir de la cuál se quieren generar los samples.

2. **Construir la CDF (Función de Distribución Acumulada)**:
   - La CDF para una distribución contínua se construye sumando las probabilidades acumuladas a lo largo del rango de la variable.
       $$
      F(x) = P(X \leq x) = \int_{-\infty}^{x} f_X(t) \, dt 
      $$

3. **Generación de Muestras Uniformes**

   Generamos una muestra de números aleatorios que siguen una distribución uniforme en el intervalo $[0, 1]$. Estos números actuarán como los valores de probabilidad que se transformarán para ajustarse a la distribución deseada.

4. **Transformación Usando la CDF**

   Para cada número generado uniformemente $u$ en el intervalo $[0, 1]$, se utiliza la inversa de la función de distribución acumulada (CDF) para encontrar el valor correspondiente en la distribución deseada (del mísmo modo en el que se hallan los percentíles). Esto se basa en la idea de que, si $U$ es una variable aleatoria uniforme en $[0, 1]$ y $X$ es una variable aleatoria con CDF $F$, entonces $X$ tiene la misma distribución que $F^{-1}(U)$. 


A continuación, ilustramos el proceso de generar samples a partir de una distribución deseada contínua. No obstante note que en ésta ocación empleamos el método `.ppf` de la distribución contínua generada con la librería `scipy.stats` debido a que éste método nos permíte traducir directamente un valor de percentil `p` en un valor de `X` ya que calcula la inversa de la CDF. Método computacionalmente eficiente. Para ilustrar el método, emplearemos la distribución normal.

```python
# Definimos la distribución continua (Normal en este caso)
mu = 0
sigma = 1
distribution = stats.norm(mu, sigma)  # Distribución normal estándar

# Creamos una función de sampleo a partir de una densidad dada
def sample_continua(distribution, num_samples):
    samples = []
    for _ in range(num_samples):
        u = np.random.uniform(0, 1)  # Genera un número aleatorio uniformemente distribuido en [0, 1]
        x = distribution.ppf(u)  # Obtiene el valor correspondiente usando la inversa de la CDF
        samples.append(x)
    return samples

# Generar muestras
num_samples = 10000
samples = sample_continua(distribution, num_samples)

# Imprimir las primeras muestras generadas
print("Muestras generadas (continuas):", samples[:10])

# Calcular el histograma de las muestras generadas
plt.figure(figsize=(8, 5))
plt.hist(samples, bins=30, density=True, color='green', edgecolor='black', alpha=0.7, label='Distribución Experimental')

# Graficar la distribución teórica
xmin, xmax = min(samples), max(samples)
x = np.linspace(xmin, xmax, 100)
p = distribution.pdf(x)
plt.plot(x, p, 'k', linewidth=2, label='Distribución Teórica')

plt.title('Generación de muestras a partir de una distribución continua')
plt.xlabel('X')
plt.ylabel('Densidad')
plt.legend()
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```
