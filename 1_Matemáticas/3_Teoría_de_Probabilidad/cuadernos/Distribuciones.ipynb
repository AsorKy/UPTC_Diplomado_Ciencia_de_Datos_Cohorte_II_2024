Variables Aleatorias Discretas

---

Una **variable aleatoria discreta** es una variable que puede tomar un conjunto finito o contable de valores posibles, cada uno con una cierta probabilidad asociada. Estas variables son el resultado de experimentos aleatorios donde los resultados pueden ser contados claramente, como el nmero de veces que un evento ocurre en una serie de ensayos o el valor que toma una tirada de dado.

Una variable aleatoria discreta se caracteriza por su **funcin de probabilidad** (tambin conocida como funcin de masa de probabilidad, **PMF**), que asigna a cada valor posible de la variable una probabilidad. La importancia y utilidad de las variables aleatorias consste en que nos permten modelar enteramente el experimento aleatorio y nos facilita el entender las propiedades del sistema de inters.

Para ilustrar que es una variable aleatoria, observemos los siguientes ejemplos donde teniendo una moneda justa, nuestr variable aleatoria se definir como *"X = Nmero de caras que resultan de lanzar una moneda N veces"*. 

```python
# Definir los posibles resultados de un lanzamiento de moneda
outcomes = ['H', 'T']  # H para "Cara" (Heads), T para "Cruz" (Tails)

# Generar el espacio muestral para tres lanzamientos
sample_space = list(product(outcomes, repeat=3))

# Definir la variable aleatoria X como el nmero de "caras" (H) en tres lanzamientos
def count_heads(trial):
    return trial.count('H')

# Calcular la funcin de probabilidad (PMF)
pmf = Counter(count_heads(trial) for trial in sample_space)

# Normalizar las frecuencias para obtener probabilidades
total_outcomes = len(sample_space)
for x in pmf:
    pmf[x] /= total_outcomes

# Mostrar la funcin de probabilidad (PMF)
print("Variable Aleatoria X: Nmero de Caras en Tres Lanzamientos de Moneda")
print("Espacio muestral = ", sample_space)

for outcome, probability in sorted(pmf.items()):
    print(f"P(X = {outcome}) = {probability:.2f}")
    
# Generamos el grfico de la distribucin PMF
graficar_distribucion(list(range(1, len(pmf) + 1)), pmf.values(), title='Probabilidad de # caras en tres lanzamientos', x_label='X', y_label='P(X)')
```

Note que las ateriores distribuciones para nuestra variable aleatoria han sido generadas a partir de generar el espacio muestral de la variable, por otra parte, es posible llegar a los mismos resultados si realizamos el experimento aleatorio como tal (lanzando una moneda mltiples veces) y deducir la distribucin PMF a partir del conteo de los resultados generados por el experimento.

```python
# Definir la funcin para simular el lanzamiento de la moneda
def simulate_coin_tosses(n_tosses):
    outcomes = ['H', 'T']
    return [random.choice(outcomes) for _ in range(n_tosses)]

# Definir la funcin para contar las "caras" en un experimento
def count_heads(tosses):
    return tosses.count('H')

# Nmero de experimentos y lanzamientos por experimento
num_experiments = 1000
n_tosses = 10

# Realizar los experimentos y contar las "caras"
results = [count_heads(simulate_coin_tosses(n_tosses)) for _ in range(num_experiments)]

# Calcular la funcin de probabilidad (PMF) basada en los resultados de los experimentos
pmf = Counter(results)
total_experiments = num_experiments

# Normalizar para obtener probabilidades
for x in pmf:
    pmf[x] /= total_experiments

# Mostrar la funcin de probabilidad (PMF)
print("Variable Aleatoria X: Nmero de Caras en Tres Lanzamientos de Moneda (1000 Experimentos)")
for outcome, probability in sorted(pmf.items()):
    print(f"P(X = {outcome}) = {probability:.2f}")


# Generamos el grfico de la distribucin PMF
graficar_distribucion(pmf.keys(), pmf.values(), x_label='X', y_label='P(X)', title='Probabilidad de # caras en tres lanzamientos')
```

Valores Esperados y Varianzas

---

Teniendo en cuenta que la distribucin de probabilidad de una variable aleatoria $X$ viene dada por.

$$
g_X (x_i) = P(X = x_i)
$$

El **valor esperado** de una variable aleatoria discreta es una medida de la centralidad de su distribucin de probabilidad. Se calcula como la suma de todos los posibles valores de la variable aleatoria, cada uno multiplicado por su probabilidad. Esencialmente, representa la media ponderada de los valores de la variable aleatoria, la cual es equivalente al *promedio* $\mu$ de la variable.

Matemticamente, para una variable aleatoria discreta $X$ con posibles valores $x_i$ y probabilidades asociadas $P(x_i)$, el valor esperado $E(X)$ se define como:

$$
E(X) = \sum_{i} x_i P(x_i) = \mu _{X}
$$

Por otra parte, la **varianza** de una variable aleatoria discreta es una medida de la dispersin de sus valores respecto al valor esperado. Indica cunto varan los valores de la variable aleatoria en torno a su media. La varianza se calcula como la suma de los cuadrados de las diferencias entre cada valor y el valor esperado, ponderada por la probabilidad de cada valor.

Matemticamente, para una variable aleatoria discreta $X$, la varianza $\text{Var}(X)$ se define como:

$$
\text{Var}[X] = E[(X - E[X])^2]
$$

Esto se puede expandir como:

$$
\text{Var}[X] = \sum_{i} (x_i - E[X])^2 P(x_i)
$$

Donde la **variacin estndard** es una mtrica de la dispersin de la variable definida en las msmas unidades de sta, y corresponde simplemente a la raiz cuadrada de la varianza:

$$
\sigma_X = \sqrt{\text{Var}[X]}
$$


Para ilustrar estos conceptos, calcularemos el valor esperado y la varianza de la variable $X$ la cual corresponde al nmero de caras posibles en 10 lanzamientos de una moneda.

```python
# Simulacin de 1000 experimentos con 10 lanzamientos de moneda
num_experiments = 1000
n_tosses = 10

# Funcin para simular lanzamientos de moneda y contar "caras"
def simulate_coin_tosses(n_tosses):
    outcomes = ['H', 'T']
    return [random.choice(outcomes) for _ in range(n_tosses)]

def count_heads(tosses):
    return tosses.count('H')

# Realizar experimentos y contar resultados
results = [count_heads(simulate_coin_tosses(n_tosses)) for _ in range(num_experiments)]

# Calcular la PMF
pmf = Counter(results)
total_experiments = num_experiments
for x in pmf:
    pmf[x] /= total_experiments

# Clculo del valor esperado (E(X))
expected_value = sum(x * p for x, p in pmf.items())

# Clculo de la varianza (Var(X))
variance = sum((x - expected_value) ** 2 * p for x, p in pmf.items())

# Resultados
print(f"Valor Esperado (E(X)): {expected_value:.2f}")
print(f"Varianza (Var(X)): {variance:.2f}")
```

Distribucin de Probabilidad Discreta

---

Una **distribucin de probabilidad discreta** es una funcin que asigna probabilidades a los valores posibles de una variable aleatoria discreta. Una variable aleatoria discreta es aquella que puede tomar un nmero finito o infinitamente contable de valores especficos, a diferencia de una variable aleatoria continua que puede tomar cualquier valor en un intervalo continuo. Las distribuciones discretas poseen las siguientes caractersticas:

1. **Lista de Valores Posibles**: La distribucin especifica todos los valores posibles que puede tomar la variable aleatoria.

2. **Probabilidades Asociadas**: A cada valor posible de la variable aleatoria se le asigna una probabilidad, que indica la frecuencia con la que se espera que ocurra ese valor en un gran nmero de experimentos.

3. **Suma de Probabilidades**: La suma de todas las probabilidades en una distribucin de probabilidad discreta es igual a 1.

Matemticamente, si $X$ es una variable aleatoria discreta con valores posibles $x_1, x_2, \ldots, x_n$ y sus respectivas probabilidades $p(x_1), p(x_2), \ldots, p(x_n)$, entonces la distribucin de probabilidad se define como:

$$
P(X = x_i) = p(x_i)
$$


Dentro de las distribuciones de probabilidad discretas ms importantes encontramos:

1. **Distribucin de Bernoulli**: Describe la probabilidad de xito de un nico ensayo de Bernoulli.

2. **Distribucin Binomial**: Describe el nmero de xitos en una serie de ensayos independientes de Bernoulli.

3. **Distribucin Binomial Negativa**: Describe el nmero de ensayos necesarios para obtener un nmero fijo $r$ de xitos en una serie de ensayos de Bernoulli independientes.

4. **Distribucin de Poisson**: Describe el nmero de eventos que ocurren en un intervalo fijo de tiempo o espacio.

5. **Distribucin Geomtrica**: Describe el nmero de ensayos necesarios para obtener el primer xito en una serie de ensayos de Bernoulli.

6. **Distribucin Uniforme Discreta**: Cada valor en el conjunto tiene la misma probabilidad de ocurrir.


Supongamos que estamos lanzando un dado justo. La variable aleatoria $X$ representa el nmero que aparece en el dado. Los valores posibles de $X$  son 1, 2, 3, 4, 5 y 6. La distribucin de probabilidad discreta para este caso es:

$$
P(X = x) =
\begin{cases}
\frac{1}{6} & \text{si } x = 1, 2, 3, 4, 5, 6 \\
0 & \text{en caso contrario}
\end{cases}
$$

En este ejemplo, cada valor tiene una probabilidad de $\frac{1}{6}$, y la suma de todas las probabilidades es 1.

Las distribuciones de probabilidad son fundamentales en estadstica y probabilidad porque proporcionan una forma estructurada de describir la incertidumbre y la variabilidad en eventos y datos. Su importancia y utilidad abarcan varias reas clave:

**1. Descripcin y Modelado de Fenmenos Aleatorios**: Las distribuciones de probabilidad permiten modelar fenmenos aleatorios y describir la distribucin de datos observados. Por ejemplo, en la vida real, podemos usar distribuciones para modelar cosas como el nmero de llamadas que recibe un centro de atencin al cliente en una hora (distribucin de Poisson) o el tiempo entre llegadas de autobuses (distribucin exponencial).

**2. Clculo de Probabilidades**: Permiten calcular la probabilidad de que ocurra un evento especfico. Por ejemplo, con una distribucin normal, podemos calcular la probabilidad de que una variable aleatoria caiga dentro de un rango determinado, lo que es til en muchas aplicaciones, como el control de calidad o la gestin de riesgos financieros.

**3. Toma de Decisiones Informadas**: Las distribuciones de probabilidad son esenciales para la toma de decisiones bajo incertidumbre. En economa, finanzas, y muchos otros campos, la probabilidad de eventos futuros se utiliza para tomar decisiones que minimicen riesgos y maximicen beneficios.

**4. Estimacin y Prediccin**: Permiten hacer inferencias sobre poblaciones con base en muestras. En estadstica inferencial, las distribuciones de probabilidad ayudan a estimar parmetros desconocidos de una poblacin (como la media o la varianza) y a realizar predicciones.

**5. Inferencia Estadstica y Pruebas de Hiptesis**: Las distribuciones de probabilidad son la base para realizar pruebas de hiptesis. Por ejemplo, se pueden utilizar para determinar si una muestra proviene de una poblacin con una media especfica o si dos muestras difieren significativamente.

**6. Medicin de la Incertidumbre**: Ofrecen una manera de cuantificar la incertidumbre. La varianza y la desviacin estndar, derivadas de una distribucin de probabilidad, son medidas comunes de la dispersin y la incertidumbre en los datos.

**7. Simulacin y Modelado Computacional**: Las distribuciones de probabilidad son cruciales en simulaciones y modelos computacionales, que se utilizan para predecir y entender sistemas complejos. Por ejemplo, en la ingeniera y las ciencias naturales, las simulaciones de Monte Carlo usan distribuciones de probabilidad para modelar y analizar sistemas que son demasiado complejos para resolver de forma analtica.

**8. Comunicacin de Resultados**: Las distribuciones de probabilidad facilitan la comunicacin de resultados estadsticos. Al describir la forma de una distribucin (simetra, curtosis, etc.), se puede dar una idea clara de cmo se distribuyen los datos y qu esperar de ellos.

Distribucin de Probabilidad Acumulada Discreta

---

Una **distribucin de probabilidad acumulada** (CDF, por sus siglas en ingls) es una funcin que describe la probabilidad de que una variable aleatoria tome un valor menor o igual a un cierto nmero. Formalmente, para una variable aleatoria $X$, la CDF $F(x)$ se define como:

$$
F(x) = P(X \leq x)
$$

donde $P(X \leq x)$ representa la probabilidad de que $X$ tome un valor menor o igual a $x$. Este tipo de distribuciones tienen las siguientes caractersticas:

1. **Montona no Decreciente**: La CDF siempre aumenta o se mantiene constante a medida que aumenta $x$.

2. **Rango de Valores**: La CDF toma valores en el intervalo [0, 1]. Es decir, $F(x)$.

3. **Punto de Inicio y Fin**: Para una variable aleatoria continua, $F(x)$ se aproxima a 0 cuando $x$ tiende a $-\infty$ y se aproxima a 1 cuando $x$ tiende a $\infty$.

4. **Uso en Probabilidad**: La CDF se utiliza para determinar la probabilidad de que una *variable aleatoria est dentro de un cierto rango*. Por ejemplo, $ P(a \leq X \leq b) = F(b) - F(a) $.

5. **Aplicaciones**: La CDF es til en estadstica y probabilidad para resumir y describir la distribucin de una variable aleatoria. Tambin se utiliza en pruebas de hiptesis y anlisis de datos.

En trminos matemticos, la distribucin de probabilidad acumulada se determina de la siguiente manera, donde $R_X$ es el rango de la variable aleatoria:

$$
F(x) = \sum _{x_i \in R_X, x_i <= x } g_{X} (x_i)

Distribucin de Bernoulli

---

**Definicin**: La distribucin de Bernoulli es una distribucin de probabilidad discreta que describe el resultado de *un solo experimento con dos posibles resultados*: xito (generalmente representado por 1) y fracaso (generalmente representado por 0). Es la distribucin ms simple y bsica en probabilidad y estadstica.

**Parmetros**:
- **\( p \)**: La probabilidad de xito (el evento que se define como 1).
- **\( 1 - p \)**: La probabilidad de fracaso (el evento que se define como 0).

**Funcin de Masa de Probabilidad (PMF)**:
$$
P(X = x) = 
\begin{cases} 
p & \text{si } x = 1 \\
1 - p & \text{si } x = 0 
\end{cases}
$$

La distribucin de Bernoulli nos permite **modelar de experimentos binarios** los cules son aquellos experimentos que tienen exactamente dos posibles resultados. Por ejemplo, lanzar una moneda (cara o cruz), verificar si un producto pasa o falla una prueba de calidad, etc. Esta distribucin corresponde la base para distribuciones ms complejas tales como la distribucin binomial. Un experimento binomial es simplemente una serie de experimentos de Bernoulli.

Algunas de las pripiedades ms importantes de sta distribucin son:

**1. Media (Valor Esperado)**: La media o valor esperado de una variable aleatoria $X$ con distribucin de Bernoulli, donde $p$ es la probabilidad de xito, se calcula como:
$$
\mathbb{E}[X] = p 
$$

**2. Varianza**: Para la distribucin de Bernoulli, la varianza se calcula como:
$$
\text{Var}(X) = p \cdot (1 - p) 
$$

**3. Desviacin Estndar**: La desviacin estndar para la distribucin de Bernoulli, se calcula como:
$$
\sigma = \sqrt{p \cdot (1 - p)} 
$$

**6. Funcin de Distribucin Acumulada (CDF)**: La funcin de distribucin acumulada para la distribucin de Bernoulli es:
$$
F(x) =
\begin{cases} 
0 & \text{si } x < 0 \\
1 - p & \text{si } 0 \leq x < 1 \\
1 & \text{si } x \geq 1 
\end{cases}
$$

7. **Propiedad de Independencia**: La distribucin de Bernoulli es una distribucin binomial especial con $n = 1$. La suma de $n$ variables aleatorias independientes con distribucin de Bernoulli (con la misma probabilidad $p$) sigue una distribucin binomial con parmetros $n$ y $p$.

**<h3 id="subsection123" style="color: #003366;">1.2.3. Distribucion Binomial</h3>**

---
Distribucin Binomial

---

La distribucin binomial es una distribucin de probabilidad discreta que describe el *nmero de xitos en una secuencia fija de experimentos independientes*, cada uno de los cuales tiene dos resultados posibles (xito o fracaso). Es una extensin natural de la distribucin de Bernoulli a mltiples ensayos. Si una variable aleatoria $X$ sigue una distribucin Binomial, se denota como $ X \sim \text{Binomial}(n, p) $, donde:
- $ n $ es el nmero de ensayos (o pruebas) independientes.
- $ p $ es la probabilidad de xito en cada ensayo.
- $ 1 - p $ es la probabilidad de fracaso en cada ensayo.

La funcin de masa de probabilidad (PMF) de una distribucin Binomial se define como:

$$
P(X = k) = \binom{n}{k} p^k (1 - p)^{n - k} 
$$

donde $ k $ es el nmero de xitos deseados $0   k    n $. Dentro de sus propiedades estadsticas encontramos:

**1. Media (Valor Esperado)**: La media o valor esperado de una variable aleatoria Binomial $ X $ es:
$$
   \mathbb{E}[X] = n \cdot p 
$$

**2. Varianza**: La varianza de una variable aleatoria Binomial $ X $ es:
$$
   \text{Var}(X) = n \cdot p \cdot (1 - p) 
$$

**3. Desviacin Estndar**: La desviacin estndar es la raz cuadrada de la varianza:
$$
   \sigma = \sqrt{n \cdot p \cdot (1 - p)} 
$$

Pero cabe preguntarse, por que la frmula matemtica de sta distribucin tiene esta forma?, para ello exploraremos los siguientes conceptos:

**Coeficiente Binomial**

El coeficiente binomial, **\binom{n}{k}**, es una parte fundamental de la frmula de la distribucin binomial. Representa el *nmero de maneras diferentes de seleccionar $ k $ xitos (o elementos) de un total de $ n $ ensayos*, es decir, mide la cantidad de *combinaciones de tamao $k$ para un para $n$ lanzamientos*.El coeficiente binomial se calcula como:

$$
\binom{n}{k} = \frac{n!}{k! \cdot (n - k)!} 
$$

donde $n!$ (n factorial) es el producto de todos los enteros positivos hasta $n$ y representa la cantidad de *permutaciones ordinarias posibles dentro de $n$ lanzamientos* (una permutacin ordinaria es aquella secuencia ordenada de elementos que tiene el msmo tamao al espacio muestral, es decir $p(r=n;n)= n!$), $k!$ es el factorial de $k$, y $(n - k)!$ es el factorial de $( n - k )$.

**Justificacin Matemtica**:

- **Permutaciones Totales Posibles**: Para $n$ elementos, hay $n!$ maneras de permutarlos ordinariamente.
- **Seleccin de $k$ xitos**: De estas $ n! $ permutaciones, queremos seleccionar solo aquellos en los que hay exactamente $ k $ son xitos. 
- **Divisin por Permutaciones Internas**:
  - Dentro de los $ k $ xitos, las permutaciones entre ellos no cambian el resultado final (es decir, son indistinguibles).
  - Dentro de los $( n - k )$ fracasos, las permutaciones tambin son indistinguibles.
- **Ajuste**: Por lo tanto, el nmero de maneras de seleccionar $ k $ xitos (o cualquier subconjunto de tamao $ k $) de $ n $ es:
  $ \frac{n!}{k! \cdot (n - k)!} $

Para una secuencia especfica con exctamente $k$ xitos y $n-k$ fracasos, la probabilidad de que ocurran dicho nmero de xitos es igual al *producto de las probabilidades de ocurrencia de cada ensayo posible* ya que cada evento es *independiente uno del otro*. As, si tenemos $k$ xitos, la probabilidad de esto es igual a $p^k$ mientras que la probabilidad de tener  $(n-k)$ es $p^{n-k}$. As, la proabilidad de describir un escenario con $k$ ensayos y $n-k$ fracasos es simplemente $p^k(1-p)^{n-k}$. 

Dado que hay \binom{n}{k} secuencias diferentes que pueden resultar en exactamente $k$ xitos, la probabilidad total de tener exactamente $k$ xitos es el producto de las probabilidades de todas esas secuencias.

Note entonces, que la distribucin de Bernoulli puede ser recuperada a partir de la distribucin Binomial si $n = k =1$

$$
P(X = 1) =  p^1 (1 - p)^{n - 1} 
$$


```python
# Definicin de una distribucin binomial
def binomial(k,n,p):
    binom = factorial(n) / (factorial(k) * factorial(n-k))
    pmf = binom * (p**k) * (1-p)**(n-k)
    return pmf

# Calcular la PMF usando nuestra funcin
n = 10  # Nmero de ensayos
p = 0.3  # Probabilidad de xito
x = np.arange(0, n+1)
pmf = [binomial(k,n,p) for k in x]

# Generamos el grfico de la distribucin PMF
graficar_distribucion(x, pmf, x_label='X', y_label='P(X)', 
                      title=f'Distribucin Binomial para k y {n} Lanzamientos de Moneda')
```

Distribucin Binomial Negativa

---

La distribucin binomial negativa es una distribucin de probabilidad discreta que generaliza la distribucin Binomial. Es til para modelar el *nmero de ensayos necesarios hasta que ocurran un nmero fijo de xitos en una serie de experimentos de Bernoulli independientes*. Vamos a desglosar su definicin, uso, casos de utilidad y justificacin matemtica en detalle. La distribucin binomial negativa describe el nmero de fracasos (ensayos fallidos) que ocurren antes de alcanzar un nmero fijo $r$ de xitos en una secuencia de experimentos de Bernoulli independientes, donde cada experimento tiene una probabilidad $p$ de xito. Posee los siguientes parmetros

1. **$ r $**: Nmero de xitos deseados.
2. **$p $**: La probabilidad de xito en cada ensayo.

El nmero de fracasos $ k $ sigue una distribucin Binomial negativa si la probabilidad de obtener exactamente $ k $ fracasos antes de lograr $ r $ xitos est dada por la funcin de masa de probabilidad:

$$
 P(X = k) = \binom{k + r - 1}{r - 1} (1 - p)^k p^r 
$$

donde $ \binom{k + r - 1}{r - 1} $ es el coeficiente binomial que representa el nmero de formas en que se pueden distribuir los $ k $ fracasos entre los $ r $ xitos.

**Justificacin Matemtica**

La funcin de masa de probabilidad (PMF) de la distribucin Binomial negativa puede derivarse a partir de una serie de experimentos de Bernoulli independientes. La probabilidad de que se necesiten exactamente $k$ fracasos antes de obtener $ r $ xitos se puede calcular considerando las siguientes observaciones:

- **Nmero Total de Ensayos**: Para obtener $ r $ xitos, necesitamos realizar $ k + r $ ensayos en total.
- **Posiciones de los xitos**: Elegimos $ r $ posiciones para los xitos entre los $ k + r $ ensayos. El resto de los ensayos sern fracasos.
- **Probabilidad de xitos y Fracasos**: Cada uno de estos $ k + r $ ensayos tiene una probabilidad $ p $ de xito y $ 1 - p $ de fracaso.

El nmero de formas en que los $ k $ fracasos pueden ser distribuidos en los $ k + r - 1 $ ensayos previos a la obtencin del $ r $-simo xito es dado por $ \binom{k + r - 1}{r - 1} $. Luego, multiplicamos esto por la probabilidad de tener exactamente $ k $ fracasos y $ r $ xitos en los $ k + r $ ensayos. 


**Propiedades Estadsticas de la Distribucin de Geomtrica**

Algunas de las propiedades estadsticas ms importantes de la distribucin Binomial negativa son:


**1. Esperanza (Media):**
$$
E[X] = \frac{r(1 - p)}{p} 
$$

**2. Varianza:**
$$
\text{Var}(X) = \frac{r(1 - p)}{p^2} 
$$

**Casos de Uso y Aplicaciones**

Algunas de las aplicaciones de la distribucin Binomial Negativa son:

- **Tiempo hasta un Nmero Fijo de xitos**: La distribucin binomial negativa es til para modelar situaciones en las que estamos interesados en el nmero de intentos necesarios hasta alcanzar un nmero especfico de xitos. Por ejemplo, el nmero de fracasos antes de lograr una cantidad determinada de ventas en un modelo de ventas.

- **Modelado de Datos Contables**: En finanzas y contabilidad, puede ser til para modelar el nmero de errores o problemas que ocurren antes de alcanzar un objetivo de calidad.

- **Nmero de Pruebas Clnicas**: Puede ser usada para modelar el nmero de pruebas fallidas hasta que se encuentren un nmero fijo de casos positivos en estudios mdicos.


- **Anlisis de Campaas Publicitarias**: Permite modelar el nmero de intentos publicitarios necesarios para lograr una cantidad fija de conversiones o ventas.


```python
# Definicin de una distribucin binomial negativa
def binomial_negativa(k, r, p):
    binom = factorial(k + r - 1) / (factorial(r - 1) * factorial(k))
    pmf = binom * (p**r) * ((1 - p)**k)
    return pmf

# Calcular la PMF usando nuestra funcin
r = 5  # Nmero de ensayos
k_max = 20 # Nmero mximo de fracasos
p = 0.4  # Probabilidad de xito
k = np.arange(0, k_max + 1)
pmf = [binomial_negativa(k_val,r,p) for k_val in k]

# Generamos el grfico de la distribucin PMF
graficar_distribucion(k, pmf, x_label='X', y_label='P(X)', 
                      title=f'Distribucin Negativa Binomial para \n k fracasos y  {r} xitos con probabilidad {p}')
```

  donde $k$ es el nmero de ensayos hasta el primer xito.

- **Modelos de espera**: En situaciones donde se espera un resultado exitoso en una serie de intentos, como en juegos de azar o pruebas de productos.
- **Simulacin de eventos**: Para modelar el nmero de intentos necesarios antes de que ocurra un evento especfico en experimentos repetidos.

**<h3 id="subsection125" style="color: #003366;">1.2.5. Distribucion de Poisson</h3>**

---
Distribucin de Poisson

---

Un **proceso de Poisson** es un modelo estocstico que describe eventos que ocurren de manera aleatoria e independiente en el tiempo o el espacio. Es ampliamente utilizado en la teora de colas, telecomunicaciones, economa y otros campos para modelar la llegada de eventos en un intervalo de tiempo o regin espacial.

Una distribucin de Poisson es una distribucin de probabilidad discreta que describe el *nmero de eventos que ocurren en un intervalo de tiempo $t$ (o de espacio) fijo*. Dichos eventos son de tipo Bernoulli en donde la situacin puede modelarse bajo la premisa sencilla de *"ocurre un evento"* o *"el evento no ocurrio"*. Un proceso de Poisson posee las siguientes caractersticas:


1. **Eventos Aleatorios e Independientes**: Los eventos ocurren de manera aleatoria e independiente unos de otros.
2. **Estacionariedad**: La tasa a la que ocurren los eventos es constante en el tiempo. Es decir, la probabilidad de que ocurra un evento en un intervalo de tiempo dado solo depende de la longitud del intervalo a considerar mas no del momento especfico. El contnuo *t* puede ser dividido en partes iguales $\delta t$ lo suficientemente pequeas como para que a lo sumo, uno de los eventos puntuales tenga lugar en dicho intervalo de tiempo. 
3. **No Acumulacin**: No hay dos eventos que ocurran exactamente al mismo tiempo.
4. **Taza de ocurrencia**: La probabilidad de que un evento puntual tenga lugar en un intervalo $\delta t$ depende de la longitud del intervalo y de una constante $\lambda > 0$ que caracterza al proceso, donde esta constante se denomina como la *taza promedio de eventos por unidad de tiempo*. 

$$
P(\text{Ocurrencia de un evento en intervalo de tiempo delta t}) = \lambda \delta t
$$

Note entonces que la probabilidad de ocurrencia de un evento no depende de la ubicacin del intervalo $\delta t$, por ende, podemos decir que las llegadas o eventos estn uniformemente distribuidas en el tiempo.

El proceso de Poisson caracterizado por una nica tasa de eventos promedio $\lambda$, representa el nmero promedio de eventos por unidad de tiempo. Si $N(t)$ es el nmero de eventos que ocurren en el intervalo de tiempo $[0, t]$, entonces $N(t)$ sigue una distribucin de Poisson con parmetro $\lambda t$. Este modelo probabilstico es equivalente a modelar la premisa de *obtener exactamente $k$ xitos en los $N$ experimentos independientes de Bernoulli que consisten en observar si ha habido o no una llegada del proceso en el intervalo de tiempo de longitud $t/N$*

La funcin de masa de probabilidad (PMF) para la variable *$N(t) = X$ = nmero de llegadas en el intervalo de tiempo $t$*, es:

$$
g_{X}(k;\lambda, t) = P(X = k) = \frac{(\lambda t)^k e^{-\lambda t}}{k!}
$$

donde:
- $k$ es el nmero de eventos.
- $\lambda$ es la tasa de eventos por unidad de tiempo.
- $t$ es el intervalo de tiempo considerado.


**Propiedades Estadsticas de la Distribucin de Poisson**

Dentro de las propiedades estadsticas ms importantes del proceso de Poisson encontramos:

1. **Esperanza (Media)**: La esperanza matemtica es $\lambda t$. Es el nmero promedio de eventos en el intervalo de tiempo $t$.

$$
E[X] = \lambda t
$$

2. **Varianza**: La varianza es tambin $\lambda t$, lo que significa que la media y la varianza son iguales.

$$
\text{Var}[X] = \lambda t
$$

3. **Desviacin Estndar**: La desviacin estndar es:

$$
\sigma _{X} = \sqrt{\lambda t}
$$



**Casos de Uso y Aplicaciones**

1. **Teora de Colas**: Modelar la llegada de clientes a una tienda, llamadas a un centro de atencin telefnica, o paquetes a un servidor.
2. **Telecomunicaciones**: Modelar la llegada de mensajes o llamadas en una red de comunicacin.
3. **Economa y Finanzas**: Modelar la ocurrencia de eventos financieros como cambios de precios, ventas de activos, etc.
4. **Ciencias Naturales**: Modelar la aparicin de mutaciones genticas, llegadas de partculas, entre otros.


Para ilustrar la aplicacin de la distribucin de Poisson, modelaremos la cantidad de llamadas que llegan a un centro de atencin al cliente en un intervalo de tiempo de $t = 60$ minutos, con una tasa promedio de $\lambda = 3$ llamadas por minuto. Utilizando una distribucin de Poisson, calculamos la probabilidad de recibir un cierto nmero de llamadas en ese intervalo de tiempo.

```python
# Definimos la funcin de Poisson
def poisson(k, lambda_, t):
    return (((lambda_ * t)**k) * np.exp(-lambda_*t)) / (factorial(k))

# Parmetros del problema
lambda_rate = 0.4  # Tasa de llamadas por minuto
t = 20  # Intervalo de tiempo en minutos

k_values = np.arange(0,16)
pmf = [poisson(k_val, lambda_rate, t) for k_val in k_values]


# Numero esperado de llamadas
print('Nmero esperado de llamadas en t = ', lambda_rate * t)
print('Varianza del nmero de llamadas en t = ', lambda_rate * t)

# Generamos el grfico de la distribucin PMF
graficar_distribucion(k_values, pmf, x_label='X = nmero de llamadas', y_label='P(X)', 
                      title=f'Distribucin Poisson con taza {lambda_rate} para t = {t}')
```

Variables Aleatorias Contnuas

---

Las **variables aleatorias continuas** son aquellas que pueden tomar un nmero infinito de valores dentro de un rango continuo. A diferencia de las variables aleatorias discretas, que se asocian con conteos de eventos y pueden tomar valores especficos y separados, las variables aleatorias continuas estn asociadas con mediciones y pueden tomar cualquier valor real dentro de un intervalo.

**Caractersticas de las Variables Aleatorias Continuas**

1. **Infinitos Valores Posibles**: Una variable aleatoria continua puede tomar un nmero infinito de valores dentro de un rango determinado. Por ejemplo, la temperatura en una ciudad o la altura de una persona pueden tomar cualquier valor dentro de un intervalo continuo dado.

2. **Funcin de Densidad de Probabilidad (PDF)**: En lugar de una funcin de masa de probabilidad (PMF) como en el caso de las variables discretas, las variables continuas se describen mediante una **funcin de densidad de probabilidad (PDF)**. La PDF, denotada como $f(X)$, **no** da la probabilidad de que la variable tome un valor especfico (lo cual es tcnicamente cero para variables continuas), sino que describe la *densidad de probabilidad por unidad de medida*. El hecho de que la probabilidad de un valor especfico sea igual a cero se debe a que si tomamos la lgica de las variables discretas, la probabilidad se calcula como casos favorables para el valor de la variable dividido por el nmero de casos posibles lo cul, para variables contnuas, es infinito.

3. **Probabilidad como rea Bajo la Curva**: La probabilidad de que una variable aleatoria continua tome un valor dentro de un intervalo especfico se calcula como el rea bajo la curva de la PDF en ese intervalo. Formalmente, si $X$ es una variable aleatoria continua, entonces la probabilidad de que $X$ est entre $a$ y $b$ es:

   $$
   P(a \leq X \leq b) = \int_{a}^{b} f(x) \, dx
   $$

4. **Funcin de Distribucin Acumulada (CDF)**: Similar a las variables discretas, las variables continuas tambin tienen una **funcin de distribucin acumulada (CDF)**, que da la probabilidad de que la variable aleatoria tome un valor menor o igual a $x$. La CDF de una variable aleatoria continua $X$ se define como:

   $$
   F(x) = P(X \leq x) = \int_{-\infty}^{x} f(t) \, dt
   $$

5. **Normalizacin de la probabilidad**: Al igual que toda probabilidad, la funcin de densidad de probabilidad es una funcin completamente normalizada, es decir:
   $$
   \int_{-\infty}^{\infty} f_X(x) \, dx = 1

Para ilustrar los conceptos anteriores, observemos cmo se comporta una variable aleatoria contnua normalmente distribuda (tipo de variable que estudiaremos ms a detalle en subsecciones siguientes) e intentaremos calcular la probabilidad de un rango de valores a partir de la PDF y la CDF.

```python
# Definimos la funcin de densidad de probabilidad (PDF) para una distribucin normal
mu, sigma = 0, 1  # media y desviacin estndar
dist_normal = stats.norm(mu, sigma)

# Generamos valores para la variable aleatoria continua
x = np.linspace(-4, 4, 1000)
pdf = dist_normal.pdf(x)

# Graficar la funcin de densidad de probabilidad (PDF)
graficar_densidad(x, pdf, title='Funcin de Densidad de Probabilidad (PDF) de la Distribucin Normal',
                  label='PDF de la Distribucin Normal',
                  x_label='X', y_label='Densidad de Probabilidad')

# Calcular la probabilidad de que la variable est en un intervalo especfico
a, b = -1, 1
probabilidad_intervalo = dist_normal.cdf(b) - dist_normal.cdf(a)
print(f"La probabilidad de que X est entre {a} y {b} es {probabilidad_intervalo:.4f}")

# Graficar el rea bajo la curva para el intervalo [a, b]
x_intervalo = np.linspace(a, b, 1000)
pdf_values_intervalo = dist_normal.pdf(x_intervalo)

# Graficar la funcin de densidad de probabilidad (PDF) en un intervalo
graficar_densidad_intervalo(x, pdf, x_intervalo, pdf_values_intervalo,
                            label='PDF de la distribucin normal',
                            title='rea bajo la curva para el intervalo [a,b]',
                            x_label='X', y_label='Densidad de Probabilidad')

```

Valores Esperados y Varianzas

---


El **valor esperado** (tambin conocido como esperanza matemtica o media) de una variable aleatoria continua $X$ se define como:

$$
E[X] = \int_{-\infty}^{\infty} x f_X(x) \, dx = \mu _{X}
$$

Por otra parte, la **varianza** de una variable aleatoria continua $X$ se calcula mediante la siguiente expresin, la cul puede expresarse en trminos de la funcin de densidad de probabilidad f_X{x}:

$$
\text{Var}[X] = E[(X - E[X])^2] = \int_{-\infty}^{\infty} (x - E[X])^2 f_X(x) \, dx 
$$

La **desviacin estndar** es la raz cuadrada de la varianza y proporciona una medida de la dispersin en las mismas unidades que la variable aleatoria, la cul se calcula simplemente como:

$$
\sigma = \sqrt{\text{Var}[X]}
$$

```python
# Clculo del valor esperado y la varianza de una variable contnua
mu = 8
sigma = 6.7
normal = stats.norm(mu,sigma)

# Calculamos el valor esperado
expected_value, _ = integrate.quad(lambda x: x * normal.pdf(x), -np.inf, np.inf)

# Calculamos la varianza
variance, _ = integrate.quad(lambda x: (x - expected_value)**2 * normal.pdf(x), -np.inf, np.inf)

#Calculamos la desviacin estandard
sigma_X = np.sqrt(variance)

# Imprimimos los valores
print(f'El valor esperado de nuestra variable X es = {round(expected_value,3)}')
print(f'La varianza de nuestra variable X es = {round(variance,3)}')
print(f'La desviacin estandard de nuestra variable X es = {round(sigma_X,3)}')
```

Distribucin de Densidad Acumulada

---

La funcin de densidad acumulada (CDF, por sus siglas en ingls) de una variable aleatoria continua es una funcin que describe la probabilidad de que la variable aleatoria tome un *valor menor o igual a un cierto valor*. Es una herramienta fundamental en estadstica y probabilidad para analizar la distribucin de probabilidades de variables continuas. Para una variable aleatoria continua $X$, la funcin de densidad acumulada $F(x)$ est definida como:

$$
F(x) = P(X \leq x) = \int_{-\infty}^{x} f_X(t) \, dt 
$$

donde $f_X(t)$ es la funcin de densidad de probabilidad (PDF, por sus siglas en ingls) de $X$. Algunas propiedades de la CDF son:

1. **No decreciente**: La CDF es una funcin no decreciente, es decir, si $a \leq b$, entonces $F(a) \leq F(b)$.
2. **Lmites**:
   - $\lim_{x \to -\infty} F(x) = 0$
   - $\lim_{x \to \infty} F(x) = 1$
3. **Rango**: El valor de la CDF est en el rango [0, 1].
4. **Continuidad**: Para una variable aleatoria continua, la CDF es una funcin continua.

Una de las principales utilidades de la CDF es que nos permte realizar el clculo de probabilidades para una variable como:

$$
P(a \leq X \leq b) = F_X(b) - F_X(a)


```python
# Clculo de la probabilidad acumulada para una distribucin normal
import scipy.integrate as integrate

# Parmetros de la distribucin normal
mu = 8
sigma = 6.7
normal = stats.norm(mu, sigma)

# Definimos una funcin para hallar la CDF
def cdf(funcion, valores_x, a):
    return [integrate.quad(lambda x: funcion.pdf(x), -np.inf, x_val)[0] for x_val in valores_x if x_val <= a]

# Generamos los valores de la distribucin
x = np.linspace(-4 * sigma + mu, 4 * sigma + mu, 300)

# Limitar el valor mximo para la CDF
a = 4 * sigma + mu 
cdf_vals = cdf(normal, x, a)

# Graficar la funcin de densidad de probabilidad (CDF)
graficar_densidad(x, cdf_vals, title='Funcin de Densidad de Acumulada (CDF) de la Distribucin Normal',
                  label='CDF de la Distribucin Normal',
                  x_label='X', y_label='Densidad de Probabilidad Acumulada')
```

Dentro de las aplicaciones de cualquier CDF encontramos el *clculo de percentiles*. Para calcular percentiles utilizando la funcin de densidad acumulada (CDF) de una distribucin, se usa la relacin entre la CDF y la funcin de distribucin inversa (tambin conocida como la funcin cuantil o percentil).

El **percentil** $p$ (donde $ 0 < p < 1 $) es el valor de la variable aleatoria tal que una fraccin $ p $ de la distribucin est por debajo de ese valor. Es decir, el percentil $ p $ es el valor $ x_p $ tal que $ F(x_p) = p $. 

Por otra parte, la **Funcin Cuantil (Inversa de la CDF)** denotada como $ F^{-1}(p) $, es la inversa de la CDF. Para un valor dado $ p $, $ F^{-1}(p) $ devuelve el valor de $ x $ tal que la probabilidad acumulada es $ p $. Matemticamente, se define como:

$$
F^{-1}(p) = x_p \quad \text{donde} \quad F(x_p) = p
$$

As pues, para calcular el percentil $p$ de una distribucin dada, se sigue el siguiente procedimiento:

1. **Identificar el valor de $p$**: Decidir el percentil deseado. Por ejemplo, para el percentil 90, $p = 0.90$.

2. **Calcular la inversa de la CDF**: Usar la funcin cuantil $ F^{-1}(p) $ para encontrar el valor de $ x $ correspondiente. Este valor es el percentil buscado.

En este sentido, algunas de las aplicaciones del clculo de percentiles mediante el uso de la CDF son:

- **Evaluacin de Pruebas y Clasificacin**: Los percentiles son usados para clasificar puntuaciones en pruebas estandarizadas, determinando cuntas personas estn por debajo de una cierta puntuacin.
- **Riesgo Financiero**: En finanzas, los percentiles (como el Value at Risk, VaR) se utilizan para medir el riesgo de inversiones.
- **Control de Calidad**: En manufactura, los percentiles ayudan a determinar lmites de tolerancia para asegurar la calidad de los productos.

Para ilustrar el clculo de cuartles mediante el uso de la CDF, consideremos el caso de la distribucin normal la cul hemos usado como nuestra distribucin contnua de ejemplo hasta ahora. Concretamente, la CDF de la distribucin normal la cual es simplemente su integral viene dada por:

$$
F(x) = \frac{1}{2} \left[ 1 + \text{erf}\left(\frac{x - \mu}{\sigma \sqrt{2}}\right) \right]
$$

donde $\text{erf}()$ es la *funcin error* dada por:

$$
\text{erf}(x) = \frac{2}{\sqrt{\pi}} \int_0^x e^{-t^2} \, dt
$$

Para encontrar el percentil $p$ realizamos la igualdad $p = F(x)$ y despejamos el valor de $x$.

$$
p = \frac{1}{2} \left[ 1 + \text{erf}\left(\frac{x_p - \mu}{\sigma \sqrt{2}}\right) \right]
$$

Despejando $x_p$:

$$
2p - 1 = \text{erf}\left(\frac{x_p - \mu}{\sigma \sqrt{2}}\right)
$$

$$
\frac{x_p - \mu}{\sigma \sqrt{2}} = \text{erfinv}(2p - 1)
$$

$$
x_p = \mu + \sigma \sqrt{2} \, \text{erfinv}(2p - 1)
$$

Este $x_p$ es el valor del percentil $p$ para la distribucin normal con parmetros $\mu$ y $\sigma$.


Para ilustar su uso, calcularemos a continuacin el valor de $x_p$ mediante el uso de la funcin error inversa, valor el cul corresponde al valor de $X$ que representa el percentil $p$ para una distribucin normal.

```python
# Clculo del percentil 90% para una distribucin normal
from scipy.special import erf, erfinv
from scipy.stats import norm

def percentile_error_function(percentile, mu=0, sigma=1):
    x_p = mu + sigma * (np.sqrt(2) * erfinv(2*percentile - 1))
    return x_p

# Hallamos el percentile de inters usando la funcin error inversa
mu = 8
sigma = 6.7
percentile = 0.9
normal = stats.norm(mu, sigma)

x_p_value = percentile_error_function(percentile, mu, sigma)
p_value = normal.cdf(x_p_value)

print(f'El valor de X correspondiente al percentile {percentile*100} es =', round(x_p_value,3))
print(f'Si evaluamos x_p={round(x_p_value,3)} dentro de la CDF normal, obtenemos el percentile = ', p_value*100)
```

**Estimacin Numrica de los Percentiles**

Notemos que el calculo de los percentiles es realmente sencillo una vez podemos hallar $F^{-1}(p) = x_p$ en el caso de estar interesados en el valor de $X$ que representa el percentil $p$ de nuestros datos. Sin embargo, muchas veces no tenemos conocimiento de la forma de la funcin inversa de la CDF, por ende, nos es posible deducir el valor $x_p$ de manera numrica si conocemos cul es nuestra funcin CDF a travs de diferentes mtodos de optimizacin tales como el *Mtodo Newton-Raphson*.

El mtodo de Newton-Raphson es una tcnica iterativa para encontrar races de una funcin (sus ceros), que en el contexto de la distribucin normal se utiliza para calcular el percentil correspondiente a un valor de probabilidad acumulada. El mtodo de Newton-Raphson se basa en una aproximacin lineal de la funcin en torno a una estimacin inicial:

1. **Definicin del Problema**:
   - Queremos encontrar el valor $x$ tal que la funcin de distribucin acumulada (CDF) de la distribucin normal con media $\mu$ y desviacin estndar $\sigma$ sea igual a un valor $p$, es decir, queremos resolver $F(x) = p $.

2. **Funcin Objetivo**:
   - Definimos una funcin $g(x)$ que representa la diferencia entre la CDF y el valor deseado $p$:
     $$
     g(x) = F(x) - p
     $$
   - Queremos encontrar la raz de esta funcin, es decir, el valor de $x$ para el cual $g(x) = 0$.

3. **Derivada de la Funcin**:
   - La derivada de $g(x)$ con respecto a $x$ es la funcin de densidad de probabilidad (PDF) de la distribucinde inters ya que dentro del contexto de $g(x)$, $p$ es una constante y al ser CDF = $F(x)$ la integral de la PDF = $f(x)$, la derivada de la CDF es la msma PDF:
     $$
     g'(x) = f(x) = \text{PDF}(x)
     $$

4. **Iteracin de Newton-Raphson**:
   - El mtodo utiliza la frmula de actualizacin:
     $$
     x_{n+1} = x_n - \frac{g(x_n)}{g'(x_n)} = x_n - \frac{g(x_n)}{f(x_n)}
     $$
   - Aqu, $x_n$ es la estimacin actual del percentil, y $x_{n+1}$ es la nueva estimacin basada en la aproximacin lineal de la funcin en $x_n$.

5. **Proceso Iterativo**:
   - Se inicia con una estimacin inicial $x_0$ (a menudo se toma la media $\mu$ como una buena estimacin inicial).
   - Se actualiza la estimacin usando la frmula de Newton-Raphson hasta que la diferencia entre estimaciones sucesivas sea menor que una tolerancia predefinida, o hasta alcanzar un nmero mximo de iteraciones.


```python
#Estimacin numrica de los percentiles de una distribucin normal
import numpy as np
from scipy.special import erf, erfinv
from scipy.stats import norm

# Definimos la CDF  y la PDF  de la distribucin
def normal_cdf(x, mu=0, sigma=1):
    return 0.5 * (1 + erf((x - mu) / (sigma * np.sqrt(2))))

def normal_pdf(x, mu=0, sigma=1):
    return (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)

# Definimos el mtodo Newton-Raphson para hallar los ceros de la funcin objetivo
def newton_raphson_cdf(p, mu=0, sigma=1, x0=0, tol=1e-6, max_iter=100):
    
    # Definimos la funcion objetivo
    def func(x):
        return normal_cdf(x, mu, sigma) - p
    
    # Definimos la derivada de la funcin objetivo
    def d_func(x):
        return normal_pdf(x, mu, sigma)
    
    x = x0
    for _ in range(max_iter):
        fx = func(x)
        dfx = d_func(x)
        if abs(dfx) < 1e-10:  # Evitar divisin por cero
            return x
        x_new = x - fx / dfx
        if abs(x_new - x) < tol:
            return x_new
        x = x_new
    return x

# Parmetros de la distribucin normal
mu = 8
sigma = 6.7

# Percentil deseado (por ejemplo, el percentil 90)
percentile = 0.90

# Calcular el valor del percentil usando Newton-Raphson
result = newton_raphson_cdf(percentile, mu, sigma, x0=mu)
print(f'El valor de X correspondiente al percentile {percentile*100} es = {result:.2f}')
```

Distribucin Uniforme

---

La distribucin de probabilidad uniforme es la distribucin de probabilidad contnua ms elemental dentro de las distribuciones de variables aleatorias contnuas. Corres ponde a la funcin de densidad de probabilidad definida en los rangos $R = [a,b]$ donde todos los posibles valores de $X$ tienen el msmo valor de probabilidad. 

$$
f_X(x) = \frac{1}{b-a}
$$

para todo $a \leq x \leq b$, fuera de ste intervalo, la PDF es nula. Por otra parte, la CDF de la distribucin contnua viene dada por:

$$
F_X(x) = \int_a^x \frac{1}{b - a} \, dt = \frac{x-a}{b-a}
$$

$$
F_X(x) =
\begin{cases} 
p & \text{si } x < a \\
(x-a)f_X(x) & \text{si } a\leq x < b \\
1 & \text{si } b\leq x
\end{cases}
$$

**Propiedades Estadsticas de la Distribucin de Uniforme**

Dentro de las propiedades estadsticas ms importantes de la distribucin uniforme encontramos:

1. **Esperanza (Media)**:q

$$

$$

2. **Varianza**: La varianza es tambin $\lambda t$, lo que significa que la media y la varianza son iguales.

$$
\text{Var}[X] = \frac{(b-a)^2}{12}
$$

3. **Desviacin Estndar**: 

$$
\sigma _{X} = \sqrt{\frac{(b-a)^2}{12}}
$$



**Casos de Uso y Aplicaciones**

1. **Generacin de Nmeros Aleatorios:** La distribucin uniforme se utiliza para generar nmeros aleatorios en simulaciones y experimentos de Monte Carlo.
2. **Sistemas de Control:** Se usa en sistemas de control y diseo experimental para representar incertidumbre y variabilidad.
3. **Pruebas Estadsticas:** En algunas pruebas estadsticas, la distribucin uniforme puede ser utilizada para evaluar la aleatoriedad y la independencia de los datos.
4. **Modelado de Ruido:** En algunos modelos de ruido y seal, la distribucin uniforme puede ser usada para representar errores o variaciones aleatorias.

```python
# Parmetros de la distribucin uniforme
a = 0
b = 10

# Crear una distribucin uniforme
dist = stats.uniform(loc=a, scale=b - a)

# Generar valores aleatorios
random_samples = dist.rvs(size=1000)

# Calcular PDF y CDF
x = np.linspace(a - 1, b + 1, 1000)
pdf = dist.pdf(x)
cdf = dist.cdf(x)

# Estadsticas principales
mean = dist.mean()
variance = dist.var()
std_dev = dist.std()

print(f"Media: {mean}")
print(f"Varianza: {variance}")
print(f"Desviacin Estndar: {std_dev}")

# Graficar la funcin de densidad de probabilidad (PDF)
graficar_densidad(x, pdf, title='PDF de la distribucin Uniforme',
                  label='PDF de la Distribucin Normal',
                  x_label='X', y_label='Densidad de Probabilidad f(x)')


# Graficar la funcin de densidad de probabilidad (CDF)
graficar_densidad(x, cdf, title='CDF de la distribucin Uniforme',
                  label='CDF de la Distribucin Normal',
                  x_label='X', y_label='Densidad de Probabilidad Acumulada F(x)')
```

Distribucin Normal

---

La distribucin normal $X \sim \mathcal{N}(\mu, \sigma^2)$, tambin conocida como distribucin gaussiana, es una distribucin de probabilidad continua que se utiliza ampliamente en estadstica, matemticas, y muchas otras disciplinas. Su forma es conocida como la campana de Gauss debido a su grfica caracterstica. La funcin de densidad de probabilidad (PDF) de la distribucin normal se define como:

$$
f_X(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp \left( -\frac{(x - \mu)^2}{2 \sigma^2} \right) 
$$

donde:
- $\mu$ es la media de la distribucin y caracteriza la ubiaccin del **centro de la distribucin normal**.
- $\sigma$ es la desviacin estndar (la raz cuadrada de la varianza $\sigma^2$) la cul caracteriza **el ancho de la distribucin normal**.

Dentro de las caractersticas ms importantes de la distirbucin normal encontramos las siguientes:

1. **Forma de la Distribucin**: La grfica de la distribucin normal es simtrica y tiene forma de campana. *La media, la mediana y la moda de la distribucin normal son todas iguales y se encuentran en el centro de la distribucin*.

2. **Media $\mu$**: Es el valor alrededor del cual la distribucin est centrada. Es la medida del "centro" de la distribucin.

3. **Desviacin Estndar $\sigma$**: Mide la dispersin o extensin de la distribucin. Una desviacin estndar ms grande significa que la distribucin es ms ancha y plana, mientras que una desviacin estndar ms pequea significa que la distribucin es ms estrecha y alta.

4. **Propiedades de la Distribucin relacionadas a los percentiles**:
   - **Simetra**: La distribucin es perfectamente simtrica alrededor de la media $\mu$.
   - **rea bajo la Curva**: El total del rea bajo la curva de la funcin de densidad es 1, lo que representa una probabilidad total del 100%.
   - **68-95-99.7 Regla**: Aproximadamente el 68% de los datos estn dentro de la primera desviacin estndar de la media, el 95% dentro de 2 desviaciones estndar, y el 99.7% dentro de 3 desviaciones estndar. Esta regla es tambin conocida como *la regla emprica*. Esta regla establece entonces que el 99% porciento de los valores de la distribucin se encuentran entre las primeras tres desviaciones estandar a partir de la media:
   $$
   P(\mu - 3\sigma  \leq x \leq \mu + 3\sigma) = 0.997

5. **Propiedad de la Suma**: La suma de variables aleatorias independientes que siguen una distribucin normal tambin sigue una distribucin normal. Esta propiedad es fundamental en la teora de probabilidad y en la inferencia estadstica.

Para ilustrar cmo una variable aleatoria puede aproximarse a una distribucin normal a medida que el tamao de la muestra aumenta, podemos realizar un experimento usando el Teorema Central del Lmite. Este teorema establece que, para una muestra suficientemente grande de variables aleatorias independientes e idnticamente distribuidas, la distribucin de la media muestral se aproxima a una distribucin normal, independientemente de la distribucin original.

En este ejemplo, vamos a simular un experimento donde generamos mltiples muestras de una distribucin uniforme (o cualquier otra distribucin) y calculamos la media de cada muestra. Luego, mostraremos cmo estas medias muestrales se distribuyen normalmente cuando el tamao de la muestra es grande.

```python
# Parmetros del experimento
num_samples = 1000  # Nmero de muestras
sample_size = 30    # Tamao de cada muestra
low, high = 0, 1    # Rango de la distribucin uniforme

# Generar muestras y calcular medias
means = []
for _ in range(num_samples):
    sample = np.random.uniform(low, high, sample_size)
    means.append(np.mean(sample))

# Visualizar el resultado
plt.figure(figsize=(8, 5))

# Histograma de las medias muestrales
plt.hist(means, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Ajustar y dibujar la distribucin normal terica
mu, sigma = np.mean(means), np.std(means)
x = np.linspace(min(means), max(means), 100)
pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
plt.plot(x, pdf, 'r', linewidth=2)
plt.title('Distribucin de la Media de Muestras')
plt.xlabel('Media de la Muestra')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

La estandarizacin de una variable es un proceso que transforma una variable aleatoria en una variable estandarizada, que tiene una *media de 0 y una desviacin estndar de 1*. Este proceso *convierte las unidades originales de la variable a una escala comn, facilitando la comparacin y el anlisis* Para estandarizar una variable $X$ con media $\mu$ y desviacin estndar $\sigma$, se usa la siguiente frmula:

$$
Z = \frac{X - \mu}{\sigma}
$$

Donde:
- $X$ es el valor original de la variable.
- $\mu$ es la media de la variable.
- $\sigma$ es la desviacin estndar de la variable.
- $Z$ es el valor estandarizado o puntaje $Z$.

**Propiedades de la Variable Estandarizada**

- **Media**: La media de los valores estandarizados es 0.
- **Desviacin Estndar**: La desviacin estndar de los valores estandarizados es 1.
- **Distribucin**: Si la variable original $X$ sigue una distribucin normal $ \mathcal{N}(\mu, \sigma^2) $, entonces la variable estandarizada $Z$ sigue una distribucin normal estndar $ \mathcal{N}(0, 1) $.

**Distribucin Normal Estndar**

La distribucin normal estndar es una distribucin normal con una media de 0 y una desviacin estndar de 1. Se denota como $ \mathcal{N}(0, 1) $. La funcin de densidad de probabilidad para una distribucin normal estndar se define como:

$$
f(z) = \frac{1}{\sqrt{2 \pi}} \exp \left( -\frac{z^2}{2} \right) 
$$

Donde $ z $ es un valor de la variable estandarizada.

**Importancia en Ciencia y Anlisis de Datos**

**1. Comparacin y Normalizacin**: La estandarizacin es crucial en ciencia y anlisis de datos para normalizar diferentes variables que pueden tener unidades o escalas diferentes. Esto permite comparar variables de manera equitativa y mejora el rendimiento de algoritmos que dependen de la magnitud de los datos, como los algoritmos de aprendizaje automtico que utilizan distancias, como K-means o mquinas de soporte vectorial (SVM).

**2. Identificacin de Valores Atpicos**: La interpretacin del valor $Z$ corresponde al ***nmero de desviaciones estandar \sigma que una observacin $x$ est por encima o por debajo de la media***. As, considerando la regla 68-95-99 tenemos que un valor $x$ se clasifica como:
  * $Z$ > 0 : $x$ se encuentra por encima de la media (donde la media representa la ubicacin del 50% de los datos).
  * $Z$ < 0 : $x$ se encuentra por debajo de la media.
  * $Z$ > 2 o $Z$ > 3 : $x$ puede ser considerado como un valor atpico dependiendo del humbral del contexto, no obstante el valor ms utilizado es  $Z$>3 ya que sto corresponde a que la observacin se encuentra tres desviaciones estandar por encma o por debajo de la media, lo que es equivalente a que es un dato que no corresponde con la tendencia del 99.7% de los datos.

**Clculo de Probabilidades con la Distribucin Normal Estndar**

Gracias a la perfeccin y simetra de la distribucin normal, es posible determinar la probabilidad de un valor de la en la distribucin original a partir de la partir de la PDF de la distribucin estandar mediante su CDF como hemos visto anteriormente. Esto, devido a que la integral de la distribucin estandar es mucho ms fcil de calcular. La CDF proporciona la probabilidad acumulada de que la variable aleatoria tome un valor menor o igual a un valor especfico. Para calcular la probabilidad de que un valor estandarizado \(Z\) sea menor o igual a un cierto valor \(z\), se usa la CDF:
$$
   P(X \leq x) = P(Z \leq z) = \Phi(z)
$$

   Donde $\Phi(z)$ es la CDF de la distribucin normal estndar.

Por otra parte, se tiene que gracias a la simetra de la distribucin estandar:

$$
  P(Z \leq -z) = 1 - P(Z \leq z)
$$

Donde $P(Z \leq z)$ es el **percentil** de inters de la variable aleatoria.

Note a continuacin, que el rango de la distribucin normal estandar se encuentra entre -4 y 4, lo que simboliza el nmero de desviaciones estandar a partir de la media cero.
