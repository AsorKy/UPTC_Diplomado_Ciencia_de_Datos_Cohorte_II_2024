Conceptos Fundamentales de la Teora de la Probabilidad


---

**Definicin de Probabilidad:**

La **probabilidad** es una rama de las matemticas que se ocupa del estudio de fenmenos aleatorios y la cuantificacin de la incertidumbre. En trminos simples, la probabilidad mide la ***posibilidad*** de que ocurra un determinado evento. Matemticamente, la probabilidad de un evento se expresa como un nmero entre 0 y 1, donde 0 indica que el evento es imposible y 1 indica que es seguro.

**Teora de la Probabilidad:**

La **teora de la probabilidad** es el campo matemtico que estudia los experimentos o procesos aleatorios, donde el resultado no puede predecirse con certeza antes de que ocurra. Se utiliza para modelar situaciones en las que hay incertidumbre y para hacer *inferencias* sobre el comportamiento de fenmenos aleatorios. Esta teora proporciona las herramientas y los principios para calcular y entender la probabilidad de eventos, as como para analizar la distribucin de los resultados posibles.

**Conceptos Bsicos de la Teora de la Probabilidad:**

1. **Experimento Aleatorio:** Es un proceso que lleva a cabo una accin que genera resultados que no pueden predecirse con certeza. Ejemplos incluyen lanzar una moneda, tirar un dado, o seleccionar una carta de un mazo.

2. **Espacio Muestral (S):** Es el conjunto de todos los posibles resultados de un experimento aleatorio. Por ejemplo, al lanzar un dado, el espacio muestral es ${1, 2, 3, 4, 5, 6}$.

3. **Evento (A):** Es un subconjunto del espacio muestral. Un evento ocurre si el resultado del experimento pertenece a este subconjunto. Por ejemplo, al lanzar un dado, el evento "obtener un nmero par" se puede representar como ${2, 4, 6}$.

4. **Probabilidad de un Evento (P(A)):** Es una medida que indica qu tan probable es que ocurra un evento. Se define como la proporcin del nmero de resultados favorables al nmero total de resultados posibles en el espacio muestral. Por ejemplo, si lanzamos una moneda justa, la probabilidad de obtener "cara" es 1/2.

5. **Evento Seguro:** Es un evento que siempre ocurre, es decir, el espacio muestral completo. Su probabilidad es 1.

6. **Evento Imposible:** Es un evento que nunca ocurre, y su probabilidad es 0.

7. **Eventos Mutuamente Excluyentes:** Son eventos que no pueden ocurrir simultneamente. Por ejemplo, al lanzar una moneda, obtener "cara" y "cruz" son eventos mutuamente excluyentes.

8. **Probabilidad Condicional (P(A|B)):** Es la probabilidad de que ocurra un evento $A$ dado que ha ocurrido otro evento $B$. Se utiliza para actualizar la probabilidad de un evento basado en nueva informacin.

8. **Probabilidad Conjunta (P(A,B)):** Es la probabilidad de que ocurra un evento $A$ y $B$ ocurran simultaneamente dentro del espacio muestral.

9. **Variable Aleatoria:** Es una variable que asigna un valor numrico a cada resultado posible de un experimento aleatorio, es decir, es aquella variable cuyos valores provienen de experimentos aleatorios. Por ejemplo, podemos definir en el experimento aleatorio "lanzar un dado", la variable aleatoria $X$ como el valor que puede adquirir el dado despus de un lanzamiento, donde un evento $A$ corresponde a un valor especfico adquirido por la variable $X$. Asi, la sentencia "Probabilidad de que el dado tome un valor igual a par" viene representado por $P(X=A)$ con $A = {2,4,6}$.

    * *Variable Aleatoria Discreta:* Variable aleatoria $X$ que puede tomar un valor finito y contable de valores.
    
    * *Variable Aleatoria Contnua:* Variable aleatoria $X$ que puede tomar infinitos valores y cuyos estados posibles se encuentran en un rango contnuo.

La forma ms rudimentaria, pero conceptualmente elemental para calcular una probabilidad, consiste en definir un evento dentro del espacio muestral donde podamos calcular la proporcin entre ocurrencias de nuestro evento y el tamao del espacio muestral. Por ejemplo, piense en una moneda donde los estados posibles de la moneda son 'cara' o 'sello'. Si el experimento consiste en lanzar la moneda nicamente una vez, el espacio muestral tendr un tamao igual a 2.

No obstante, podemos ampliar nuestro espacio muestral a partir de un espacio muestral ms pequeo como el de nuestra moneda. Piense ahora que nuestra variable aleatoria es ahora *$X$ = combinacin de resultados de la moneda despus de N lanzamientos.* Con esto en mente, ahora nuestro estado en cuestin no es solo un nico estado de la moneda a la salida de nuestro experimento aleatorio sino la combinacin de sus dos posibles estados a lo largo de N lanzamientos.
Es decir, cuando nos preguntamos por una combinacin de  estados sucesivos de la moneda, calculamos la probabilidad $P(X=A) = P(S_{1}, S_{2}, S_{3},...,S_{N})$.

```python
# Conteo de probabilidad de un evento complejo

def calcular_probabilidad(evento, experimentos, resultado_deseado, mostrar_espacio=False):
    espacio_muestral = list(product(evento, repeat=experimentos))
    numero_resultado_deseado = espacio_muestral.count(resultado_deseado)

    if mostrar_espacio:
        print('Espacio muestral del experimento:')
        for combinacion in espacio_muestral:
            print(combinacion)

    probabilidad = numero_resultado_deseado / len(espacio_muestral)

    return probabilidad


# Definimos un experimento aleatorio

evento = ('cara', 'cruz')
lanzamientos = 3
resultado_deseado = ('cara', 'cara', 'cara')

# Calcular la probabilidad de obtener "cara, cara, cara" en tres lanzamientos
probabilidad = calcular_probabilidad(evento, lanzamientos, resultado_deseado,True)
print(f"La probabilidad de obtener {resultado_deseado} es {probabilidad:.4f}")
```

Propiedades de la Probabilidad

---

Las propiedades fundamentales de la probabilidad, tambin conocidas como axiomas de probabilidad, son principios bsicos que forman la base de la teora de la probabilidad. Estas propiedades son consistentes con la intuicin y los conceptos bsicos de probabilidad y se aplican a cualquier espacio muestral y conjunto de eventos. Aqu estn las propiedades fundamentales:

1. **No negatividad (Axioma de no negatividad):**
   La probabilidad de cualquier evento $A$ es un nmero no negativo entre cero y uno. Esto significa que:
   $$
   1 \geq P(A) \geq 0
   $$
   para cualquier evento \( A \).

2. **Probabilidad del evento seguro (Axioma de normalizacin):**
   La probabilidad del evento seguro (es decir, el evento que incluye todos los resultados posibles) es igual a 1. Esto se puede expresar como:
   $$
   P(S) = 1
   $$
   donde $S$ es el espacio muestral.

3. **Probabilidad Complementaria:**
   Corresponde a la probabilidad de que el evento en cuestin **NO** ocurra. Esto se puede expresar como:
   $$
   P(B) = 1 - P(A)
   $$
   donde $A$ es el evento de inters, y $B$ son TODOS los eventos dentro del espacio muestral $S$ que no corresponden a $A$ tal que $S = A \cup B$, es decir, $B$ es el complemento de $A$.

4. **Aditividad (Axioma de aditividad o suma de probabilidades):**
   Para cualquier secuencia de eventos mutuamente excluyentes (es decir, eventos que no pueden ocurrir simultneamente), la probabilidad de la unin de estos eventos es igual a la suma de las probabilidades de los eventos individuales. En otras palabras, si $ A_1, A_2, \dots, A_n $ son eventos mutuamente excluyentes, entonces:
   
   $$
   P(A_1 \cup A_2 \cup \dots \cup A_n) = P(A_1) + P(A_2) + \dots + P(A_n)
   $$

Adems de estos axiomas bsicos, existen otras propiedades derivadas de ellos que son tiles en el anlisis de problemas probabilsticos:

5. **Probabilidad del evento vaco:**
   La probabilidad del evento vaco (un evento que no puede ocurrir) es 0:
   $$
   P(\emptyset) = 0
   $$

6. **Aditividad para eventos no mutuamente excluyentes:**
   Para cualquier dos eventos $A$ y $B$, la probabilidad de la unin de estos eventos es:
   $$
   P(A \cup B) = P(A) + P(B) - P(A \cap B)
   $$
   donde $A \cap B$ es la interseccin de $A$ y $B$.

7. **Probabilidad condicional:**
   La probabilidad de un evento $A$ dado que ha ocurrido otro evento  $B$ se define como:
   $$
   P(A|B) = \frac{P(A \cap B)}{P(B)}
   $$
   siempre que $P(B) > 0 $.



Ya que en parrafos anteriores hemos visto como la probabilidad, al construirse como una proporcin entre los estados favorables y el tamao del espacio muestral resulta en una probabilidad positiva menor o igual a uno, exploremos a continuacin un ejemplo de la probabilidad complementaria.

```python
# Definimos un experimento aleatorio

evento = ('cara', 'cruz')
lanzamientos = 3
resultado_deseado = ('cara', 'cara', 'cara')

# Calcular la probabilidad de obtener "cara, cara, cara" en tres lanzamientos
probabilidad_A = calcular_probabilidad(evento, lanzamientos, resultado_deseado,True)

print('*'*30)
print(f"La probabilidad de obtener {resultado_deseado} es {probabilidad_A:.4f}")
print(f"La probabilidad de NO obtener {resultado_deseado} es {1- probabilidad_A:.4f}")
```

Diagramas de Venn y la Identificacin de los Espacios Muestrales

---

Los **diagramas de Venn** son una herramienta visual y conceptual til para representar conjuntos (sub esopacios muestrales) y sus relaciones en un espacio muestral. Un espacio muestral es el conjunto de todos los posibles resultados de un experimento aleatorio, y los eventos dentro de este espacio pueden representarse como subconjuntos. Los diagramas de Venn nos ayudan a ilustrar la relacin entre diferentes eventos y a calcular probabilidades teniendo en cuenta el tamao de estos conjuntos y sus posibles intersecciones.

**Representacin de Espacios Muestrales y Eventos**

1. **Espacio Muestral (S):**
   En un diagrama de Venn, el espacio muestral se representa como un rectngulo que contiene todos los posibles resultados de un experimento. Cada punto dentro del rectngulo representa un resultado nico.

2. **Eventos:**
   Los eventos son subconjuntos del espacio muestral y se representan como crculos u otras figuras geomtricas dentro del rectngulo del espacio muestral. Por ejemplo, si tenemos un experimento donde lanzamos un dado, el evento "obtener un nmero par" puede representarse como un crculo que incluye los resultados {2, 4, 6}.

**Relacin entre Eventos y Clculo de Probabilidades**

Los diagramas de Venn facilitan la visualizacin de varias relaciones entre eventos, que son cruciales para calcular probabilidades:

1. **Unin de Eventos (A  B):**
   La unin de dos eventos $A$ y $B$ representa todos los resultados que pertenecen a al menos uno de los eventos. En el diagrama de Venn, esto se visualiza como la combinacin de las reas ocupadas por los dos eventos. La probabilidad de la unin de dos eventos se puede calcular utilizando la frmula:
   
   $$
   P(A \cup B) = P(A) + P(B) - P(A \cap B)
   $$

   donde $P(A \cap B)$ es la probabilidad de la interseccin de los eventos, que se representa como el rea comn de los dos crculos.

2. **Interseccin de Eventos (A  B):**
   La interseccin de dos eventos representa todos los resultados que pertenecen a ambos eventos. En un diagrama de Venn, esto se muestra como el rea donde los crculos se superponen. La probabilidad de la interseccin se calcula directamente observando la frecuencia relativa de los elementos en esta rea comn.

3. **Complemento de un Evento (A^c):**
   El complemento de un evento $A$ incluye todos los resultados en el espacio muestral que no estn en $A$. En el diagrama de Venn, esto se visualiza como el rea fuera del crculo que representa $A$ pero dentro del rectngulo del espacio muestral. La probabilidad del complemento de $A$ se calcula como:
   
   $$
   P(A^c) = 1 - P(A)
   $$

4. **Eventos Mutuamente Excluyentes:**
   Dos eventos son mutuamente excluyentes si no pueden ocurrir simultneamente, es decir, su interseccin es el conjunto vaco. En un diagrama de Venn, esto se visualiza como dos crculos que no se superponen. Para estos eventos, la probabilidad de la unin es simplemente la suma de las probabilidades individuales:
   
   $$
   P(A \cup B) = P(A) + P(B)
   $$
   ya que $P(A \cap B) = 0$.

**Probabilidad de eventos mutuamente excluyentes**

Para ilustrar el clculo de la probabilidad entre eventos excluyentes, generaremos un espacio muestral de $N \times M$ donde $N$ es el nmero de estados mnimo posbile dentro de nuestro experimento y $M$ es el nmero de veces que el experimento base se ejecuta. Note entonces que para que nuestro experimento sea el "lanzamiento de dos dados a la par o, el lanzamiento de un dado dos veces consecutivas" necesitaremos $N=$6 y $M$=2. Pasos a seguir:

* Generar el espacio muestral $N \times M$.
* Definir $A$ y $B$.
* Calcular la probabilidad de $A$ y $B$.
* Comprobar si $A$ y $B$ son ,mutuamente excluyentes entre si.
* Calcular $ P(A \cup B) = P(AB) = P(A) + P(B)$.

```python
# Generamos un espacio muestral con N estados para E experimentos
def generar_espacio_muestral(num_estados, num_experimentos):
  estados = list(range(1, num_estados + 1))
  espacio_muestral = list(product(estados, repeat=num_experimentos))
  return espacio_muestral


# Calculamos la probabilidad
def calcular_probabilidad(evento, espacio_muestral):
  num_eventos = len(evento)
  num_total   = len(espacio_muestral)
  p = num_eventos / num_total
  return p

# Especificamos el experimento
num_estados_min = 6
num_experimentos = 2

# Generamos el espacio muestral
espacio_muestral = generar_espacio_muestral(num_estados_min, num_experimentos)
print('*'*40)
print(f"Espacio Muestral ({len(espacio_muestral)} elementos): {espacio_muestral}")

# Definimos dos eventos A y B mutuamente excluyentes como subconjuntos de S
evento_A = [(1,1), (1,2), (1,3)]
evento_B = [(6,6), (6,5), (6,4)]
print('*'*40)
print('Evento A = ', evento_A)
print('Evento B = ', evento_B)

#Verificamos que los eventos sean mutuamente excluyentes
mutuamente_excluyentes = len(set(evento_A) & set(evento_B)) == 0
print('*'*40)
print(f"Son los eventos mutuamente excluyentes? {mutuamente_excluyentes}")

# Calculamos la probabilidad para cada evento
prob_A = calcular_probabilidad(evento_A, espacio_muestral)
prob_B = calcular_probabilidad(evento_B, espacio_muestral)

# Calcular la probabilidad de la unin de eventos mutuamente excluyentes
prob_union = prob_A + prob_B
print('*'*40)
print(f"Probabilidad de la unin de A y B (mutuamente excluyentes): {prob_union}")
```

```python
# Definimos los estados independientes de manera que cumplan cierta condicion

def definir_estados_por_suma(num_estados_min, num_experimentos, valor_A, valor_B):
  espacio_muestral = generar_espacio_muestral(num_estados_min, num_experimentos)
  evento_A = []
  evento_B = []
  complemento_AB = []

  for estado in espacio_muestral:
    if sum(estado) == valor_A:
      evento_A.append(estado)
    elif sum(estado) == valor_B:
      evento_B.append(estado)
    else:
      complemento_AB.append(estado)

  return evento_A, evento_B, complemento_AB


# Especificamos el experimento
num_estados_min = 6
num_experimentos = 2

# Generamos el espacio muestral
espacio_muestral = generar_espacio_muestral(num_estados_min, num_experimentos)
print('*'*40)
print(f"Espacio Muestral ({len(espacio_muestral)} elementos): {espacio_muestral}")

# Definimos los eventos
valor_A = 7
valor_B = 10

evento_A, evento_B, complemento_AB = definir_estados_por_suma(num_estados_min, num_experimentos, valor_A, valor_B)
print('*'*40)
print('Evento A = ', evento_A)
print('Evento B = ', evento_B)

#Verificamos que los eventos sean mutuamente excluyentes
mutuamente_excluyentes = len(set(evento_A) & set(evento_B)) == 0
print('*'*40)
print(f"Son los eventos mutuamente excluyentes? {mutuamente_excluyentes}")

# Calculamos la probabilidad para cada evento
prob_A = calcular_probabilidad(evento_A, espacio_muestral)
prob_B = calcular_probabilidad(evento_B, espacio_muestral)

# Calcular la probabilidad de la unin de eventos mutuamente excluyentes
if mutuamente_excluyentes:
  prob_union = prob_A + prob_B
  prob_complemento = calcular_probabilidad(complemento_AB, espacio_muestral)
  print('*'*40)
  print(f"Probabilidad de la unin de A y B (mutuamente excluyentes): {prob_union}")
  print(f"Probabilidad de que NO ocurra A o B: {prob_complemento}")
else:
  print('Los eventos son mutuamente compatibles, debes calcular P de una manera distinta.')
```

**Probabilidad de eventos compatibles**

Para ilustrar el clculo de la probabilidad entre eventos compatibles, generaremos un espacio muestral de $N \times M$ donde $N$ es el nmero de estados mnimo posbile dentro de nuestro experimento y $M$ es el nmero de veces que el experimento base se ejecuta. En esta situacin, la probabilidad de un evento conjunto $P(AB)$ debe tener en cuenta, que la parte del espacio muestral de $A$ se intersecta con la parte de $B$ Y por ende la interseccin de ambos estados representa aquella situacin donde los estados del espacio muestral $S$ pueden en efecto, pertenecer a ambos estados. De esta manera si sumamos las probabilidades de $A$ y $B$ sin tener en cuenta el aporte de la interseccin, estaramos contando **dos veces la msma porcin del espacio muestral**, y por ende, nos es necesario restar dicha fraccin.

* Generar el espacio muestral $N \times M$.
* Definir $A$ y $B$.
* Calcular la probabilidad de $A$ y $B$.
* Comprobar si $A$ y $B$ son ,mutuamente excluyentes entre si.
* De no ser mutuamente excluyentes, calcular $P(A \cap B)$.
* Calcular $ P(A \cup B) = P(A) + P(B) - P(A \cap B)$.

En el siguiente ejemplo, se ilustra un caso no excluyente: calcular la probabilidad de que al lanzar dos dados de seis caras, se tenga $A$ = la suma de los resultados de los dados es mayor que siete, $B$ = al menos uno de los dados presenta un dado con cada par.

Independencia entre eventos

---

Dos eventos $A$ y $B$ son independientes si la ocurrencia de uno no afecta la probabilidad de ocurrencia del otro. En trminos matemticos, esto significa que la probabilidad conjunta de ambos eventos es igual al producto de sus probabilidades individuales.

$$
P(AB) =  P(A \cup B) = P(A) P(B)
$$

Donde si exsten mltiples eventos independientes $A_1, A_2, A_3 , ... A_n$, la probabilidad conjunta es:

$$
P(A_1 \cap A_2 \cap \dots \cap A_n) = \prod_{i=1}^{n} P(A_i)
$$

A continuacin, te presento un cdigo en Python que ilustra el concepto de independencia entre eventos mediante el clculo de la probabilidad conjunta de eventos independientes. Utilizaremos un ejemplo simple donde se lanzan dos dados, y se consideran los siguientes eventos:

* Evento A: Obtener un nmero par en el primer dado.
* Evento B: Obtener un nmero mayor a 4 en el segundo dado.
Estos eventos son independientes porque el resultado de un dado no afecta el resultado del otro.

```python
# Definimos las funciones que caracterizan nuestro experimento aleatorio

def es_par(num):
    return num % 2 == 0

def mayor_a_cuatro(num):
    return num > 4

def lanzar_dado():
    return random.randint(1, 6)

# Nmero de experimentos
num_experimentos = 10000

# Contadores de eventos
contador_A = 0
contador_B = 0
contador_A_y_B = 0

# Realizar experimentos
for _ in range(num_experimentos):
    dado1 = lanzar_dado()
    dado2 = lanzar_dado()

    evento_A = es_par(dado1)
    evento_B = mayor_a_cuatro(dado2)

    if evento_A:
        contador_A += 1
    if evento_B:
        contador_B += 1
    if evento_A and evento_B:
        contador_A_y_B += 1

# Calcular probabilidades
prob_A = contador_A / num_experimentos
prob_B = contador_B / num_experimentos
prob_A_y_B = contador_A_y_B / num_experimentos

# Probabilidad conjunta asumiendo independencia
prob_A_y_B_independiente = prob_A * prob_B

# Resultados
print('*'*30)
print(f"Probabilidad de A (nmero par en dado 1): {prob_A}")
print(f"Probabilidad de B (nmero > 4 en dado 2): {prob_B}")
print(f"Probabilidad conjunta observada P(A y B): {prob_A_y_B}")
print(f"Probabilidad conjunta bajo independencia P(A) * P(B): {prob_A_y_B_independiente}")
print('*'*30)
```

Probabilidad Condicional

---

Contrario a los eventos independientes donde la ocurrencia de un evento $A$ no determina o afecta la probabilidad de ocurrencia de un evento $B$, encontramos los eventos no independientes para los cules la probabilidad de ocurrencia de un evento afecta la del otro. Sea $P(A|B)$ la probabilidad de $A$ dada una informacin previa, es decir, la certeza de $A$ bajo una restriccin de ocurrencia de otro evento $B$, viene dada por la siguiente igualdad.

$$
P(A|B) = \frac{P(A \cap B)}{P(B)}
$$

Para entender como calcular desde el espacio muestral la probabilidad condicional, pensemos en que si la probabilidad condicional ajusta la probabilidad de
$P(A)$ al considerar solo los casos donde $P(B)$ ha ocurrido, vemos que el condicionamiento tiene el efecto de restringir el espacio muestral original. Cuando consideramos una probabilidad condicionada, estamos interesados en un **subconjunto** del espacio muestral donde el evento condicionante $B$ **particiona** el espacio muestral en dos partes: los resultados en los que $B$ ocurre y en los que no. En este sentido, el conjunto de resultados donde $B$ ocurre se convierte en el *"nuevo espacio muestral restringido"*, por ende, $P(A|B)$ es calculado con base a este conjunto de estados y no sobre el conjunto $S$ original.

```python
# Conteo de probabilidad de un evento complejo

def definir_espacio_muestral(estados_base, num_experimentos, mostrar_espacio=False):
    espacio_muestral = list(product(estados_base, repeat=num_experimentos))
    if mostrar_espacio: print('El espacio muestral es = ', espacio_muestral)
    return espacio_muestral


# Definimos un experimento aleatorio

estados_base = ('cara', 'cruz')
resultado_deseado_A = ('cara', 'cara')
resultado_deseado_B ='cara'
lanzamientos = 2
espacio_muestral = definir_espacio_muestral(estados_base, lanzamientos,True)

evento_A = [res for res in espacio_muestral if res == resultado_deseado_A]
evento_B = [res for res in espacio_muestral if res[0] == resultado_deseado_B]
A_dado_B = [res for res in evento_B if res == resultado_deseado_A ]

P_A_dado_B = len(A_dado_B)  / len(evento_B)

# Calcular la probabilidad de obtener "cara, cara" DADO QUE el primer lanzamiento es 'cara'
print('Evento A = ', evento_A)
print('Evento B = ', evento_B)
print('Evento A dado B = ', A_dado_B)
print(f"La probabilidad de obtener {resultado_deseado_A} dado {resultado_deseado_B} es {P_A_dado_B:.4f}")
```

Ahora, ilustremos un ejemplo de probabilidad condicional usando un dado de $N$ estados con $M$ lanzamientos: Cul es la probabilidad de obtener una suma igual a 7 al lanzar dos dados **dado que** el primer dado es par?
* Evento A: Obtener un nmero par en el primer dado.
* Evento B: La suma de los dos dados es mayor a 7.

```python
# Probabilidad condicionada para un dado

def es_par(num):
    return num % 2 == 0

def suma_mayor_a_siete(dados):
    return sum(dados) > 7

def lanzar_dado(num_estados=6):
    return random.randint(1, num_estados)

# Generar espacio muestral para dos dados
num_estados = 6
num_experimentos = 2
espacio_muestral = list(product(range(1, num_estados + 1), repeat=num_experimentos))

# Definir los eventos A y B
evento_A = [(d1, d2) for (d1, d2) in espacio_muestral if es_par(d1)]
evento_B = [(d1, d2) for (d1, d2) in espacio_muestral if suma_mayor_a_siete((d1, d2))]
evento_A_y_B = [(d1, d2) for (d1, d2) in evento_B if es_par(d1)]

# Calcular probabilidades
prob_B = len(evento_B) / len(espacio_muestral)
prob_A_y_B = len(evento_A_y_B) / len(espacio_muestral)
prob_A_dado_B_manual = len(evento_A_y_B) / len(evento_B)

# Probabilidad condicional P(A | B)
if prob_B > 0:
    prob_A_dado_B = prob_A_y_B / prob_B
else:
    prob_A_dado_B = 0

# Resultados
print(f"Probabilidad de B (suma > 7): {prob_B}")
print(f"Probabilidad conjunta P(A y B): {prob_A_y_B}")
print(f"Probabilidad condicional P(A | B): {prob_A_dado_B}")
print(f"Probabilidad condicional P(A | B) manual: {prob_A_dado_B_manual}")
```

Relacin entre Diagramas de Venn, Interseccin de Conjuntos y Probabilidad Condicional

---

La **interseccin** de dos conjuntos $A$ y $B$, denotada como $ A \cap B $, representa los elementos que son comunes a ambos conjuntos. En trminos de eventos, $ A \cap B $ corresponde a la ocurrencia **simultnea** de ambos eventos. La probabilidad de la interseccin de dos eventos es una medida de cun probable es que ocurran ambos eventos al mismo tiempo y viene dada por $P(A \cap B)$.

Por otra parte la **probabilidad condicional** de un evento $A$ dado un evento $B$ (denotada como $P(A \mid B)$) se puede visualizar en un diagrama de Venn como el **rea del conjunto $ A \cap B $ en relacin con el rea total del conjunto $ B $**. Es decir, cuando consideramos $P(A \mid B)$, estamos restringiendo nuestro "espacio muestral" al conjunto $ B $ y luego examinando qu fraccin de ese conjunto tambin pertenece a $ A $. Es decir, mientras que la probabilidad conjunta se calcula como:

$$
P(A \cap B) = \frac{|A \cap B|}{|S|}
$$

Donde $|A \cap B|$ denota el tamao del conjunto de interseccin entre $A$ y $B$ y $|S|$ es el tamao del espacio muestral. Por otra parte, la probabilidad condicional se calcula con respecto al conjunto de interseccin como:

$$
P(A|B) = \frac{|A \cap B|}{|B|}
$$

Donde $|B|$ es el tamao del conjunto $B$. En resumen, la probabilidad conjunta de $A$ y  $B$ corresponde a la **proporcin del conjunto de intersccin con respecto al espacio muestral S**, mientras que la probabilidad condicional corresponde a la **proporcin del conjunto de interseccin con respecto al evento condicionante B**. $P(A \mid B)$ se interpreta como la probabilidad de $A$ dentro del "nuevo" espacio muestral definido por $B$. Esto significa que al calcular $P(A \mid B)$, estamos midiendo el rea relativa de $A \cap B$ en comparacin con el rea total de $B$ dentro de un diagrama de Venn reflejando la "condicin" impuesta por el evento \( B \).

Regla del Producto General

---

En prrafos anteriores, vimos que si dos eventos $A$ y $B$ son independientes, la probabilidad conjunta de ambos eventos es simplemente el producto de sus probabilidades:

$$
P(A \cap B) = P(A)P(B)
$$

Cuando los eventos no son independientes, la probabilidad de uno de los eventos puede afectar la probabilidad del otro. En este caso, se utiliza la probabilidad condicional para ajustar la probabilidad conjunta donde tenemos la siguiente relacin:

$$
P(A \cap B) = P(A)P(B|A)
$$

Para entender el por que de esta relacion, vimos que $P(A\cap B) = |A \cap B|/|S|$, por otra parte si $P(A)=|A|/|S|$ siguiendo la anterior definicion llegamos a:

$$
P(A)P(B|A) = \frac{|A|}{|S|} \frac{|B\cap A|}{|A|} = \frac{|B\cap A|}{|S|}
$$

Lo cual restaura la definicin natural de la probabilidad conjunta. Note que lo anterior puede interpretarse como que la probabilidad conjunta es equivalente a "el producto del evento 'independiente' $A$ llevado a cabo sobre el espacio muestral $S$, por el evento 'independiente' $B$ llevado a cabo sobre el subespacio muestral $A$". Debido a que el conjunto de interseccin es el msmo para los dos conjuntos, se cumple que:

$$
P(A \cap B) = P(A)P(B|A) = P(B)P(A|B)
$$

Finalmente, despejando de la relacin anterior la probabilidad condicional, tenemos:

$$
P(A|B) = \frac {P(A \cap B)}{P(B)}
$$

Y equivalente mente, tenemos el teorema de Bayes dado por:

$$
P(A|B) = \frac {P(A)P(B|A)}{P(B)}
$$

Principio de Probabilidades Totales y rboles de Probabilidad

---

Para entender el clculo de ste tipo de probabilidades, cmo podemos calcularla a partir del espacio muestral finito y su interpretacin, necesitamos comprender primero que un conjunto de eventos $N_{i}$ constituye una particin del espacio muestral $S$ si son incompatbles entre si de modo tal que su union de cuenta del espacio muestral. Dicho esto, un evento $M$ constituido como una coleccin de estados pertenecientes a las particiones $N_{i}$ del espacio muestral $S$ donde $S=N_{1} \cup N_{2} \cup ... \cup N_{n}$, define una probabilidad $P(M)$ que es calculable a partir de determinar aquellas probabilidades $P(N_{i})$ y $P(M|N_{i})$ relacionadas con el evento $M$, donde $MN_{i}$ es la porcion de estados del espacio muestral que pertenecen tanto a $M$ como a $N_{i}$ y cuya union $M=MN_{1} \cup MN_{2} \cup ... \cup MN_{k}$ define el evento $M$ siendo $M \neq S$ y $M \in S$. En este sentido, la probabilidad de $M$ puede calcularse de la siguiente manera teniendo en cuenta la igualdad de la probabilidad condicional.

$$
P(M) = \sum_{i}^k P(M|N_{i})P(N_{i})
$$

Ahora, pensemos en que si la particin o **categora** $N_{i}$ puede estar compuesta por otras categoras o particiones $N_{i}=G_{1} \cup N_{2} \cup ... \cup N_{r}$, se cumple que

$$
P(N_{i}, G_{j}) = P(G_{j})P(N_{i}|G_{j})
$$

Tenemos, que la suma llevada a cabo sobre todas las particiones de $M$ las cuales incluyen las subparticiones $G_{j}$ debe cumplir la restauracin del espacio muestral $M$ tal que:

$$
P(N_{i}) = \sum_{j} P(N_{i}, G_{j}) = \sum_{j} P(G_{j})P(N_{i}|G_{j})
$$

$$
P(M) = \sum_{i} P(N_{i}) = \sum_{ij} P(N_{i}, G_{j}) = 1
$$


El principio de probabilidades totales es una regla fundamental en la teora de la probabilidad que permite calcular la probabilidad de un evento basndose en un conjunto de eventos mutuamente excluyentes y exhaustivos. Este principio es til cuando un evento puede ocurrir de varias formas distintas, y estas formas son mutuamente excluyentes y cubren todas las posibilidades.



**rboles de Probabilidad**


Los rboles de probabilidad son una herramienta grfica que facilita la representacin y el clculo de probabilidades para una secuencia de eventos. Son particularmente tiles cuando se aplica el principio de probabilidades totales, ya que permiten visualizar cmo se descompone un evento en varios casos o categoras mutuamente excluyentes.

Un rbol de probabilidad comienza con un nodo inicial que se ramifica en diferentes ramas, cada una representando un posible evento. A cada rama se le asocia una probabilidad. Las ramas pueden continuar dividindose en sub-ramas, representando eventos condicionales.

* **Nodos:** Representan eventos o situaciones posibles.
* **Ramas:** Representan las probabilidades de los eventos que parten de un nodo.

arbol_probabilidad.png

En general, cuando tenemos un espacio muestral $S$ que puede ser dividido de acuerdo a un evento cuyos posibles estados pueden ser de la categora $N_{i}$ y a su vez, estos estados pueden ser divididos de acuerdo a un segundo evento donde los posibles resultados son las categoras $G_{j}$, nos es posible realizar un rbol de probabilidad que grficamente permita deducir las probabilidades de los diferentes escenarios definibles en el sistema siguiendo los siguientes pasos:
1. Identificamos el espacio muestral de inters y el subespacio muestral de cada evento.
2. Construimos un rbol de probabilidad que represente los eventos de inters, siendo cada evento una familia de rammas en el rbol, y los nodos, los posibles resultados del evento.
3. Completamos el rbol de acuerdo a la informacin disponible, donde el conjunto de nodos debe reflejar todas las combinaciones posibles en el espacio muestral.
4. Identificamos las probabilidades disponibles en el enunciado del problema, pesando las ramas del rbol con stas probabilidades.
5. Calcularmos la probabilidad de un evento $M$ de inters, como el producto de las probabilidades encontradas a lo largo de la rama que dan cuenta del evento $M$.


Vamos a ilustrar el concepto de un rbol de probabilidad utilizando el ejemplo de una prueba mdica. En este ejemplo, consideraremos una poblacin de 1,000,000 personas, con un cierto porcentaje de personas enfermas y sanas, y calcularemos las probabilidades asociadas con los resultados de la prueba (diagnstico).

Asumamos los siguientes datos:

* Poblacin total: 1,000,000 personas.
* Prevalencia de la enfermedad ($P(E)$): 1% (porcentaje de personas enfermas).
* Sensibilidad de la prueba ($P(Positivo | Enfermo)$): 99% (probabilidad de que la prueba sea positiva si la persona est enferma).
* Especificidad de la prueba ($P(Negativo | Sano)$): 95% (probabilidad de que la prueba sea negativa si la persona est sana).
* Objetivo: Calcular la probabilidad de que una persona sea sana o enferma dado un resultado positivo o negativo en la prueba.

sick_test.jpg

```python
# Datos iniciales
poblacion_total = 1000000
p_enfermo = 0.0001
p_sano = 1 - p_enfermo

# Sensibilidad y especificidad de la prueba
p_positivo_dado_enfermo = 0.99
p_negativo_dado_sano = 0.95

# Calculamos probabilidades complementarias
p_negativo_dado_enfermo = 1 - p_positivo_dado_enfermo
p_positivo_dado_sano = 1 - p_negativo_dado_sano

# Calculamos el nmero de personas en cada categora
num_enfermos = poblacion_total * p_enfermo
num_sanos = poblacion_total * p_sano

verdaderos_positivos = num_enfermos * p_positivo_dado_enfermo
falsos_positivos = num_sanos * p_positivo_dado_enfermo
verdaderos_negativos = num_sanos * p_negativo_dado_sano
falsos_negativos = num_enfermos * p_negativo_dado_enfermo

# Probabilidad de que la persona sea enferma dado un resultado positivo
p_enfermo_dado_positivo = verdaderos_positivos / (verdaderos_positivos + falsos_positivos)

# Probabilidad de que la persona sea sana dado un resultado negativo
p_sano_dado_negativo = verdaderos_negativos / (verdaderos_negativos + falsos_negativos)

# Mostrar los resultados
print(f"P(Enfermo | Positivo): {p_enfermo_dado_positivo:.4f}")
print(f"P(Sano | Negativo): {p_sano_dado_negativo:.4f}")

# Resultados detallados
print("\nResultados detallados:")
print(f"Personas enfermas: {num_enfermos:.0f}")
print(f"Personas sanas: {num_sanos:.0f}")
print(f"Verdaderos positivos: {verdaderos_positivos:.0f}")
print(f"Falsos positivos: {falsos_positivos:.0f}")
print(f"Verdaderos negativos: {verdaderos_negativos:.0f}")
print(f"Falsos negativos: {falsos_negativos:.0f}")
```

Teorema de Bayes

---

El **Teorema de Bayes** es un resultado fundamental en la teora de la probabilidad que describe cmo actualizar nuestras creencias o conocimientos sobre un evento a la luz de nueva evidencia. Este teorema permite calcular la probabilidad condicional de un evento dado otro evento que ha ocurrido.
Dado un evento $A$ y un evento $B$, la probabilidad condicional de $A$ dado $B$ se puede calcular como:

$$
P(A \mid B) = \frac{P(B \mid A) \cdot P(A)}{P(B)}
$$

donde:
- $P(A \mid B)$ es la **probabilidad posterior** de $A$ dado $B$.
- $P(B \mid A)$ es la **probabilidad de ver la evidencia $B$ si $A$ es verdadero (verosimilitud)**.
- $P(A)$ es la **probabilidad previa (apriori)** de $A$.
- $P(B)$ es la **probabilidad marginal (evidencia)** de $B$, la cual se puede calcular como:

$$
P(B) = P(B \mid A)P(A) + P(B \mid \neg A)P(\neg A)
$$

Note que en la relacin anterior, $A$ es el evento referente a que nuestra hipotesis sea verdadera mientras que $\neg A$ es el evento complementario. 

El teorema se usa comnmente en contextos donde se tiene una hiptesis $A$ y se observa nueva evidencia $B$. Permite actualizar la probabilidad de la hiptesis basndose en esta evidencia. Es ampliamente utilizado en reas como estadstica, aprendizaje automtico, inferencia estadstica, entre otras.


Supongamos que en una poblacin, el 1% de las personas tiene una enfermedad. Una prueba para esta enfermedad tiene una sensibilidad (probabilidad de dar positivo si la persona est enferma) del 99% y una especificidad (probabilidad de dar negativo si la persona est sana) del 95%. Queremos calcular la probabilidad de que una persona tenga la enfermedad dado que ha obtenido un resultado positivo en la prueba.

```python
# Probabilidades previas
p_enfermo = 0.01  # Probabilidad de estar enfermo P(E)
p_sano = 1 - p_enfermo  # Probabilidad de estar sano P(E)

# Sensibilidad y especificidad
p_positivo_dado_enfermo = 0.99  # P(Positivo | E)
p_negativo_dado_sano = 0.95  # P(Negativo | E)

# Probabilidad de positivo dado que est sano (falsos positivos)
p_positivo_dado_sano = 1 - p_negativo_dado_sano  # P(Positivo | E)

# Probabilidad total de un resultado positivo
p_positivo = (p_positivo_dado_enfermo * p_enfermo) + (p_positivo_dado_sano * p_sano)

# Aplicacin del Teorema de Bayes para calcular P(E | Positivo)
p_enfermo_dado_positivo = (p_positivo_dado_enfermo * p_enfermo) / p_positivo

# Mostrar resultado
print(f"La probabilidad de estar enfermo dado un resultado positivo es: {p_enfermo_dado_positivo:.4f}")
```

Tcnicas de conteo y el clculo de probabilidades en espacios finitos

---

En un espacio muestral $S$ finito, las tcnicas de conteo o sampleo constituyen a una serie de estrategias que nos permiten deducir el nmero de estados del espacio de muestras pertenecientes a un evento en especfico. Esto se hace an ms necesario no solo en situaciones donde debamos determinar una probabilidad de manera rpida y precisa sino a dems, en espacios muestrales donde la cantidad de estados posibles para el sistema es demasiado grande como para manualmente, generar el espacio $S$ y contar uno por uno aquellos estados que pueden cumplir con los requerimientos de la situacin de inters.
Con esto en mente, segn sea la situacin a estudiar, los eventos definibles dentro de espacio finitos pueden categorizarse como una de cuatro configuraciones del espacio muestral:

* Muestras de tamao $r$ de $S$.
* Permutaciones de tamao $r$ de $S$.
* Combinaciones de tamao $r$ de $S$.
* Particiones de tamao $r$ de $S$.


**1. Muestras de tamao $r$ de $S$**


Las **muestras de tamao $r$** de un conjunto $S$ consisten en seleccionar $r$ elementos de $S$, donde el ***orden y la repeticin no son relevantes***, dependiendo del problema. En concreto, una muestra es un conjunto de estados ***ordenado con remplaxo (es posible repetir estados)***, es decir, se permite seleccionar el mismo elemento ms de una vez y el orden de seleccin importa
$$
M(r;n) =  n^r
$$
     
Donde $n$ es el nmero de estados posibles para cada arreglo ordenado o muestra de tamao $r$. **Ejemplo**: Lanzar un dado (6 caras) tres veces, se tiene $6^3 = 216 = |S|$ posibles resultados.


**2. Permutaciones de tamao $r$ de $S$**


Las **permutaciones** son arreglos de todos los elementos de un conjunto en un orden particular. Cuando se trata de permutaciones de tamao $r$ de un conjunto $S$ de tamao $n$, se refiere a ordenar determinar los arreglos ***ordenados sin repetir***de $r$ elementos de $S$. La cantidad de permitaciones de tamao $r$ en un espacio de tamao $n$ es igual a:

$$
p(r;n) =  \frac{n!}{(n-r)!}
$$

Un **ejemplo de** permitacin es Ordenar tres personas de un grupo de 5, tenemos $P(5, 3) = \frac{5!}{(5-3)!} = 60 $ permutaciones posibles.


**3. Combinaciones de tamao $r$ de $S$**


Las **combinaciones** corresponden a colecciones de tamao $r$ del conjunto $S$ donde el ***orden de los elementos no interesa y no es posible la repeticin***. El nmero de combinaciones posibles dentro de un espacio muestral viene dado por:

$$
C(r;n) = \frac{n!}{(n-r)!r!}
$$

Un **ejemplo** de combinacin, es el seleccionar 3 cartas de una baraja de 52 cartas, donde C(52, 3) = \binom{52}{3} = \frac{52!}{3!(52-3)!} = 22,100. Por otra parte, si se permite la ***repeticin de elementos***, pero el orden no importa, se tiene que:

$$
C(r;n) = \frac{(n+r-1)!}{(n-r)!r!}
$$

Un **ejemplo** de una combinacin con remplazo es el elegir 3 bolas de colores de una bolsa con 4 colores diferentes, donde se permite la repeticin, se tiene $\binom{4+3-1}{3} = \binom{6}{3} = 20$.

** 4. Particiones Ordenadas de $S$*

Las **particiones ordenadas** de un conjunto $S$ consisten en dividir $S$ en ***grupos disjuntos donde el orden de los grupos importa***. Se usa para contar la cantidad de maneras en que se pueden repartir $n$ elementos en $k$ grupos con un orden especfico de los grupos. Si $n_1, n_2, \ldots, n_k$ son los tamaos de los $k$ grupos, la cantidad de particiones realizables viene dada por la siguiente expresion, donde $n = n_1 + n_2 + \cdots + n_k$:

$$
G(r,k;n) = \frac{n!}{n_1! n_2! \cdots n_k!}
$$

Un **ejemplo** de uso es dividir 5 personas en 3 grupos, con el primer grupo de 2 personas, el segundo de 2 personas y el tercero de 1 persona, se tiene $ \frac{5!}{2!2!1!} = 30 $ maneras.


Estas tcnicas de conteo son esenciales para calcular probabilidades en espacios muestrales finitos porque proporcionan un mtodo sistemtico para enumerar los posibles resultados de un experimento aleatorio. Al conocer el nmero de resultados posibles y el nmero de resultados favorables, se puede determinar la probabilidad de eventos de inters usando la definicin clsica de probabilidad:

$$
P(A) = \frac{\text{Nmero de resultados favorables}}{\text{Nmero total de resultados}}
$$

Estas herramientas son particularmente tiles en situaciones donde se desea calcular probabilidades sin realizar un experimento aleatorio real, lo cual puede ser costoso, imprctico o imposible.

**Muestra $M(r;n)$**

```python
import random

# Conjunto S
S = list(range(1, 11))  # Nmeros del 1 al 10

# Tamao de la muestra
r = 3

# Muestra con reemplazo
muestra_con_reemplazo = [random.choice(S) for _ in range(r)]
print("Muestra con reemplazo:", muestra_con_reemplazo)

# Muestra sin reemplazo
muestra_sin_reemplazo = random.sample(S, r)
print("Muestra sin reemplazo:", muestra_sin_reemplazo)
```

**Permutaciones**

```python
import math

# Parmetros
n = 10  # Tamao del conjunto S
r = 3   # Tamao de la permutacin

# Clculo de permutaciones
permutaciones = math.factorial(n) // math.factorial(n - r)
print("Cantidad de permutaciones de tamao", r, "de", n, "elementos:", permutaciones)
```

**Combinaciones**

```python
import math

# Parmetros
n = 10  # Tamao del conjunto S
r = 3   # Tamao de la combinacin

# Clculo de combinaciones
combinaciones = math.factorial(n) // (math.factorial(r) * math.factorial(n - r))
print("Cantidad de combinaciones de tamao", r, "de", n, "elementos:", combinaciones)
```

**Particiones Ordenadas**

```python
from itertools import permutations, combinations

# Conjunto S
S = ['a', 'b', 'c', 'd']

# Tamao de los subgrupos
group_sizes = [2, 2]

# Generar todas las combinaciones de tamao 2
combs = list(combinations(S, group_sizes[0]))

# Generar todas las permutaciones de estas combinaciones
partitions = []
for comb in combs:
    remaining_elements = [item for item in S if item not in comb]
    remaining_comb = combinations(remaining_elements, group_sizes[1])
    for other_comb in remaining_comb:
        # Ordenar los grupos y agregar a las particiones
        partitions.append([comb, other_comb])

# Imprimir las particiones ordenadas
print("Particiones ordenadas:")
for partition in partitions:
    print(partition)
```
